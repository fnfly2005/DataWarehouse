# 11 商业智能应用程序介绍
![11](../syn_pic/data_warehouse_life_cycle/11.png)  
主要的前台开发活动包括设计和构建报表及分析的初始集，称为商业智能BI应用程序  
分析周期提供了一个框架，用于了解BI应用程序如何使用以及BI应用程序设计和开发的意义  
整个团队都应阅读本章和下一章的内容，通过本章的学习您将了解到：  
+ 分析周期及其对设计商业只能应用程序的意义 5
+ 有关BI应用程序的类型、价值以及提交平台选择的基本概念
+ 常见BI应用程序的描述和对机构的意义，这其中包括标准报表、分析性应用程序、仪表板和记分卡、数据挖掘以及运营BI
+ BI门户的角色为业务用户访问BI应用程序的主要接入点

## 11.1 商业智能应用程序的重要性
在您的结构里，大部分知识工作者并不会建立自己的报表，所以必须向这些人提供恰当的方式以便于他们访问数据仓库中的信息，而这正是BI应用程序的基本任务  
思考一下这句话的含义：对于多数业务用户来说，BI应用程序是他们访问DW/BI系统的唯一方式 5  
最终标准是，BI应用程序必须要做得非常好，他们必须满足下列设计需求：  
+ **正确**。与其他系统的不同之处在于，BI应用程序必须提供准确的审计报表和文档  
+ **性能良好**。平均响应时间不到5秒钟是一个比较好的目标  
+ **易于使用**。那些完全能够进行在线订书的用户应该至少可以使用标准报表，一个不错的目标是，定制后的结果应当不超过10次鼠标点击就能获取  
+ **外观漂亮**。工具和报表应看起来干净利落并且富有吸引力 5  
+ **是一项长期投资**。应用程序必须适当地进行存档、维护、增强和扩展

在满足机构的一些了数据访问需求的过程中，BI应用程序填补了重要的空白。图11-1显示了从战略报表需求到运营报表需求的一系列需求  
![11-1](../syn_pic/data_warehouse_life_cycle/11-1.png)  
在最极端的信息访问情形中，一些用户需要从零开始构建对数据仓库的查询，这就是所谓的即席访问  
这些业务分析员或超级用户通常负责创建复杂分析，寻找比较紧迫的问题或时机。通常运营系统中所有事务处理都会利用运营BI，或者说运营BI会为大量的用户所使用，例如呼叫中心的所有代理人 5  
这两个极端之间的差异非常大。根据我们的经验，DW/BI系统用户中有很大的比例位于中端，或者位于范围的低端，其百分比可能高达80%-90%。这些用户需要预先建立的、已参数化了的报表  
注意：对于很多业务内行来说，即席仅仅是通过改变报表上的参数来创建个性化报表。对大多数业务用户来说，BI应用程序的标准报表集是主要的访问工具  
在成熟的DW/BI系统中，通常有几十个甚至数百个BI应用程序，用户需要的东西太多了，仅凭一个简单的列表是无法满足的。这需要您提供一个高级界面，也成为BI门户，它为应用程序提供了合理的顺序  
大多数BI应用程序都是通过BI门户访问的，至少刚开始时需要通过这种方式，一旦用户找到感兴趣的报表，他们很可能直接通过工具平台或者托管的应用程序来访问具体内容  
在图11-1中，只按照一个消费模式对用户进行归类比较困难。多数即席用户也经常使用标准报表和其他分析性应用程序。您需要提供全方位的应用程序类型，同时用户会被那些能够帮助他们解决具体问题的应用程序所吸引 5  
BI应用程序需求背后的巨大推动力是人们在技术功能的深度和质量方面的需求存在很大差别。因为数据仓库需要满足各种层次的技术技能和各种程度的分析经验  
## 11.2 商业智能分析周期
业务分析是一个常见过程，从监视活动开始识别某个问题或时机，进而采取行动，最终回到监视该行动产生的结果上来。分析周期把业务分析过程分为5个不同的阶段，如图11-2所示
![BI分析的分析周期](../syn_pic/data_warehouse_life_cycle/11-2.png)  
### 11.2.1 第一阶段：监视活动
标准报表是分析周期的起点。用户使用这些报表对当前结果和前期结果或计划进行对比，以便提供一份关于业务状态的绩效鉴定。BI应用程序需求在监视活动阶段主要关注呈现层和相关技术，如仪表板、门户、记分卡  
### 11.2.2  第二阶段：识别异常
这一阶段的重点是确定“除了什么问题”或“问题在哪里”。这一阶段主要识别相对于正常性能来说出现的异常情况和时机。识别异常的阶段可以帮助用户跳过海量数据，使他们集中经理关注最值得关注的时机  
识别异常阶段需要额外的功能，如用于向用户设备发布警报的分发服务器，帮助用户以多种创造性方式来查看数据的可视化工具，包括趋势性、火花线、地图或聚类 5  
### 11.2.3 第三阶段：确定构成原因的因素
这一阶段是为了理解“为什么”或识别异常的根本原因。其关键在于识别驱动异常性能变量之间的可靠关系和相互作用  
在这一阶段要成功支持用户的工作，将需要您的DW/BI系统架构中还含有其他软件，包括统计工具和/或数据挖掘算法，如关联、排序、分类和分割。您可以使用这些工具来量化问题和原因之间的因果关系  
### 11.2.4 第四阶段：模型供选方案
在这一阶段，您在已有因果关系的基础上设计模型以评价决策供选方案。数据仓库是洞察历史信息的宝库  
当使用典型的分析周期时，最终目标是具有对一系列潜在决策进行假设分析和仿真的功能   
在模型供选方案阶段，您的数据仓库架构可能也需要采用其他技术，包括用于模型评估的统计工具和数据挖掘算法，如敏感性分析、蒙特卡洛仿真以及单变量求解优化 5  
### 11.2.5 第五阶段：采取行动并跟踪结果
理想情况下，您可以在BI应用程序中使用闭环过程并向运营系统反馈建议的行动  
采取行动/跟踪结果阶段对您的数据仓库架构提出了额外要求。您需要通过API从数据仓库访问运营系统，增强已有的维度模型或者构建性能管理来跟踪特定业务决策的结果，并且确定哪些决策有效，哪些决策无效  
为了连续不断地对分析过程、业务规则和分析模型进行适当调整，应该捕捉并分析行动结果。这就又使您回到了分析周期开始的监视活动阶段  
### 11.2.6 分析周期的更多意义
分析周期帮助我们了解业务用户如何使用DW/BI系统，告诉我们应该提供什么样的工具来使他们能够简单、无缝地进行操作  
分析周期隐含于快速查询中，其中大部分周期处理都发生在分析员的大脑里。但是，周期必须被明确地应用于您构建地所有分析性应用程序中 5  
这样工作平台通过业务形势的分析主动地“指导”用户，最终帮助他们做出富有见地和深思熟虑的决策。分析周期的目标是：  
+ 主动指导业务用户超出基本报表的范围
+ 识别和了解异常的绩效情况
+ 为每个异常绩效情况捕获制定决策的“最优方法”
+ 在机构内部共享产生的“最优方法”或者知识资本 5

理解分析周期有助于确保为每个在业务需求定义过程中识别出的分析性时机都提供完整的解决方案  
## 11.3 商业智能应用程序的类型
回顾图11-1，我们看到BI应用程序可以分为几个主要的子集，包括：  
+ **直接访问查询和报表工具**允许用户直接查询维度模型并定义结果集。在这种情况下，即席工具也被其他使用这些工具的用户作为标准报表的开发工具  
+ **数据挖掘**包含于直接访问逻辑框中，因为经常有少数统计分析员使用数据挖掘工具来探查数据仓库中有用的模式和关系。产生的模型通常嵌入到其他的BI应用程序中  
+ **标准报表**是预定义的、预先编排好格式的报表，一般提供某个层次的用户交互，比如输入参数，向下钻取较低层次的细节数据，以及链接到相关报表  
+ **分析性应用程序**是一个嵌入领域专业知识的报表集合，根据对具体业务过程的分析来进行管理。分析性应用程序也常常使用数据挖掘模型 5  
+ **仪表板和记分卡**一般是由报表和图表组合而成的，使用突出异常和向下钻取功能对多个业务过程的数据进行分析  
+ **运营BI和闭环应用程序**包括使用比典型运营报表更加复杂的应用程序。这些应用程序充分利用丰富的历史上下文跨越数仓多个业务过程来指导运营决策的制定。这些应用程序也常带有回访源系统的交互界面  

这些BI应用程序、BI门户的基本组织框架将在本章单独的一节里讨论  
### 11.3.1 直接访问查询和报表工具
即席查询和报表工具为用户提供了直接访问维度模型的工具，通常是通过一个元数据层，提供额外的列的子分组、业务描述以及联接的定义  
由于这些工具比较复杂并且需要了解数据的细节内容，因此即席工具大部分情况下都是供那些既有专业技术又有商业头脑的专家、超级用户使用的。一般来说，这些用户能将列拖拽到报表区，设置约束，定义计算并添加格式 5  
即席工具复杂的一个原因是SQL还不是真正意义上的报表编辑器语言。OLAP语言更为完善，但是却很难构建查询。在这两种情形下，查询工具必须克服将业务问题转换成查询语言语法形式时所面临的根本性困难  
一般来说，即席查询工具应该包含四个方面的功能：构建查询、分析和表示、用户体验以及技术特性  
1. 查询构建  
    顾名思义，查询工具的首要工作是构建查询。您可能需要的查询构建功能包括：
    + **多通道或多重集查询**  
        为了在报表中断行里计算比较关系或正确计算非加性度量，查询工具必须将报表分解为几个简单查询，然后分别交给RDBMS来处理  
        多通道查询也允许横向钻取几个不同业务过程一致性维度模型中的事实表，也可能是对几个不同的数据库进行查询 5  
        最后，多通道查询查询为数据库优化器和聚集导航器提供了一个加速生成报表的机会，因为每个原子查询都很简单并且易于分析  
        注意OLAP引擎使用的分析性语言能够只用一条语句就完成和多通道查询等价的工作  
    + **警报**  
        由于数据量成指数级增长，查询工具需要帮助用户识别那些与众不同的记录，通常可以基于一个比较规则，例如：“显示上个月销售上升或下降了10个百分点的地区”。人们常常在仪表板和记分卡上使用警报来度量绩效  
    + **逐次约束**  
        某个查询结果被用作子查询的限制或者过滤条件。在您创建一个行为研究分组来识别一个伙伴并且将其作为一个单位来检查其行为时，这是一种特别重要的功能  
        这些约束列表可能过于庞大以至于无法在工具中存储，由此可能需要传回数据库以便执行远程联接 5  
        这种情况下，您还可以通过创建临时表或在数据库里定义的视图保存研究分组参与者列表，虽然提到将列表或者视图写入数据库会让数据库管理员感到紧张，但是在某些BI工具中这是一种标准功能  
    + **半可加性求和**  
        在公共业务事实表中有一类重要的数值型独立并不是完全可加的。任何关于强度的度量在所有的维度上都是不可加的，尤其是日期维度  
        遗憾的是，您无法使用基本的SQL函数AVG进行整个时间段上的平均计算。AVG会对所有的维度进行平均，而不仅仅是时间  
        其实这里所需要的就是将一个通用的SUM操作符改成AVGTIMESUM。这种特性使所有包含库存水平，账户余额以及其他强度度量的应用程序明显简化  
    + **ANSI SQL 99的支持**  
        SQL 99对SQL添加了基本的OLAP功能，例如，包括一个WINDOW构造子，允许您对一个定义好的查询子集进行聚集，产生移动平均数 5  
    + **直接查询字符串记录**  
        作为最后一种手段，您可能需要查看并更改由工具生成的SQL语句或OLAP语句。如果您发下自己需要经常这么做，那么不是工具出了问题就是数据模型设计得有问题  
2. 分析和呈现功能  
    在大多数情况下，您的查询工具除了获取数据并以表格形式提交到桌面上之外还要做更多的工作。该工具必须支持操纵数据的业务需求并使用观感较佳的格式对其进行呈现  
    + **结果集的基本计算**。包括一系列数学函数、统计函数、字符串函数、顺序处理函数、条件函数和报表函数。这些计算常常用于克服工具中的其他缺陷  
    + **旋转结果**。旋转是多维度或交叉报表分析的基础  
    + **向下钻取**。这项功能允许用户通过包含额外的列向结果集添加更多的细节 5
    + **有关旋转结果的列计算**。这些计算创建一个计算列，是两个或多个旋转列的函数。单列计算，比如列的百分比、累积和n-tiles，都属于这一类    
    + **列与行计算**。有些计算是很有用的比如将一行的值显示为另一行值的百分比。份额计算和比率都依赖于这种功能  
    + **排序**。排序是很重要的。尤其是对于不显示出来的元素。例如，一个财务报表可能按与所显示信息无关的特定顺序来显示条目，而不是按字母顺序。在这种情况下，维度中的排序列指定了适当的显示顺序    
    + **复杂格式编排**格式编排往往超过其应有的重要性。您需要有全方位的图形设计工具，包括线、方框、阴影、图片、字体和尺寸，并且需要对报表元素位置拥有像素级的控制  
    + **图表和图形**。图表和图形是进行分析的基本要素。几乎每份报表最后的形式都是一幅图表，这一点只要做些“和眼球相关”的分析或预测就知道了 5  
    + **复合文档**。仪表板界面实际上就是一个由几个单独的报表、图形及着色工具组成的复合文档。这些可以通过一个共享参数联系起来（比如日期）  
    + **用户可更改变量**  用户可更改变量或参数可以位于查询文档中从查询过滤条件到报表首部的任何位置  
3. 用户体验  
    如果工具满足所有的技术需求，但用户不会使用或者不喜欢使用它，该工具等于没用，以下功能有助于改进用户对分析环境的体验：
    + **易用性**。工具应有自然而直观的界面。您必须让您的用户参与到对这一问题的评估中来。让他们参与BI工具评估，并了解他们是如何看待工具可用性的  
    + **元数据访问**。工具应该向用户提供上下文敏感的帮助，不仅对于工具本身要作出说明，对数据也是如此。这意味着工具必须提供灵活的方式从元数据目录里抽取描述性数据 5  
    + **选择列表**  
        该工具应提供一种方法来查找能在查询里用作约束或过滤条件的值列表    
        对一个7500万行的维度表种的高基数属性直接提交选择列表的请求，可能永远也不会返回结果，因此在维度内必须存在一个层次结果，以便在更高的层次约束用户查询，进而限定随后的选择列表中的结果数量  
    + **与其他应用程序的无缝集成**。集成也包括使报表在其他应用程序中可用。它们也具有一些可以使BI开发人员将报表嵌入到所有应用程序中去的API  
    + **导出多种文件类型**。理想情况下，工具中包含比较全面的发布功能，可以将最终报表和/或图表发布到文件目录、电子邮件或者直接发布到Web上  
    + **嵌入式查询**。用户应当能够从其他应用程序发出查询。例如，应该能够从一个电子表格调用查询并将结果行返回到指定区域，然后填入一个复杂的财务模型中 5  
4. 技术特性  
    有时，像多任务查询和取消查询这样的功能都是体现工具可用性的基本功能，如果缺失，您的用户可能会觉得不满：  
    + **多任务**。当一个查询在运行的时候，用户必须能够运行其他程序，也可以建立和运行其他查询  
    + **取消查询**。用户应该能中止一个由他们启动的正在运行的某个查询而不会终止所有查询。这种取消应该在数据库服务器上对中断进行管理，并且不需要重启桌面计算机  
    + **脚本**。一个脚本语言界面和命令行界面对于报表的自动执行很重要  
    + **连接性**。弄清楚您是否可以进入所有想要进入的数据库平台。连接性包括连接到其他数据源——文本，电子表格，XML文件，其他关系数据库以及OLAP引擎 5  
    + **调度**。工具需要提供或利用某种形式的调度系统。用户可能希望延迟查询，将其安排到夜晚执行，或者将其设置为定期处理
    + **元数据驱动**  
        管理员应该能定义简单的数据仓库子集，比如仅包含在单个业务过程维度模型内的那些表。这可能包含预定义的联接路径、列分组、业务描述、计算列，以及选择列表的来源  
    + **软件管理**。这是一个随着采用WEB作为应用程序平台而逐步消失的问题
    + **安全性**。理想情况下，该工具应参与到所有可用的用户验证系统中。BI工具可能需要提供验证功能，限定用户仅能访问允许他们访问的报表或数据子集  
    + **查询**  
        对数据库的直接查询应当不需要管理层的支持，或者要花费最少的工作量（即初始设置少于10分钟） 5  
        对数据仓库团队来说，这是特别有价值的，业务他们常常检查新的数据源，而且经常是在不同的平台上  

### 11.3.2 标准报表
标准报表位于BI应用程序的基础部分。它们是典型的参数驱动的报表，有预定义的、预格式编排的输出。听起来这些报表似乎很单调，但对业务来说，它们通常是起骨干作用的BI应用程序  
这对“按钮用户”的价值是明确的：标准报表提供了一种易于使用的方式，只需要短期学习就能够获取所需信息  
这些报表是企业度量的官方来源，并为这些度量提供了一个质量保证参考点  
标准报表标题的一些典型的例子是：  
+ YTD销售量vs销售代表的预测 
+ 按照服务计划的月客户流失率 5
+ 学校的5年辍学率时间序列
+ 产品和促销的直邮响应率
+ 按照网络、周几和时刻的用户计数和用户总数百分比
+ YTD索赔 vs 按照车辆类型的预测
+ 按照产品占全部产品销售的百分比的电话通话量 2

在第4章有关数据访问服务的那一节，我们讲述过标准报表的关键功能需求
### 11.3.3 分析性应用程序
分析性应用程序比标准报表更复杂。它们通常都是针对具体业务过程，封装了一定量的关于如何分析和解释该业务过程的领域专业知识。常见分析性引用程序包括：  
+ 促销效果
+ Web路径分析
+ 亲和计划分析 5
+ 货架空间规划
+ 欺诈探测
+ 销售人员管理
+ 类别管理 2

一个促销效果分析应用程序可能允许用户调查几个变量对信用卡促销响应率的影响。促销分析屏幕包括几个窗格，每个窗格都载有图表或报表  
主窗格显示总响应率，几个相关窗格显示与该活动的主要变量相关响应率，包括5个不同的活动区域、优惠（比如额外信誉加分，零展期率等），以及使用的渠道，如直接邮寄、电子邮件和Web等  
这类应用程序真正优势来自于用户与变量的交互以及寻找产生最大响应率组合的功能。促销分析筛选与其他应用程序组件（如活动计划）紧密相关 5  
注意：分析性应用程序的针对性越强，就越容易被作为一个独立的系统来实现。如果分析性应用程序有足够的业务价值，那么它就应该是优先级最高的DW/BI项目之一  
如果底层数据对于需要它的各个机构部门都有效，那么长期来看在支持分析性应用程序上花费的工作会产生更大的价值  
1. 预先构建好的分析性应用程序  
    从主流BI工具厂商以及有行业和功能经验的公司那里可以获取广泛的预先构建好的应用程序  
    预先构建好的分析性应用程序重点关注具体时机领域，能够封装可能需要在内部开发好几年的领域专业知识  
    这些预构建应用程序常常附带ETL组件和底层数据结构（通常为维度模型），这能使它们更容易安装和运行 5  
2. 读/写分析性应用程序  
    有一类独立分析性应用程序同时需要读写存取。这些应用包括规划、预算、预测以及假设分析建模  
    预测过程往往涉及多个版本的多次迭代，可以使用来自机构不同部分的输入。您也需要建立一个独立的事实表，通常具有比实际销售更粗的粒度。当预测存储到数据仓库后，可以进行版本比较  
    如果DW/BI团队最终拥有预测系统，将会又引入一层复杂性，预测系统本质上是一个事务处理过程，同时要求具备管理数据库升级、安全性以及事务监视等功能  
    大多数数据访问工具是设计用于读取，因此构建此类系统需要更传统的应用程序开发技能和工具  
    这些规划和预测应用程序提供了与实际情况做对比的目标，这可以作为评估性能的起点。如果不知道目标是什么，就无法知道是否已经达到了目标 5  

### 11.3.4 仪表板和记分卡
仪表板和记分卡应用程序涉及了一系列的报表界面和应用程序。横跨这些报表应用程序的公共主题是信息合并，异常突出，以及易于使用  
仪表板提供的界面工具支持状态报表，并在较高的层次上为多个数据源提供了警报，而且允许向下钻取更详细的数据  
警报产生于与实际数据的对比，比如将销售量和一个确定的数值或百分比进行对比，或者在许多情况下，也可以与一个独立的目标或目标数值进行对比。仪表板通常针对具体的用户类型，如销售经理  
图11-3提供了一个销售仪表板的例子。通常，使用火花线可以简单地挑出异常绩效，无论该绩效是好还是坏    
![avater 5](../syn_pic/data_warehouse_life_cycle/11-3.png)  
一个相关的被称为记分卡的应用程序是一种管理整个机构绩效的手段，这一类应用程序主要用于将公司目标和具体度量关联起来  
__平衡记分卡__  
记分卡方法的一个例子是平衡记分卡，由两个博士定义。在平衡记分卡方法中，管理层必须首先确定机构的目标和战略  
使用这些来定义度量，从四个视角来衡量机构的战略绩效：从内部业务过程、学习和成长、客户和财务  
这些措施在其他方法中也被称为关键绩效指示器（KPI)，然后通过它们跟踪和趋向的记分卡来使其有效  
无论如何，设计良好的仪表板或绩效记分卡程序都有巨大的价值，它对DW/BI程序还有几个重大的意义 5  
首先，这些应用程序几乎总是综合性的应用程序，因为在它们彻底实现之前，需要从几个业务过程中获取数据  
在多数关键业务过程维度模型就位之前，如果您试图实现仪表板和记分卡，那么它们通常都无法正常工作  
一类需要由仪表板和记分卡来满足的需求在20世纪80年代被称为经理信息系统（EIS）  
这些应用程序最近已经开始强劲复苏了，这很大程度上是因为许多大型机构的核心数据仓库已经到位，并构建了有效的标准报表集  
### 11.3.5 运营商业智能
对多数运营来说，运营系统提供了职员需要的信息以完成他们的基本工作职责。运营BI针对的是在事务层运行业务的那一类人 5  
例如，一个服务代表处理客户账单问题时可以用事务处理系统中的历史账单和支付数据来回答咨询  
然而，该代表也可以通过向客户提供“特殊优惠” 以升级其账户、增加新服务、提供下一订单免费送货或任何去其他吸引人的条件来潜在地增加收入并提高客户满意度  
底层规则引擎可能会把整个数据仓库中的信息都列入考虑范围。例如，具体服务或提供优惠可能基于客户以前有过类似采购行为并且表现出对此感兴趣的东西  
每次使用运营BI应用功能程序对局部的影响相对来说很小，也许就值几美元或更少。然而，如果广泛使用这样的系统方法，则总体影响可能时巨大的  
任何时候报表需求都包含对即时或接近即时数据的需要时，这就是一个暗示，即需求可能在事务处理系统环境下能够得到更好的服务 5  
部分困难来自于运营BI和运营系统间的界限迅速变得模糊的事实。不论哪种情况，服务层对DW/BI系统都意味着更多的用户、更短的数据装载周期，以及更好的系统可用性  
注意：与我们所提到的分析性应用程序一样，在没有在中间层花费多少精力的数据仓库之上也可能创建运营BI应用程序  
这种情况下，您的运营BI应用程序必须有具体的、范围较窄的、能直接从源系统查询的数据需求，而且不需要更多的清洗、集成或历史上下文，也不会让源系统的负担过重  
一定要确保您的运营BI应用程序不会承担越来越多类似ETL的任务，而且不会最终构建许多独立的数据存储来支持各个独立的运营BI应用程序  
### 11.3.6 数据挖掘
虽然数据挖掘一般不作为一个独立的BI应用程序，但数据挖掘模型往往被嵌入到BI应用程序中以提升其效率和准确性 5  
更进一步，数据挖掘成为多数DW/BI系统的重要的组成部分之一主要有三个理由  
首先，数据挖掘比以前更容易实现了。其次，当前这一代数据挖掘工具更容易使用，用户界面更友好，并且可以自动地完成许多数据挖掘算法所需的数据准备。第三，数据挖掘工作确有成效  
将数据挖掘模型包含到您的BI应用程序中作为数据准备过程的一部分，甚至作为事务处理系统的一部分有助于提升销售量、降低欺诈和客户流失，并且能够更准确地预测收入流的情况  
1. 数据挖掘技术概述  
    数据挖掘是试图发现对机构有用的模式和关系的一种数据探索过程。从业务的角度来看，数据挖掘可以帮助您理解和预测行为，识别关系或将各个项（比如客户或产品）归类并放入一致的集合  
    这些模型可以通过分析性应用程序或运营BI应用程序来访问，向用户提供指导、警示或标识 5  
    数据挖掘领域可以更宽泛地称为知识发现和数据挖掘，或更具体一些，可以称作数据库中的知识发现  
    您通过使用自动化工具掘取大量的数据来识别和发现可以在业务过程中利用的有价值的模式或关系  
    人类的大脑就是很好的数据挖掘工具的例子。在人类的一生中积累了大量的经验。某些情况下，人们会根据经验识别模式并生成模型用以预测未来  
    数据挖掘已经法则成熟并在许多业务中都起到了关键作用。大家每天都要精力好几十遍地数据挖掘——从邮箱内地垃圾邮件、到在杂货店使用地会员卡，再到检查每次信用卡支付地欺诈探测算法    
    对数据挖掘地全面讨论超出了本书的范围。数据挖掘已经衍生出五个主要的类别：聚类、分类、估算和预测、相关性分组、异常探测 5  
    1. **聚类**  
        聚类使一个纯粹的无向数据挖掘的例子，用户没有具体的议程，只是希望数据挖掘工具能够揭示一些有意义的结构  
        具体用于聚类的技术包括统计、基于记忆的推理、神经网络以及决策树等  
        **命名的力量**。当Claritas最初创建称之为PRIZM的客户划分系统时，他们能够使用聚类技术识别大约60种不同的客户分组  
        Claritas中有人意识到数值不是描述性的，而且也并没有导致好的营销效果。因此，他们为每个聚类都相处了一个灵巧的名称，每一个聚类一个速记方式以显示其唯一特征  
    2. **分类**  
        分类的一个例子是检查候选客户并且指派该客户到一个预先确定的聚类或分类，比如，用不感兴趣、漠不关心、一般、积极或渴望等离散值来指定每位客户的兴趣水平，这也可能是销售自动化应用程序的一部分功能 5  
        可用于分类的技术包括标准统计、基于记忆的推理、基因算法、链接分析、决策树和神经网络  
    3. **估计和预测**  
        估计和预测是两种类似的活动，通常都以数值型量度作为结果  
        用于估计和预测的具体技术包括用于数值型变量的标准统计和神经网络  
    4. **相关性分组**  
        相关性分组是一种特殊的聚类，用于识别同时发生的事件或事务  
        可以用于相关性分组的具体技术包括标准统计、基于记忆的推理、链接分析和特殊目的的市场分析工具 5  
    5. **异常探测** 
        一些业务过程依赖于识别那些明显偏离正常情况的实例。异常探测可以利用任何数据挖掘算法。聚类算法也可以创建出包含数据异常值的聚类，这种聚类有别于模型中的其他聚类  
2. 应用程序架构中的数据挖掘  
    数据挖掘，比如标准报表程序，通常是一个单独的系统附带着单独设计工具以便适应各种不同形式的统计分析  
    数据挖掘应用程序可能需要来自数据仓库的服务更像ETL服务，正如第9章讲述用于数据传播的子系统21时所提到的  
    从BI应用程序的角度看，数据挖掘工具本身被调用作为服务提供者。也就是说，它们可能被视为前端工具和数据库之间的一个应用层  
    DW/BI团队应对数据仓库和数据挖掘工具之间可能存在的不匹配非常敏感 5  
    这种不匹配的解决方法时像批处理我呢见那样运行横向钻取报表，然后把这些报表提交给数据挖掘团队进行分析，并且从服务器离线  
    注意：尽管数据挖掘工具正在不断地改善，成功的数据挖掘仍然不容易实现  
    如果您地机构很早就开始使用DW/BI系统，那么您的第一个BI应用程序不太可能包含数据挖掘  
    然而，如果您觉得要用到数据挖掘了，那么我们建议您阅读以下Michael Berry和Gordon Linoff的书，Data Mining Techniques for Marketing,Sales,and Customer Support 2nd Edition 5  

## 11.4 通过BI门户导航应用程序
如果对于多数业务用户社区来说，BI应用程序都是主要的访问工具，那么BI门户就是这些BI应用程序的主要访问界面  
构建BI门户的目标是给用户一个组织良好的、有用的、易于理解的场所来寻找需要的工具和信息  
注意：我们在此引用这些资料是因为门户的概念是完全独立于平台的，并且对所有的DW/BI系统都很关键  
由于企业用户的第一印象很重要，一个成功的BI门户网站必须满足以下要求：  
+ **可用的**。人们必须能够找到他们所需要的信息 5  
+ **内容丰富的**。门户应包括大量的内容而不仅仅是报表，它应该尽可能多地包含支持信息、文档、帮助、例子和意见  
+ **清爽**。门户应该布局良好，这样人们就不会被搞糊涂或者不知所措  
+ **当前的**。应有人专职负责保持内容是最新的，不能损坏链接，也不能将12个越之前的内容仍然标为“新”  
+ **交互**。门户应含有一些功能吸引用户并鼓动他们再次登录。一个好的查询工具、元数据浏览器和面向支持的新闻组都是用户与门户互动的方式  
+ **面向价值**。这是组织变更目标。您希望所有来过BI门户的人最后都认为DW/BI系统是有价值的资源，有些能够帮助他们更好地工作 5  

创建BI门户的过程需要仔细地权衡两个基本设计原则：密度和结构  
+ **密度**。我们地大脑可以处理不可思议地信息量。浏览器给了我们这样一个低分辨率的平台，我们必须尽可能认真而高效地使用它。每个像素都必须充分利用  
+ **结构**。虽然我们需要为门户主页填充信息，但如果在主页上填充了数以百计无序描述和链接地话，它一样也不能工作。您地大脑只在信息经过有序组织的情况下才能处理所有这些信息  

**参考** Edward Tufte的书提供了一个用于结构涉及和信息显示的一般性参考，可以参考Visual Explanations:Images and Quantities,Evidence and Narrative这本书  
### 11.4.1 考虑密度
有关密度的思想通过两种方式迁移到BI门户。首先，这意味着我们打破了信息的层次结构。浏览器中类别是以层次结构表示的 5  
将层次结构平面化意味着可以把尽可能多的信息方到顶层页面。以前按分层页面组织的信息现在通过缩进式类别列表和子类别标题的形式显示在一个页面上

把两个层次置入同一页面实际上给用户更多的信息，因为这两个层次有助于互相定义。例如，销售有助于把与销售相关的子类别组合到一起，但是同时，每个子类也有助于用户理解销售包括哪些活动

### 11.4.2 基于业务过程的导航结构
BI门户的首要职责是为BI应用程序提供导航框架。导航框架是组织BI应用程序的手段，这样人们就能找到他们所要的东西 5  
一般来说，组织门户的最好方式是使用机构的业务过程作为主轮廓  
参见图11-4中的一个业务过程的观点。“Standard Reports”下的列包含了机构的主要业务过程，正如数据仓库总线矩阵所描述的那样，该结构本质上遵循了机构的价值链  
也有些产生信息的内部业务过程对于整个机构都是有用的，比如来自人力资源部门的员工总数，或者来自财务部门的成本数据  
除了业务过程分类，BI门户也要有标准化的布局来使人们比较容易找到他们要寻找的东西  
### 11.4.3 附加门户功能
除了类别和报表清单，您还需要提供几项常见功能： 5  
+ **搜索**  
    如果业务过程类别不够全面，搜索工具就可以作为一个供选的报表定位器，好的搜索工具可以找到BI站点上的所有文档和页面，以及所有的报表元数据（包括报表名和报表描述）  
+ **元数据浏览器**  
    元数据浏览器可以简单到只是一个动态网页或者报表，只要允许用户浏览有关数据库、模式、数据表、列、业务规则、装载统计、报表使用以及报表内容的元数据描述即可  
+ **用户论坛**  
    在BI门户上建一个面向支持的论坛会很有意义。对用户来说，当他们需要的时候，这是一种很好的获取帮助的方式。也可以创建问题记录和问题解决方案作为今后的参考资料  
+ **个性化**  
    用户应该能够把报表或报表链接按照自己选择的任何参数设置保存到其个人网页上。个性化是对人们每天浏览门户的有力刺激  
+ **信息中心**  
    定期地出现一些新地内容有助于保持用户地兴趣。可以提供一份民意调查，也可以让人们注册有关工具的培训，或张贴通知告知用户即将举行的论坛会议 5  

您还需要提供一整套支持和管理BI门户内容的工具，其中包括在线培训/教程、帮助页、元数据浏览器、样例报表、数据欺诈表、帮助申请书和DW/BI团队地联系信息  
在图11-4中，这些信息位于右下角屏幕中最没有价值的区域  
创建和维护BI门户所需地工作远比多数DW/BI团队期望的要多  
### 11.4.4 应用程序界面供选方案
正如前面所讲述的，BI应用程序中基于门户的界面很流行，这是由于其具有丰富的工具集来管理用户的感受  
除了门户界面，供选的发布机制还包括： 5  
+ 直接基于工具的界面  
    您只需开发一套报表并通过标准报表访问工具界面直接将其提供给用户。这虽然不是很灵活，但是很迅速，而且当您试验其他选项时这也提供了一个良好的开端  
+ 自定义编码的界面  
    许多数据访问工具都提供API，并且很多情况下都有完整的软件开发包（SDKs)，这允许您开发自己的界面，然后调用核心工具功能并显示由此产生的报表元素  
    多数运营BI应用程序使用自定义编码构建，这使得它们更容易与运营系统集成。而且，这种界面也提供了对合适的报表组件的结构化导航  

最后，自定义编码主题的一个很有意义的话题时开发面向手持设备的应用程序的趋势 5  
## 11.5 小结
BI应用程序是一座桥梁，它为绝大多数机构利用DW/BI系统中的信息提供了便利。BI应用程序包含了一系列包括标准报表、分析性应用程序、仪表板和记分卡，以及运营商业智能  
几乎所有的BI应用程序都是通过BI门户发布给用户的。门户必须满足与BI应用程序同样严格的设计要求  
既然您已经对BI应用程序有了一定的了解，第12章将会讲述如何定义和构建您的BI应用程序

现在您必须清楚，BI应用程序时整个DW/BI系统的一个关键部分。您必须有一个可观的资源集专门用于生产和维护BI应用程序的各个方面  
每个团队都会遇到两个主要的资源问题：BI应用程序开发人员的角色时生命，以及由谁来充任这个角色  
### 12.1.1 BI应用程序开发人员的角色
显然，BI应用程序开发人员必须开发、维护、加强和扩展核心的BI应用程序集  
除了基本的开发角色，也要有人对其他报表或应用程序的业务需求做出响应  
### 12.1.2 谁来完成商业智能工作
我们推荐将商业智能开发人员中的一个中心小组作为整个DW/BI团队的一份子，而其他的商业智能开发人员则定位到业务上 5  
该中心团队是核心项目团队的一部分，参与过需求定义过程和维度模型设计  
我们也经常发现BI应用程序开发人员位于局部或部门层面上。我们也试图在BI应用程序开发过程开始的时候引入这些关键的超级用户  
虽然这样做并不一定能够让工作进展得更快，但总可以为这些人提供一次了解数据和工具的机会，以便使他们在本轮迭代部署的时候能在其业务领域变得卓有成效  
显然，构建和支持BI应用程序是一项巨大的任务。一个好的指导原则是：DW/BI系统中需要在面向用户的前台投入和后台一样多的人手  
除了应用程序开发任务，面向用户的团队也需要新的报表、增强、维护、编制文档、用户支持和业务元数据 5  
### 12.1.3 生命周期时间安排
本章反映了一个不连续的时间框架，因为BI应用程序过程发生在生命周期的不同时刻  
在需求完成后应立即指定报表模板。即使有详细的需求结果文档，仍然有很多与应用程序相关的信息保留在访谈参与人员的大脑里  
一旦捕捉到了规范，在后面的时间里您就不需要返回BI应用程序开发过程，直到项目生命周期较晚的阶段为止  
根据不同的时间选择，我们将本章划分为两个小节：规范和开发 5  
## 12.2 BI应用程序规范
BI应用程序规范包含以下几个任务：设计报表模板、识别和规范应用程序本身，以及设计导航框架和门户  
这项工作必须在需求定义过程之后立即进行，因为此时应用程序相关的需求还清晰地留在您的脑海中  
创建BI应用程序是真正的软件产品开发工作  
在深入规范的过程之前，花点时间寻找并且引入一些核心的关键业务用户，这些业务用户具有关于目标业务过程的领域专业知识并且理解那些分析时机  
应该让这些人参与到创建报表清单和为报表进行优先级排序的过程中来，明确规范单个应用程序并评估不同的导航策略  
BI应用程序规范可以分解为五个主要活动，我们将在下面的小节里分别讨论 5  
### 12.2.1 创建应用程序标准和模板
在开始创建单个报表规范之前要制定一系列命名标准和一个模板，该模板应当具有您的报表和应用程序的通用外观和效果  
1. 确定命名标准  
    对于DW/BI团队来说，花几个小时讨论各个数据元素和数据表的名称是经常要做的事，但大多数机构会完全忽视需要开发数据访问环境标准。  
    这些标准涵盖了命名、数据格式和公共字段等领域。他们对于帮助用户快速理解报表的本质尤其重要——包括内容、来源和时间选择。当您进入开发阶段后，要随时准备根据需要修改标准  
    在需要标准的领域中，命名报表可能是最明显的。您应该考虑把报表类型和数据类型置入报表名中  
    根据数据访问工具的不同，可能需要为多种对象命名，比如业务独立、过滤条件或约束条件，以及模板等。对数据访问环境内的每个对象都要制定一致的命名策略 5  
    提示：虽然完整的名称很好，但他们在狭窄的计算机屏幕或纸质报表上难以管理，所以缩略语会很有用。要和业务用户一起制定可接受的缩略语词汇表  
    建议您在元数据中同时保留包含短名和长名的字段。短名对于报表很有用，而长名在描述报表的内容和意义上很有用  
2. 创建应用程序模板  
    一旦您确定了命名标准，就可以把注意力转到创建一个BI应用程序物理布局的标准模板上  
    注意：我们在这个领域提出的建议与所有的工具和平台都相关  
    如果使用一致的风格进行表现，那么人们就能更快地找到信息，因此创建模板是很重要的 5  
    BI应用程序团队在发布业务中的工作是富有成效的。您必须有自己的格式和内容标准，并且在使用它们时保持一致。在门户层和独立地文档层上您需要制定标准  
    在单个报表的层面上，您应该创建一个模板来识别将在所有的报表里显示的包括其未知和风格的标准元素  
    在开始规范单独的应用程序之前就定义好标准模板是很有帮助的，因为模板会为您提供一个对所有应用程序都通用的结构和上下文  
    大多数情况下，所有来自于DW/BI系统的商业智能报表或应用程序都需要定义以下的标准元素：  
    
# 编辑点

    + 报表名，建立明确的描述性的名称，商业门户上显示的报表应创建一个较短的名称  
    + 报表标题，包括标题以及子标题，所有参数选择和过滤条件都应在标题、行列头或报表头尾显示出来 5
    + 报表主体，数据行/列布局，包括以下选项：
        + 数据对齐：数值右对齐，行首左对齐，列首右对齐或居中
        + 显示精度：根据数据不同，确定所有数值字段的精度
        + 行列标题格式
        + 背景填充和背景颜色
        + 为完全或部分断开的行编排格式 5
    + 页面和页脚，以下各项都在页眉或页脚里出现：
        + 报表名
        + 使用的参数和过滤条件，用户必须理解报表内容并了解它如何定义
        + 导航目录
        + 报表就任何不寻常的定义或内容进行的注释
        + 页编号
        + 报表运行时间和日期 5
        + 数据源，即报表数据来自于哪个维度模型，是关系数据库还是OLAP
        + 保密规定
        + DW/BI引用：名称和商标
    + 报表文件，基于标准文件命名约定

### 12.2.2 确定初始应用程序集

1. 识别候选报表  
    经过整理需求、对访谈的总结、和业务用户一起脑暴，可以识别出一系列潜在报表

    一旦有了经过批准的候选报表列表，要让其他参与设计会议的用户来审查它  
2. 合并候选列表  
    当列表完成后要合并报表，从业务角度出发会使分类变得明晰  
3. 确定报表列表的优先级  
    确定优先级需要考虑业务价值、数据可用性、开发的困难程度、使用范围和对用户的重要性  
    初始报表是要用有限的资源向尽可能多的用户提交最大的价值

    + 很早时就预料新报表和旧报表不一致，创建描述其不同指出的详细文档  
    + 与其重建报表集，不如考虑对它们从数据仓库中抽取的数据进行重定向 
    + 如果您需要将报表转换到DW/BI，让关键业务用户参与到您的数据验证阶段，然后停止老系统
    + 确定您有足够的资源同时处理转换现有报表和构建新的报表  

### 12.2.3 制定详细的应用程序规范
这个过程必须始终让业务用户参与进来 5  
规范过程的第一个目的是捕捉足够的信息以便您可以如实地创建应用程序  
规范过程的第二个目的是确保DW/BI系统能为业务社区提供一些有价值的信息  
规范单个报表/应用程序的方法是将目标报表列表分割开来，让每个人通过下面步骤完成自己那部分目标列表子集

+ 实体模型 是报表的主要通信工具、物理示例，包含所有可视组件
+ 用户交互 列出所有用户与报表交互的节点，包括变量、参数、列表、限制条件、向下钻取路径和链接 5
+ 数据集 描述填充报表各个小节和组件的数据集，包括报表主体和支持用户交互的选择列表，标注数据仓库外部的数据源

## 12.3 BI应用程序开发
应用程序开发过程是一个标准地软件开发流程：准备、构建、测试和部署
### 12.3.1 准备应用程序开发
审查并更新您之前在项目里制定地BI应用程序规范，另外需要安装和测试BI工具并设置用户安全性、使用情况和性能跟踪  
1. 安装和测试BI工具  
    大型BI工具有许多组件，一但组件到位，要定义元数据来访问数据库
    1. 将元数据填入BI工具  元数据定义的丰富程度和复杂性是高度依赖于工具的
    2. 创建业务元数据   
        业务元数据是DW/BI系统中最重要的元数据，确定您填入的所有业务元数据元素对BI工具都是可用的 5  
        整个报表导航框架应构建为元数据，报表能简单地属于多个分类，改变导航结构也可以通过简单地编辑元数据实现
    3. 测试BI工具  
        您应该试用工具全部功能，作为初始测试的一部分  
        调试BI工具的元数据定义需要对工具的元数据有深入的理解
2. 验证您的BI应用程序界面策略   使用合适的BI工具来制定访问方式的最后决策  
3. 设置用户安全性  包括限制报表用的浏览和执行 5  
4. 建立报表过程元数据系统  
    过程元数据是用于跟踪一个给定活动持续运行情况的数据，在开发过程的早期设立这样的日志系统是比较好的做法  
    如果BI工具不提供，就应该想其他办法，应有汇总层的表记录用户的日常访问的度量  
    最好创建一个针对报表过程元数据运行的BI应用程序集，在问题变得严重之前就识别出它们

### 12.3.2 构建应用程序
建议在前期规划上多花时间，以后可能会您省下几天甚至几个星期的工作量  
构建BI应用程序的基本步骤：  
1. 沿着核心过程前进  
    大多数工具开发BI应用程序都包括：5
    1. 定义用户交互 
    2. 创建查询
    3. 对结果进行格式编排
    
   应用程序的性质、使用的工具和选取的应用程序方法决定了要在这三个方面做多少工作  
   1. 定义用户交互  
       多数BI应用程序都允许用户指定参数并且查看结果  
       审查应用程序规范并找出用户交互，确定最适合的交互工具  
       您也需要想办法向用户展现这些选项  
       好的BI应用程序要使用户界面易于使用，点击鼠标次数少于三次 5  
   2. 创建查询  
       主要查询由报表设计规范中的各种参数组成，各个参数通常都由对应的能够产生相关选择列表的查询  
       可能需要做一些横向钻取查询，或者自己构建查询  
       在构建分析性应用程序或运营BI应用程序时可以编写并测试一些自己的查询  
   3. 结果的格式编排  
       要保证报表打印出来也很好看，当导出EXCEL、PDF文件或任何其他对您的组织很重要的目标格式时，要检查是否进行了恰当的格式编排  
       注意您必须至少有足够用于开发BI应用程序的数据，哪怕只是测试数据 5
2. 设计和开发分析性应用程序  
    分析性应用程序一般以具体业务过程和封装领域专业知识为典型目标  
    任何分析性应用程序都是一个程序，该程序能够绘出用户界面（UI)、管理用户交互、通过API从BI工具中显示报表和图形组件  
    根据应用程序复杂性和系统环境的不同，使用更标准的编程环境开发您的应用程序内核可能会更有意义  
    分析性应用程序经常嵌入到一个被设计用于具体业务过程的综合产品中  
3. 设计和开发运营BI应用程序  
    通常是自定义开发的，用于查询数据仓库和运营系统，使用户能够在BI应用程序内调用事务处理过程 5  
    运营BI通常所需的服务级别比较高，需要提高DW/BI系统的性能服务级别  
    这些应用程序往往需要数据接近实时  
    实时层和运营系统监视都是专门设计用于满足这些运营BI应用程序需求的组件  
4. 包含数据挖掘模型  
    通常采取的形式是带有具体参数调用预定义数据挖掘模型  
    创建用于分析性应用程序底层数据挖掘模型通常是独立的活动，是一个过程 5   
    ![12-6](12-6.png)  
    运营BI系统也常常嵌入数据挖掘模型作为其规则引擎的一部分，找出事务处理流程中的时机和问题  
5. 验证数据和数据模型   要在项目规划中为查出这些问题并采取争取的行动留足时间
6. 创建导航BI门户  
    门户有多个层次，顶层主要是一个导航工具，用于组织标准BI应用程序和仪表板，最清晰的方式是在业务过程之后跟一个类别  
    第二个层次的门户通常关注基于所选业务过程数据的报表，可以在BI门户的多个区域交叉链接一些报表和仪表板 5  
    ![12-7](12-7.png)
7. 安排报表调度  
    每当ETL过程结束时应当立即运行和存储完成的报表  
    多数BI工具提供了根据预定义参数调度标准报表的服务  
    报表分发和管理依赖于数据访问工具，访问工具支持报表缓存  

### 12.3.3 应用程序和数据的测试和验证
测试时必须步骤：针对每个报表检查您的逻辑，创建简单的查询，并把报表数据复制到某个电子表格并重新进行手工计算 5  
根据正确的数值检查百分比和差值，改变参数并重新运行  
一旦确信报表已经正确运行就需要看看数据本身了，下面时一些很可能会遇到的问题：  
+ 无意义的描述  源系统提供的名称和描述可能需要修改 
+ 重复的维度信息  主要源数据中两个不同键可能会有相同的名称
+ 错误的维度关系  在钻取的时候进行检查以确保引入的是正确信息 5
+ 数据不平衡  不能保证数据和另一个已有系统的平衡，可能是因为BI应用程序计算业务度量的方式不同，也可能因为遗留的系统里使用不同业务规则创建事实
+ 性能调整  复杂报表和分析中，初始集无法很好的运行是常见情况  

### 12.3.4 完成文档
除了业务级的文档应当已经安全地存储在源数据目录中，为应用程序模板基础架构信息（命名标准、主模板的位置信息）编制文档也很重要
### 12.3.5 部署计划
BI应用部署不像数据库或事务处理那么严格，它没有那么高的风险 5

## 12.7 保证质量
测试是主要的质量保证工具：
+ 测试计算——加起来做除法运算，计算比率，确保报表本身内部一致性和正确性
+ 在层次结构的不同层次进行测试，那些汇总层加起来正确的结果，向下钻取细节时就不一定正确了
+ 测试边缘参数，比较年份之间的数据时，如果选取最早的年份，没有上一年，那么测试一下会产生什么结果 
+ 与已有的数据源进行比较，如果结果应当匹配，要确保它们是匹配的
