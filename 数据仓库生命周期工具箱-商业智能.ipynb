## 11.1 商业智能应用程序的重要性
在您的结构里，大部分知识工作者并不会建立自己的报表，所以必须向这些人提供恰当的方式以便于他们访问数据仓库中的信息，而这正是BI应用程序的基本任务  
思考一下这句话的含义：对于多数业务用户来说，BI应用程序是他们访问DW/BI系统的唯一方式 5  
最终标准是，BI应用程序必须要做得非常好，他们必须满足下列设计需求：  
+ **正确**。与其他系统的不同之处在于，BI应用程序必须提供准确的审计报表和文档  
+ **性能良好**。平均响应时间不到5秒钟是一个比较好的目标  
+ **易于使用**。那些完全能够进行在线订书的用户应该至少可以使用标准报表，一个不错的目标是，定制后的结果应当不超过10次鼠标点击就能获取  
+ **外观漂亮**。工具和报表应看起来干净利落并且富有吸引力 5  
+ **是一项长期投资**。应用程序必须适当地进行存档、维护、增强和扩展

在满足机构的一些了数据访问需求的过程中，BI应用程序填补了重要的空白。图11-1显示了从战略报表需求到运营报表需求的一系列需求  
![11-1](../syn_pic/data_warehouse_life_cycle/11-1.png)  
在最极端的信息访问情形中，一些用户需要从零开始构建对数据仓库的查询，这就是所谓的即席访问  
这些业务分析员或超级用户通常负责创建复杂分析，寻找比较紧迫的问题或时机。通常运营系统中所有事务处理都会利用运营BI，或者说运营BI会为大量的用户所使用，例如呼叫中心的所有代理人 5  
这两个极端之间的差异非常大。根据我们的经验，DW/BI系统用户中有很大的比例位于中端，或者位于范围的低端，其百分比可能高达80%-90%。这些用户需要预先建立的、已参数化了的报表  
注意：对于很多业务内行来说，即席仅仅是通过改变报表上的参数来创建个性化报表。对大多数业务用户来说，BI应用程序的标准报表集是主要的访问工具  
在成熟的DW/BI系统中，通常有几十个甚至数百个BI应用程序，用户需要的东西太多了，仅凭一个简单的列表是无法满足的。这需要您提供一个高级界面，也成为BI门户，它为应用程序提供了合理的顺序  
大多数BI应用程序都是通过BI门户访问的，至少刚开始时需要通过这种方式，一旦用户找到感兴趣的报表，他们很可能直接通过工具平台或者托管的应用程序来访问具体内容  
在图11-1中，只按照一个消费模式对用户进行归类比较困难。多数即席用户也经常使用标准报表和其他分析性应用程序。您需要提供全方位的应用程序类型，同时用户会被那些能够帮助他们解决具体问题的应用程序所吸引 5  
BI应用程序需求背后的巨大推动力是人们在技术功能的深度和质量方面的需求存在很大差别。因为数据仓库需要满足各种层次的技术技能和各种程度的分析经验  








## 11.2 商业智能分析周期
分析周期把业务分析过程分为5个不同的阶段，如图所示
![BI分析的分析周期](11-2.png)
### 11.2.1 第一阶段：监视活动
标准报表是分析周期的起点，用户使用这些报表来鉴定业务状态的绩效，如仪表板、门户、记分卡 5  
### 11.2.2  第二阶段：识别异常
这一阶段主要识别相对于正常性能来说出现的异常情况和时机，识别异常的阶段可以帮助用户跳过海量数据，使他们集中经理关注最值得关注的时机  
识别异常阶段需要额外的功能，如用于向用户设备发布警报的分发服务器，帮助用户以多种创造性方式来查看数据的可视化工具  
### 11.2.3 第三阶段：确定构成原因的因素
这一阶段的关键在于识别驱动异常性能变量之间的可靠关系和相互作用，需要系统架构还含有统计工具或数据挖掘算法（关联、排序、分类、分割）
### 11.2.4 第四阶段：模型供选方案
这一阶段，您在已有因果关系的基础上设计模型以评价决策供选方案  
目标使具有对一系列潜在决策进行假设分析和仿真的功能 5  
数据仓库架构需要采用包括用于模型评估的统计工具和数据挖掘算法，如敏感性分析、蒙特卡洛仿真、单变量求解优化  
### 11.2.5 第五阶段：采取行动并跟踪结果
可以在BI应用程序中使用闭环过程并向运营系统反馈建议的行动  
需要通过API从数据仓库访问运营系统，增强已有的维度模型或者构建性能管理来跟踪特定业务决策的结果  
应该捕捉并分析行动结果，回到了监视活动阶段，形成循环过程  
### 11.2.6 分析周期的更多意义
分析周期帮助我们了解业务用户如何使用DW/BI系统 5  
分析周期的处理必须明确地应用于您构建地所有分析性应用程序，通过业务形式地分析主动指导用户做出决策  
分析周期的目标是：
+ 主动指导业务用户超出基本报表的范围
+ 识别和了解异常的绩效情况
+ 为每个异常绩效情况捕获制定决策的“最优方法”
+ 在机构内部共享产生的“最优方法”或者知识资本 5

理解分析周期有助于确保为每个在业务需求定义过程中识别出的分析性时机都提供完整的解决方案

## 11.3 商业智能应用程序的类型
+ 直接访问查询和报表工具允许用户直接查询维度模型并定义结果集
+ 数据挖掘产生的模型通常潜入到其他的BI应用程序中  
+ 标准报表是预定义的，预先编排好格式的报表，提供某个层次的交互，如输入参数或向下钻取
+ 分析性应用程序是一个嵌入领域专业知识的报表集合，常常使用数据挖掘模型 5  
+ 仪表板和记分卡一般是由报表和图表组合而成，使用突出异常和向下钻取功能对多个业务过程的数据进行分析
+ 运营BI和闭环应用程序跨越数仓多个业务过程来指导决策，常带有回访源系统的交互界面

### 11.3.1 直接访问查询和报表工具
通常是通过一个元数据层，提供额外的列的子分组、业务描述以及联接的定义，用户能将列拖拽到报表区，设置约束定义计算并添加格式  
即席工具复杂的原因是业务必须克服将业务问题转换成查询语言语法形式时所面临的根本性困难  
即席查询工具应该包含四个方面内容：构建查询、分析和表示、用户体验以及技术特性 5
1. 查询构建  
    您可能需要的查询构建功能包括：
    + 多通道或多重集查询  
        查询工具必须将报表分解为几个简单查询交给RDBMS来处理  
        多通道查询也允许横向钻取几个不同业务过程一致性维度模型中的事实表  
        OLAP引擎使用的分析性语言能够只用一条语句就完成和多通道查询等价的工作
    + 警报  
        查询工具需要帮助用户识别那些与众不同的记录，设定警报阈值  
    + 逐次约束  
        某个查询结果被用作子查询的限制或者过滤条件 5  
        这些约束列表可能过于庞大需要传回数据库  
        还可以通过创建临时表或在数据库里定义的视图保存研究分组参与者列表 
    + 半可加性求和  
        需要自定义函数可以自动按相关查询时间约束来计算平均值  
    + ANSI SQL 99的支持  
        SQL 99对SQL添加了基本的OLAP功能，允许您对一个定义好的查询子集进行聚集  
    + 直接查询字符串记录  
        作为最后一种手段，可能需要查看并更改由工具生成的SQL语句或OLAP语句 5  
2. 分析和呈现功能  
    该工具必须支持操纵数据的业务需求并使用观感较佳的格式对其进行呈现  
    + 结果集的基本计算   包括一系列数学函数、统计函数、字符串函数、顺序处理函数、条件函数和报表函数，用于克服工具中的其他缺陷
    + 旋转结果   旋转是多维度或交叉报表分析的基础
    + 向下钻取   这项功能允许用户通过包含额外的列向结果集添加更多的细节
    + 有关旋转结果的列计算   这些计算创建一个计算列，是两个或多个旋转列的函数，如环比、列的百分比 5  
    + 列与行计算   份额计算和比率，如占比
    + 排序   排序对于不显示出来的元素尤其重要
    + 复杂格式编排   格式编排往往非常重要
    + 图表和图形  图表和图形是进行分析的基本要素
    + 复合文档  仪表板界面实际上就是一个由几个单独的报表、图形及着色工具组成的复合文档 5  
    + 用户可更改变量  用户可更改变量或参数可以位于查询文档中从查询过滤条件到报表首部的任何位置  
3. 用户体验  
    如果用户不会使用或者不喜欢使用，该工具等于没用，以下功能有助于改进用户对分析环境的体验：
    + 易用性 工具应由自然而直观的界面，必须让用户参与到对这一问题的评估中来  
    + 元数据访问 工具应该向用户提供上下文敏感的帮助，要对数据和工具本身做出说明  
    + 选择列表 应提供一种方法来查找能在查询里用作约束或过滤条件的值列表 5  
    + 与其他应用程序的无缝集成  集成也包括使报表在其他应用程序中可用，使BI开发人员将报表嵌入到所有应用程序中去的API
    + 导出多种文件类型  
    + 嵌入式查询  用户应当能够从其他应用程序发出查询，实现可交互查询
4. 技术特性  
    多任务查询和取消查询这样的功能都是体现工具可用性的基本功能，如果缺失，用户可能会觉得不满：
    + 多任务 用户必须能够同时允许其他程序，也可以建立和运行其他查询 5  
    + 取消查询 用户应该能种植一个由他们启动的正在运行的某个查询而不会终止所有查询
    + 脚本  一个脚本语言界面和命令行界面对于报表的自动执行很重要
    + 连接性  弄清楚您是否可以进入所有想要进入的数据库平台，包括其他数据源
    + 调度  工具需要提供或利用某种形式的调度系统
    + 元数据驱动  管理员应该能定义简单的数据仓库子集 5  
    + 软件管理  这是一个随着采用WEB作为应用程序平台而逐步消失的问题
    + 安全性  该工具应参与到所有可用的用户验证系统中  
    + 查询  对数据库的直接查询应当不需要管理层的支持  

### 11.3.2 标准报表
标准报表是BI应用程序的基础部分，是参数驱动的、由预定义的、预格式编排的输出  
这对按钮用户的价值是明确的：标准报表提供了一种易于使用的方式，只需要短期学习就能够获取所需信息5  
这些报表是企业度量的官方来源，并为这些度量提供了一个质量保证参考点  
### 11.3.3 分析性应用程序
通常都针对具体业务过程，设计用于解决具体问题，常见分析性引用程序包括：
+ 促销效果
+ Web路径分析
+ 相关性分析
+ 货架空间规划
+ 欺诈探测
+ 销售人员管理
+ 类别管理

一个促销效果分析屏幕包括几个窗格，主窗格显示总响应率，相关窗格显示与该活动的主要变量相关响应率 5   
这类应用程序真正有时来自于用户与变量的交互以及寻找产生最大响应率组合的功能   
1. 预先构建好的分析性应用程序  
    预先构建好的分析性应用程序重点关注具体时机领域，能够封装可能需要在内部开发好几年的领域专业知识  
    这些预构建应用程序常常附带ETL组件和底层数据结构（通常为维度模型）
2. 读/写分析性应用程序  
    有一类独立分析性应用程序同时需要读写存取，包括规划、预算、预测以及假设分析建模  
    预测过程往往涉及多个版本的多次迭代，可以使用来自机构不同部分的输入 5  
    DW/BI引入预测系统会增加复杂性，因为预测系统本质上是一个事务处理过程  
    大多数数据访问工具是设计用于读取，构建此类系统需要更传统的应用程序开发技能和工具  
    规划和预测应用程序提供了与实际情况做对比的目标，这可以作为评估性能的起点  

### 11.3.4 仪表板和记分卡
横跨这些报表应用程序的公共主题是信息合并，异常突出，以及易于使用  
仪表板允许下钻，并提供了警报，通常针对具体的用户类型 5  
销售仪表板的例子，使用火花线可以简单地挑出异常绩效  
![avater](11-3.png)
记分卡是一种管理机构绩效的手段，主要用于将公司目标和具体度量关联起来  
__平衡记分卡__  
管理层首先确定机构的目标和战略，使用这些来定义度量  
从内部业务过程、学习和成长、客户和财务来衡量机构绩效 5  
设计良好的仪表板或绩效记分卡程序，通常需要在关键业务过程维度模型就位之后  
经理信息系统通常需要仪表板和记分卡来满足
### 11.3.5 运营商业智能
运营BI针对的是在事务层运行业务的那一类人  
底层规则引擎可能会把整个数据仓库中的信息都列入考虑范围  
广泛使用运营BI应用程序对总体的影响可能是巨大的 5  
任何时候报表需求都包含对即时或接近即时数据的需要时，暗示可能在事务处理系统环境下能够得到更好的服务  
当运营BI和运营兄台你过的界限变得模糊时，意味着更多的用户、更多的数据装载周期以及更好的系统可用性  
一定要确保您的运营BI应用程序不会承担越来越多类似ETL的任务
### 11.3.6 数据挖掘
数据挖掘称为多数DW/BI系统的重要的组成部分之一主要有三个理由：
+ 数据挖掘比以前更容易实现了
+ 当前这一代数据挖掘工具更容易使用 5 
+ 数据挖掘工作确实有成效

将数据挖掘模型包含到您的BI应用程序中作为数据准备过程的一部分  
1. 数据挖掘技术概述  
    数据挖掘是视图发现对机构有用的模式和关系的一种数据探索过程  
    数据挖掘可以帮助您理解和预测行为，向用户提供指导、警示或标识  
    数据挖掘通过使用自动化工具掘取大量的数据来识别和发现可以在业务过程中利用的有价值的模式或关系 5   
    人类大脑根据经验识别模式生成模型用以预测未来  
    数据挖掘在识别垃圾邮件、信用卡欺诈探测、交叉销售等业务中起到了关键作用  
    数据挖掘主要有五个类别：聚类、分类、估算和预测、相关性分组、异常探测  
    1. 聚类  统计、基于记忆的推理、神经网络、决策树
    2. 分类  标准统计、基于记忆的推理、基因算法、链接分析、决策树和神经网络 5
    3. 估计和预测  具体技术包括用于数值型变量的标准统计和神经网络  
    4. 相关性分组  标准统计、基于记忆的推理、链接分析和特殊目的的市场分析工具
    5. 异常探测 聚类算法
2. 应用程序架构中的数据挖掘  
    数据挖掘通常是一个单独的系统附带着单独设计工具以便适应各种不同形式的统计分析，更像ETL服务  
    数据挖掘工具本身被调用作为服务提供者，可能被视为前端工具和数据库之间的一个应用层 5  
    应对数据仓库和数据挖掘工具之间存在的不匹配非常敏感  
    成功的数据挖掘仍然不容易实现
    

## 11.4 通过BI门户导航应用程序
构建BI门户的目标是给用户一个组织良好的、有用的、易于理解的场所来寻找需要的工具和信息  
一个成功的BI门户网站必须满足以下要求：
+ 可用的，人们必须能够找到他们所需要的信息
+ 内容丰富的，门户应包括大量的内容而不仅仅是报表，包括支持信息、文档、帮助、例子和意见
+ 清爽，门户应该布局良好
+ 当前的，应有人转置保持内容更新，不能损坏链接 5
+ 交互，门户应含有一些功能吸引用户并鼓动他们再次登录
+ 面向价值，希望所有来过BI门户的人最后都认为DW/BI系统是有价值的资源

创建BI门户的过程需要仔细权衡两个基本设计原则：密度和结构
+ 密度，浏览器这样一个低分辨率的平台，必须尽可能认真而高效地使用它
+ 结构，大脑只在信息经过有序组织的情况下才能处理所有这些信息 5

### 11.4.1 考虑密度
可以通过将层次结构平面化来把尽可能多的信息放到顶层页面  
把两个层次置入同一页面实际上给用户更多的信息

### 11.4.2 基于业务过程的导航结构
BI门户的首要职责是为BI应用程序提供导航框架  
组织门户的最好方式是使用机构的业务过程作为主轮廓 5  
BI门户也要有标准化的布局来使人们比较容易找到他们要寻找的东西  

### 11.4.3 附加门户功能
除了类别和报表清单，还需要提供几项常见功能：
+ 搜索 搜索工具可以作为一个供选的报表定位器
+ 元数据浏览器 允许用户浏览有关数据库、模式、数据表、列、业务规则、装载统计、报表使用以及报表内容的元数据描述
+ 用户论坛 在BI门户上建一个面向支持的论坛会很有意义
+ 个性化  用户应该能够把报表或报表链接按照自己选择的任何参数保存到个人网页上 5
+ 信息中心  定期地出现一些新地内容有助于保持用户地兴趣，可以提供一份民意调查

还需要提供包括在线培训/教程、帮助页、元数据浏览器、样例报表、数据欺诈表、帮助申请书和DW/BI团队地联系信息  
创建和维护BI门户所需地工作远比多数DW/BI团队期望的要多  
### 11.4.4 应用程序界面供选方案
除了门户界面，供选的发布机制还包括：
+ 直接基于工具的界面
+ 自定义编码的界面 5

## 11.5 小结
BI应用程序包含了一系列包括标准报表、分析性应用程序、仪表板和记分卡，以及运营商业智能  
BI引用程序都使通过BI门户发布给用户的

# 12 设计和开发商业智能应用程序
设计BI应用程序应在需求定义步骤结束和高层管理人员就最高优先级达成一致之后立即启动  
路线开始于一个详细的规范，包括设计一个报表模板、识别和描述应用程序本身，以及设计一个导航框架  
可以取出规范文档并且开始按照开发步骤进行工作，创建和测试目标BI应用程序  
在部署BI应用程序之前，必须仔细测试其准确性和性能，需要指定一些人员负责维护和增强BI应用程序，这项工作会越来越繁重  
本章将学会：
+ 设计和开发BI应用程序的结构化方法 5
+ 用于规范、构建和部署应用程序的具体提示和技巧
+ 关于创建更高级的分析性应用程序的补充指导，包括数据挖掘模型和运营BI应用程序

## 12.1 商业智能应用程序资源规划
必须有一个可观的资源集中专门用于生产和维护BI应用程序的各个方面  
### 12.1.1 BI应用程序开发人员的角色
BI应用程序开发人员必须开发、维护、加强和扩展核心的BI应用程序集  
除了基本的开发角色，也要有人对其他报表或应用程序的业务需求做出响应 5
### 12.1.2 谁来完成商业智能工作
商业智能开发人员需参与需求定义过程和维度模型设计  
业务部门分析人员页开始参与到BI应用程序开发，有助于使他们在其业务领域变得卓有成效  
DW/BI系统中需要在面向用户的前台投入和后台一样多的人手，并提供新的报表、增强、维护、编制文档、用户支持和业务元数据
### 12.1.3 生命周期时间安排
BI应用程序过程发生在生命周期的不同时刻，即使有详细的文档，仍然有很多与应用程序相关的信息保留在脑中  
一旦建立了规范，在后面的时间里就不需要返回BI应用程序开发过程，直到项目生命周期较晚的阶段为止 5
## 12.2 BI应用程序规范
包含设计报表模板、识别和规范应用程序本身、设计导航框架和门户  
需要让业务用户参与到创建报表清单和为报表进行优先级排序的过程中来，明确规范单个应用程序并评估不同的导航策略  
### 12.2.1 创建应用程序标准和模板
1. 确定命名标准  
    开发数据访问环境标准涵盖了内容、来源和时间选择，当进入开发阶段，随时准备根据需要修改标准  
    考虑把报表类型和数据类型至于报表名中，对数据访问环境内的每个对象都要制定一致的命名策略  
    在狭窄的计算机屏幕或纸质报表上缩略语会很有用，可以和业务用户一起制定可接受的缩略语词汇表 5  
2. 创建应用程序模板  
    使用一致的风格，人们就能更快地找到信息  
    在门户层和独立地文档层上您需要制定标准  
    在开始规范单独的应用程序之前就定义好标准目标是很有帮助的  
    大多数情况下，都需要定义以下的标准元素：
    + 报表名，建立明确的描述性的名称，商业门户上显示的报表应创建一个较短的名称  
    + 报表标题，包括标题以及子标题，所有参数选择和过滤条件都应在标题、行列头或报表头尾显示出来 5
    + 报表主体，数据行/列布局，包括以下选项：
        + 数据对齐：数值右对齐，行首左对齐，列首右对齐或居中
        + 显示精度：根据数据不同，确定所有数值字段的精度
        + 行列标题格式
        + 背景填充和背景颜色
        + 为完全或部分断开的行编排格式 5
    + 页面和页脚，以下各项都在页眉或页脚里出现：
        + 报表名
        + 使用的参数和过滤条件，用户必须理解报表内容并了解它如何定义
        + 导航目录
        + 报表就任何不寻常的定义或内容进行的注释
        + 页编号
        + 报表运行时间和日期 5
        + 数据源，即报表数据来自于哪个维度模型，是关系数据库还是OLAP
        + 保密规定
        + DW/BI引用：名称和商标
    + 报表文件，基于标准文件命名约定

    标准模板例子
    ![12-1](12-1.png) 5

   您也需要为每个报表创建以下信息（不显示在报表上）：
    + 用户变量以及和其他用户交互，比如向下钻取路径和链接
    + 报表元数据，包括描述、估计、导出、作者和创建日期
    + 安全性需求，包括那些可以查询报表的安全分组的一个列表或描述
    + 执行周期，报表是否按一定的周期自动运行
    + 执行触发时间，报表是否针对某个特定系统事件执行 5
    + 发布机制，比如电子邮件或网站
    + 发布列表，如电子邮件分发列表
    + 标准输出格式，比如文本格式、HTML、EXCEL
    + 页面方向、尺寸和空白设置

   这些信息是必须的，应该存放到BI工具可访问的元数据知识库里供创建报表时使用，可以用电子表格记录 5  
3. 创建仪表板和分析性应用程序模板  
    仪表板和分析性应用程序标准模板包含与12-1仙童的基础元数，也需要考虑子组件的图形布局，通常由报表、图表和具体的仪表板控件组成  
    仪表板的视觉效果可以让那些具有图形设计和用户界面设计经验的人员来参与

### 12.2.2 确定初始应用程序集
BI规范过程的下一步是制定报表的目标列表，这个过程包含三个任务：
1. 识别候选报表  
    经过整理需求、对访谈的总结、和业务用户一起脑暴，可以识别出一系列潜在报表  
    ![12-2](12-2.png) 5  
    一旦有了经过批准的候选报表列表，要让其他参与设计会议的用户来审查它  
2. 合并候选列表  
    当列表完成后要合并报表，从业务角度出发会使分类变得明晰  
3. 确定报表列表的优先级  
    确定优先级需要考虑业务价值、数据可用性、开发的困难程度、使用范围和对用户的重要性  
    初始报表是要用有限的资源向尽可能多的用户提交最大的价值  
    如果无法对报表进行排序，至少要分成三组：A(高价值容易做的)、B（高价值不容易做的）、C 5   
    如何处理已有报表：  
    + 很早时就预料新报表和旧报表不一致，创建描述其不同指出的详细文档  
    + 与其重建报表集，不如考虑对它们从数据仓库中抽取的数据进行重定向 
    + 如果您需要将报表转换到DW/BI，让关键业务用户参与到您的数据验证阶段，然后停止老系统
    + 确定您有足够的资源同时处理转换现有报表和构建新的报表  

### 12.2.3 制定详细的应用程序规范
这个过程必须始终让业务用户参与进来 5  
规范过程的第一个目的是捕捉足够的信息以便您可以如实地创建应用程序  
规范过程的第二个目的是确保DW/BI系统能为业务社区提供一些有价值的信息  
规范单个报表/应用程序的方法是将目标报表列表分割开来，让每个人通过下面步骤完成自己那部分目标列表子集  
每个报表/应用程序都应具有以下文档集：
+ 实体模型 是报表的主要通信工具、物理示例，包含所有可视组件
+ 用户交互 列出所有用户与报表交互的节点，包括变量、参数、列表、限制条件、向下钻取路径和链接 5
+ 数据集 描述填充报表各个小节和组件的数据集，包括报表主体和支持用户交互的选择列表，标注数据仓库外部的数据源  
+ 算法，高级计算和业务规则  包括对任何应用程序所需的高级分析或数据挖掘模型的描述
+ 与其他报表/系统的交互 如果一个报表链接到另一个，就描述这些交互并把链接吸入API文档

如果您设计一个仪表板，就把界面分解成若干组件，每个组件对应一部分，使用以上步骤编制各部分文档，将整体实体模型和仪表板自身文档包含到组件部分的文档中  
1. 指定应用程序内容  
    每个标准报表实体模型提供了一个通常由参数集驱动的报表布局 5  
    ![12-3](12-3.png)  
    报表对各个被选择的实体都将产生一个新页面  
    除了用户输入参数，还有下钻等交互功能对于指明实体模型也很有帮助  
    每个BI应用程序规范都分为两部分：实体模型和定义
    + 定义提供了报表或应用程序的基本信息
    + 实体模型提供了一种关于内容如何显示到页面上的直观表示 5

### 12.2.4 设计导航框架和门户
用户报表的增长将很容易超过100个或更多的报表  
需要设计一种方法来组织这些应用对象，机构的业务过程对于结构化该导航框架是很好的出发点  
通过一个简单的报表集的顶层放置多重组织框架，可以有效帮助这个分组和组织过程  
![12-5](12-5.png) 5  
BI应用程序的元数据对支持导航极为有用  
### 12.2.5 审查以及确认应用程序和模型
BI应用程序规范提供了逻辑上测试维度模型的方法，验证模型中的熟悉、约束、层次结构、度量和合适的粒度  
如果您有可用的资源和数据，您可以创建一些原型让您的用户更加相信应用程序规范的正确性  
### 12.2.6 与业务人士一同审查
与核心用户一起审查确认应用程序规范，以及维度模型是否支持该规范  
当进入开发阶段后，这些规范会不可避免地改变 5

## 12.3 BI应用程序开发
应用程序开发过程是一个标准地软件开发流程：准备、构建、测试和部署
### 12.3.1 准备应用程序开发
审查并更新您之前在项目里制定地BI应用程序规范，另外需要安装和测试BI工具并设置用户安全性、使用情况和性能跟踪  
1. 安装和测试BI工具  
    大型BI工具有许多组件，一但组件到位，要定义元数据来访问数据库
    1. 将元数据填入BI工具  元数据定义的丰富程度和复杂性是高度依赖于工具的
    2. 创建业务元数据   
        业务元数据是DW/BI系统中最重要的元数据，确定您填入的所有业务元数据元素对BI工具都是可用的 5  
        整个报表导航框架应构建为元数据，报表能简单地属于多个分类，改变导航结构也可以通过简单地编辑元数据实现
    3. 测试BI工具  
        您应该试用工具全部功能，作为初始测试的一部分  
        调试BI工具的元数据定义需要对工具的元数据有深入的理解
2. 验证您的BI应用程序界面策略   使用合适的BI工具来制定访问方式的最后决策  
3. 设置用户安全性  包括限制报表用的浏览和执行 5  
4. 建立报表过程元数据系统  
    过程元数据是用于跟踪一个给定活动持续运行情况的数据，在开发过程的早期设立这样的日志系统是比较好的做法  
    如果BI工具不提供，就应该想其他办法，应有汇总层的表记录用户的日常访问的度量  
    最好创建一个针对报表过程元数据运行的BI应用程序集，在问题变得严重之前就识别出它们

### 12.3.2 构建应用程序
建议在前期规划上多花时间，以后可能会您省下几天甚至几个星期的工作量  
构建BI应用程序的基本步骤：  
1. 沿着核心过程前进  
    大多数工具开发BI应用程序都包括：5
    1. 定义用户交互 
    2. 创建查询
    3. 对结果进行格式编排
    
   应用程序的性质、使用的工具和选取的应用程序方法决定了要在这三个方面做多少工作  
   1. 定义用户交互  
       多数BI应用程序都允许用户指定参数并且查看结果  
       审查应用程序规范并找出用户交互，确定最适合的交互工具  
       您也需要想办法向用户展现这些选项  
       好的BI应用程序要使用户界面易于使用，点击鼠标次数少于三次 5  
   2. 创建查询  
       主要查询由报表设计规范中的各种参数组成，各个参数通常都由对应的能够产生相关选择列表的查询  
       可能需要做一些横向钻取查询，或者自己构建查询  
       在构建分析性应用程序或运营BI应用程序时可以编写并测试一些自己的查询  
   3. 结果的格式编排  
       要保证报表打印出来也很好看，当导出EXCEL、PDF文件或任何其他对您的组织很重要的目标格式时，要检查是否进行了恰当的格式编排  
       注意您必须至少有足够用于开发BI应用程序的数据，哪怕只是测试数据 5
2. 设计和开发分析性应用程序  
    分析性应用程序一般以具体业务过程和封装领域专业知识为典型目标  
    任何分析性应用程序都是一个程序，该程序能够绘出用户界面（UI)、管理用户交互、通过API从BI工具中显示报表和图形组件  
    根据应用程序复杂性和系统环境的不同，使用更标准的编程环境开发您的应用程序内核可能会更有意义  
    分析性应用程序经常嵌入到一个被设计用于具体业务过程的综合产品中  
3. 设计和开发运营BI应用程序  
    通常是自定义开发的，用于查询数据仓库和运营系统，使用户能够在BI应用程序内调用事务处理过程 5  
    运营BI通常所需的服务级别比较高，需要提高DW/BI系统的性能服务级别  
    这些应用程序往往需要数据接近实时  
    实时层和运营系统监视都是专门设计用于满足这些运营BI应用程序需求的组件  
4. 包含数据挖掘模型  
    通常采取的形式是带有具体参数调用预定义数据挖掘模型  
    创建用于分析性应用程序底层数据挖掘模型通常是独立的活动，是一个过程 5   
    ![12-6](12-6.png)  
    运营BI系统也常常嵌入数据挖掘模型作为其规则引擎的一部分，找出事务处理流程中的时机和问题  
5. 验证数据和数据模型   要在项目规划中为查出这些问题并采取争取的行动留足时间
6. 创建导航BI门户  
    门户有多个层次，顶层主要是一个导航工具，用于组织标准BI应用程序和仪表板，最清晰的方式是在业务过程之后跟一个类别  
    第二个层次的门户通常关注基于所选业务过程数据的报表，可以在BI门户的多个区域交叉链接一些报表和仪表板 5  
    ![12-7](12-7.png)
7. 安排报表调度  
    每当ETL过程结束时应当立即运行和存储完成的报表  
    多数BI工具提供了根据预定义参数调度标准报表的服务  
    报表分发和管理依赖于数据访问工具，访问工具支持报表缓存  

### 12.3.3 应用程序和数据的测试和验证
测试时必须步骤：针对每个报表检查您的逻辑，创建简单的查询，并把报表数据复制到某个电子表格并重新进行手工计算 5  
根据正确的数值检查百分比和差值，改变参数并重新运行  
一旦确信报表已经正确运行就需要看看数据本身了，下面时一些很可能会遇到的问题：  
+ 无意义的描述  源系统提供的名称和描述可能需要修改 
+ 重复的维度信息  主要源数据中两个不同键可能会有相同的名称
+ 错误的维度关系  在钻取的时候进行检查以确保引入的是正确信息 5
+ 数据不平衡  不能保证数据和另一个已有系统的平衡，可能是因为BI应用程序计算业务度量的方式不同，也可能因为遗留的系统里使用不同业务规则创建事实
+ 性能调整  复杂报表和分析中，初始集无法很好的运行是常见情况  

### 12.3.4 完成文档
除了业务级的文档应当已经安全地存储在源数据目录中，为应用程序模板基础架构信息（命名标准、主模板的位置信息）编制文档也很重要
### 12.3.5 部署计划
BI应用部署不像数据库或事务处理那么严格，它没有那么高的风险 5  

## 12.4 BI应用程序维护
BI应用程序初始集以及所有随后的附件都需要维护和加强，您需要做以下工作：
+ 添加由业务用户和DW/BI团队一起构建的新的BI应用程序
+ 升级BI应用程序使之包含新数据源或对已有数据源的修改
+ 监视BI应用程序性能
+ 基于监视系统去除无用的BI应用程序，这需要在过程元数据中根据报表名捕捉其使用情况

## 12.5 小结
BI应用程序是向机构传递价值的主要方式，设计和开发是生命周期中两个独立阶段，需求定义完成后应立即捕捉BI应用程序规范  
构建健壮的、可用的和正确的BI应用程序时一项重大的任务，应该由专人负责，且要让它们参与到需求收集和维度模型设计过程中来
## 12.6 管理工作并降低风险
风险是没有集中关注传递业务价值并低估了工作量  
回到分析周期并确保您为用户提供了一个完整的环境  
安排一点实际向您的BI工具专家进行咨询
## 12.7 保证质量
测试是主要的质量保证工具：
+ 测试计算——加起来做除法运算，计算比率，确保报表本身内部一致性和正确性
+ 在层次结构的不同层次进行测试，那些汇总层加起来正确的结果，向下钻取细节时就不一定正确了
+ 测试边缘参数，比较年份之间的数据时，如果选取最早的年份，没有上一年，那么测试一下会产生什么结果 
+ 与已有的数据源进行比较，如果结果应当匹配，要确保它们是匹配的

## 12.8 关键角色
+ BI应用程序设计师和开发人员是主要驱动者
+ 业务分析员紧密地与开发者一起开发规范并测试应用程序
+ 具有高级报表和分析技能的业务用户也可能参与设计和开发应用程序
+ DW/BI质量保证分析师通过应用程序检查数据并验证数据和准确性
