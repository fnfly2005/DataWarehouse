本章讲述了典型DW/BI技术架构中应当包含的所有功能组件，首先讲述了创建一个明晰的技术架构的价值，在较高层面对DW/BI进行综述

第4章和第5章主要面向团队里的技术架构设计人员和技术支持专家，要想创建DW/BI技术架构，这些人员就必须深入了解这两章内容 5  
## 4.1 架构的价值
架构在项目创建过程中的价值和设计图在很多方面类似，优良的DW/BI技术架构可以带来如下好处：
+ 使满足业务需求的可能性更大  
    业务需求是决定系统功能最主要的因素，尽管所有的DW/BI系统基本架构相同，但是他们在具体功能上有根本性差异
+ 便于沟通  
    架构计划在多个层次上都是一种比较好的沟通工具，让参与沟通的人员明确地了解到自己的职责和需要完成的工作  
    架构计划也可以促进对上面的沟通，帮助管理层了解DW/BI系统项目群和项目的规模及复杂性，如果管理人员可能没有认识到DW/BI系统的复杂性，他们可能会认为系统实现比较容易，会纳闷为什么会花费这么多资金和时间
+ 规划  
    架构有助于项目规划的交叉检验，架构使得系统细节集中到一起，规定了它们如何才能形成一个整体 5  
    在架构计划中还可以反映出某个具体项目规划过程中没有考虑到的技术需求和依赖关系，例如在安装软件过程中突然发现所选择的数据访问工具需要一台独立服务器支持，这种意外情况需要花费更多的资金  
+ 灵活性、生产率和维护  
    创建架构实际上就是尽可能多地预先考虑各种潜在的问题，并且在次基础上创建一套体制来处理这些问题，而不是等到这些问题浮出水面并且造成项目群停工和返工之后才进行解决  
    本章讲述的架构主要依赖于元数据，为数据仓库增加了一个语义层次，该语义层定义和描述了数据仓库的内容和过程，用于数据仓库的创建、导航和维护等过程  
    架构有助于选择合适的工具对数据仓库的部分过程实现自动化处理，而不是手工编写代码构建语义层和定制层，提高了生产率和复用性  
    数据模型和元数据有助于分析主要变更带来的影响，可能还会提供某个发生变更的节点来支持这种转换 5  
+ 学习  
    架构在系统中还扮演着重要文档的角色，它可以帮助团队中的新成员尽快熟悉系统组件、内容以及业务客户等情况  

## 4.2 技术架构综述
业务需求回答“需要做什么”的问题，技术架构回答“怎样做”的问题，技术架构是一个总体规划  
技术架构主要包括三个部分:数据架构、应用程序架构和底层设施  
DW/BI系统的核心功能是从源系统抽取数据，通过清洗、对准并且使数据一致，将数据传输到当前呈现服务器上，进而使最终业务用户的客户端能够有效使用这些数据   
DW/BI高层功能和处理过程比较类似，但是在技术架构的实现中提供的技术模型集和功能说明需要根据满足机构的需求和实现资源的有效利用来进行定制  
每个机构的需求和自身具有的资源都不同，每个机构的DW/BI系统都应该具有特有的技术架构 5  
因为业务需求是确定技术架构的内容和优先级的基本指导方针  
技术经常会变化，兼并是我们这一行业的一大趋势，第二种趋势就是利用DW/BI系统中的数据的方法越来越多  
图4-1 给出了DW/BI技术架构的高层要素，是学习本章内容的路线图  
![4-1](../syn_pic/data_warehouse_life_cycle/4-1.png)  
本节将讲述DW/BI系统架构的总体流程，并且讨论数据仓库内部工作区（后台），和数据仓库公共商业智能界面（前台）之间的自然分割 5  
分离的两部分提供了一个十分有用的架构化框架，关注这两个区域的是不同的角色和不同的人员  
应当明白图4-1是一个逻辑模型，实际的物理服务器和运行于其上的应用程序根据多种因素会有所不同  
### 4.2.1 从源系统到用户桌面的流程
图4-1左半部分的后台区显示了如何在ETL过程中使用ETL服务层中提供的功能对源系统的数据进行处理  
该流程由元数据驱动，这些元数据描述了源系统和目标系统的位置和定义，以及数据转换、时间安排和依存关系  
一旦数据在ETL过程中经过清洗和对准以后，ETL系统就会对数据进行选择、聚集和重构，形成业务过程维度模型 5  
这些数据集会被装载到图4-1中所示的呈现服务器平台，并且通过企业总线架构中描述的一致性维度和一致性事实关联到一起  
我们规定业务用户、查询工具、报表书写程序或者软件应用程序通过任何方式访问的数据都属于一个或者多个业务过程维度模型  
这种规定有助于将数据表现方面的问题和ETL方面的问题加以区分，因为产生这两类问题的原动力是不同的  
ETL主要是面向装载的，BI应用程序则是业务查询驱动的  
后台并不为业务用户提供查询服务，只有呈现服务器才支持查询服务，并且它们按维度格式存储和表现数据 5  
图4-1右半部分的前台部分表名了用户如何通过各种BI工具和应用程序来访问呈现服务器上的数据  
不同的BI应用程序需要不同的工具，技术架构中常常要有一整套通用工具，用户前端界面则是在应用程序开发环境中生成的  
这些工具利用查询管理服务来协助管理用户登陆、查找恰当的数据和收集有关系统使用和执行情况的信息  
这些服务中大部分都是由元数据定义的，元数描述数据仓库的位置和内容、业务说明和分组、用户访问权限和数据过程（聚集导航）  
在图4-1中，环绕呈现服务器四周的虚线代表现实世界的单个逻辑视图，对于用户社区来说，所有的业务过程维度模型都是可以直接访问的 5  
可以通过多种方法实现这一的结果，最常见的就是物理上将数据放在同一平台中，如果无法使用这种方法，还可以使用如中间件、应用程序服务器、元数据、物理表和视图等方法组合，为用户提供一个隔离层  
### 4.2.2 常见架构特征
所有技术架构都由一些共同特征  
1. 元数据驱动  
    元数据是DW/BI系统领域极为重要的主题，但我们对它的认识是模糊的  
    我们认为定义和描述DW/BI系统的结构、操作和内容的所有信息都是元数据  
    DW/BI行业常常提到两类主要的元数据：技术元数据和业务元数据，以及我们提出的第三类：过程元数据 5  
    技术元数据是定义性的，而业务元数据和过程元数据一般都是描述性的，需要注意三类元数据会有一些重叠  
    + 技术元数据  
        从技术角度定义了构建DW/BI系统的对象和过程，包括表、字段、数据类型、索引、分区，以及数据库、维度、度量和数据挖掘模型  
        在ETL过程中，为特定的任务定义了来源和目标、转换过程（包括业务规则和数据治理筛选）以及频率
        在前台过程中，它定义了数据模型和将模型展现给用户的方法，也定义了报表、时间表、分配表和用户安全权限等方面的信息  
        有些技术元数据元素对业务用户非常有用，如表和字段名，而对于另一些如表分区函数用户一般没有兴趣 5  
    + 业务元数据  
        使用多种用户可访问的形式描述了数据仓库的内容，它主要告诉您有什么样的数据、数据的来源、数据的含义、以及该数据在数仓中和其他数据的关系  
        列出名称和内容描述字段就是一个基本业务元数据例子，通常以DW/BI系统文档的形式存在  
        业务元数据可以将表按照面向业务过程分组的方法分为多个子集，可以将相关联的列划分到同一个维度中  
        当用户通过浏览元数据来获知数据仓库含有什么信息时，他们一般都是在使用业务元数据  
    + 过程元数据  
        过程元数据描述了数据仓库中各种操作的结果 5  
        在ETL过程中，每个任务都会记录有关任务执行的关键数据，如开始时间、结束时间、占用CPU秒数、读盘、写盘和处理行数，用户查询数仓时，也会有类似的过程元数据产生    
        这一类数据最初主要用于发现和解决ETL过程或查询过程中的问题，还可以是进行系统性能监控和改进的重要输入参数
        这类元数据还可以用于监视用户访问情况，既可以显示数仓普及程度，又可以解决系统安全性和合规性方面的问题  
        过程元数据是在建立和使用DW/BI系统的过程中用于事件度量的数据  
        如果机构的核心业务是信息提供者负责收集某个行业的销售数据，随后将这些数据的访问权卖给多个客户，那么我们一般称作过程元数据的那些信息就会成为数据仓库中事实表和维度表的业务过程数据 5  
    
   元数据好比是数据仓库的DNA，它定义了数据仓库的各种元素和这些元素协同工作的规则  
   元数据驱动了数据仓库，并且能够将系统各个部件隔离开来，这位数据仓库提供了一定的灵活性  
   元数据在架构中扮演着重要的角色，因此我们将架构描述为由元数据驱动，元数据为应用程序完成工作提供了很多参数和信息  
   数据仓库技术中，最理想的情况就是使单一综合的元数据知识库为DW/BI系统中的所有服务所共享，这种情况下，元数据知识库中的某种变化会立刻通过架构和所有可用的服务反映出来  
   大多数工具都智能管理自身的元数据且无法和其他工具共享，所以将所有的信息都存储到同一个位置是不现实的，因此元数据一般都存放在多个知识库中，这些知识库由维护的DW/BI系统运行的工具、程序和技术设备所创建 5  
2. 灵活的服务层  
    我们将服务层定义为某种基本功能或任务，服务很可能非常简单，如在某个数据库中创建一个表或者为BI应用程序请求返回结果，如何提供某种服务取决于业务需求、资源和优先级  
    通过网络服务来发布产品越来越常见，这种面向服务的架构（SOA)方法使您可以更为灵活地将所需功能集成到DW/BI系统中，不必再封装到某个特定的工具中  
    SOA的发展得到了很多IT机构的支持，提供可重用的服务，解决复杂机构中各种独立技术的集成问题，DW/BI行业可以从SOA的发展中学到很多，他们已经对下列事项进行了规划和详细说明：
    + 服务接口定义
    + 服务质量特征 5
    + 业务策略知识库
    + 服务的管理
    + 生命周期管理、包括从开发到测试再到生产的整个过程
    + IT和业务的协作模型
    
   但是还有一些问题没有得以解决，例如安全性，是一种很有前途但是又并不成熟的方法 5  
   ETL服务和BI应用程序服务都提供了一个间接的层次，增加了架构的灵活性  
   集中式的服务（如用户身份验证）只需使用单一资源，不需要反复使用每个工具中的功能  

### 4.2.3 DW/BI 架构评估
架构计划是会不断演化的，因为业务需求不断变化，可用的新数据在逐渐增加，各种新技术也不断上市  
如果业务用户催促您进行重大的变更，说明您已经取得了成功  
本书的技术都有一定的适应性，可用来应对不断变化的架构 5  
例如我们通常使用尽可能最低粒度层次上的数据，过早的进行数据聚集是架构灵活性的天敌
在解决业务需求改变和更强大的技术手段的可用性问题时，原子层数据具有非常好的灵活性  
本书介绍的另一种避免架构变化的重要方法就是维度建模方法，一旦定义好了总线架构，就能够保证业务过程维度模型连接到一起创建出一个真正的企业信息资源库  
下面几节将详细讲述DW/BI系统的各主要方面  

## 4.3 后台架构
后台时进行ETL处理的地方，后台是一个复杂的、性能不稳定的系统 
本节考察了后台的架构，从一个一般性的需求入手，讨论有关标准ETL流程的问题，随后介绍ETL过程的四个关键步骤:抽取、清洗与一致化、提交、管理  
### 4.3.1 ETL一般性需求
先了解一下ETL系统的一般性需求：
+ 生产率支持  
    需要提供基本开发环境相关的功能：代码库管理、check in/check out、版本控制、以及生成和开发系统构件
+ 可用性  
    任务本身十分复杂，ETL系统必须尽可能便于使用，大多数ETL工具借助图形用户界面GUI来定义ETL任务，从而减少学习成本、加速开发过程、辅助编制系统文档 5  
    系统文档也是可用性的一个组成部分，ETL系统必须能够帮助开发方便地获取他们当前创建地过程信息，系统文档也是合规性的关键要素，合规性审计经常会问：您怎样处置自己保管的这些数据  
+ 元数据驱动  
    支持ETL过程的服务最重要的一个特征是：它们应当是元数据驱动的  
    这些服务应当由一个有关表、列、映射、转换、数据质量筛选、作业和其他组件信息的知识库所驱动的，而不是嵌入到ETL工具或SQL代码中的，因为很难查找和更改这些信息  

### 4.3.2 创建与购买
市场上的大多数DW/BI系统通用工具都可以提供本节讲述的大部分服务,ETL工具能够提高生产率，但不能指望马上得到回报  
根据经验，数据仓库项目群的第一次迭代过程中，这些工具并不能改进生产率，因为安装好这些工具并学会使用他们所花费的时间和他们所为您节省的时间差不多 5  
此后随着新业务过程的添加和对初始模式的不断充实，生产率提高才逐渐会明细和可观  
高端单机版的ETL工具对于小型数据仓库不会产生预期的效果，小型项目应当从那些与DBMS或者BI工具套件捆绑发布的工具入手，并在此基础上添加这些工具缺少的功能  
### 4.3.3 后台ETL流程
在后台，数据从源系统经过ETL过程传输到呈现服务器，后台服务就是支持ETL过程所用到的工具和方法  
本节重点描述典型的ETL流程，包括常见源系统和主要服务，服务列表和大多数ETL系统中常见的那34个子系统紧密相关  
如 图4-2给出了后台的更为详细的架构模型 5  
![4-2](../syn_pic/data_warehouse_life_cycle/4-2.png)
ETL处理流程包括四种主要操作：从源系统中抽取数据、通过一系列清洗和一致化转换过程，将数据提交到呈现服务器上，还要对ETL过程和后台环境进行管理  
有一条经验法则是：数据仓库开发过程有70%花费在ETL过程上，要考虑到对源数据格式的理解、数据探查、数据清洗、单元测试、系统测试、进入生产环境和文档编制等工作  
几乎每项ETL作业的起点都是源系统，我们将首先讲述一些关于常见源系统的内容  
### 4.3.4 源系统
数仓一般都会从多个源系统抽取数据，企业层的数仓更是如此，数据来自于多个系统，系统又往往由多个平台的多个数据存储所组成 5  
一般初始业务需求是为了访问核心工作系统，有些情况可能无法直接访问源系统，需要定期提供合适的数据，通常是生成一个抽取文件并将其传送给数仓  
要确保准备使用的ETL工具能够和关键源系统配套使用，否则就需要自己编写抽取程序，将数据转换为比较中立的格式  
1. 企业资源规划系统  
    这些企业资源规划（ERP）系统一般都是由许多模块组成，涵盖了如订单记录、人力资源、产品交易和生产制造等主要功能  
    从数仓角度来看，这种绝妙的工具也存在两个重要缺陷  
    + 首先源系统常常会有数以千计的表，它们的命名可能会使用某种您不了解的语言，要找出哪些表记录着对数仓有用的数据以及这些表之间的连接关系对ETL开发来说非常困难    
    + 其次这些系统一般都无法支持重要的分析查询装载，因为事务处理系统很难进行大范围分析查询，而对DW/BI来说这将是一项主要的合理性证明 5  
   
   ERP系统厂商很快认识到，开始生产和提供完全定制的DW/BI系统  
   尽快ERP数仓模块很有用，但很多公司并没有把所有的数据都放在ERP系统中，将外部数据加载、访问和使用这些数据是很困难的    
   有些公司使用厂商提供的DW/BI系统进行第一阶段ETL处理，然后从ERP系统的DW/BI系统中抽取数据，并且将结果同其他数据进行集成，最终创建一致的企业DW/BI系统  
   即时所有的数据都存放在ERP系统，并且机构没有对系统进行自定义，使用厂商提供的数仓也不能抛开业务需求、设计和架构  
2. 运营数据存储  
    运营数据存储简称ODS，基于详细的事务层数据，在数仓发展早期，单独创建一个和数据仓库主题分开的、面向决策支持的ODS是有必要的 5  
    随着技术发展，现在我们将最低层次的事务数据按照多年历史时间序列的方式存储，突破了以前对ODS的限制，可以将大型业务中的单个事务存储到某个历史时间序列里  
    我们将面向决策支持的ODS看作数据仓库的前沿部分，不再指定专门名称和提供单独环境，独立的ODS只有在为集成后的事务数据提供运营需求时才会遇到  
    也有一些运营事务处理系统的数据存储被称作ODS，这样的事务处理ODS是事务处理系统的一部分，用于响应事务处理亲求和满足事务处理服务层的要求，不是DW/BI系统的一部分  
    如果您已经确信自己拥有了ODS，仔细对其进行检查并确定其用途，如果它具有报表或者决策支持功能，那就应当在原子层将其集成到数据仓库，并使其和企业总线架构一致  
3. 报表ODS  
    有些机构基于客户/服务器的事务处理系统创建业务数据库的一个单独副本，将其作为工作系统的报表环境 5  
    报表ODS非常有用，因为它支持和事务处理系统数据模型相对的运营报表  
    另一方面，它也有缺陷，它并不具有DW/BI系统所具有的数据清洗、变化跟踪、历史数据、企业一致性维度、性能调整和便于使用的维度模型等功能，意味着机构会同时拥有两个版本的事实  
    如果需要报表ODS，就应当在适当的服务层上将其作为工作系统的核心部件进行管理，对其进行监视、调整和规划，从数据仓库的角度来看，它是另一种源系统  
4. 主数据管理  
    主数据管理(MDM)系统是集中式工具，用于保存关键实体的主副本，这类系统是为解决表的扩散问题而建立的，表的扩散是由于使用了多个键对同一客户或产品进行多次描述  
    将主数据管理系统作为事务处理系统环境的一部分对DW/BI系统来说是很重要的，因为ETL系统中数据清洗和一致化阶段的大部分工作就是要创建单一版本的一致性客户或产品信息 5  
    主数据管理系统可以帮助我们把数据集成问题转移到产生这些问题的源系统加以解决，如果客户信息或产品信息集成面临严重问题，应当开始着手建立一个主数据管理系统，而不是等到数仓的ETL系统中再修正  
    未来许多主数据管理实施方案是基于Web服务SOA架构的，在大型机构中，使用分布式客户端访问单一版本的关键实体是最佳方案  
5. XML源  
    平面文件XML文件比平面文件有更好的结构，但XML在处理和存储方面开销较大  
    XML文件常常用于跨越机构界线交换数据，能使数据从特定计算机实现中独立出来，在www.xml.org 可以看到不同行业的XML schema示例  
6. 消息队列、日志文件和Redo文件  
    消息队列以及读取事务处理日志文件可以使得数据仓库的操作性和实时性更强 5  
7. 专有格式  
    在运营领域有数十种转悠格式，有些数据库有丰富的文档并且带有商用的可用源适配器，有些情况下ETL团队必须手工进行日常的抽取工作

    + 批参数为所有ETL工作确定了当前活动的源表和目标表  
3. 业务元数据  
    + 数据筛选说明包括数据质量测试的代码、潜在错误的严重程度评价，以及当错误发生时采取的措施
    + 数据字典描述了数仓中所有列和表的内容

## 4.4 呈现服务器架构
呈现服务器是进行数据存储的目标平台，业务用户、报表系统和其他BI应用程序都可以直接到这里查询数据  
在讨论架构细节之前，了解一下信息方面的业务需求是如何有助于定义恰当的方法  
### 4.4.1 信息方面的业务需求
几乎所有的BI用户都有下列需求：
+ 访问所有主要业务过程的数据  
    每个人都想浏览系统中的所有资源，比如浏览订单，不过各自角度不同，销售关注订单顾客，市场关注订单产品，所有这些都来自于相同的业务过程维度模型  
    对于管理人员，他们可能希望看到所有业务过程的关键性能指示器 5  
+ 访问汇总数据和原子数据  
    每个人都想知道总体情况和具体细节，管理层则希望深入了解关于所有性能指示器的细节  
+ 单一来源的分析数据  
    这样做的目标是将管理层的讨论聚焦到决策制定，而不是关注谁拥有最准确的数据  
    
注意：
    每个人都希望所有主要业务过程原子层上的数据来源都是一致的，那么提交给部门数据集市的数据会迅速增加，给数据管理带来极大的困难  
### 4.4.2 细节原子数据  
上述三种核心需求简单来说就是无法预测的分析性查询，来自机构的每个角落，并且要求能够立即得到细节或汇总数据  
为满足这些需求，就有了图4-3所示的呈现服务器架构模型，呈现服务器的基础是基于原子层的业务过程维度模型 5  
这些原子层数据集是使用企业总线矩阵中定义的一致性维度创建的，而且一般都存储在关系数据库而不是OLAP引擎中  
![4-3](../syn_pic/data_warehouse_life_cycle/4-3.png)  
如果您的数据集较小，或者硬件性能较强，那么一定需要这样的原子层，数据库能够即时汇总，通过横向钻取共享一致性维度的事实表查询可以很容易获得整个业务过程的分析结果  
### 4.4.3 聚集
对于大数据集，为了改善汇总层的性能，我们引入了呈现服务器的第二个要素:聚集，在装载过程中预先对数据进行聚集是提高分析查询性能的主要方法之一，聚集占据了一个单独的逻辑层    
聚集每天都会被创建和重建，理想状况下，聚集导航系统会追踪聚集使用情况，并且自动跳转它所创建的聚集 5  
我们将现在所谈的这个层面称作聚集，但是实际的数据结构可能会因为性能的原因而包含一些细节层的数据，这种情况本质上可以将聚集层视为一个胖索引  
聚集和OLAP引擎定义了完全不同的架构角色，聚集和OLAP引擎都可以看做是BI淋雨的一部分，归属于核心数仓之外的业务用户  
我们通过汇总数据明确了聚集组件中的内容是单个无缝业务过程的一部分，内置于后台，关注焦点必须从部门转向企业，不同部门使用同样的聚集  
这些聚集后的数据集应当和原子层数据具有相同的维度结构，当上升到更加概况性的层次时，有些维可能会被缩减或者省略  
#### 聚集导航
有了聚集和原子数据，就增加了数据环境的复杂度，因此我们引向呈现服务器的第三个重要部件：聚集导航器 5  
聚集导航器接收基于原子层维度模型的用户查询，它对查询进行检查以便确定是否能够使用较小的聚集表应答该查询，如果可以则按聚集表要求重写并且提交给数据库引擎，查询结构随后返回给用户    
在实现层，有很多技术可以提供聚集导航功能，包括：
+ OLAP引擎
+ 基于优化器导航的关系数据库物化视图
+ 关系OLAP(ROLAP)服务
+ BI应用程序服务器或查询工具

这些技术中有很多含有创建和管理聚集的而功能 5  
如图4-3所示，原子层业务过程维度模型、性能聚集和聚集导航器这三种部件的联合可以使来自己各个业务部门的用户透明地查询所有可用的业务过程在所有细节层次上的数据  
#### 原子层数据模型
从原子层业务过程维度模型名称来看，应该明确的是我们建议将原子数据存储到一个维度模型中，而不是一个规范模型中  
维度模型是最好的用户方位内模型，因为用户希望访问最低层次的细节数据，应该将其存储在同一维度架构中  
如果使用类型2方法跟踪维度变化，就会发现每个人都需要使用类型2方法跟踪一些属性变化，都会在原子层用到维度模型  
### 4.4.4 呈现服务器设计规定
技术架构中的呈现服务器是为满足向BI应用程序提交维度模型的要求而设计的，这里罗列一下这些设计规定，呈现服务器包括以下具体内容： 5  
+ 具有单一字段代理键的反规范化维度表
+ 类型1、2和3的缓慢变化维
+ 多值维和层次结构维
+ 基于企业总线架构的一致性维度和事实
+ 事务处理、周期快照和累积快照事实表 5
+ 面向性能的聚集、星型索引、位图索引和OLAP
+ 可选的实时BI应用程序专用热分区

### 4.4.5 调整呈现服务器架构
对于大多数大型机构来说，规模是最大的技术挑战之一，需要对核心呈现服务器架构的物理实现进行调整  
垂直分割意味着将呈现服务器的架构分解成几个独立的平台，为原子层数据分配一个服务器，为聚集管理和导航分配一个服务器  
水平分割意思是基于数据集分配装载，应当用独立的呈现服务器作为具体业务过程维度模型的主机 5  
### 4.4.6 机构考虑事项
理想情况下，DW/BI核心团队应当将表示层作为企业资源来建立，同时要兼顾到来自企业总线矩阵和聚集的原子数据，以便满足每个查询的需求  
现实中业务过程的复杂性和数据质量问题会极大地延长创建表示层的时间，一般可以通过两种方式来了解DW/BI系统的细节部分  
+ 源系统划分常常存在一些机构重叠的情况下进行，如果独立的业务管道没有共享一致性维度，最好使用并行且专用的DW/BI系统  
+ 如果它们是共享维度的，则它们必须使用相同的一致维度表协同工作  

业务群体划分常常在一个群体对BI应用程序在数据、计算或聚集方面都有特定要求，而核心DW/BI太忙以致于无法做出响应的情况下进行 5   
一旦要进行业务群体划分，则IT人员可能会为业务群体单独建立和管理局部分析性数据存储，这种方法会造成灾难性后果，导致数据集市烟囱的产生    
在大型机构中，核心数据仓库团队的职责就演化为负责主一致性维度表、数据仓库工具和数据集成  
### 4.4.7 呈现服务器元数据
在我们的技术架构中，呈现服务器自身所拥有的元数据相对比较有限  
+ 过程元数据  
    + 数据库监视系统表包括呈现服务器上所有表的使用信息
    + 聚集使用统计包括OLAP情况 5 
+ 技术元数据  
    + 数据系统表包含标准RDBMS表、列、视图、索引和安全信息
    + 分区设置包括分区定义和逻辑，以便随着时间推移对它们进行处理
    + 用于创建分区、索引和聚集的存储过程和SQL脚本，也可以用于安全管理方案
    + 聚集定义包含系统实体（物化视图）的定义，以及其他与聚集导航器的查询重组功能相关的信息
    + OLAP系统定义包括专门面向OLAP数据库的系统信息 5
    + 目标数据策略和步骤包含保留、备份、存档、恢复、所有权和安全设置
+ 业务元数据  
    由BI应用程序语义层、OLAP定义以及数据库系统表和列定义直接提供  

### 4.4.8 呈现服务器总结  
大量范围宽泛的分析性查询是难以预料的，而呈现服务器层必须具有响应分析性查询的功能  
本讲述的逻辑架构包含一组原子业务过程维度模型，可以产生用来提高系统性能的聚集  
聚集导航器通过重写查询实现对聚集的即时使用，并且使其复杂性对用户不可见  
## 4.5 前台架构
前台是数据仓库的公共界面，是业务用户每天工作的平台，对于用户来说，用户界面就是数据仓库，他们仅仅想知道答案  
DW/BI系统的主要目标之一就是让人们尽可能容易地访问信息，系统在用户和信息之间建立一个BI应用程序层，将诸多复杂操作隐藏起来，图4-4给出了前台的主要存储和服务  
![4-4](../syn_pic/data_warehouse_life_cycle/4-4.png)  
我们首先回归一下BI应用程序的多种类型，以及前台向用户提交信息和管理环境所需的服务，随后讨论前台的数据存储和BI元数据  
### 4.5.1 BI应用程序类型
很多BI应用程序是从呈现服务器中获取数据，必须以一种灵活、对称、可伸缩的方法来设计呈现服务器 5  
最重要的BI应用程序类型有以下几种：
+ 直接访问查询：通过查询工具应用程序生成的常规用户即席查询
+ 标准报表：通过BI门户或者以电子表格格式向在线库提交定期报表
+ 分析性用程序：除了标准数据库查询，还应包含带有高性能分析算法的应用程序，一般包含预算、预测和业务活动监视等功能    
+ 仪表板和记分卡：使用文本或图形显示关键绩效指标（KPI)的多主题用户界面
+ 数据挖掘和模型： 将一个很大的数据集从数仓下载到数据挖掘软件中分析，也用于创建供一些分析和运营BI应用程序使用的底层模型 5  
+ 运营BI：运行状态中的实时或近实时查询，常常同时带有事务回写界面  

这些BI应用程序通过各种应用程序界面提交给业务用户，包括如下一些界面：
+ BI门户和自定义前端: 特殊用途的用户界面，为用户访问BI应用程序提供便利，或者用于特定的复杂查询和结果筛选
+ 手持设备界面：为手持屏幕和输入设备设计的特殊BI应用程序
+ 即时BI：实时数据仓库架构的一种终极形式，直接连接源事务处理系统和用户屏幕 5  

### 4.5.2 BI管理服务
DW/BI市场的购买力为数据库销售商改进自己的产品使其支持数据仓库技术提供了动力  
简化客户端工具的要求引起了BI工具销售商对桌面客户端的瘦身，他们将大部分共享功能都转移到应用程序服务器上  
IT届近年来对面向服务的架构（SOA)研究热潮起到了很大的推动作用  
数据访问服务是真正的服务，应当独立于专门的工具，可以为所有用户所使用  
BI管理服务包括从呈现服务器和用户之间的共享服务到桌面服务的整个过程，这里所指的桌面服务是在用户层，主要复杂报表定义和结果显示 5  
1. 共享服务  
    包括元数据服务、安全服务、使用监视、查询管理、企业报表，以及Web服务和门户服务，都在图4-4管理服务方框中  
    1. 元数据服务  
        元数据模型把用户视图限制到表和列的子集、预定义的连接规则（包括列、类型以及路径引用）、预定义的计算值和标准过滤条件上，并提供额外的描述性元素和组织好的类目，从而使数据的复杂性对用户不可见，
        这些起简化操作的元数据一般都存储在BI工具的元数据知识库中  
        如果某种BI工具是由元数据驱动，用户可能会需要额外的元数据元素，您需要提供一定的功能来创建、管理和存储相关元素  
        利用该元数据知识库进行元数据报表处理可以帮助用户查找和访问所需的信息，理想情况下，需要业务信息的用户应当通过交互式报表，详细了解DW/BI系统元数据，从中寻找感兴趣的元素 5  
        交互式的元数据报表应动态链接到元数据知识库，以显示当前主题领域和相关的数据元素，该元数据报表应当能够提供各种数据元素的来源和定义，显示一些里包含这些元素的标准报表  
        元数据报表听起来似乎工作量很大，但是其回报却是用户通过自己的努力就能够满足自身的需求  
    2. 安全服务 
        安全服务促进了用户和数据库的连接，是设计和管理中所面临的主要挑战之一  
        安全服务包括授权、认证服务  
        认证意味着通过某种方法验证用户的身份，可以分为不同层次，认证是企业底层服务的一种，DW/BI系统应该充分利用这种服务 5  
        授权是一种确定用户可以访问哪些特定内容的方法，授权在DW/BI系统中是一个比认证更为复杂的问题  
    3. 使用监视  
        使用监视就是捕获那些有关数据仓库使用情况的信息，有以下5个理由来说明为何要在前台架构中增加使用监视：  
        + 性能，利用使用信息来有效地调整DW/BI系统
        + 用户支持，监视查询活动，对那些格式较差的查询进行干预，从中可以发现哪些用户需要在构造更有效的查询方面获得帮助  
        + 市场营销，发布简要的使用统计情况来告诉管理层他们的投资都用在哪里，一份好的增长曲线图是极棒的市场营销工具 5  
        + 规划，通过分析使用增长情况、平均查询事件、并发访问的用户数量、数据库大小和装载事件等，可以对需求进行量化，并且择机增加功能  
        + 合规性，对于有些行业（如卫生保健）或者有些数据（如人力资源或财务数据），还需要提供对每个用户查询活动的审计追踪  
        
       可以自己编写一个简单的使用监视程序或者购买比较完整的工具包  
       
   4. 查询管理  
       查询管理服务是系统中对整个查询过程进行管理的一项功能，下面强调的内容描述了架构中主要的查询管理服务：  
       + 查询重组  
           服务需要解析一个接收到的查询请求，并且判断出怎样解决最好，应当能够用目标服务器的语言生成复杂查询 5  
       + 查询重定向和横向钻取  
           查询重定向解析到来的查询请求，在元数据中查找它们的实际位置，随后根据需要改变查询方向或者改变查询的一部分，这一功能使查询能够对分离的硬件平台上单独的事实表进行操作  
           横向钻取帮助用户查询两个事实表中的数据，通过横向钻取可以将查询结果无缝地集成到客户报表中并返回给客户  
       + 聚集导航  
           通过聚集导航可以找到一个可用的聚集表来满足查询，而不是马上就计算详细记录地数目
           聚集导航器可以处在中间件的位置上，创建跨平台甚至跨技术的查询  
       + 查询管理和优先级区分  
           创建一个效率低下的查询很容易，但常常会减缓数据仓库的运行速度，应该在发生之前就意识到这种情况并且尽量避免 5  
           最简单的查询管理是限制查询执行花费的时间或者返回记录的行数，问题是它们都在事实发生之后  
           为了更有效地管理查询，查询管理服务需要在查询真正开始执行之前就估计该查询执行所需的时间，并告知用户  
           有时也会对具体用户和查询区分优先级，执行标准报表的账号可能需要较高的优先级，这些功能中的大多数在数据库层面上都很有意义  
           
   5. 企业报表服务  
       企业报表服务提供了创建和提交产生样式的预定义报表的功能，这种报表中含有某种层次的用户交互，使用非常广泛，而且定期执行  
       如果ERP不能完成运营事务处理和报表的工作量，就需要使用大型企业报表系统，这是一项很大的作业，有自己的一系列需求、服务和服务层次 5  
       大多数DW/BI系统中的查询活动都来自于企业报表，需要一种功能将这些报表集成到一个报表管理环境中去，企业报表工具的功能需求包括：  
       + 报表设计环境，在这个环境中应当含有大多数具有功能性和可用性的特定工具  
       + 灵活的报表定义，主要包括复合文档布局  
       + 报表执行服务器，它是运行报表的集中式资源，它将报表存储于登台区以备提交给用户使用  
       + 参数驱动或者变量驱动的功能，可以改变区域名称参数，使整个报表集都基于该新参数值运行 5  
       + 基于时间和基于事件安排报表的执行，一份报表可以安排在一天的某个特定时间执行，或者基于系统事件执行  
       + 迭代执行或结果分割  
           如果给出了一个区域列表，服务器将为每个区域创建一份报表的副本  
           结果分割和报表节选或者分页符概念类似，我们常常将这个过程叫做分页  
       + 灵活的报表传送  
           报表可以通过多种方式（E-mail、Web）来传送，也可以以多种类型的结果提供，如excel，csv，PDF  
       + 用户易于访问的发布和订阅功能  
           用户自己创建的报表应当能够被其部门或整个公司的人员所使用，同样他们也应当能够订阅其他人的报表，并且每当报表更新或者改进之后能够接收到副本或者通知 5    
       + 报表链接，这是一种能够向下钻取的简单方法  
       + 带有浏览功能的报表库，描述报表库中每一份报表的情况的一种元数据，可以记录其内容和上次执行事件以供查询使用，应当提供用户界面，方便用户按照不同的条件在报表库中进行查询  
       + 大众分发，使用简单、便宜的访问工具进行大众分发  
       + 报表环境管理工具，管理员应当能够凭借管理员模块、监视和解决报表相关问题，提出那些无人使用的报表  
       + 用户访问安全，需要对用户进行认证，并且确定他们是否有查看所请求报表的权限 5  
   6. Web访问  
       前台架构中还需要为用户提供基于Web浏览器的数据访问服务  
       除了内部访问，很多公司还利用因特网为客户或供应商提供对特定信息的直接访问，如果必须要提供外部访问，那么安全、维护和底层设施方面的问题就显得非常重要  
       Web访问功能可以封装到客户所使用的产品中，这一DW/BI系统才可以真正带来盈利  
   7. 门户服务  
       门户工具常常利用Web服务提供一个更加通用的用户界面，用于访问机构的组织机构信息、沟通和产品演示服务  
   8. 运营BI回写  
       尽管事务处理工具一般不能看作是BI环境的一部分，但是近来它们也成为运营BI应用程序的一个重要类别 5  
2. 厂商的特定架构选择  
    大多数主流BI工具厂商都创建了自己的三层架构，并且它们都将大多数服务放置在位于桌面前端界面和数据库中间的应用程序服务器上
    好处是用户可以将注意力集中于格式化的查询和报表生成环境，而不比关注多余的数据访问服务  
    遗憾的是，目前并没有针对应用服务器的标准，这些应用程序服务器相对专有，且很难为其他BI工具所用  
    数据库销售商也开始将一些服务集成到核心数据库引擎中，避免陷入只有单个BI工具可以使用服务的境地，但这种方式对使用应用程序服务器的方法来说有一定局限性  
    您需要明白它们的价值，以及应当规数架构中的哪个环节，建议和工具厂商沟通您对这些服务的需求情况 5  

### 4.5.3 BI数据存储
理论上来说，在前台不应当有任何数据存储，数据一般存储在呈现服务器中，事实上，当具体数据请求的应答集离开呈现服务器时，会被存放到某个云端，并生成文件形式以供下载  
1. 存储的报表  
    数据达到前台并接近用户后，就变得分散了，大多数单个的查询得到的结果集都会少于10000行，这些结果集至少会临时存储在BI工具中  
    有些数据访问工具运行在中间层的应用程序服务器上，使前台架构中又增加了一个新的部件和层次  
    这种服务器可以提供辅助数据存储，缓存用户的查询结果和标准报表，提供更快的响应速度  
2. 应用程序服务器缓存  
    在前台有多种面向数据的服务，图4-4显示了用于运营BI、分析性应用程序和企业报表的服务，所有这些服务可能有自己的数据存储，往往会存储到本地缓冲区为应用逻辑提供服务或者更快的响应 5  
3. 局部用户数据库  
    进行局部或者部门数据存储的思想就和个人电脑的历史一样古老，增强型的数据库工具已经使局部数据存储能够为本地的桌面程序所使用，而本地机器在这里就充当服务器了  
    实际上局部数据存储是由应急分析需求驱动的，这些需求又是业务人士所确定的  
    从架构来看，应当考虑到这种实际情况，使局部数据存储的开发人员能够方便地使用标准数据仓库工具和过程  
    从组织结构上看，应该支持局部数据存储的发展，应当尽量了解各个部门建立长期的局部存储需求，探寻为什么这些部门不能够直接从DW/BI系统中获取信息    
    最好从企业的角度出发，将所有有用的资源存放到DW/BI系统中而不是存放到使用受限的局部存储中，以便所有用户都能够直接访问 5  
4. 一次性的分析性数据存储  
    一次性的分析性数据存储是一个用于支持特定短期业务的数据集，和局部数据存储类似，但它有一定的存在周期限制  
    实际上建立一个单独的数据存储要更好，常常会有一些临时需求用到一些并不存储在DW/BI系统中的内部数据或外部数据    
    一次性的数据存储有助于对数据进行面向分析性需求的特定设计，将业务规则和过滤条件应用其中，为分析人员创建一个简单的沙盒  
5. 分析性应用程序产生的结果  
    分析处理过程常常运行于单独的机器上，并使用他们从数据库中钻取的属于自己的数据  
    将分析性应用程序或数据挖掘过程得到的结果以维度属性形式存放会数据仓库中是一件有意义的工作 5  
6. 下游系统  
    当DW/BI系统成为分析和报表所使用的权威源系统时，下游系统的基本需求是报表  
    尽管这些系统一般都是面向事务处理的，但它们对数据仓库的数据清洗、集成和历史信息挖掘等工作都有重要价值  
    客户关系管理系统是利用DW/BI系统数据的好例子，这些应用程序都从数据仓库获取数据，但是获取数据后又在单独的环境中使用  
    因此需要在数据仓库架构中考虑到对这些下游系统的支持，特别是需要确保满足可用性和响应时间等服务层要求，此外，还要考虑到数据等待时间和访问方法  
7. 数据存储安全  
    数据的分散配置不可避免地会带来安全性和合规性问题，需要对数据访问进行不同层次的跟踪和控制 5  

### 4.5.4 桌面工具架构方法
仅有少数功能是存在于或者至少运行于客户端桌面上的，他们帮助用户访问数据仓库中的数据，用户对数据仓库质量的整体感觉取决于这些工具是否能够满足他们的需求  
桌面工具直接和数据库相连接，虽然它们对于快速查询十分有用，但不适用于企业层，因为无法共享服务和元数据  
在规划有关呈现服务器层和BI工具的策略时，最好研究一下厂商的架构方法，他们的BI工具的架构将会对您的DW/BI系统整体架构有重大影响  
### 4.5.5 BI元数据  
前台BI元数据包括下面所列出的元素：
+ 过程元数据  
    + 报表和查询执行统计包括用户、列、表、应用程序使用情况跟踪、运行时间和结果集行数 5  
    + 网络安全使用统计包括登录企图、访问企图和用户ID定位  
+ 技术元数据  
    + BI 语义层定义包括所有表和映射到呈现服务器对象上的列的业务名称、联接路径、计算列和业务分组、也可能包括聚集导航和横向钻取功能  
    + 标准查询和报表定义包括查询逻辑、执行调度和提交说明  
    + 应用程序逻辑用于复杂BI应用程序的细节处理步骤中  
    + BI门户设置包括默认设置和特定用户的自定义及参数 5  
    + 安全分组和用户指派连接到基于网络的目录服务或者本身就是目录服务的一部分，也可能包含有外部用户指派  
+ 业务元数据  
    BI语义层定义包括基本的面向业务的元数据，这里补充BI业务元数据包括以下部分：
    + 一致性属性和事实定义与业务规则包括缓慢变化维度、空值处理，以及每个列的错误处理  
    + 用户文档和培训材料包括完整的在线图书馆和业务用户能够获得的文档  

### 4.5.6 前台总结
就可用性、种类和涉及到的工具功能来说，前台是数据仓库中内容最丰富的一部分 5  
前台是业务用户所看到和使用的数据仓库的一部分，为前台提供最好的功能是至关重要的  
数据仓库可以为广大用户群的多样需求提供服务，在前台设计时就要需要考虑到对这些需求的支持  
前台服务包括元数据、安全、监视、报表、管理、查询和其他桌面服务  

## 4.6 底层设施  
底层设施为本章所讲述的所有架构元素提供了基础，包括硬件、网络和一些能够为高层组件提供基础的底层功能（如安全性）   
为给定的系统实现确定合适的底层设施需要考虑多种元素，而且其中大多数都并不是技术方面的因素，DW/BI应当是IT机构中底层设施提供者的一个重要客户端  
### 4.6.1 底层设施驱动因素  
业务需求对底层设施仍然起主导作用，业务需求决定了数据仓库中历史数据的细节层次和数量，这决定了底层设施的数据容量、装载数据的频率和转换业务规则的复杂度  
数据仓库实现人员的技能和经验同样会影响底层设施决策  
规章制度和其他组织问题同样对确定底层设施起重要作用，通常，他们都是资本支出的临时性限制条件，需要通过更富有创造性的方法来掌控底层设施 5  
最后，底层设施的选择也会受预期增长率影响，不论是DW/BI系统的数据量和用户使用量，头24个月都增长很快  
一些实际的数据仓库系统在生命周期一个逻辑框就能够成功实现；然而，即使是小规模也需要应用程序服务器来支持基于Web的数据访问  
### 4.6.2 后台和呈现服务器底层设施因素
确定后台和呈现服务器底层设施需求的主要因素包括：
+ 数据规模，受所要解决的业务问题驱动  
+ 易失性  
    度量了数据库本身具有的动态特征，包括数据库多长时间被更新一次，有多少数据发生变化或者每次要置换多少数据，以及装载窗口的长度，例如使用客户流失率去判断客户维度的改变量，数据仓库得适应业务和技术的变化 5    
+ 用户数量  
    用户的数量、他们的积极程度、并发活动的用户数量和他们活动中的所有周期性峰值，都是平台选择中的重要因素  
+ 业务过程的数量，随着业务过程数量的增加，数据仓库的源系统复杂度会显著增加  
+ 使用特点  
    前端BI工具及其使用特点取决于平台选择,即使只有少量活动用户也可能极大地消耗呈现服务器资源  
    对这种使用情况进行优化十分困难，因为分析员会生成不可预测的即席查询以找出所有可能进行优化的机会  
    另一方面，一个经常生成标准报表的系统可以对那些报表进行优化，大多数标准报表的生成避开了系统使用的高峰时间 5  
    大规模数据挖掘同样对硬件平台有较高要求，一般对数据粒度和I/O可扩展性都有一定要求  
    即席查询、报表和数据挖掘都有不同的查询要求配置文件，并且在不同的平台上可能会表现更好  
+ 服务层协定  
    性能和可用性方面的需求对所需硬件的大小和数量有所影响  
    如果用户期望获取毫秒级响应，就需要一个较大的工作平台，如果用户遍布全世界，可能需要系统全天候可用  
    如果运行系统是集中式的-数仓也可能是集中式的，如果运行系统是非集中式，则也需要使用非集中式的呈现服务器组件 5  
+ 技术准备  
    服务器又很多方面的支持需求，包括基础的硬件和系统软件管理、连通性（包括于用户的连通，也包括同源系统的连通）、数据库管理和备份与恢复，IT支持人员的数量、质量和经验会对平台选择产生巨大影响  
+ 软件可用性  
    通常、需求分析会显示出对某种特定功能的需求，如能够将数据仓库数据显示到地图上的地理信息系统  
+ 财务人员  
    项目投资总额常常也是业务用户希望DW/BI系统能提供的一项功能  
    对业务需求理解得越透彻，同时将DW/BI系统的目标定位在解决高价值的、相对简单的需求上，就会越容易得到建设DW/BI系统所需的经费  

### 4.6.3 并行处理硬件架构  
在服务器市场上有三种基本的并行处理硬件：对称多处理(SMP)、大规模并行处理（MMP)和非均匀存储器架构（NUMA) 5  
如图4-5所示，这些架构在处理器与磁盘存储器、内存的配合工作方式上有所不同  
![4-5](../syn_pic/data_warehouse_life_cycle/4-5.png)  
1. 对称多处理SMP  
    SMP架构是单机配置多个处理器，由一个系统统一调度，而且所有处理器都访问同一的磁盘存储器和内存领域  
    SMP架构的机器非常适合即席查询，对于即席查询，SMP架构的共享、集中式等特点使得系统具有为整个数据库分配处理器资源的能力    
    随着CPU数量的增加和时钟速度的加快，使用单一共享的通信通道进行CPU、内存和I/O之间的通信会成为系统性能的瓶颈，此外也造成了数据仓库运行过程中会出现单点故障  
2. 大规模并行处理MPP  
    MMP系统本质上是一组相对独立的计算机，每台机器都有自己的操作系统、内存和磁盘存储器，通过来回传递消息实现相互协调 5  
    大规模并行处理的优势是能够将成百上千个机器节点连接到一起，并且利用它们使用强力的方法解决问题  
    MPP系统一般应用于大规模的、有海量数据的数据仓库，可以对MPP进行配置，通过多个节点上的镜像数据提高系统的可用性  
    MPP系统支持一般性查询或者标准报表环境，或者将那些输入聚集数据包含的原子数据存储在呈现服务器层  
3. 非均匀存储器架构(NUMA)  
    从本质上来看，非均匀存储器架构是SMP和MPP两种架构的混合，既具有SMP架构中共享存盘存储其的灵活性，也有MPP架构中并行处理的速度  
    如果将习惯相同的用户分为一组，进而依据不同用户分组将数据仓库分成相对独立的部分，每个群组都拥有自己的节点，NUMA架构就会非常高效 5  
4. 集群(cluster)  
    可以将几台计算机连接到一起形成一个集群，共同充当一个单独的服务器，集群的主要作用有高可用性和扩展服务性能，后者也称作向外扩展架构（scale out)  
    在高可用性的环境中，两个或多个系统被连接起来，共享存储器并进行主从配置，一旦出现故障就自动进行故障切换  
    向外扩展这个术语是说通过增加更多的设备来扩展服务器的性能，向上扩展是通过增大单机规模来提高服务器的性能  
    集群面临的最大挑战就是系统中出现了一个节点的查询需要另一个节点的数据存储的情况，此外各节点的通信开销也非常大，需要妥善解决  
    数据库和操作系统解决这一问题的方法有多种，主要包括共享缓存、联邦数据和复制数据等方法： 5  
    + 共享缓存，当集群环境下存储器被所有机器共享时，系统必须对未被授权的事务处理进行持续跟踪，系统需要具有分布式锁定管理的功能，以便协调各个协作节点之间的资源    
    + 联邦数据，联邦数据环境需要多个节点的数据被分割或者进行联邦，这一查询就可以通过多个节点查找数据，并且将找到的数据整合后返回  
    + 复制数据，数据复制环境要求数据被复制到多个不同的节点，这一仅使用本地关系引擎就可以满足查询请求，每个节点都访问自己的数据副本，不必在考虑协作问题  
    
   操作系统、数据库和通信技术的发展极大地影响了这个领域，将几台相对便宜地计算机连接起来构成集群，并且将其连接到一个高性能磁盘存储器子系统就可以获得很好地效果  
5. 数据仓库设备  
    数仓平台地高端市场都是面向特定用户的，因为主流关系数据库产品在设计时，目的并不是像DW/BI系统那样对大量数据做综合处理 5  
    有些公司这些年开始提供预先配置好的专门面向数据仓库的硬件和数据库平台，这些产品的最新版本都基于通用技术和开源数据库  
6. 划分硬件  
    大多数情况下，可以将一个大型的服务器从逻辑上或者物理上划分为几个较小的服务器，它们拥有自己的操作系统和资源  
7. 所有并行架构都需要考虑的事项  
    对于所有的平台都有必要针对软件可用性和系统管理复杂性做特别考虑，尤其要考虑操作系统的版本和类型  
    还需要确定应用程序、技术设备和驱动器在该版本的操作系统上是否可用  
    一定要考虑软件能否与最新版本的RDBMS、ETL工具、开发环境、数据仓库功能设备和应用程序服务器等相兼容 5  

### 4.6.4 硬件性能推进器
磁盘的速度和内存对数据仓库来说尤为重要，因为查询是数据密集型的，数仓的查询可能需要对多个表中的上百万条记录进行扫描或者聚集  
1. 磁盘问题  
    磁盘驱动器对数据仓库平台的性能、灵活性、扩展性有很大影响，对于典型的数仓环境来说，I/O子系统的带宽是限制因素，和OLTP的工作相反，在OLTP中存在着很多小事务，着意味着每秒钟的I/O数量是关键度量  
    高端驱动器一般都做成一个单独的计算机或者一个被称作存储区域网络（SAN）的磁盘存储器子系统来管理对磁盘的访问  
    这些磁盘驱动器系统速度快、易扩展且可移植，可通过建立冗余存储器来提高容错性能（RAID1到RAID5)，进而使数据仓库具有更高的可用性  
    也可以通过分带RAID0将读/写工作分配给多个磁盘驱动器和控制器执行，以提高效率 5    
    由于磁盘驱动器是比较容易出故障的部件，因此具有容错功能的磁盘系统允许进行热交换功能是十分重要的  
    注意数据库需要较大的临时存储空间，主要用来进行排序、联接和聚集等操作，一般需要由高性能磁盘和磁盘控制器提供这些空间，但是并不需要保存临时空间的镜像——节省成本  
    磁盘启动器子系统比较昂贵，但它们是物有所值的，刚开始只要满足一两年存储需求，等到价格降低时再适时依据需求进行扩展  
2. 内存  
    就内存来说，DW/BI系统的内存当然越大越好，这是商业智能和事务处理之间的另一个区别，因为BI查询的规模大得多，需要对较大的数据表进行多次遍历  
    DW系统中所有主要组件都对内存有较高的要求，对64位平台就是很大优势，因为64位平台支持更大内存，最好将主服务器设计为64位的 5  
    使内存与磁盘工作相协调的观念已经有数十年了，但是速度差别问题仍然没有得到解决，内存要比磁盘驱动器的速度快100倍，尽管如此，我们仍然可以通过简单地位数据库增加大量内存将数据仓库性能再提高10到30倍  
3. CPU  
    一般来说CPU的数量越多越好，速度越快越好  
4. 辅助存储器  
    在系统配置中还要确保有支持备份和存档的设备，如果可能，尽可能配置一个速度快的备份系统，以便在引导过程中完成备份  
    从经济角度来看，进行磁盘到磁盘的备份是比较合算的  

### 4.6.5 数据库平台因素
选择什么样的产品应该依据实际情况的不同而有不同的考虑  
对大多数DW/BI系统来说，除非您的数据量很大，否则机构对数据库的选择主要应考虑功能的完备程度和使用许可方面的因素  
数据库普通的选择过程中要考虑的一个主要因素是聚集管理和导航问题  
1. 关系引擎的特征  
    关系数据库的优点在于能够存储大量的细节数据
    如果数据仓库规模比较大，数据容量在几十TB左右，那么最好在购买前做一些研究工作 5  
2. OLAP引擎特征  
    多维数据库管理系统，也称作OLAP引擎，是为了分析工作提供专门支持而设计的专有数据库管理系统  
    OLAP主要有两种用途，通过预先对数据进行聚集来改进查询性能，以及通过更加面向分析的查询语言来支持复杂分析  
    OLAP能够处理更多更大型的维度和更多的数据量  
    OLAP技术的进步之一就是一些主要的产品是真正按照维度组织数据的，关系维度可以直接一对一地转换到OLAP层  
    主流OLAP引擎是关系普通中卓越地组件，能够进行聚集管理和导航，并提供增强地分析功能 5  
    如果在关系数据库中以及有了聚集管理和导航，那么BI工具就可以直接支持对数据库的复杂分析，也不再需要OLAP引擎  
    对多维度引擎的评估不能脱离对BI数据访问工具的评估而单独进行，将OLAP引擎引入架构会限制BI工具的选择  
    DW/BI团队需要认真评估产品的规模限制和功能，因为基于桌面系统的轻量级OLAP解决方案只是短期有效，最终可能会带来更多要做的工作  

### 4.6.6 前台底层设施要素  
前台的底层设施需求和后台有很大差异，因为前台的底层设施更依赖于业务和工具  
1. 应用程序服务器的考虑事项  
    前台的服务器数量在使用过程中会急剧增加，用于处理来自Web的数据访问请求、查询管理、企业报表生成、认证、元数据数据库等，最好的方法是向厂商询问详细的配置信息并且从internet上收集用户反馈回来的问题： 5    
    + 内存，系统需要配置多大内存，配置多大的内存使系统的性能最好  
    + 磁盘存储器，使生命影响磁盘使用的情况，一般需要多大的磁盘，磁盘的扩容速度一般有多快  
    + 平台共享，能够在同样的硬件普通上运行多种服务吗？影响普通性能的因素是什么？是不是有些产品的兼容性很差
    + 瓶颈，造成系统不稳定的因素有哪些？是什么让系统的速度慢下来？系统真的是多线程的吗？系统能否产生独立的过程并且让它们开始运行直到完成？  
2. 用户桌面考虑事项  
    这里给出一些和桌面电脑相关的问题以供参考： 5  
    + 跨平台支持  
        支持多个桌面普通意味着前端团队要做更多的工作，这种对多桌面普通支持的需求会简化数据访问工具的选择过程  
    + 桌面操作系统和软件  
        即使每个人都使用同样的桌面硬件普通，仍然可能无法支持客户端软件，业务还要考虑桌面操作系统的版本是否正确，是否按照当前版本的办公应用程序套件，浏览器版本是否已经陈旧等因素  
    + 内存，现在大家都已经知道内存能够极大地影响桌面电脑的性能，如果花时间和精力来研究网络环境中的情况，就会发现性能瓶颈在于进出桌面电脑虚拟内存的分页数据和程序  
3. 连通性和组网因素  
    连通和组网将后台和前台连接到了一期，连通性是底层设施应该具有的基本功能，它是使用任何类型客户/服务器应用程序的前提条件，连通性问题主要涉及以下几点：  
    + 带宽，如果拥有带有高速局域网的源系统，对定位数据库和应用程序服务器很有帮助，对数据装载和查询性能也是如此 5  
    + 远程访问  
        如果用户位于远程，他们显然也需要和本地用户一样访问数据仓库，并享用相同的功能，意味着在数据库和应用服务器所在的局域网以及远程用户所在的局域网之间必须要有可能的、高带宽的连接  
        一旦对报表需求有所了解，就可以同网络团队沟通，以确定连接带宽是否足够，并了解何时能够进行远程访问  
    + 文件传输  
        远程访问时可能会遇到很多文件传输鞋业，并且有专门实现文件传输的程序，首要的是FTP这种通用传输工具，其次SSL可以对通信过程进行加密  
    + 数据库连通性  
        大多数数据库、ETL和BI工具厂商都会提供将其他厂商的数据库产品和遗留源系统连接到当前数据库的驱动程序，第三方中间件产品也可以提供这样的连通性，还可以将多个源的数据集成起来  
    + 目录服务  
        组网底层设施中需要某种形式的目录服务器来提供一个单独用于登陆和管理的入口点，这些目录可以用于罗列服务器位置、用于数据提交的用户目录以及分发标准报表所用的E-mail列表 5  

### 4.6.7 底层设施总结
DW/BI系统的底层设施中有大量组件，涵盖了硬件平台、数据库平台、连通性和组网，以及用户桌面等，大多数运行系统都有相似的底层设施，很多情况下DW/BI系统可以直接在已有的底层设施上构建  
底层设施为DW/BI系统的创建提供了基础，动力不足或性能不好的底层设施会削弱整个系统的功能  
## 4.8 元数据
元数据和其他底层设施组件相比更加无形，但是它为DW/BI系统的工具、用户和过程提供了相同种类的基础层  
大多数工具都创建和管理自己的元数据，并且将这些元数据存储在一个称作知识库的局部物理存储区域中，导致DW/BI系统中有多个元数据知识库，多个知识库的联合和内容重叠引起了很多问题  
### 4.8.1 元数据集成的价值  
为DW/BI系统元数据提供集成的知识库，在影响分析、审计、文档编制和元数据质量管理等方面很有价值  
1. 影响分析  
    集成的知识库能够帮助您确定某项变更对DW/BI系统的影响，如果所有元数据都存放在统一位置，或者至少通过公共键和ID连接，就很容量了解这一变更所产生的影响  
2. 审计和文档编制  
    沿袭和影响分析能够使用审计元数据来确定任何给定的事实行或维度行的来源，在有些涉及合规性的场合，它是必须的信息 5  
3. 元数据质量和管理  
    不同系统中保存的元数据元素的多个副本也不可避免的会出现不同步的情况  
    单一的知识库中只有一个对这些领域进行描述的入口，任何地方需要元数据都会到该数据库中进行查找，因此，这种情况下每个人都能够使用更新后的值  

### 4.8.2 元数据集成的供选方案
大多数DW/BI系统都是基于多个厂商的产品而建立起来的，在一个多厂商的环境中维护一个单一标准的知识库是不可能的，尽管如此，还有一些办法可实现某些层次上的元数据集成  
1. 单一来源的DW/BI系统厂商  
    主流数据库、ERP和BI工具厂商都在不断进去，建立起完备的、可靠的DW/BI系统技术群，进而提供端到端的BI，单一来源的DW/BI技术提供者能够最大程度地为您提供一个单独的、集成的元数据知识库    
2. 核心厂商产品  
    在一个大型机构中将不得不在DW/BI系统中使用多个厂商的产品，应该仔细看看核心厂商提供元数据管理和知识库工具 5  
    如果不行，就需要使用元数据管理系统作为中央元数据知识库，为大多数主流DW/BI工具的元数据知识库提供读写模块  
3. 自动动手实现  
    只对某种元数据要素进行重点管理而让其他元数据要素自行进行管理并非是不切实际的想法  
    作为DW/BI系统的创建者和管理者，您的职责就是要确保技术元数据是正确的，但不管用什么方法，对业务元数据也要进行管理  

### 3.8.3 元数据总结
元数据遍布于DW/BI系统的每个角落，在这里给出一些统一的原则：  
+ 对所有元数据编制目录，这样就可以知道自己有什么 5
+ 用标准软件库形式管理元数据:版本控制、版本迁移和可靠的备份和恢复
+ 为元数据编制文档，形成关于元数据的元数据
+ 对元数据资源的价值进行评估

## 4.9 安全性
DW/BI系统必须向那些需要使用数据的用户发布数据，同时还要保护数据，系统的安全性主要涉及识别合法用户，并且给予合法用户特定的权限以让他们访问一部分数据，而非每个人都访问所有数据  
数据安全令人担忧，而且很难预见到，建议在DW/BI团队中选择某人担任安全经理  
### 4.9.1 安全方面的弱点
安全性问题涉及的面很大，而且无法进行评估  
系统弱点源于多种不同因素，不仅有来自黑客的微信，还有来自内部源系统的威胁  
1. 对物理资产的威胁  
    物理资产最主要的弱点包括可能被偷、故意或无意损坏、电磁干扰等方面的自然破坏、技术陈旧或遗漏，以及通过内部或外部的暴力行动劫持控制某种资产 5  
    为了降低这些风险因素造成的危害，所有服务器都应当按照到一个安全、限制访问的环境中，这种集中式方式统一也可以使对软件和信息资产的控制更加方便  
2. 对信息和软件资产的威胁  
    信息资产几乎包括除物理资产和软件资产意外的所有从外部获得的有价值的东西  
    软件资产也具有很多弱点，软件弱点包括目标代码或源代码的盗窃，通过黑客渗入，证书失效或者病毒等  
    信息资产的偷窃或者损伤会导致一系列麻烦，包括机密信息、敏感信息、商业计划或代码的泄露、金融或服务资产的失窃、隐私的丢失或者用户身份的失窃  
    恶意对手会通过待机攫取和挖掘无意中从广播泄露的信息、窃听、对数据存储资产的物理偷窃、冒名顶替、使用陷阱、或者干脆使用行贿、抢劫和敲诈等方式 5 
    为了保护信息资产，集中式的服务器设施应当提供核心信息资产保护功能，如能够保护局部媒介安全的备份和恢复功能  
    服务器软件应当在管理服务器的团队的掌控之中，一般都保存在机房，并且加了锁和密钥  
    机构必须制定恰当的信息安全规定，并且进行广泛的日常教育，还必须对系统的使用情况进行监视    
3. 对业务的持续威胁  
    还有一些弱点可能会影响到对业务任务的控制，包括：
    + 拒绝服务攻击，这种情况下信息系统会因为一个或多个组件受到攻击而被阻塞或被迫停止 5  
    + 无法重构一致的软件快照，由于有预谋的破坏、偶然性的丢失或者麻痹大意等情况对系统造成的破坏，往往无法重建系统的一致性快照  
    + 恐怖主义，拒绝服务式攻击中最严重的一种是攻击者试图将主要的公共基础设施瘫痪掉，由于DW/BI系统承担着很多经验活动，因此它必将成为吸引恐怖分子进行恐怖活动的目标  
    
   对于上述的每一种攻击都有专门的防御方法，拒绝服务攻击可以在网络层处理  
   一致性快照是详细备份策略的一部分，包括文件系统对象和元数据存储，除了正常的数据库备份之外，还可能包括系统镜像  
4. 网络威胁  
    网络威胁的种类只取决于那些想要攻破网络的人员的创新能力，并不仅仅是黑客，可能攻击的还有不满的雇员、顾客或者其他业余黑客、搜寻利益的职业罪犯、竞争对手、信息经纪人、以及想要证明自己技能的系统高手 5  
    普通类型的威胁包括：    
    + 拒绝服务DOS攻击：使用大量未完成的连接请求对主机进行压倒性攻击，迫使主机一直向这些请求开放直到超时   
    + 病毒、蠕虫和特洛伊木马程序：能够在主机上与性能并且造成机器功能障碍的程序  
    + 劫持：通过窃听用户话路或程序来获得认证信息，或者通过向应用程序中插入代码来实现这一目的  
    + 欺骗：冒充用户或者一台计算机以获取对网络的访问权 
    + 网络仿冒：冒充一个合法的企业来获取用户ID、密码和其他信息 5  
    + 设置后门或陷门：这是一种不会留下文档记录的方式，用于获得对程序、服务或者计算机系统的访问权  
    
   组成机构通信主干的核心部件也针对网络威胁为您提供了基本防御机制  
   如果您担任了DW/BI安全管理员的角色，很可能将需要获得机构中IT安全专业人士的帮助  

### 4.9.2 安全性总结
安全性是一个巨大的、神秘的和复杂的主题，我们将讲述一种控制信息资产访问和认证的方法，完成这一安全工作是DW/BI系统安全管理员不可推卸的责任  

## 4.10 小结
当设计一个DW/BI系统时，会有很多问题需要考虑：
+ 后台架构，包括ETL过程、源系统和ETL数据存储
+ 呈现服务器层，包括原子数据和聚集数据，以及聚集导航功能
+ 前台架构，包括元数据、安全性、管理、企业报表、建模、规划和性能管理、Web和门户服务以及运营BI，此外还有BI应用程序有关查询和报表的功能  
+ 底层设施，包括硬件平台、数据库平台、组网和用户桌面等方面的问题 5
+ 元数据，包括对一个集成元数据知识库的价值的讨论 
+ 安全性，包括DW/BI系统几个主要的弱点  

您不需要对每一门技术都了如指掌就能创建一个可行的DW/BI技术架构

![5](../syn_pic/data_warehouse_life_cycle/5.png)  
创建架构计划的主要工作是确定对机构而言最重要的功能是什么，在深入了解业务需求之前就开始进行产品选择是一种错误的做法，很难取得成功  
事实上，90%的情况下产品选择都不是整个DW/BI系统成功与否的关键因素，如果能使用一般的产品解决高价值问题，才能体现出您的能力  
整个团队必须熟悉架构计划，明确他们在实现过程中所承担的具体角色 5  
## 5.1 创建架构  
如果您是一位IT系统设计师，至少对架构开发中较为严格的综合方法比较熟悉了，除非需要构建超大型DW/BI系统，否则直接应用企业IT架构方法一般都会超出您的需求  
用于设计架构的方法主要基于80-20规则，用20%工作量获取80%收益，这种方法也是一个迭代过程，需要将整个设计方案研究成熟，确定好各部分的优先级，并且将整个实现过程分成若干阶段  
### 5.1.1 架构开发过程  
创建架构是一个自上而下的过程，首先要了解业务需求，从整体视角出发建立一个描述解决方案的高层模型，随后将方案分解成多个子系统和子系统描述，接下来选择产品并进入系统实现阶段  
如图5-1顶部，业务需求为应用程序架构确定了边界和基线，前台和后台的设计是并行推进的，它们都有各自的一系列应用程序需求

创建架构的方法很多，使用哪种方法取决于DW/BI系统的整体规模和范围，以及架构设计想要达到的细节水平，图5-3给出了架构设计工作中的主要驱动因素矩阵

### 5.1.2 设计应用程序架构计划
DW/BI应用程序架构设计过程可以分为8个步骤，包括最后创建架构计划文档  
1. 第一步:组建架构特别工作组  
    特别工作组是专门为了创建DW/BI架构而组件的一个小组，由DW/BI技术设计师领导，最好尽早任用一位有经验的DW/BI设计师以开始架构的设计工作 5    
    架构特别工作小组中应当既有具有后台工作经验的人员，又有具有前台工作经验的人员，小组应当首先对自己的人物、期限、角色和提交内容进行严谨和讨论  
2. 第二步:收集架构相关的需求  
    架构设计的大多数要求是基于业务需求的，提前和主要访谈人确定访谈计划是一种很好的做法，如果需要更多信息，还可以随后进行一些跟踪访谈  
    此外还需要了解已有的技术环境，明确自己可使用的各种要素以及现有技术环境的局限和强制边界，图5-4给出了针对IT人员样例问题集，而不是针对业务用户的  
    ![5-4](../syn_pic/data_warehouse_life_cycle/5-4.png)  
3. 第三步:创建一个架构含义文档草稿  
    这种文档罗列了具有架构含义的业务需求，并且列出了各种含义，主要是为了寻找那些能够标识需要某种功能的业务需求或服务，要实现上述目标，比较好的做法是使用如图5-5所示的表格形式来进行工作 5  
    该图在第一列列出了需求和时机，随后第二列中给出了每个时机的功能含义，一般每个时机都有好几种含义，如有必要还可以加入业务优先级或者业务等级方面的信息  
    ![5-5](../syn_pic/data_warehouse_life_cycle/5-5.png)  
4. 第四步:创建架构模型  
    这一步的目的是借助一个安静的场所，让特别工作组成员构建出DW/BI架构模型，该模型和图4-1类似，但也可以定制  
    对图5-5中“架构/功能含义”列中的所有内容都进行考虑，并且将它们安排在架构模型中适当的位置上  
    在此可以先看一下图5-7的例子，了解模型是什么样子。您的模型基于您自己的业务需求，可以没有图5-7所示的某些具体要素 5  
5. 第五步:确定架构实现的各个阶段  
    一般来说，架构的实现阶段应当和生命周期的迭代过程相对应，必须为架构每个阶段都明确相应的功能需求和提交事项   
    首先要审查架构主要组件的优先级，并且估计实现它们所需的人员和资金，回顾第四步，看看要满足最高优先级的业务需求需要哪些功能支持  
    最好将这些最高优先级的业务需求用一种颜色表示出来，将下一优先级的业务需求用另一种不同的颜色表示出来，能够清楚地将二者区别开来即可  
6. 第六步:设计和描述子系统  
    功能一般根据机构的特定和内部数据系统的情况会有所不同，因此您需要将每个子系统都定义得足够详细，这样可以请人来实现这些子系统或者选择符合需求的产品  
7. 第七步:创建应用程序架构计划文档  
    现在，可以将前面每个阶段的成果综合成一个主计划文档，应用程序架构计划描述了DW/BI系统的形式和功能，即描述了初始阶段的情况，也对其预期的整个时间跨度进行了描述 5  
    架构计划一般都有50多页，对需要的各种组件、功能、业务规则，以及恰当的定义、优先级、人员和实现时间安排等都进行了详细描述  
    制定这份文档是系统IT机构承担各种角色的人都能够阅读它，要尽量通俗易懂，不要奢望每个阅读者都有足够的专业知识  
    文档要以DW/BI业务需求概述开头，重点讲清楚这些需求是如何对架构施加影响的，还要阐述用户类型和不同的访问需求，并对重要的数据问题加以讨论  
    架构计划一般不包括具体的产品，指定一份好的架构计划十分重要，有了它，就可以进一步创建各种模块部件，并对如何将新产品和新服务集成到现有框架中进行规划和设计  
8. 第八步:评审草案  
    这一阶段，需要和DW/BI团队成员一起召开一个设计评审会议，对当前的架构草案进行评审，并且期望对方案进行改进 5
    如果能够广泛听取意见并且保持方案的灵活性，就能够得到大量非常好的反馈信息和意见  
    将这些反馈信息加入到最终的架构计划文档中，并随后向主要管理人员和干系人介绍相关情况，与他们一同审查并敲定架构计划

    本示例包括一个有关架构计划文档内容的表（如图5-6所示），以及一个应用程序架构模型 5
    ![5-6a](../syn_pic/data_warehouse_life_cycle/5-6a.png)  
    ![5-6b](../syn_pic/data_warehouse_life_cycle/5-6b.png)

## 5.2 选择产品
一旦DW/BI架构就位，需要借助两个关键需求来驱动产品选择过程：业务需求和功能需求  
### 5.2.1 保留一个业务关注点
在开始产品选择之前，一定要牢记业务需求驱动整个评估过程，在了解业务需求和优先级前选择产品是错误的  
技术产品评估的时机选择是非常重要的，如果没有对需求有所了解就过早地去做这些工作，那么测试产品性能的时候就无法建立它们相对应的度量，但如果过晚地去做这项工作，就会拖项目后退  
### 5.2.2 主要DW/BI评估领域 
一般DW/BI系统有四个主要的领域需要在购买时加以注意，每个领域都会涉及到技术评估过程中的很多工具  
+ 硬件平台，包括用于ETL系统地不同服务器/平台、呈现服务器和应用程序服务器等 5  
+ DBMS平台，大多数DW/BI系统中，原子层数据仓库地DBMS是一个关系引擎，OLAP数据库也是一个DBMS平台，对它的评估和关系引擎密切相关  
+ ETL工具，单机版的是DW/BI系统的主要成本，ETL工具的选择也需要尽早进行的一项工作，且知道维度模型构建完毕后必须完成  
+ BI工具，大多数DW/BI系统有多种工具，可以支持不同形式的数据访问和BI应用程序运行  

至少在硬件平台和数据库平台层面上所做出的的选择大体上是一致的，对大多数小到中型的数据仓库来说，主流BI工具厂商和数据库厂商都有比较合适的平台  
如果您希望自己的数据仓库很大，就必须认真评估所有领域 5  
### 5.2.3 评估供选方案并挑选产品
在评估过程中有一些基本的步骤是通用的  
如果需要一个彻底而富有意义的产品评估过程，那么就必须为这项工作分配足够的人员，而且时间安排也应当比较合理  
现在我们把注意力放到产品选择过程的8个步骤上  
1. 第一步：了解购买过程  
    首先需要做的事情是了解公司里高价商品是如何购买的。您可能需要提供附有成本合理性证明和投资回报率(roi)预测结构的详细业务计划  
    购买过程是生命周期中需要重点关注业务价值收益的多个环节之一，如果在花费上遇到了巨大阻力就说明您还没识别出一个高价值的业务时机 5  
2. 第二步：设计产品评估矩阵  
    从您的架构计划的细节部分直接为正在评估的领域抽取所需功能的清单，把它们罗列在评估矩阵的左边  
    当进行市场调查时，会了解到更多先前没有意识到的有关这些功能的情况，这个清单可能还会增加  
    所有功能性产品的需求都应当进行优先级排序，至少需求上应当标上'必需'或'最好有'等标记  
    优先级排序可以使您为不同的需求分配不同的权重因子，我们使用0-100的值来衡量

3. 第三步：进行市场调查  
    市场调查的目的就是使您变成一个消息更为灵通的顾客，产品的基本信息可以借助下述方法来收集  
    + 在Internet上搜索，查找产品信息、博客、讨论小组、论坛和用户集团，初级形式的调查能够帮助您确定候选产品，但不可避免地会带有某种倾向性  
    + 人际关系网，参加DW/BI课程、会议和研讨班，在班上寻找那些正在使用您所评估的工具的人，向他们咨询经验，并且获取他们的联系方式 5  
    + 阅读商业性出版物，有时也会有一些比较好的产品评估文章，但多数都是促销传单  
    + 从行业分析员那里咨询意见，大多数IT公司都和一个或者多个从事信息技术咨询的公司签订合同，有一些也做产品对比，但也会从厂商那里得到回扣  
    
   您对上述信息进行收集和筛选后，将了解到的信息填写到产品评估矩阵中  
   评估这些产品时，也需要评估销售这些产品的厂商，在此要考虑到市场占有率的问题，工具安装所需要的基础部件越多，就越需要在出现问题的时候有人帮您解决  
   + 厂商支持  
       他们提供了哪些可以帮助您成功的资源？大多数公司都在其网站上发布了自己的文档资料、支持知识库和教程，详细内容如下所示： 5  
       + 文档资料，这些材料应当可以在线获得，并且以某种便于查找和浏览的形式提供 
       + 培训，公司应当提供定制的培训或者材料以便于学员自学 
       + 技术支持，您可以向相关人员咨询他们对厂商技术支持服务方面的经验  
       + 咨询，公司应当有一个由多名顾问组成的机构，该机构中的顾问都是工具方面的专家，这些顾问也应当准备一份本领域咨询机构的清单  
   + 外部支持  
       能够提供外部支持的有在线技术支持论坛、鉴定过这种产品的本地顾问以及使用过该产品的本地用户集团 5  
   + 与厂商的关系  
       首先其业务代表应当是理性的、通情达理的人，该公司的产品价格应当可以变动，能够提供常规的升级服务，该公司应当经营文档，发展前景良好，此外还要考察该公司收入、盈利能力、成长前景和产品发布历史     
4. 第四步：缩小选择范围形成简明的清单  
    当您已经依据选择标准对每种产品的情况都进行了考察之后，得分能够很明确地显示出哪个好哪个差  
    这种情况下，首要的事情是再次公正地评价它们的优先级权重因子，一定要注意，不要因为您喜欢的产品得分较低而有意改变权重因子  
    上述步骤的目的是选择少数几个产品进行更为详细的评估，市场调查结果会帮助您尽快缩小产品选择范围，只留下那些符合业务、技术和组织标准的产品  
    将产品从清单上删除时，要使主要干系人参与决策过程 5  
    应当听取干系人的意见，能够确保我们的评估确实针对可进行比较的产品，如果您认为自己还没有足够能力做出决断，你们就去寻求更多的专业人士帮助  
    __为什么咨询信息总是有问题__  
    产品选择中有一种古老的方法是创建一个咨询征求说明或者建议征求说明，这种方法将您认为重要的问题和特性编制成文档  
    一般这类响应并不会比那些具有较好市场调查能力并能够提出建议的咨询机构好多少  
    如果您所在的机构采用的标准运作方式需要用到这类RFI，那么可以为清单中每个厂商都提供同样的商业报告，并让他们给出解决方案 5  
    该商业报告应当简短，但其中要包含精心定义的关键需求示例，可以要求每个厂商给出一个解决该业务问题的方案演示  
5. 第五步：评估候选对象  
    当准备好经过缩减的清单后，就需要更详细地对这些候选对象进行评估了，主要有两种方法：深入研究和亲自体验评估  
    1. 深入研究  
        和每个候选厂商进行会谈，要给他们一个关于需求、注意事项以及您想提前完成的事项的清单，按照下述要求来做：  
        1. 倾听厂商的介绍，确保他们能够解决您所提出的问题 
        2. 提出一些有针对性的问题，包括他们在您需求中遗漏的问题 5  
        3. 让每位竞争者阐述他们如何才能提交关键功能  
        4. 要为详细的现场演示留够时间，并且询问他们如何解决具体问题，例如如何实现ETL系统中类型2变化跟踪，对于功能清单中的高优事项，可以问问厂商要实现它们是否需要一些其他的额外工作    
        5. 在会议中将评估矩阵作为指导，当相关问题得以解决时，及时填写评估矩阵  
        
       在和多为相互竞争的厂商会谈时您会了解到很多新的信息，因此您也可能会和每个厂商进行多次会谈，便于从他们的竞争者那里了解到的信息进行追踪调查  
       如果您评估的是BI工具，那么最好让那些将要使用这些工具的关键业务人员也参与会谈 5  
       __调查参考客户__  
       调查参考客户是非常有价值的，因为只有从产品的实际用户那里才能知道这些产品在现实中表现如何，对于厂商推荐的参考客户所提供的所有信息都要持怀疑态度  
   2. 亲自体验评估  
       很多情况下，只能通过直接体验产品才能获得自己所需的信息  
       如果已经将选择范围缩小到两家厂商，但是无法确定该选择哪一家，那么最好针对每种产品都开发一个原型系统  
   3. 产品比较原型系统（可选）  
       下面给出的指导原则可以帮助您完成一个比较原型系统：  
       + 定义原型系统范围，包括完成标准 5
       + 避免一次性涉及太多内容
       + 使用一个比较有针对性的小数据子集
       + 仅为BI工具的比较提供快速装载过程即可，并且可以直接使用脏数据
       + 在原型系统开发过程中，使用数据探查工具来尽可能多地了解源系统数据问题
       + 在原型系统开发过程中，尽可能多地了解软硬件性能，做笔记记录配置情况、设置情况和其他技巧 5  
       + 在原型系统开发过程中要尽可能多地了解支持需求
       + 在评估BI工具时要收集业务用户地反馈信息，不要仅将这些工具摆到他们桌上，过一两周再去问题他们感受，要和他们进行会谈，肩并肩地讨论
       + 查找参考用户并且打电话给他们
       + 不要让原型系统直接进入生产阶段
       + 要紧紧抓住期望，不要让期望失去控制 5
      
      最重要的时，在开发原型系统时要真实地对待您所看到和学到的东西，进行客观的分析会为后续工作节省大量的时间和资金  
    4. 评估捷径  
        一般很少对DW/BI项目进行全部四个关键领域的详细评估。通常，会有下面的某个因素介入评估过程，使评估过程变得无关紧要，例如：  
        + 已有IT产品标准
        + 已有资源，通常使指空闲的机器或者软件许可证  
        + 团队经验，可以依据已有的经验进行指导，这种已知的内部经验对团队来说非常宝贵 5  
        + 行政因素，听起来难以理解但最终又成为实际决策的方案往往都是基于行政因素  
        
       虽然此时的产品选择并非注定是失败的，但需要您进行差距分析，按照这个强制使用地系统重新审查您地需求，立刻报告由于这个决策造成地范围缩减或功能减少  
       这种折衷的选择应当是业务上决策，向业务管理层提交供选方案，并且和他们一起做出决定  
6. 第六步：推荐产品  
    应当请那些受产品选择影响较为显著的人员参加决策制定过程，其中对BI工具来说，涉及到的人员相当多  
    在完成评估矩阵、调查结果、参考用户查证、厂商信息搜集和亲自体验评估反馈等步骤后要和产品选择涉及到的人员进行一次会谈，并达成共识 5  
    最后的产品推荐应当以报告的形式提交，简要说明评估过程中都了解到了哪些信息，以及团队进行推荐的理由  
    无论选择了什么样的产品，一定要记住DW/BI市场的变化非常快，应当在两到三年之后对自己的选择再评估一次  
7. 第七步：试用  
    要谨记一点，不深入到实现过程的细节就无法了解产品的所有缺陷，但真到了那个时候再发现问题则往往为时已晚  
    我们的建议正是厂商们痛恨的，一定要按照我们所讲述的评估过程一步步走下来，做出最好的选择，并且告诉厂商您需要90天或者更长的试用期  
    在此之后，就像已经购买了该产品一样开始进行系统实现，要注意记录在使用过程中所发现的所有问题，然后在90天后您就可以拿出一个重要问题清单和厂商进行商谈了 5  
    您应该在签订合同之前就和厂商协商如何解决这些问题，否则您可能会遇到一些非常严重的问题甚至可能不得不推翻自己的首选产品去尝试另一种  
8. 第八步：洽谈合同  
    一旦推荐书获得批准而您的产品试用期又即将结束，就要考虑合同桥谈的问题了  
    要密切关注维护、服务、支持和桥谈附加事项的成本，要确保对以下补充事项有足够的关注：  
    + 开发系统和测试系统的许可证，尤其是基于服务器的产品
    + 咨询服务
    + 培训课时
    + 培训材料和文档资料
    + 增强型支持水平 5
    
   洽谈合同时您身边最好有一位专家，该专家应当做过主流IT产品的采购，并且有良好的口碑  
   在谈判桌上您能够获得多少主动权直接取决于合同有多大，掌握好时机也可以为您赢得更多主动权    

### 5.2.4 后台和呈现服务器方面的考虑事项
在后台和呈现服务器领域，主要工作是对硬件、数据库和ETL工具的评估  
1. 硬件平台  
    如果您创建了一个小到中型的DW/BI系统，您的厂商应该可以按照所需数据规模、并发查询载荷量、查询工作量情况和服务层次给出恰当的参考DW/BI平台配置  
    此外，还应该想办法到那些和您打算使用的平台相类似的实际系统上做一些测试，确保该平台能够满足需求 5  
    如果正在进行深度测试，那么尽可能利用厂商的设备和资源，应当包括前台和后台的活动，还要进行压力测试，硬件测试一般都需要经过下述步骤：
    1. 基于架构、需求和数据库测试来确定配置，关键部件有：
        + 内存，数仓对内存需求里很大，64位架构可以支持非常大的内存配置  
        + CPU，至少配置双核的处理器，即使实现的是小型系统也应选择双核
        + 磁盘存储器 
        + 数据库优化器，验证数据库是否已经配置成能够支持数据仓库特性的状态 5
    2. 设计和执行用于收集测试结果的程序  
    3. 设计物理数据库(磁盘、索引和聚集）
    4. 装载测试数据
    5. 编码并且执行测试程序集(包含对ETL和查询情况的测试）,并对结果集进行处理
    6. 进行基线测试
    7. 调整
    8. 执行最终测试 5 
2. DBMS 平台  
    在您的架构中需要考虑两种类型的DBMS平台：关系引擎和OLAP引擎  
    任何住了关系数据库都足够用于装载和管理一个维度数据仓库，但涉及到查询问题时，关系引擎的差别就很大了，问题也开始变得复杂  
    OLAP引擎也是一种DBMS平台，但OLAP产品比关系引擎更多样化，OLAP产品在伸缩性、性能和易用性方面差别显著  
    如果计划将OLAP用作主要的用户访问呈现服务器，你们评估工作应该将关注焦点放在如何挑选最能满足您需求的OLAP服务器上  
    如果您计划使用仅包含关系引擎的架构，那么就必须认真关注所选关系引擎的特性，包括星型联接优化、位图索引、聚集 管理以及其他专门为数仓系统设计的特性，它们对查询性能的影响很大 5  
    如果系统规模非常大，那么RDBMS平台选择至关重要，因为您不能让OLAP承担所有的查询服务  
3. ETL工具  
    如果您没有ETL工具方面的使用经验，那么您会发现ETL工具的工作方式和其他大多数IT开发工具都有一些不同  
    ETL功能测试中最直接的出发点就是我们在第4章给出的34个子系统的清单  
    创建一个能够覆盖装载和转换过程的各种场景的测试集，也就是说要考虑比初始仓库集更多的内容  
    您的DW/BI项目可能还要考虑调度、源码管理、BUG跟踪、连通性和web服务器软件等方面的事项 5  

### 5.2.5 前台考虑事项
在前台也有很多工具供您选择，为了避免被形形色色的工具淹没，可以将关注焦点放在满足顶级业务优先级所需的工具和功能上  
所有的厂商都会为他们的解决方案做漂亮的演示，您的人物就是从他们的宣传中找出事实，然后从功能或者架构的角度来了解不同的工具是如何工作的  
选择BI工具是一个麻烦的过程，有些一流厂商完全在自己的代码基础上不断改进，创建出更为集成的、连贯的生产线  
谨慎对待推销员所说的没问题，要根据您自己的业务需求和数据实际来考察产品 
如果您还没有认识到数据访问需求涉及的面很广，你那么工具选择就变得更有挑战性了，一种工具是不太可能满足所有需求的 5  
大多数工具厂商都只擅长某个方面的功能，但是迫于市场竞争的压力，他们必须在自己的产品中集成更多（甚至所有）的功能  
在BI工具评估中不仅要询问这些工具是如何实现的，还要了解这些工具的规模如何、该工具需要多少管理员和，以及这一级别的管理员工作报酬是多少  
BI工具评估是几乎所有的数据仓库都需要经历的评估领域，尽管之前讲述的产品选择的一般过程仍然适用，但此针对BI工具还有一些细微差别需要予以说明和建议：
+ 需求一定要具体，编制一份工具应该支持的功能综合清单（例如今年/上一年、多个事实表的横向钻取），清单还应当包括如Web的数据访问、复杂格式编排和报表调度    
+ 对自己和团队都进行基础培训，这样厂商就不能再糊弄您，还要检索有关产品信息的文献，获取行业分析员就BI工具的对比报告，并且联系您所知道的所有人，问问他们是否有一些直接经验 5  
+ 审查数据访问工具的特性，确保自己没有忽略重要特性  
+ 确定候选产品并且按照一定优先级排序  
+ 设计一个能够描述合理场景的商业报告，要留意已有的BI应用程序说明文档，它们一般都包含了很多有度量、计算、布局和交付格式等方面的信息  
+ 安装和测试排名最靠前的候选对象，这个过程应当要求一些关键的超级用户参加，如果进行现场评估，就建立一个实验室，花费一定时间按照教程一步步学习和使用工具，创建一些报表样例    
+ 举行用户审查会议，用户的参与是非常宝贵的，了解他们的看法和获得他们对最终产品的选择和支持 5  
+ 紧密追踪参考用户，应该去寻找类似的业务问题、技术环境、数据和用户量的站点，同一个参考用户的系统不会在上述所有特征上都与您的系统相似，需要与多个站点进行沟通和访谈    
+ 收集和整理测试结果以及用户反馈信息
+ 当在自己的开发环境中选择和安装工具时，要做好应对意外情况的心理准备  

### 5.2.6 管理元数据
从本质上来看，第4章的元数据清单就是数据仓库的基本框架  
弄明白捕捉哪些元数据、从哪里捕捉、如何捕捉、如何集成、如何在数据仓库中使用、如何进行维护并保持同步等事项是关于元数据的主要工作 5  
### 5.2.7 任命元数据管理员
需要任命团队中的某位成员来担任元数据管理员的角色，因为如果没有人对此负责，那么相关问题就无法得以解决  
元数据管理员负责创建和实现元数据策略，理想的元数据管理员候选人应该是无所不知的  
如果有人被指定负责元数据，那么他就必须具有SQL和DBA的技能，应当懂得如何在开发环境中编程、如何创建和发布报表，而且还要深入了解业务  
### 5.2.8 创建元数据策略 
下面将要介绍的方法是一种较为折中的方法，折中方法在建立一个不考虑元数据管理的DW/BI系统和建立一个企业元数据系统这两种极端之间找到了平衡点  
制定元数据策略的工作关键就是不要做得过火了，下面是为元数据管理员提供的任务分配清单： 5  
1. 要知道系统中的元数据情况  
    通过调查来确定整个DW/BI系统中元数据的位置、格式和使用情况
    任选一种元数据工具来探查系统中的元数据情况，如果没有，就需要创建查询或者有计划地访问其他源系统，探究它们并且进行跟踪  
    为找到的元数据元素创建一个清单，说明它们在什么位置、来源是哪里、归谁所有、您是如何看待和改变它们的、哪里需要使用这些元数据元素以及如何使用它们  
    这份清单就是您的元数据目录，目录本身不是知识库，只是对集成知识库的一个逻辑映射  
2. 和数据管理员一起对DW/BI团队及关键业务用户进行培训，让他们认识到元数据和元数据策略的重要性，在刚创建元数据内容的时候让他们参与进来，分配有关元数据创建和更新的职责 5  
3. 确定或者定义需要捕捉和管理的元数据  
    这一阶段确定和定义的元数据只是第一步元数据目录的一部分，在系统实现过程中需要不断更新和发布元数据  
    您需要管理的元数据取决于很多因素:机构对于主动管理元数据的偏好、对DW/BI团队主动管理元数据的支持程度、可用于解决元数据问题的人员等  
4. 当您开始元数据管理工作的时候，要为每一种需要管理的元数据确定明确的位置和版本标识  
    这就是说要确定元数据在何处存放和编辑，有些存放在关系数据库，另一些元数据则可能存放在ETL或BI工具中  
5. 创建系统以捕捉哪些尚未被归类整理的业务元数据或过程元数据 5  
    在建立自己的元数据表时，尽量用到所有已经存在的可用元数据结构  
    您可能找到很多需要整理和存放的字段，数据模型中的注释字段或者较长的描述都是这样的例子   
    我们很多客户都创建单独的数据库，由业务人员来维护这些元数据表和所有增值内容表
    创建一个简单的web前台来让他们管理这些表的内容并不是一件多难的事情  
6. 创建程序或者工具以按照需求共享和同步元数据  
    首先应当能够将元数据从其主位置上复制到任何所需的子系统上，这就需要在在知识库中创建一个关系表以将多种工具知识库中的元数据字段关联起来 5  
    要确保将初始数据库中所有表、扩展属性和对象模型中的描述字段、源字段和业务名称字段都输入到前端工具中  
7. 设计和实现将业务元数据提交给用户社区的方法  
    通常需要为业务元数据创建一个简单的元数据知识库，并且为用户提供某种浏览知识库的方式，以便于他们查找DW/BI系统中的有用信息  
8. 管理并且监视其使用情况和合规性  
    不仅要确保人们都知道系统有哪些信息以及如何使用它们，而且要确保元数据是完整的、最新的和正确的  
    虽然上述的是一种介于什么也不做和做得过的之间的平衡策略，但是仍然需要相当大的工作量 5

## 5.3 保护系统安全
当产品定下来之后，就可以设计和实现DW/BI安全策略了，本节所讲述的工具和方法可以帮助您避开第4章所指出的系统弱点，确立一个既能使用户访问所需数据又不至于完全开放的系统  
DW/BI系统中管理安全性的第一步是任命一位安全管理员,他必须设计和编制一份安全计划，并且和管理人员一起基于新的安全策略购买相关产品  
安全管理员应当参与架构设计，验证DW/BI系统的实际安装和使用，很多机构都需要在部署过程中由安全管理员发布一个强制性通告  
定义和实现安全措施的任务贯穿生命周期的架构、数据和BI应用这三条主线，还必须考虑到部署和维护安全措施的开销  
在为系统安装防盗门的时候，必须记住DW/BI系统的价值就在于人们可以访问和使用其中的信息 5  
当然，您需要确保只有授权用户才能访问DW/BI系统，并且对每个用户看到的数据都进行适当的限制  
### 5.3.1 保护硬件和操作系统的安全
在DW/BI系统中，访问重要信息最直接的方法是对支持系统运行的计算机进行物理访问，所有测试服务器和产品服务器都必须放置在一个小的数据中心里-上锁的房间  
发给我DW/BI系统的第二种最直接的方式是通过操作系统，应当对您的开发、测试和生产系统中的所有服务器都实现下述措施： 
+ 限制登陆访问，只有系统管理员能够登陆到服务器以运行DW/BI组件，其他人都只能通过网络来实现接入访问  
+ 限制网络接入
+ 确保数据目录的安全，包括数据库文件
+ 坚持及时为操作系统打上安全补丁 5

### 5.3.2 保护开发环境的安全
很多团队的安全标准都比较宽松，有的甚至根本就没有为管理开发环境制定的安全标准，尽管需要采用与测试系统和生产系统不同的标准，但对开发环境和服务器也应当进行专业的管理  
为了便于部署，可以使开发过程使用的机器在安全环境方面与生产系统相类似  
### 5.3.3 保护网络安全
Internet是20世纪60年代设计的一个巨大的面向UNIX的系统，Internet使用传输协议（TCP协议）/网际协议（IP协议）  
TCP协议确保了两台主机之间能够建立和维护一个连接，在连接过程中进行往复式会话  
IP协议负责告诉两台主机如何在Internet上找到对方，该协议可以看做是一个信封，含有目标地址和返回地址 5  
使用TCP/IP也可以实现所需的安全性和可靠性，这并不是使用外部方法胡乱修补一气就能够奏效的，而是通过确保每个网络部件都发挥其正常功能，并且通过加密保护“信封”的内容来实现的  
1. 网络部件  
    大多数机构的内部网都是用和Internet相同的TCP/IP协议，大多数网络都有一个包含路由器、防火墙和目录服务器的设备和功能集合  
    1. 路由器和防火墙  
        要将一个网络连接到另一个网络就必须使用路由器，这种设备负责在两个网络之间传输数据包  
        路由器可以检查每个数据包中的目标地址和源地址，所以可以作为包过滤器来使用，有下述基本功能:
        + 拒绝未知主机的连接请求 5  
        + 拒绝通过电子欺骗手段伪装成内部用户的外部请求
        + 仅接收来自某台机器的流量
        + 隔离污染并防止窃听  
        
       注意：包过滤防火墙是系统安全的基本要素，数据仓库技术架构设计人员和安全管理员应该制订一个具体的计划，将技术架构中的关键环节隔离起来加以保护  
       在一个连接到Internet的DW/BI环节中，屏蔽子网防火墙配置是一种很好的整体安全配置 5  
       屏蔽子网防火漆实际上就是两个包过滤器，第一个过滤器作为一个基本防火墙，将所有输入网络流量转发到一个堡垒服务器的特定内部主机上  
       除了已经获得访问资格的用户外，所有通信都必须经过运行于堡垒服务器上的代理服务器应用程序，包过滤防火墙能够谨慎选择外部主机，使之通过堡垒机访问内部服务器  
       如果将这种配置和目录服务器提供的有效验证机制以及加密系统配合使用，那么机构的安全策略就是比较优良的了  
    2. 目录服务器  
        目录服务器是一种数据仓库资源，可以从互联的网络上获取，这些资源包括数据库所在的机器、单个数据库、文档知识库、事务处理系统、文件存储区、打印机和人员等  
        目录服务器是有用的、集中式的，用于查找如何与互联的网络进行通信的控制资源 5  
        在目录服务器通信方面，很多厂商都遵守了LDAP标准，因此在实现目录服务器时也隐含了这一标准  
        对DW/BI安全管理员来说，目录服务器非常重要，它必须是数据仓库中所有用户验证和授权的唯一中心  
        目录服务器有以下几个特征：
        + 单点接入，所有访问都经过单一的控制点，这样做可以为一致性和完整性管理以及审计带来便利
        + 单点授权，不论连接到DW/BI资源的方式如何，每个人都必须按照相同的方式获得授权  
        + 单点管理，目录服务器是一个单一的管理控制器，访问规则就是在这里定义和强制执行的 5  
       DW/BI安全管理员需要和企业网络安全人员进行沟通，并且在目录服务器上实现适当的用户分组和用户分配，此外还要确保您的BI工具、数据库和其他服务能够从目录服务中得到验证和授权信息  
2. 加密  
    在前面的小结，我们努力去减少安全性方面的问题，这主要通过安装多道防火墙和将所有数据仓库接入请求都交给单一的目录服务器处理来实现的  
    除了通信传输机制，真正的保护必须要应用于通信的内容，加密是一种非常强大的技术，能够极大的保护通信的内容  
    对加密最广泛的理解是使用密码来对消息进行更改，这样就只有那些知道密码的人才能读懂消息了  
    DW/BI安全管理员一般需要了解六七种加密形式，但是没有任何一种加密方案能够解决所有的安全问题 5  
    如果确实需要对通信进行保护，就必须对内容进行保护，而不是对选址模式进行保护  
    作为DW/BI系统安全管理员，需要了解下述能够对通信进行某种层次加密的要素：
    + 私有和公共密钥加密
    + 可信证书授权
    + 签名的软件证书
    + 安全套接字协议层SSL
    + 虚拟专用网络VPN
    + Kerberos验证 4

### 5.3.4 用户验证  
大多数DW/BI系统都使用两部分（two-part）安全机制 5  
第一部分验证用户-确认用户身份，如果身份相符则允许其接入系统，系统管理员应当具有设置和管理验证部件的专业知识  
第二部分保护数据，这一工作场次都是由BI团队负责的，因为面对的是业务用户，涉及到数据库和数据访问工具的功能实现过程  
### 5.3.5 数据保护  
安全计划中最优意义的一部分内容：在保证用户能够查询数据的前提下保护数据，DW/BI系统安全需要考虑的主要事项就是查询  
1. 为内部用户提供开放式访问  
    在设计数据访问策略时，强烈建议应当为公司内部用户提供相对开放的访问，隐藏的信息越多，则以后DW/BI系统价值也就越打折扣  
    理想情况是将数据仓库中的敏感信息减到最少，但如果您必须要在DW/BI系统中存放一些敏感信息，就可以将其作为数据可用性一般规则的一个特例记录下来 5  
    为了实现自由访问策略，您必须：
    + 从执行发起人那里获取敏感数据元素的清单和访问方法
    + 按照规定条款开发系统，用户必须签名之后才能获得访问的权限
    + 发布详细的安全规定，包括敏感数据元素的清单
    + 监视使用情况 2  
2. 列举敏感数据  
    不管采用的思路是“大多数数据都是可用的”还是“大多数数据都是敏感的”，都必须为那些要保护的（或者可用的）数据设计一个矩阵，还要说明这些数据的归属和去向  
    ![5-9](../syn_pic/data_warehouse_life_cycle/5-9.png)  
3. 尽量屏蔽敏感数据  
    每当您列举出一个敏感数据元素时，应当自问一下，该元素是否确实是数据仓库所必需的 5  
    公民ID号是能够唯一确定一个人的标识符，但必须保密，您可以在ETL过程中生成一个没有含义的键来标识一个人  
    要严密控制对敏感信息的访问，例如职员工资和客户信用卡编号等，大多数数据库引擎都能在列的层次上对数据加密  
4. 保护数据访问  
    对数据的限制越多，则越难以提供对DW/BI系统的即席访问，因为需要在聚集层实现开放式访问，但同时在某个细节层次又限制访问  
    如果您允许即席访问底层数据，就需要为数据库添加复杂的访问规则，用户对部分限制数据的即席访问需求越多，则越需要使用高级的面向BI的呈现服务器，如OLAP数据库  
    大多数客户端访问工具都具有一些安全特性，在决定使用这些功能的时候，要仔细检查每种客户端工具的安全特性 5  
    很多BI工具都可以——甚至是必须——配置一个用于访问数据库服务器的管理ID，您必须特别注意不要泄露管理ID  
    大多数系统实现都有多种安全措施，除了报表，您还需要支撑即席访问  
1. 非受限报表
    非受限报表是一种预定义的报表，所有能够访问报表门户的人都可以访问这些报表，唯一的安全措施就是进入BI门户的验证措施  
2. 受限报表  
    只有机构中的一部分人员能够访问受限报表，也是预定义的，报表对每个有权访问它的人而言都是一样的，是一个很好的安全模型，但是对于最敏感的数据来说使用首先报表还不够  
3. 过滤报表  
    过滤报表是一种预定义报表，允许报表的人员不同，返回结果集也不同，一般也是受限的，并且适用于机构中的大多数敏感数据 5  
    为了实现过滤报表，需要使用某种方式将用户名连接到部门列表、列和其他用户有权查看的数据元素  
    最一般的方法是将源查询作为一个存储过程，将用户身份识别看作输入参数，或者在数据库中使用行级安全措施，此外还有一些报表工具能够在报表服务层上过滤报表  
    如果您只有少数过滤报表，就可以为每个部门都创建一个单独的报表，并且仅将合适的报表版本授权给访问用户，这种方式只是权宜之计或者应用于小型公司    
    过滤报表的开发和维护开销很大，需要开发一个安全底层设施，并且对每个用户有权访问的数据元素清单进行维护，可能还要编写一个简单的程序来辅助管理这些过程  
4. 即席访问  
    需要即席访问的用户显然需要登陆特权，在含有敏感信息的数据库中，就意味着要在数据库层进行授权 5  
    在关系数据库中，容易为表、试图以及表和试图中的列定义权限，但是定义航机安全性则需要创建额外的视图或者行级安全规定  
    这种方法的问题在于要保护细节数据的访问安全是比较困难的，不过它同时也提供了对聚集数据的自由访问  
5. 外部报表  
    外部报表是一个BI报表，可以为机构外部的人员所使用  
    满足标准外部报表安全需求最简单的方法是使用一个推模型-邮件形式，如果您的系统支持外部报表，要确保BI技术的产品评估矩阵中优先考虑数据驱动的订阅功能  
    数据驱动的订阅功能无法满足所有的外部访问需求，我们也将这种方式称作运营报表 5    
    外部用户一般只能访问过滤报表，使用前面讨论的方法就可以实现这一要求  
    有些公司常常也会为外部用户提供即席访问，一般来说这些公司都是信息提供商  

### 5.3.6 监视使用情况和保证合规性
即使采用门户开放策略，监视系统的使用情况也十分重要，可用于审计，每个DW/BI系统都必须监视下述事件：  
+ 用户连接、包括登录和退出系统的日期/时间
+ 失败的用户连接请求 5
+ 用户报表执行，包括报表名称、版本和用户提供的参数
+ 使用即席查询的活动，一般捕捉查询开始和结束的日期/时间
+ 进行的处理活动，如数据库事务处理
+ 开发活动，如发布新的报表

捕捉即席查询文本也非常有价值，这一信息对周期性性能调整十分重要，然而将查询文本存放在日志记录数据库中要比存放在其他所有组合日志中占用更多空间 5  
如果已经捕捉到了有关哪些用户使用系统和如何使用系统的信息，就应当生成和发布有关这种使用情况的报告，高级管理人员常常会发现使用情况报告是非常有意义和有价值的  
### 5.3.7 备份和恢复计划
可靠的备份和恢复系统是所有安全计划的关键要素，当灾难性故障发生时，您只有两种选择：
+ 重新抽取合适的源数据子集并重新执行ETL过程  
    大多数情况下是完全不可能的，因为数据已经被覆盖或者不再可用了  
+ 从系统备份中恢复丢失的数据  
    必须创建和实施一份备份计划，一开始要按类别列出呈现服务器和ETL领域的所有表，然后仔细检查并且估计重建这些表的文件所需的时间和工作量 5  
    
### 5.3.8 创建底层设施图
底层设施图描述了DW/BI系统的具体细节：服务器、网络和桌面系统，我们主要用它和IT操作人员进行沟通——这里所指的IT操作人员主要协助系统维护  
如果没有花时间来考虑底层设施，并且将这些想法和那些负责提交底层设施的人员进行沟通，那么由于一些不起眼的原因而导致项目延期的风险将增加  
确定底层设施需求时一个鸡和蛋的问题，您无法预先确定所有的底层设施需求  
图5-10显示了一个底层设施，和图5-7所示的架构模型相同，这里仍然以区域零售链为例  
![5-10](../syn_pic/data_warehouse_life_cycle/5-10.png) 5  
底层设施图最好按照一定的结构组织好规划，这样分组的需求就可以形成独立的一节，图中要涵盖五个主要领域、服务器硬件、磁盘子系统、网络、安全和桌面系统：  
+ 服务器硬件  
    项目到了这一步，应当重点关注底层并且认真考虑服务器硬件方面的问题，基于业务需求和初步的规模来确定主要的数据仓库平台（硬件和操作系统）  
    同样也应该考虑您的开发、测试和产品服务器，在这一领域该做哪些工作取决于DW/BI系统的规模和所需的服务水平  
    使用虚拟化的方式在操作方面有明显好处,这主要是指需要管理的硬件较少而且系统备份也比较容易  
+ 磁盘系统  
    在大多数IT环境中使用存储区域网络SAN是很普通的，即使在小型公司也很常见，其灵活性、管理和性能都要远远由于直连式磁盘 5   
+ 网络  
    网络中有两个部分需要考虑，在源系统、ETL区、呈现服务器之间需要有较大的容量通道，而且呈现服务器和用户桌面系统之间也要有比较大的吞吐量  
+ 安全  
    和安全人员会谈以确保他们明白您的计划，并能够帮助您理解这些安全措施的深层内容  
    您还需要了解机构中基于网络的验证和授权服务，以及如何通过它们定义用户、分组和DW/BI系统中需要的特性
+ 桌面系统  
    要确定DW/BI系统支持何种数据访问应用程序，指定项目支持的桌面操作系统和其他标准组件，确定您需要哪种本地组件  

由于数据仓库随着事件的推移不断演化，应当不断维护底层设施文档，使之保持最新状态，当团队中加入了新的成员时，该文档就更加有用了 5  

## 5.4 安装硬件和软件  
不同的软件产品其安装问题也有所不同，硬件安装和软件安装显然就更加不同了，要保证成功安装，最好的办法是在评估过程中加倍注意，并且做详细的笔记  
在安装过程中设置的参数对每个系统组件的性能都会有重要影响，下面几个主要步骤可以帮助您争取地解决问题：  
+ 预备场地  
    要保证有足够地空间来放置系统中的各种部件，包括硬件和软件，一定要和工作人员沟通，事先未系统找到相应的空间  
+ 培训  
    工作组最好对引入的部件有一定的使用经验，要在安装计划中考虑到培训问题，包括硬件和操作系统管理方面的培训  
+ 安装、检查和测试  
    很多产品都很容易安装，安装向导十分有用，很多情况下使用默认配置也不错，然而每种产品都有自己的特点，不知道它们将会带来不必要的麻烦 5    
    之前我们建议您在产品选择过程中进行评估和试验做详细的笔记，现在可以拿出来和负责管理服务器的工作人员一起研究这些笔记了  
    产品安装结束后，要对其进行测试，以确保没有明显问题  

## 5.5 小结 
创建一个DW/BI系统必须先进行架构设计
本章我们回顾了创建和实现DW/BI架构的过程  
创建一个DW/BI架构会面临巨大的困难，强烈建议查看本章最后给出的项目计划任务清单，它总结了创建架构计划、产品选择、元数据管理和安全措施实现等过程中的任务和工作  
## 5.6 管理工作和降低风险  
在架构计划的设计过程中由一些常见的高风险行为模式，需要予以关注：
+ 工作量不足-团队没有对架构进行周密的思考
+ 工作量过多-花费数月甚至一年多时间来追求架构计划
+ 独立工作-团队和机构其他部门隔离开来，必须让负责管理、控制和与DW/BI架构交互的兴趣方及早地参与进来  
+ 产品选择过程中的最大风险是对硬件、数据库和ETL工具进行测试（尤其是对大型系统来说更是如此），以及没有让将要使用BI工具的人员参与进来 5  
    次要风险是没有评估产品的功能是否能够满足自己的需求

元数据方面的风险是没有任命元数据管理员进行直接而实际的管理，对安全性来说也有类似的风险  
## 5.7 质量保证
让团队外的IT专家参与设计过程，是确保创建高质量架构的好方法,让这些设计架构人员和DW/BI团队一同参与增量审查，并且让他们在提交架构计划文档时进行完整审查也是很好的方法  
按照步骤进行彻底的产品选择对于取得好的结果十分重要  
IT和业务领导者需要强调捕捉和维护准确性的、描述性的元数据重要性，这能够激励参与这项工作的成员  
对产品进行可靠的测试，请一位自主的安全专家十分重要，能够确保安全措施高质量地实现  
## 5.8 关键角色
创建架构计划、进行产品选择、解决元数据和安全性问题这一系列工作中关键角色包括：
+ DW/BI技术架构师必须与承担创建架构核心任何的成员协作，需要源系统、服务器管理、网络和安全等领域的专家都参与进来  
+ DW/BI安全管理员必须在营造系统安全环境的过程中担当领导重任，确保安全标准的贯彻和维护，并以此来教育机构职员  
+ DW/BI元数据管理员面临的挑战是要将整个DW/BI系统中的元数据规划成一个复杂网络，为元数据处理创建全局策略，并且鼓励元数据所有者输入和维护元数据  
+ 待定的技术专家基于战略和战术计划实现对技术环境的变更 5
+ 产品选择过程需要技术专家参与，也需要那些团队中经常使用该产品的成员参与进来  
+ 安装过程也需要核心DW/BI团队和技术专家联手进行，技术专家拥有必备的技能，可以将每个部件都正确地安装  

## 5.9 关键提交内容
创建架构计划、选择产品、确立元数据和安全性策略的过程中，主要提交材料包括：
+ 架构计划
+ 产品评估指标和建议
+ 元数据策略
+ 元数据详细目录
+ 元数据捕捉和同步工具
+ 安全策略和数据敏感性
+ 安全监视和合规系统
+ 底层设施图

## 5.10 工作量估计
### 5.10.1 创建架构计划
如果已经确定了高优先级的业务需求，就应该能够在合理的时间里创建一份涵盖接下来2到3年的架构计划  
小型系统1-2星期，中等规模2-6星期，大型系统2-3月，这其中还不包括产品选择、元数据与安全性方面的工作量
### 5.10.2 选择产品
如果您经历了一个完整的产品选择过程，每个主要领域应该估计2个月工作量  
### 5.10.3 元数据
对中等规模的DW/BI系统来说 元数据目录整理需要1-2星期，较大系统则需要两倍以上的时间  
建立一个捕捉和管理主要元数据元素的系统需要2-4个星期，建立同步系统的工作也需要2-4周，运行过程中的元数据管理每周需要1天时间 5  
管理元数据工作量会迅速膨胀，要把精力集中在最重要的20%的元数据上   
### 5.10.4 安全性
估计建立一个安全环境需要花多长时间，必须考虑非常多的文化因素和网站特有的因素，这个过程常常要花费数月时间
