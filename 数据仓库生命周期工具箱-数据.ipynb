## 6.1 使用维度建模的场合
在深入研究设计维度模型的具体指导方针之前，我们将其与事务处理源系统中遇到的规范化模型先做个比较  
### 6.1.1 什么是维度建模
维度建模一种将数据结构化的逻辑设计方法，因此他们对业务用户来说很直观，并且能够提供较快的查询性能    
在数据库设计中个，维度建模具有不可抗拒的魅力，当设计人员将可理解性和性能作为最高目标时，维度方法总是首选  
维度建模将客观实际划分为度量(measurement)和上下文(context)  
度量是由机构的业务过程和支持它们的业务源系统来捕捉的，常常以数值形式出现，我们将它们称作事实 5  
事实由大量文本形式的上下文包围着，这些文本形式的上下文只有当事实被记录时才为真，这种上下文被直观地分割成多个独立的逻辑块，我们将其称作“维”  
机构的每一个业务过程都可以用维度模型来描述，维度模型由一系列含有数值型独立的事实表组成，而事实表中的数值型度量则被一系列带有文本形式上下文的维度表所环绕，如图6-1所示  
![6-1](../syn_pic/data_warehouse_life_cycle/6-1.png)  
关系数据库平台中存储的维度模型一般都被称作星型模式，存储在多维联机分析处理（OLAP）的结果中的维度模型被称为多维数据集  
### 6.1.2 怎样进行规范化建模
第三范式3NF建模和维度建模差异很大 5  
3NF建模是一种想法设法消除数据冗余的设计，数据被分成很多离散的实体，每一个实体在关系数据库中都对应一个数据表，这种方式下事务的加载和更新会比较简单和迅速  
因为规范化建模捕捉了数据元素之间的微观关系，所以即使简单的订单业务过程也会产生好几十个表，这些表通过令人困惑的、蜘蛛网式的联接关系相互联系到一起，一般来说，这些实体中的每一个都会转换成一个单独的物理表  
两种模型都有同样的信息和同样的数据关系，只不过是结构不同罢了 
业界有时将符合第三范式模型称作ER模型，第三范式模型和维度模型都可以用ER图的形式来描述，所以我们尽量避免将符合第三范式的模型称作ER模型，而是将其称作规范化模型来防止混淆  
尽管产生的表结构不同，无论使用哪一种方法，表都具有同样的关系数据库属性 5  
从规范化结构转换到维度模型的过程相对比较简单，维度表常常使用很多低基数描述符将符合第二范式的表组合到一起  
维度建模方法要比规范化建模方法出现得更早，在20世纪60年代，在由General Mills 和Dartmouth大学牵头的联合研究项目中就提出了一组含有“维度”和“事实”的词汇  
### 6.1.3 维度建模的好处
可理解性是维度模型得以广泛应用的一个主要原因，这也促使维度模型成为将提交到商业智能层面上的数据进行组织和结构化的最优方法，建模过程中要让所有的事项尽可能简单，但是有不能简单过了头  
查询性能是维度建模的第二个主要驱动因素
除了易于使用和查询性能之外，还有一个优势就是每个维都是一个等效的事实表入口点，这种对称性使维度模型能够承受查询模式中的意外变化 5  
那些产品具有优势的ETL和BI工具厂商早已将维度智能集成到自己的产品中了，如果预先建立ETL向导来辅助用户使用标准的维度功能（处理缓慢变化维）  
最后，对维度模型进行扩展也比较容易，这样可以容纳那些事先未预料到的新数据  
每隔一段时间，业界总会有一些人想要绕开有关维度建模和数据仓库数据库可行性的问题而直接去查询和分析业务数据  
厂商们声称自己的中间件能够屏蔽源系统复杂性，事实上，它并不能解决用户对性能有不同需求这一固有的问题，也无法跟踪维度属性随时间推移变化的情况  
## 6.2 维度建模入门
在简单介绍了维度建模的基本情况后，我们进一步来探究维度建模规定中的一些标准术语  
### 6.2.1 事实表
事实表存储了从机构业务活动或者事件中提炼出来的性能度量，术语“事实”其实就是指每种性能度量，事实的价值一般都是连续的，也就是说事实在相当广泛的范围内都具有有效值  
事实表一般都和业务过程度量事件相对应  
几乎每一个事实都是数值型的，大多数有用的事实既是数值型的又是具有可加性的  
然而以数值型数据表示的某一时点的强度度量，如库存量或者账户结余的事实就是半可加的 5  
有些数值型度量是完全不可加的，例如某个单位价格或者温度等，因为所有维度中的这里事实是独立的  
大多数文本型的上下文都是从一个离散的值清单中抽取出来的，这样的上下文应当存储到维度表中，而不是存储到某个文本事实中  
在重点关注一致性维度时，如果事实表的定义是相同的，则这些事实表也是一致的，一致的事实在不同的表中应有相同的标准名称，如果事实并不一致，那么必须为事实的不同解释赋予不同的名称  
并非所有的数值都存放在事实表，一个比较好的经验性法则是看看属性是否是离散取值的，以及是否用于过滤和标记，满足这些条件就应当放在维度表中，如果属性有许多值可选，并且可用于计算，则应该将其放入事实表  
事实表一般都很大，有几百万行甚至几十亿行的记录，但是使用事实表仍然有很高的效率，事实表通常存储维度模型中90%以上的数据，我们对于最小化事实表中的冗余数据比较谨慎 5  
1. 事实表键  
    事实表是由一个多方键标识的，许多方键是由来自同一业务过程中若干相交维度表的外键所组成的，事实表描述的是多对多关系    
    事实表中的每个外键必须和对应维度表中的唯一主键相匹配，事实表的外键应当不为空，为空则违背了参照完整性  
    事实表的主键一般是维度表外键的子集，有时事实表中唯一标识某一行的主键可能包含退化维  
2. 事实表粒度  
    事实表的粒度是产生事实行的度量事件的业务定义，粒度确定了事实表的主键，而粒度本身是以业务术语表达的，要预先检查是否存在可能出现的外键    
    事实表中的所有表都应当有统一的粒度，粒度就是您的锚点：维度设计中必须包含足够的维度上下文来实现粒度，而且您也必须将那些未被该粒度的维度上下文排除在外 5  
    事实表应当包含从业务过程中捕捉最底层的、最细节的原子粒度，粒度原子事实表不受变化影响，具有良好的扩充性能，可以通过添加最近获得的事实及维度属性进行扩充，也可以通过添加整个新维来实现扩充（假设粒度不会被新的维所改变）  
    提交的维度模型如果具有最细节的数据，那么就可能使模型具有最大的灵活性和可扩展性  
    
### 6.2.2 维度表
维度表不具有健壮性和完整性，它们当中充满了“大而笨重”的描述性字段
维度表的属性一般有两个主要用途:查询约束/过滤和标记查询结果集  
您可以通过“按照”(by)这个词来区分维度或其属性（如By年、By产品、By区域等），在请求查询或者报表时，业务人士经常使用这一的字眼 5  
维度属性一般都是文本字段，或者是作用和文本相同的离散数字，维度属性应当具有以下特点：
+ 是冗长的（名称由完整的词汇组成）
+ 描述性的  
+ 是完整的（没有遗漏值）2  
+ 是离散取值的（维度表中的每一行仅取一个值）  
+ 质量是有保证的（没有错误拼写、不可能的取值、舍弃的值和同一属性的不同版本） 

维度属性列中的空值常常会使业务用户感到困惑，因此推荐使用描述性字符串来代替空值，如“未知”、“不适用”等   
尽管维度表并不一定总有几十个属性，但是这种情况也是很常见的，可以使用代码和缩写形式作为维度属性，最明智的做法是使用相应的描述性字段 5  
存储为维度属性的操作代码常常具有嵌入式的含义，应当为每个嵌套于其中的部分都建立单独而详细的维度属性，这样它们就可以支持过滤和标记  
维度表几乎必然会用于表示层次关系，在一个单独的维里面解析几个多对一的层次关系的情况很常见，这也是反规范化的必然结果  
维度表由那些联系紧密的属性块组成，这些属性块分布描述主要业务对象，如产品、客户、职员和设备等，那你可以将它们看作是数据仓库的名词  
大多数维度模型只有8-15个维度表，储存事务处理细节的事实表构成了维度表的大部分维，而维度表还有少数维用于捕捉周期性的快照  
1. 维度表键  
    尽管事实表有一个多方键，但是维度行却是由单一键字段唯一标识的，建议维度表的主键应指定为从1开始的简单证书序列 5  
    大多数情况下一个4字节的证书可以表示大量的代理键，使用代理维度来取代运营自然键有以下好处：
    + 性能  
        代理键排列紧密而高效，紧凑的代理键带来较高的性能  
    + 运营键管理实践提供的缓冲  
        在运营系统中，数据常常会保留一个相对较短的周期，保存数据的能力很有效，运营键可能会在一个休眠周期后被重用或者重新指派，对于数仓来说就会出大问题  
    + 通过映射集成不同数据源  
        在一些运行环境中，同样的实体被不同的源系统赋予了不同的自然键  
    + 处理未知或者不适用的条件  
        事实表的外键不能为空，使用代理键代替运营自然键可以使我们很容易将某个代理键值分配给那些描述这里空条件的行 5
    + 跟踪维度属性值的变化  
        捕捉维度变化最长久和最合规的一种方法是创建一个新的维度行以表达新的属性特征，并且指派一个可用的代理键作为其主键  
    
   使用代理键的代价，指派和管理维度表的代理键，以及适当地使用代理键替代事实行中的运营自然键都给ETL系统带来了一定的负担  
2. 一致性维度  
    规范化的一致性维度是所有精心设计的数据仓库都在追求的目标，有时也将其称为主维度或者公共参照维度，为整个企业数据仓库环境所共享，其中加入了代表了不同业务过程的多个事实表  
    一致性维度有两种情况，最基本的情形是，如果两个一致性维度是完全相同的，那么它们有相同的键、属性名、属性定义和域值，不管它们联接了什么样的事实表  
    当一个维度是另一个更为详细的粒度维度表的完全子集时，这两个维度也是一致的 5  
    ![6-2](../syn_pic/data_warehouse_life_cycle/6-2.png)
    如图6-2所示，虽然整个详细产品维和缩小的品牌维并不完全相同，但是两个维度中的公共属性应当在每个维度表中都是用相同的标识、定义和取值  
    注意：缩小的一致性维度表是为了描述事实表，这些事实表既可以较为容易地捕捉较高细节层次的度量，又可以捕捉那些聚集成较小粒度的事实，这是视性能而定的  
    一致性维度对于DW/BI系统的企业极为重要，因为它们具有以下优点：  
    + 一致性，一致性维度确保了每个事实表都是用一致的过滤条件，并且产生的查询结果集合也被唯一地标识 5  
    + 集成，严格遵守一致性维度的做法使DW/BI环境形成了一个综合而完整的系统，一致性维度使查询可以横向钻取描述不同业务过程的事实表  
    + 缩减了开发周期，使系统得以快速上市，如果机构的某些投资用来定义和管理一致性维度，一旦这些维度建立起来，它们就能极大地缩减项目的开发时间，因为已经有了可用的公共维就不必再重复创建了  
    
   近来，很多行业都对主数据管理很感兴趣，一致性维度就是DW/BI环境中的描述性主数据  
   尽管试图使面向不同行业的部门在所有描述其业务的属性上达成共识无疑是在蛮干，但是即使在跨多种行业的综合性大公司内，一般也有少数核心产品或者客户属性应当在整个企业内部保持一致  

### 6.2.3 四步维度设计过程
在理解了主要词汇和核心概念之后，现在最重要的就是讲述我们是如何将那些凌乱的维度和事实拼凑到一起的，维度模型的逻辑设计是由四个主要决策所驱动的 5  
1. 第一步:选择业务过程  
    设计过程的第一步是确定要建模的业务过程或者度量事件，这一步一般都发生在同高级业务管理人员一起制定业务优先级的过程中  
2. 第二步:声明粒度  
    一旦业务过程被确定下来，设计团队就必须声明事实表的粒度，清楚地定义事实表的行到底代表什么在提出业务过程维度模型的过程中至关重要  
    在业务协商过程中定义事实表粒度时要非常仔细，不要跳过这一步，粒度就是从业务角度对”准确地来说,什么是事实表行”这一问题作出的回答    
    一般的，事实表粒度的选择应当尽可能是最小的原子单元，在最小粒度的数据之上设计出来的事实表是最健壮的设计  
3. 第三步:确定维度  
    一旦事实表的粒度已经稳固地确立下来，对维的选择就相当简单了，一般来说，粒度本身就能确定一个基本的或者最小的维度集合，设计过程就是在此基础上添加其他维 5    
4. 第四步:确定事实  
    四步设计过程的最后一步是仔细选择适用于业务过程的事实或指标，事实都必须是实际存在的，不要将那些和明确声明粒度不相匹配的其他时间段里的事实或者其他细节层次上的事实混杂进来

## 6.3 企业数据仓库总线架构
对DW/BI整体环境的构建进行规划是一项极为重要的活动，从历史角度来看，争议热点主要集中在两种都不怎么有效的方式上  
到底应该从建立一个集中式的、规划好的架构的角度为整个企业建立数据仓库，还是为每个具体的业务单位建立小型的独立解决方案  
面向部门的度量解决方案常常被称作数据集市，鉴于对这一术语的定义差异很大，理解也不一致，我们将这一术语去掉了  
### 6.3.1 规划危机
设计企业DW/BI数据架构的人物令人望而生畏，大型企业往往面临两个巨大的、看起来似乎并不相关的挑战  
一方面，项目群经理必须了解企业最复杂的资产-源数据有关内容，并且要知道这些源数据的具体位置 5  
另一方面，项目群经理必须知道哪些事项让管理人员夜里都睡不好觉，因为人们总是希望DW/BI系统提供的数据能够解决当前亟待解决的问题    
为了缓解这方面的压力，DW/BI项目群经理可以先将解决方案一小块部署到某个部门，并且尽量将其完善  
遗憾的是，在很多仅需要建立独立点解决方案而并非一定要建立一致的企业数据仓库的情况下，人们往往忽略应该设计一个框架将那些独立的系统集成起来形成一个耦合的整体  
无效有效地将孤立的、单独的数据烟囱式的解决方案集成到一起已经成为DW/BI系统的致命伤，这比错失数据分析的时机要严重得多，烟囱式解决方案的存在严重阻碍了集成式的企业数据仓库开发进程  
如果一次性创建数据仓库过于困难，需要将整个目标划分为若干独立小块予以实现，那么应该做哪些工作呢？ 5  
### 6.3.2 总线架构
要解决这样一个两难的问题，首先要迅速而简洁地定义整个企业DW/BI系统的数据架构  
企业数据仓库总线矩阵是DW/BI系统的一个总体数据架构  
通过总线矩阵开发数据架构是一种理性的方法，可以将企业数据仓库设计过程中那些令人生畏的工作分解开来  
一致性维度是企业数据仓库的总线，如图6-4所示

### 6.3.3 价值链的意义
很多业务都通过一系列业务过程步骤来监督内部活动逻辑流程  
在制造业，一个产品的流动过程是从获取原材料到加工、再到最终交付给客户的整个过程  
从原始起点到消费终点的流程被称作是”供应链管理“有一系列的运营系统和该链上的逻辑步骤相对应：  
+ 原材料购买
+ 原材料交货
+ 原材料库存
+ 材料账单
+ 生产制造 5
+ 将产品运输到仓库
+ 制成品库存
+ 客户订单
+ 为客户送货
+ 货品计价
+ 付款

### 6.3.4 通用矩阵的常见问题
当拟定数据仓库总线矩阵的时候，人们有时会为每个行和列所表达的细节参差而头痛，矩阵行的常见问题一般可以分为以下两类：
+ 涉及整个部门的行和包含内容过多的行  
    公司组织结构图中的方框标识职能单位，而非业务过程，矩阵行不应当和它们相对应  
+ 以报表为中心的行和定义过窄的行  
    这里问题与上述情况恰好相反，总线矩阵也不应当类似于被请求报表的细目清单  

在定义矩阵时，设计师很自然地会受到类似的限制，即定义的列既不能过宽也不能过窄，矩阵列的场景问题如下：  
+ 过于一般化的列  
    一个总线矩阵中person列可能会涉及到多种类型的人员，这些群体之间实际上是不重叠的，将它们都汇总到一个一般性的维中会造成混淆 5  
+ 为层次结构图中的每一个层次都分配单独的列，总线矩阵的列应当表示维的最细节层次 
    和创建独立的矩阵列（如产品、产品分组和营业范围等）不同，我们提倡在产品层次结构的每个层次上都建立一个单独的产品列  
    因为矩阵中阴影部分的矩阵单元显示出列和某个业务过程相交，所以可以在该单元格内指明细节层次（如果不是在最小粒度的层次上）

## 6.4 对维度的深入讨论

现在，我们主要关注的问题是如何识别和创建维度表的样式，到第9章再讨论如何创建这些表和属性  
### 6.4.1 日期和时间
在每个数据仓库中日期维都占有特殊的地位，这是因为几乎每个事实表都是一个观测值的时间序列，事实表总是有一个或者多个日期维  
图6-6显示了一个以天为粒度的标准日期维度表，该表抱哈了很多用于导航、过滤和描述日程的属性  
![6-6](../syn_pic/data_warehouse_life_cycle/6-6.png) 5  
1. 代理日期键  
    使用代理键这项规定对日期维来说也同样适用，使用SQL日期格式的字段作为日期维度表和事实表之间的联接键是错误的做法  
    对大多数维来说，代理键可能是毫无含义的，对于日期维，我们建议您按照时间顺序为代理键分配序号  
    设计人员有时希望使用带有更多含义的整数作为日期维度表的主键，通常是将整数按照YYYYMMDD的形式组织来代替代理键的序号  
    如果使用YYYYMMDD形式的键的动机是为了避开日期维并直接将其作为事实表中的一个可识别字段进行过滤，那么这种做法就是坏主意  
    如果使用YYYYMMDD形式的键是为了确立和维护基于日期的事实表分区，那么只要保证所有查询都仍然可以利用日期维度表，我们就可以将其称作为智能键 5  
2. 时刻(Time of Day)  
    对单个事务的精确跟踪可能会要求数据具有更明确的日期/时间戳含义，一般可以精确到分甚至是秒，日期分析常常和一天内的时间段分析区别很大，因此我们将它们在事实表中区分为两个不同的维    
    只有当一天的时间段由具有一定含义的文本描述时才能将时刻作为一个维，例如午饭时间、第三个班次等，也可以按照层次分组  
3. 日期/时间戳  
    如果业务中需要计算和分析两个特定日期/时间戳之间的差值,那么这些计算应当在ETL系统中执行,并且可以作为事实供用户进行简单而快捷的访问  
4. 多时区的情形   
    对于那些跨越时区的业务来说，可能需要用当地时间来表达日期和时刻，也可能同时使用全球标准时间来表达，例如世界协调时间（UTC)  
    可以通过为每个当地时区和UTC建立单独的日期和时间维来解决这一问题，如图6-7所示 5  
    ![6-7](../syn_pic/data_warehouse_life_cycle/6-7.png)  
    拥有两个明确的时区解释可能会非常有用  

### 6.4.2 退化维
很多维度设计都在绕着事务处理控制文档转圈，如订单、发票、提货单或者飞机票等  
这些编号正是设计父-子关系的事务处理系统时所用到的父标题键，不要忽略这些标题编号，而是将它们直接添加到事实表中  
标题票号可以作为维度键，但是如果像事务日期和客户维那样将标题记录信息分解成独立的维度，那么这个维度就没有其他属性可，我们将其称作退化维 5
退化维常常出现在事务事实表的设计过程中，这种表本身就具有一种父-子关系的数据结构  
退化维很常见也很有用，退化维的键应当是实际的事务编号，它本身就是事实而不需加入任何东西，退化键可以作为一条连接到运营系统的纽带  
### 6.4.3 缓慢变化维
根据定义来看，维度表属性变化相对来说并不频繁，总的来说，共有三种处理缓慢变化维（SCD）的基本方法，这三种方法可以配合使用形成更多混合方法  
注意：在一个单独的维度表中，可以根据业务需求的不同找出几种跟踪属性变化的方法  
1. 类型1：覆盖维度属性  
    使用类型1缓慢变化维解决方案时，一旦属性值发生变化，则仅使用最新的值来更新或者覆盖旧值 5  
    尽管类型1方法实施起来比较简单，但是类型1的属性变化会造成无法预料的复杂性  
    不要因为感觉到类型1方法执行起来很简单就立即将其作为默认的处理方式  
2. 类型2：添加一个新的维度行  
    缓慢变化维的第2种解决方案是准确跟踪维度属性值变化最常用和最有效的方法  
    使用类型2方法后，维度表中奖会插入一个带有新代理主键的行来反映新的属性值，如图6-8所示  
    类型2 方法在维度发生明显变化时使用，而且这种方式的优点在于可以通过改变属性来划分历史记录，每个代理键都对应维度行在某个时间跨度内的唯一版本 5  
    类型2方法需要使用一个代理键来唯一标识新的信息，一些涉及人员会建议您将生效日期和自然键连起来唯一标识变化的维度行，我们强力反对这种行为

3. 类型3：添加一个新的维度属性  
    类型3缓慢变化维度属性的处理方案相对来说并不常用  
    使用类型3方法时，应该向维度表添加一个新列，这种方式仅适用于少数这样的软变化 5  
    如果一个维度属性按照某种可预测的规律进行变化（如按照年度来变化），同时业务不仅要考虑历史值准确性和当前值情况，还需要基于属性的所有历史值来汇总事实，你们就可以在维中建立一系列类型3属性  
4. 微型维：添加一个新维  
    类型2是处理缓慢变化维最主要的方法，但它对于跟踪大型维度表（具有数百万行）中的属性变化是不适用的  
    遗憾的是，大型客户维似乎比中型的产品维更容易变化，这将您至于两难境地  
    解决困境的方法是将那些经常变化的客户属性分解开来放到独立的维度表中，这里将其称作微型维度，如图6-9所示  
    ![6-9_5](../syn_pic/data_warehouse_life_cycle/6-9.png)
    尽管微型维方法可以跟踪一个大型维度表中相对常见的变化，但是采用微型维也需要作出一些妥协：
    + 分带值，客户统计属性会集中到一个离散值的带状区域内，这就对数据的特征进行了限定    
    + 受限的增长，微型维自身无法增长太多，一条常用的经验是微型维含有的行不应当超过二十万，解决这一问题的方法是建立一个次级微型维
    + 与核心维度分离  
        微型维度在物理上与那些较稳定的客户描述符是分离开的，这使其浏览数据非常困难，仿佛是单独的维一样  
        如果需要，可以在客户维里为微型维属性的当前值增加类型1的列，这样不需要经过事实表就可以对客户信息库进行独立的统计分析 5   
    
   最后，这种方法仅在事实表行出现时才允许您将统计信息和客户信息关联起来，如果在上述过程中遗漏与统计信息的关联，可以使用一个无事实的事实表随时间推移不断捕捉客户维和微型维直接的关系，每当出现一个关系变化就装载一行  
5. 混合缓慢变化维方法  
    上述三种基本的缓慢变化维方法对于大多数情形都足够用了，然而，有时还需要将这三种方式组合起来使用  
    如果业务用户有时想要保留以前和某个事实相关联的准确维度属性，同时有希望能够基于当前维度特征来积累历史事实，那么这三种基本缓慢变化维都无法单独满足整个需求  
    一种解决方案是通过为基本维度表添加一个行来捕捉类型2的属性变化，此外，所有当前行和以前的类型2行都有一个“当前”属性被覆盖（类型1），如图6-10所示  
    不要将历史属性和当前属性放到同一个物理表中，当前属性可以加入到一个同维度自然键或者维度表的永久标识符相联接的支架表中 5  
    最后，如果您的维度表有上百万行，并且需要对很多属性进行历史跟踪和当前跟踪，那么除了类型2跟踪所使用的代理键之外，还应该考虑将维度自然键或者永久标识符作为事实表的外键  
    这种方法的本质是通过事实将两个维度表联系起来，当事实行装载的时候，类型2维具有基于有效值过滤或分组的历史准确属性，维的自然键联接到一个具有当前类型1值的表中  
    您可以使用该维度表的属性来汇总或者过滤基于当前配置文件的事实，而不必考虑维度行载入的时候这些值是否真正有效

### 6.4.4 角色扮演维
当同样的物理维度表在维度模型中扮演不同的逻辑角色时，我们将其称作“角色扮演维”，最常见的例子就是日期维  
事实表中的每个日期外键都指向某个特定的日期维度表，如图6-11所示，不能将这些外键都联接到同一个表中，业务SQL语句会认为，通过同时联接到同一个表中的两条路径进行查询所得到的日期也是相同的，而实际上却并不是这样  
![6-11](../syn_pic/data_warehouse_life_cycle/6-11.png)  
您需要骗过SQL语句，让它认为每个独立的日期维度表都具有唯一的列名称，而不要出现两次联接  
即使确实无法使用如图6-11所示的单个日期维，也可以在后台创建和管理单独的物理日期表 5  
根据行业不同，除了日期维之外，角色扮演还可以出现在多种类型的维度上，例如：
+ 客户（收货人、付款人和订货人）
+ 工厂或者港口（起点、终点）
+ 供应商（参照、执行）
+ 代表（卖主、服务商）3

注意：单独的维在一个事实表中可以同时扮演多种角色  
### 6.4.5 杂项维
在确定了明显的业务维后，常常会有大量乱七八糟的标记和文本属性无法在现有维度表找到安身之所，这种情况在对一个复杂的事务处理业务过程进行建模时表现得更为明显 5    
面对这种情形，设计人员常常会提供一些拙劣的供选方案，应该尽量避免使用如下的方案：
+ 将它们留在事实表中，将标记和文本属性留到事实表中是一种糟糕的做法，因为这样做可能会造成事实表行惊人地膨胀
+ 将他们构建成独立维，将每个标记和属性都构建成单独的维也不是一种好方法，这样会导致事实表中维度外键的数量膨胀
+ 清除它们，尽管将标记和指示符完全清除掉是一种吸引人的做法，但是如果最终证明业务是与这些标记和指示符相关的，那么就可能会出现错误  

解决这一问题的关键在于仔细研究标记和文本属性，并且将它们放到一个或多个杂项维中，如图6-12所示，而不是将它们乱堆到事实表中 5

### 6.4.6 雪花型和支架
当冗余属性和译码都被转移到单独的表，并且通过人工键连接到原始表时，维就变成雪花型了，图6-13给出了一个实例产品的雪花型维度表  
![6-13](../syn_pic/data_warehouse_life_cycle/6-13.png)
一般情况不建议将维度模型构建成雪花型,因为雪花型总是会让用户感到十分复杂，很不容易弄清楚  
任何认为维度模型的雪花化可以辅助对维度模型进行维护的观点都是站不住脚的，维护工作应当在ETL系统中进行，而且要在被装载到呈现服务器上的维度模式之前进行  
有些设计人员采用雪花型的维是为了节约存储空间，这种微不足道的节约却需要额外花费更多的精力额开销来创建和使用人工键的过程进行管理，而且叉属性的流量几乎必然会使性能降低 5  
尽管上述的一般性建议是和雪花型相冲突的，但是有些情况下，我们也承认构建一个类似初始雪花型的“支架”维可能会比较有好处，如图6-14所示，如果底层维非常大，这种方式就真的能够节省空间了  
![6-14](../syn_pic/data_warehouse_life_cycle/6-14.png)  
另一个经常会遇到的支架是从基本维分类出来的日期维，如客户生日或者产品引进日期，如果确实需要日期支架，那么要注意支架日期应当在标准日期维度表中所存储的日期范围之内  
尽管支架是雪花型的一种允许的形式，但是维度模型中不应当到处都是支架，因为这样做可能会为模型的易读性和查询的执行性能带来负面影响  
### 6.4.7 处理层次结构
层次结构是维度模型中不可避免的，本节我们回顾一下常见的固定层次关系和深度可变层次结构 5  
1. 固定层次结构  
    作为维度表反规范化的结果，维度表中简单的多对一层次结构比较常见，而且是以维度中的一组属性来表现的  
    层次机构常常和数据中的钻取路径相关，向下钻取是DW/BI系统中用户使用的最古老也是最受推荐的策略  
2. 通过桥接表实现深度可变层次结构  
    遗憾的是，并非所有的数据层次都是简单和固定的  
    在关系环境中描述一个深度可变组织结构或者部件层次结构是一件较困难的任务  
    传统的计算机科学解决这种层次问题时，在每个包含其父节点客户键的客户维度行中设置一个递归循环的指针 5  
    正如我们在数据仓库工具箱第二版中解释的那样，您可以使用桥接表来导航一个深度可变层次结构，如图6-15所示    
    ![6-15](../syn_pic/data_warehouse_life_cycle/6-15.png)
    <center>使用导航桥接表的商业客户维，联接后沿着树下降</center>
    
    机构的桥接表中将每个客户实体到每个下属子公司的路径都存储为一行，在有的行中路径长度也可能为零——如客户自己到自己的路径  
    当您沿着结构层次结构向下，可以加入如图6-15所示的表  
    另一方面，当您想要沿着机构层次结构向上时，联接就会改变，父客户键现在应当联接到事实表，而子公司客户键应当联接到客户维度表 5   
    尽管桥接表允许SQL在一个不可预测的深度层次的上查询和聚集事实，但是这种方法也有一些不利因素，业务它对我们所提的可用性和查询性能这两个目标都有消极影响  

### 6.4.8 使用桥接表的多值维
到目前为止，我们的设计中分析了事实表和相关维度表之间的简单多对一关系，事实表中很多行都联接到维度表中单一行  
卫生保健为多值维提供了一个生动的实例，假设我们对医院的账单业务过程建模，在事实表中为账单上的每一个条目都键一个新的行，设计师的问题是如何处置诊断维，求诊的人经常要进行多于一项的诊断  
为了处理这种开放性问题，不能将多个诊断维外键添加到事实表中，也不能将事实表的粒度改为让账单上与条目相关的每个诊断都占一行，因为账单度量在这样的粒度上是没有意义的  
为了解决设计问题和提供任意数量的诊断，在事实表和诊断维之间插入一个桥接表，如图6-16所示 5  
![6-16](../syn_pic/data_warehouse_life_cycle/6-16.png)  
使用这种结果，用户或BI应用程序可以执行两种类型的查询  
+ 第一，用户可以请求对诊断分组的账单费用进行正确的加权求和  
+ 第二，用户可以要求提供一个在所有表中执行同一种联接操作的影响报告，但是并不使用权重因子  

描述银行账户余额的事实表也有类似问题，如图6-17所示  
![6-17_5](../syn_pic/data_warehouse_life_cycle/6-17.png)
围绕商业客户行业分类代码的分析同样面临这样的多值问题，您也可以在客户维上添加一个属性来指明它们的基本业务类别，从而避免使用桥接表，简化某些分析  
在离开桥接表这一主题之前，我们想重申一下：不到万不得已，不要使用桥接表来解决多对多关系的问题，因为它对可用性和查询性能都有潜在的影响  
## 6.5  更多关于事实的讨论
### 6.5.1 三个基本粒度
不论行业和与其相关的业务过程如何，所有的度量事件都可以被刻画为事务、周期快照或累积快照  
1. 事务事实表  
    大多数基本事实表和公共事实表都是面向事务的，其粒度是每一行对应一个事务，或者一行对应事务中的一个条目  
    事务数据在其最低层次上一般都有数目巨大的维与之相关联，如图6-18所示  
    仅当活动发生时，事务事实表中才会插入行  
    正如本章早些时候讲述的那样，细节事务模式是非常强大的，分钟度量使您能够监视大多数粒度的业务活动 5  
2. 周期快照事实表  
    第二种最常见的事实表类型是周期快照，这种方法给了设计人员很大的自由度，在该时间跨度上的所有描述活动的事实都是可以获取的  
    有些业务过程天生就是和被描述为周期快照的形式，例如，银行账户的值可以很自然地看作是每天或者每月的快照，图6-18举例说明了汽车贷款的一个月度周期快照  
    ![6-18](../syn_pic/data_warehouse_life_cycle/6-18.png)  
    不言而喻，周期快照是细节事务事实的补充，而并非替代品  
    和事务事实表类似，一旦周期快照行被装载到事实表中后，一般都不会再次对它们进行访问 5   
3. 累积快照事实表  
     第三种类型的事实表是最不常见的，然而，它却在特定应用程序中发挥着非常重要的作用  
     周期快照按照规定时间捕捉度量，累积快照与此不同，它用于描述业务过程（有明确定义的开端和结尾）中某个不确定时间跨度里的活动，如图6-19  
     ![6-19](../syn_pic/data_warehouse_life_cycle/6-19.png)  
     和前两种类型的事实表相比，累积快照事实表具有一些独特的特征：累积快照包含多个日期外键，很多日期值在行首次装载时是未知的  
     除了典型的数量和总额之外，累积快照常常还包括另外另种类型的事实：首先是里程碑计数器，其次是累积快照常常用延迟事实来描述两个里程碑事件之间的时间差 5  
     当然，累积快照事实表和上述两种事实表的主要区别在于它的事实行是要更新的  
     累积快照事实表不能取代基本的事务事实表，因为在累积的过程中，不可避免地会丢失一些维度细节  
     事务和快照是数据仓库的阴阳两面，事务为我们展现了细节行为最完整的视图，而快照使我们迅速而容易地监视总体执行情况  
     
### 6.5.2 不同粒度的事实及其分配
有时单一的业务过程也会产生不同细节层次上的度量  
一个单独事实表中所有的度量都必须对表规定的粒度而言是合法的 5  
那么在面对度量原本就存在于不同粒度上的情形时，维度模型师应当如何去做？  
第一种方法是试着将所有事实都强制转化到最低粒度层次上，这个过程也被看作是将较较高层次的事实分配到较低的粒度上，和其他捕捉到的度量对齐，如图6-20所示  
![6-20](../syn_pic/data_warehouse_life_cycle/6-20.png)  
在分配之前，数据管理员需要定义分配算法，如果业务人士无法就某一种分配方式达成共识，可能最终要加入两种唯一标识的事实  
分配的妙处是所有由一个业务过程生成的度量限制可以共存于同一个事实表中了，每个维度属性都可以平等地作为过滤和分组度量地候选项 5  
标题事实和运营值的精确匹配是至关重要的，因为运算过程中不可避免地会遇到舍入误差，从而影响分配  
如图6-21所示,除了存储那些本身在较高细节层次的事实之外，事实表也可以用来存储各个订单条目的合计总数  
![6-21](../syn_pic/data_warehouse_life_cycle/6-21.png)  
### 6.5.3 多种货币和度量单位
有时需要以多种货币或者度量单位的形式来表达对事实的度量  
就货币转换而言，最常见的需求是以本国货币的形式对捕捉到的事实进行度量，并且为机构将其转换为标准国际货币，如图6-22所示 5  
![6-22](../syn_pic/data_warehouse_life_cycle/6-22.png)
现在您可以通过一个视图将这个事实表提交给用户和BI工具，该视图以推算的方式使用兑换率，进而显示两个货币事实集合  
同样，在包含几个业务过程的价值链中，通过系统或者多种库存度量从不同的角度监视产品流通情况时，有时也会在表示总额的时候出现矛盾  
设想一个例子中由13个基本量事实、5个度量单位解释和4种定价，错误的做法是将13个基本量事实方到事实表种，然后和让用户或者BI应用程序开发人员从远程维度表种查找争取的转换因子  
正确而这种的方法是建立一个底层物理行，具有13个基本量事实、4个度量单位转换因子和4个计价因子，如图6-23所示 5  
![6-23](../syn_pic/data_warehouse_life_cycle/6-23.png)  
有时设计者会将度量单位转换因子存储到产品维中，将这项因子封装到事实表中可以减轻产品维度表的压力，便于加入新的产品行来反映这项因子的微小变化  
现在可以通过一个或者多个视图将该事实表提交给用户了  
### 6.5.4 无事实的事实表
有时公司的业务过程中虽然有事件发生，却没有创建或者生成可以量化的度量，这些没有事实的事件产生了“无事实的事实表”  
可以针对捕捉学生出勤率的事件设想一个无事实的事实表，如图6-24所示 5  
![6-24](../syn_pic/data_warehouse_life_cycle/6-24.png)  
我们一般使用一个哑元计数器事实，它的值总是为1，您甚至不需要以物理形式存储该事实  
另一种事实表的粒度是每个学生每天注册（不是出席）一门课程占一行，而且学生参加课程则计数器为1，否则为0，这种情况下，1或者0这样的取值也是一种度量  
无事实的事实表在数据仓库中并不是记录事件的最佳方法，因为在该表中没有和事件相关的自然的数值型度量  
### 6.5.5 合并事实表
到目前为止，本章的重点是维度模型，大多数风险都是因为在ETL实现过程中一次的工作量过大而造成的 5  
当几个单源维度模型顺利实现后，最好将它们整合成一个综合的维度模型  
如图6-25所示，在ETL系统中一次性进行合并比在查询时再进行合并拥有更快的查询响应，而且结果的表示没有那么复杂  
![6-25](../syn_pic/data_warehouse_life_cycle/6-25.png)
再次重申，合并事实表应该是现有细节模式的有益补充，您不可避免地要将细节信息聚合起来形成一个横贯多个业务过程地公共粒度，所以一开始就建立一个合并事实表是不合适的  
## 6.6 有关维度建模的错误和误区
尽管业界已经接受了维度建模的方法，但是整个行业对此的理解仍然还有一些错觉和误区  
### 6.6.1 将关注点集中在部门报表上导致的错误观点
第一类错误观点是将关注点集中到部门报表和分析结果而没有将设计工作集中到度量过程上  
将全部精力投入到报表或者查询模板的需求分析上很容易在数据建模过程中产生一份具体的报告，但是却无法捕捉到相关维度和主要度量  
以报表为中心方法的一个漫画式的例子是这样的，一家公司建立了一个一百多个维度模型，使用星型模式和OLAP多维数据集进行组合，其目的是支持订单管理分析  
和集中精力于具体的报表需求或者部门需求不同，我们建议维度设计的关注点应当放在物理性能度量过程中，才能够避免下述的错误观点： 5  
1. **错误观点**：建立维度模型是为了提交一份具体的业务报表或应用程序  
    **真实情况**：维度模型应当围绕物理度量过程或者事件来建立  
2. **错误观点**：认为维度模型是部门解决方案，使得多个维度模型从同一个源中重复抽取数据  
    **真实情况**：维度模型不应当被限定于部门范围之内，面向部门的方法很容易造成不一致、非集成的解决方案
3. **错误观点**：如果不重建原有星型模式或者创建独立的事实表，就无法合并新的数据源 5  
    **真实情况**：如果新的数据源是呈现服务器上已有度量过程的另一个捕获系统，在粒度相同的前提下，不需要改变已有报表应用程序，新数据就可以很好地同原有数据相合并  
4. **错误观点**：使用维度建模，事实表只能是单一粒度的，是不可改变的  
    **真实情况**：规定创建事实表时采用单一细节层次确保了度量指标不会被错误地双重计数  
    如果度量本来就是不同粒度的，最简单的方法是为每个细节层次都建立一个事实表  

### 6.6.2 提前汇总导致的错误观点
下面要讲述的一组错误观点是由于在维度模型中混淆了恰当细节的层次而产生的  
1. **错误观点**：维度模型对业务问题进行了假设，因此是固定不变的，当需求变化时，模型也必须进行修改 5  
    **真实情况**：当您提前汇总信息时，已经对业务问题进行了假设  
    然而具有原子信息的维度模型和业务问题是独立的，这些维度模型可以应答新的、以前没有指定过的问题，且不需要改变数据库  
2. **错误观点**：维度模型仅在可预知的使用方式下才是适用的，维度模型不适用于探索性查询  
    **真实情况**：规范化模型和维度模型都包含相同的信息和数据关系，尽管二者的差异很大，但是都能够解决相同的问题  
3. **错误观点**：维度模型不是可伸缩的，如果将细节数据存储到维度模型种，性能将会下降 5  
    **真实情况**：星型模型的伸缩性能非常好，维度建模并不禁止存储质量好的历史信息  
4. **错误观点**：维度模型不可扩展，无法满足数据仓库今后的需求  
    **真实情况**：在最低细节层次上表示数据的维度模型具有最大的灵活性和可扩展性  
5. **错误观点**：维度模型不支持复杂查询，它消除了实体间的多对多关系，仅允许多对一关系  
    **真实情况**：维度模型和规范化模型的逻辑内容是相同的，一种模型里的所有数据关系在另一种模型中都能够进行准确表达 5  

### 6.6.3 过于重视规范化导致的错误观点
有些人认为规范化能够解决数据集成问题，事实上将数据规范化对集成并没有多少好处  
数据集成是一个不同于任何具体建模方法的过程，它需要标识出公司内部使用的不兼容的名称和度量，然后在企业范围内就建立和管理公共的名称和度量达成一致  
1. **错误观点**：一致性维度这样的维度建模概念为ETL增加了过多的负担  
    **真实情况**：数据集成取决于标准的名称、值和定义  
    在整个公司范围内达成共识，以及随后实现对应的ETL系统规则都是较为困难的规则，但是无论您使用规范化模型还是维度模型，都无法避免做大量的工作 5  
2. **错误观点**：当单独的源系统超过2个时，维度模型就不适用了，因为从多个数据源集成数据是非常复杂的  
    **真实情况**：数据集成所面临的挑战和建模方法没有什么关系  
3. **错误观点**：对维度属性的改变仅仅是维度模型中的问题  
    **真实情况**：每个数据仓库都必须处理时间差异问题，但一个实体的特征变化时，需要一种系统的方法来记录这种变化    
4. **错误观点**：多个数据集市无法进行集成，它们都是自下而上构建的，针对的是单个部门的需求，而不是整个企业的需求，因此混乱不可避免 5    
    **真实情况**：一点也不错，要集成面向部门建立的维度模型是一项艰巨的任务，这些独立的解决方案没有构建一致性维度  
    要在整个企业范围内对定义、业务规则和采用的实施方法达成一致，克服组织结构上的差异和文化障碍是不可避免的

我们建议引入一位或者多位分析型业务用户（也叫做超级用户）作为核心建模团队的成员，这些超级用户可以显著增加团队的洞察力 5  
核心团队和源系统开发人员一同紧密工作可以了解源系统中的内容、含义、业务规则、时限和其他需要填入维度模型的复杂事项  
DBA实现物理数据库，而ETL设计师和开发者也应当参与建模过程  
重复地将额外的计算或数据限定步骤加入到ETL过程中是有利的，而且不会增加BI应用程序的负担  
在进入建模过程之前，花时间考虑一下DW/BI环境的运行管理和数据管理的意义  
分立的业务单位并不一定要放弃他们的私有属性和名称，但是他们需要和其他业务单元就企业的一个一致属性和事实集合达成共识，而该集合是数据集成的基础，如果有必要，该企业集合完全可以是崭新的 5  
除了前面已经讨论的角色，在第9章还描述了在数据仓库环境的运行管理中ETL团队所需的两个特殊职位：维度管理员和事实提供者  
维度管理员负责定义、创建和发布一个或者多个一致性维度  
同时，当接收到维度管理员传过来的维后，事实提供者负责设计、建立和发布一个具体的事实表  
尽管过多的人员参与设计过程会增加设计过程延期的风险，但是这样做会使设计过程的内容更加充实和完备，是值得花费额外开销的  
### 7.2.2 回顾需求
在建模开始之前，建模团队自身必须熟悉业务需求 5  
在第3章，我们阐述了详细的项目层业务需求文档——该文档主要关注高优先级的业务过程和相关分析结果，还阐述了业务分析员有兴趣解决的问题和难题  
此时，最大的诱惑是跳过业务需求审查过程而直接进入到设计阶段，请不要跳过业务需求审查过程，设计一个业务驱动的数据模型比源驱动的数据模型要难得多  
业务基于属性和状态得组合定义了有关数据得新分类，要在维度模型中揭示和包含这些隐含的业务规则，而不要仅限于满足业务用户的需求，这是提交一个超出用户需求的工作平台的秘诀  
在深入到设计维度模型的工作中之前，必须详细了解业务需求

### 7.2.4 确定命名约定
在创建维度数据模型的过程中，不可避免地会出现命名约定问题  
设计维度模型的过程中有一项工作就是公共定义和公共名称达成共识 5  
不过，确立命名约定并不一定很花时间，大多数大型机构都具有某种功能（如数据管理），专门负责命名约定的问题  
如果您的机构尚未建立一组能够适用于DW/BI的命名约定，就需要在设计维度模型的时候确立命名约定  
一种常见方法是使用一个由三部分构成的列名标准：*原始词_0个或多个限定词_类词*  
原始词就是一个分类词，一般和列所属的实体相对应，有些情况下不必要使用限定词，例如：销售事实表中描述销售量的字段可以命名为sales_dollar_amount  
接下来，核心建模团队要和数据管理员一起设计一个初步的名称合集，并给出相应的理由 5  
在最终模型审查会议上，和主要干系人一起就最后的命名约定达成一致  
一旦就最终的名称集合达成了共识，要立刻认真地编制文档，并且将其应用于最后的数据模型

数据探查使用查询和报表工具来探查被调查系统中的内容和关系 5  
### 7.2.6 获取场所和用品

除了开会的地方，团队成员还需要基本的用品，如活动挂纸  
## 7.3 再论四步建模过程
四步过程设计维度数据模型不会出现在单个设计会议中，它是一个全局策略，指导几个星期的设计工作，涉及到一系列会议  
### 7.3.1 第1步：选择业务过程
第一步是选择要建模的业务过程  
注意：一种测试是否正确选择了业务过程的方法是转到第2步，为业务过程中捕捉的最细节的数据声明事实表粒度  
业务过程是维度数据仓库的基石，理想情况下，您选择的由初始业务过程所捕捉的核心数据都来自于单一的源系统或模块  
### 7.3.2 第2步：声明粒度
创建一个高层维度模型的第二步工作是声明粒度，或者说是为所选择的业务过程声明事实表的细节层次，声明粒度的例子有： 5    
+ 当使用扫描设备度量时，零售票据上的单个条目
+ 保险单中的单笔交易
+ 从医生那里收到的账单上的条目
+ 航空公司的登记证
+ 每周每个商店里每件产品的库存度量 2  

注意，这些粒度都以业务术语声明，第3步和第4步所作出的设计决策取决于如何正确地将粒度可视化  
最常见的设计错误是没有在设计过程的一开始就声明事实表的粒度  
选择粒度时既要考虑到用什么来满足业务需求，又要考虑基于源系统收集的数据可以得到什么 5  
一般都能够将原子数据综合到一起生成更高层的汇总数据，而如果没有细节数据可用的话，将无法从汇总层向下钻取  
在强调原子数据的好处时，我们也是现实主义者，某些行业中一些过程的原子数据代表着巨大的数据量，因此个别案例中必须使用细节程度较低的数据  
尽管设计团队仔细考虑和声明了事实表的粒度，但是设计团队在设计过程的某一点上改变对粒度的看法也是合理的  
例如，如果粒度声明为配货中心单个存货条目的日库存量，而设计团队发现遗留系统能够正确跟踪批号，那么他们可能决定转而跟踪批号，向上钻取到存货条目  
在设计过程的任何节点上，设计团队都必须能够清楚地描述目标事实表的粒度 5  
### 7.3.3 第3步：识别维度
第3步是在规定的粒度层次上确定应用到事实表的维度，总线矩阵中的所有维度都应当针对粒度进行测试  
有时事实表粒度以主要维的形式给出  
粒度声明可以使您创造性地思考如何将维添加到一个显然在源数据中不存在的事实表  
事实表的主键几乎从不会声明为完全由维度外键构成的集合  
正是在这一步，现实世界的复杂事物开始逐步从建模过程中浮现出来，找出这些隐藏的维是很有难度的，因为它们迫使人们对数据和数据的用途重新进行思考 5  
选择维同样也会使您对粒度的声明重新进行思考，如果一个提议的维度无法和事实表粒度相匹配，既可以把维度放一边，改变事实表的粒度，也可以考虑使用多值的设计方案  
检查维度并确保它们都是有意义的  
### 7.3.4 第4步：识别事实
建模过程的最后一步是识别业务过程中的事实（或者称做度量），事实必须适合于粒度，要坚决抵制匆匆忙忙将与事实表粒度不匹配的事实添加到事实表的做法  
大多数面向事务的业务过程只有少了的基础事实，如数量和总额，但是却有很多从这些基础事实导出的组合和计算结果（一般用于监视业务过程）  
## 7.4 设计维度模型
和所有数据建模工作一样，维度模型的开发也是一个迭代的过程  
构建维度模型的过程一般要经过三个阶段，如图7-1所示：高层设计时期，详细模型开发时期，进行模型的审查、再设计和验证等一系列工作  
对单独的业务过程(如销售订单) 或者几个紧密相关的业务过程（如卫生保健索赔，它将职业索赔、设备索赔和药物索赔等索赔包含到一组相互区别但是又紧密关联的事实表中）来说，整个建模过程一般要花费3到4个星期  
建议核心团队在模型开发期间安排一系列设计会议，在这段时间里数据模型师会不断更新设计文档，以保证每次会议的文档都是最新的  
### 7.4.1 建立高层维度模型
设计会议最初的任务是为目标业务过程创建一个高层维度模型图 5  
记住一点非常重要——大多数DW/BI项目任务都有双重目标：既要完成工作也要通过项目锻炼队伍  
1. 召开初始设计会议  
    初始设计会议应着急所有的核心建模成员，并且还要邀请受认可的外部团队成员或者源系统和ETL团队中的相关人员参加，初始设计会议的主要目的有两个：  
    第一，是为了将设计团队的所有成员首次召集起来，确立对项目目标和设计过程的共同理解，并且树立起团体协作精神  
    第二，通过初始设计会议创建最初的高层设计，而且这个会议应当由主数据模型师来主持  
    直接目标是创建高层维度模型图，它是对业务过程中的维度表和事实表的图形描述 5  
    开发初始维度模型最有效方法是使用活动挂图，要记住大多数初始建模都需要采用头脑风暴法，即让大家自由发表意见  
    当您一步步完成初始模型的设计过程时，团队成员很可能会发现很多问题和挑战，设计过程的推动者需要立刻明确问题并将它们记录到问题列表中  
    这时应该立刻展开高层设计，后面还有很多机会去讨论细节内容和相关问题  
2. 为高层模型图编制文档  
    设计初稿应当汇总起来使用图形表示，形成一份可以称之为高层模型图的提交材料，如图7-3所示，高层模型图是一个实体层上的数据模型  
    高层模型显示了应用于一个给定业务过程的具体事实表和维度表，这种图形描述有以下几个用途： 5  
    + 当和干系人沟通有关项目、范围和内容方面的事项时，它是一份随手可得的设计介绍  
    + 它方便了设计团队和业务伙伴之间的讨论，它是沟通详细设计的切入点，也为设计审查提供而文档支持，保证了对业务需求的满足  
    + 它是一种有用的培训工具，对数据模型进行了介绍，甚至是最不认真的用户也可以理解  
    
   ![7-3](7-3.png)
    为了帮助理解，在高层模型图中保持一致性也很有用的 5  
    每个业务过程可能有不同的维数，当某个维没有应用于具体事实表时，我们偏向与完全将其取出，并且为它们留下位置，确保视觉上的连续性  
3. 识别属性和度量  
    当高层维度模型完成后，设计团队就应该开始关注维度属性和事实度量的初始集合  
    这个过程中提交的材料叫做初始属性和度量清单，使用图7-2所示的电子表，为模型中每个表创建一个电子表格，并且填充其属性或者度量  
    这时就要进行集体讨论了，写下想法并且将注意事项和有争议的问题做笔记，该文档有助于记住所有与模型相关的难题的细微之处，也有助于您记住自己是打算如何解决这些问题的  

### 7.4.2 开发详细的维度模型
完成高层设计之后，现在要开始关注细节了，此后，核心团毒应当在非常正规的基础上工作，一个表接一个表，一个列接一个列地定义一个详细的维度模型 5  
详细的维度模型建模过程是为高层模型填补缺失的信息，解决涉及问题，并且不断测试模型能够满足业务需求，确保模型的完备性，您也需要确定信息来源的位置、定义，确立有关这些属性和度量如何填入模型的初步业务规则  
1. 识别数据源  
    建模过程的一个重要组成部分是找出最适用于填充目标设计模型的源数据  
    详细数据分析是设计工作的一个重要部分：首先必须立即支持业务需求所需的数据，其次必须对每个候选数据源都有足够深入的了解  
    从业务需求来看，可以识别出正式数据源和非正式的数据源，尽管非正式数据并不存储于公共的数据存储系统中，但是业务用户需要将正式数据和非正式数据综合起来使用  
    一般可以通过开发一个新的IT过程来获取非正式数据，从而使非正式数据源也包含于数据仓库中，数据必须经过严格的ETL过程，要像将其他数据源导入数据时那样对这些数据进行清洗和操作 5  
    注意：不要让用户简单地直接将数据导入到数据仓库中  
2. 了解候选数据源  
    当您创建了高层设计模型之后，要想在维度表和事实表设计的过程中取得长足进步，就需要为每个表明确数据源，图7-4给出了数据源定义的示例

3. 数据探查和选定数据源  
    有了候选数据源列表之后，设计团队需要评估每个数据源并且确定最好的数据源  
    收集和审查所有可用的源系统文档：数据模型、文件定义、记录格式、书面文档和源系统程序 5  
    研究源系统数据时，对工作本身的期望和实际情况是有一点出入的，和源系统的所有者以及核心业务用户沟通会有很多好处，核心业务用户一直使用候选数据源中的数据，他们会给您一定的启示  
    数据探查是对数据源内容的系统分析，其所涉及的范围从计算字节数、检查基数到检查数据是否满足DW/BI的要求等更为深入的分析，每个数据源都应当进行详细的数据探查  
    探查分析被分成一系列测试：开始于单个字段，结束于包括扩展数据库在内的整个表集  
    查看到底有多少行赋有空值或者背离定义域的内容是非常有意义的工作  
    数据探查可使用结构筛选来调查同一个表中字段间的关系，可以将数据表中作为键的字段显示出来，和高层的多对多关系一起形成层级关系 5  
    检查键值的唯一性也非常有用，如果违背了这一规则（键字段有重复的实例）就会出现非常严重的错误，反映了某个业务规则没有在源数据中强制执行  
    在数据探查阶段还要检查表和表之间的关系，包括检查外键和主键之间的关系以及是否出现了没有子类的父类等  
    最后，数据探查可以检查企业中特有的复杂业务规则，这些最后的测试可以称作业务规则筛选  
    数据探查过程主要有以下结果：  
    + 经过考虑后对候选数据源做出基本的"可行/不可行"决策 5  
    + 在项目继续进行之前，源系统数据质量问题必须得以修正，这些修正产生了大量的外部依赖关系，必须对这些关系进行恰当的处理    
    + 当从源系统抽取数据之后，在ETL处理流程中就可以修正数据质量问题，但是请不要期望有很高的数据质量，因为ETL系统只能修正相对较少的数据质量错误  
    + 未预料到的业务规则、层次结构和外键-主键关系，在细节层次理解数据有助于找出那些需要合并到维度数据模型中的问题  
    
   基于上述分析结果，可以选择用于填充维度设计的最优源系统，对基本数据源的选择常常由业务驱动，考虑的时候可以参考的标准包括：  
    + 数据可访问性，可访问性越强越合适 5  
    + 数据供给的持久性，通常，在处理流程中数据仓库的数据流随时随地地可能会中断，数据源持久性越强越合适  
    + 数据准确性，当数据在系统之间传递时，不可避免地会有一些更改，如果从这些下游系统提供数据，用户会质疑其准确性，此时，为了确保数据准确性，最好返回最原始的数据源  
    
   您应该将所有考虑到却没有选择的数据源都记录下来，当数月之后重新提出选择最合适的数据源问题时，这个文档就有用了  
   在此期间，核心设计团队要缠住源系统专家向他们请教相关问题，使用数据探查功能增进对源数据的理解，并且确定将该数据源引入模型中是否适合  
4. 确立一致性维度  
    在这些详细设计会议中，还应当定义关键一致性维度，在促使企业各部门就表及属性的命名、描述和定义等达成共识的过程中，数据管理员和业务分析员是主要的人力资源 5  
    毫无疑问，决策者（如业务发起人/驱动者或者管理委员会）必须参与解决一致性维度的定义和命名问题  
    有些维度属性信息可能包含有事务事件数据，不过所占的量一般非常少，而且常常是以代码的形式出现的  
    有些维度表可以有多个数据源，尤其是客户维，这些数据源中所含有的客户可能是不同的，每个客户的属性也可能是不同的，那些公有信息很有可能不相匹配，应当努力将这些数据源联合并集成为一个单独的一致性维度  
5. 确定基本事实和导出事实  
    数据探查工作能够确定由度量事件的源系统生成的计数和数量，然而可能有很多业务用户想要分析的度量是从这些基本事实中推导而来的，您也需要将这些导出事实整理存档  
    有两种类型的导出事实，具有可加性的导出事实完全可以由同一事实表行的其他事实计算生成 5  
    第二种类型的导出事实是非可加性的事实，如比率和累积事实，这些事实一般都在与基本事实不同的细节层次上进行描述  
    定义度量的整个范围是一个极具挑战性的行政过程，要指定数据管理员来协助这个过程  
    尽管在ETL过程设计之前，导出事实不需要是完备的，但是确保要考虑到所有基本事实很重要，直到支持导出事实计算所需的所有基本事实都明确下来，维度模型才是完备的    
    当业务用户对自己所需要的东西有了明确的想法之后，就可以召开一个专门的设计会议来确定导出事实，如果没有，您仍然需要设计一个导出事实的初始集合  
    通常，直到设计团队准备开发BI工具的时候，维度设计中才会开始考虑所需的导出事实  
    如果导出事实的数量很少，简单地使用维度设计工作表来捕捉和导出事实相关的信息即可，如果很多，最好捕捉并记录，然后将其存放在一个单独的文档中，例如图7-5所示： 5

### 7.4.3 审查和验证模型
如果已经对模型有了足够的信息，就需要开始图7-1所示的模型审查阶段和验证阶段，至少，设计团队应当计划与下述三个群体进行讨论：
+ 源系统开发人员和那些常常能够在模型中快速定位错误的数据库管理员 5
+ 没有直接参加模型开发过程的核心业务用户或者超级用户
+ 范围较大的用户社区

核心建模团队可以从审查和验证过程中获取宝贵的反馈信息，来自会议的反馈应当被集成到最终设计文档中  
当然，贯穿整个设计过程，建模团队偶尔也应该将自身从详细设计活动中解脱出来退一步，从整体上来审查团队的设计  
1. 实施IT数据模型审查  
    对详细维度模型的首次公众设计审查一般是和IT机构中的同僚一起进行的，准备充分、安排合理的情况介绍非常有利于这次会议的成功 5  
    这些IT人士对事物处理系统和其所表达的业务过程非常了解，如果需要建立8-10个维度，那么可能需要好几个设计会议来完成IT数据模型的审查  
    1. 准备  
        IT审查是比较有挑战性的，因为审查者常常对维度建模缺乏足够的了解，最好在审查过程中最好安排一些关于维度建模的知识讲座  
    2. 会议流程  
        会议开始时对基本的维度建模概念进行概述，确保他们明白事实表和维度表之间的区别，并且解释清楚维度模型为什么是反规范化的    
        当每个人都了解基本概念之后，开始进行总线矩阵的审查，这个过程向每个人展示了模型的实体层图，可以作为后续讨论的指南    
        审查会议的大部分时间都应当逐个检查维度表和事实表，这些与会人员能够处理详细维度模型的电子表格版本，因此所有的详细工作表都要有副本 5  
        当仔细检查模型的时候，可以对每个表中尚未解决的问题进行审查，通常与会的某个人员给出答案，至少也会知道从哪里找到答案  
2. 与核心用户一同审查  
    在很多项目中不需要进行上述审查，因为核心业务用户本身就是建模团队的成员，早已经对数据模型非常了解了  
    核心业务用户比一般业务用户在技术方面懂得更多，能处理有关模型的很多细节问题，通常IT审查会议和核心用户审查会议是合到一起举行的  
3. 向业务用户介绍  
    设计会议其实也是一次讲座，您应当使他们理解维度模型，而不是去压制他们的想法，同时还要告诉他们维度模型是如何支持他们的业务需求  
    进行一次和IT审查会类似的情况介绍，简单阐述维度的概念和定义，阐述总线矩阵，审查高层维度模型图，审查重要的维度 5  
    我们一般使用高层模型图来向用户介绍单个维度表，而不是使用详细设计工作表  
    分配大约三分之一的世界来说明如何使用模型解决业务过程中遇到的问题，从需求文档中抽取一些例子，一步步讲解如何解决这些示例问题，超级用户马上就会明白  
    到了这一步，对模型的调整幅度就很小了，当他们第一次看到真实数据时，必然会报以欣喜和欢呼的  

### 7.4.4 设计文档定稿
当处理完问题列表的所有内容或者为项目团队的人员安排完任务的事实（先完成二者中的哪一个都可以），可以认为已完成维度建模设计工作，现在最重要的是模型能否支持业务需求和源数据的验证，以及验证后的数据能否恰当地填入到模型中去  
当使用其他数据建模时，可能要花费数月时间来单独跟踪每个问题，但花费那么多时间显然是不可行的，实践证明，解决最后几个问题所花费的世界和解决前50-60个问题所花费的时间一样多，但这些问题实际影响很小 5  
当模型的最终形式确定下来后，就可以编制设计文档了，该文档一般包括：  
+ 当前正在设计的业务过程的简要描述  
+ 支持业务需求的高层讨论，这些讨论涉及的问题指向详细需求文档  
+ 高层数据模型图  
+ 每个维度表和事实表的详细维度设计工作表 5  
+ 公开问题列表，要突出标识出那些尚未解决的问题  
+ 模型涉及对项目范围和业务需求支持情况的局限性论述  
+ 其他有意义的事项，如设计过程中采取的折中处理或源数据的相关事项

因为设计团队已经从详细维度设计工作表中（图7-6）捕捉到了源数据探查信息、数据问题和转换规则，这为项目的源到目标映射文档的编制奠定了基础，这种映射概要描述了哪些工作必须要做，但是没有详细给出如何做这些工作  
最终的详细维度设计工作表捕捉到了建模团队在设计过程中了解到的有关数据的所有信息，还包含了关于有效值、译码、缓慢变化维规则，以及更加复杂的转换 5  
## 7.5 拥抱数据管理
我们已经讲述了维度设计过程中的战术策略，在离开这个主题前，重申一下在机构中确立数据管理工作计划的重要性，实现数据的一致性是大多数DW/BI项目群的关键目标，对大多数机构而言，确立数据质量和集成的职责非常困难  
识别和处理数据质量问题需要机构运作一个连续的质量改进过程，确立一个有效的数据管理工作计划对推动这件工作的完成非常关键  
数据管理工作计划的基本目标是创建有关数据资源的公用知识，为企业提供清晰、一致、准确、及时和备有文档的信息，还包括确保公用数据集得以正确和充分地发挥作用，且只有那些得到授权的个人才可以使用这些数据  
整个机构缺乏一致的数据是很多DW.BI系统中的祸害，数据管理是解决数据一致性问题的关键要素  
根据我们的经验，大多数有效的数据管理都是由业务社区完成的，借助技术手段，DW/BI团队通过确定问题和时机来推动和应用数据管理，然后实现已经达成共识的决策，包括创建、维护和分配一致的数据 5    
## 7.6 小结
维度建模是一个迭代的设计过程，需要具有不同技能的人员协同工作，设计工作一开始从总线矩阵中得到一个初始图形，并在实体层进行表现  
在早期的高层设计会议上创建该模型，并用批判的眼光对其进行严格检查，同时为每个需要补充调查的表和问题产生一个属性或者度量的初始列表  
详细建模过程逐个表地创建模型，并且向下钻取定义、来源、关系、数据质量问题和所需的转换  
基本目标是创建一个能够满足业务需求的模型，验证数据是否能够填入模型，为ETL团队提供一个稳固的起始点和明确的方向  
确定列名和表名的任务此时相互交织地进入设计过程，整个机构必须就维度模型中每个表和列的名称、定义及导出达成共识 5  
详细建模过程有几次审查会议，其中最常见的三个审查会议中参加人员分别是IT机构中其他有利害关系的成员、在技术上比较老练的业务用户和一般用户社区的代表  
最后的结果就是生成了维度模型，并且已经根据业务需求和数据实际情况进行了测试，而且该模型是经得起考验的  
## 7.7 管理工作和降低风险
根据下述最优方法来保证维度建模的工作量和风险在控制之中：
+ 确保领导设计过程的数据模型师 是一位专家级的维度模型师，否则的话，可以考虑引入外援  
+ 设计团队每个成员都必须彻底完全理解业务需求   
+ 一定要把超级用户作为设计团队的一份子  
+ 使用数据探查工具不断地探测建议使用的数据源，保证支持数据模型的数据是可用的  

## 7.8 保证质量
确保质量的关键步骤有：
+ 坚持使用维度建模最优方法  
+ 坚持让超级用户积极地参与设计过程  
+ 进行大范围的数据探查  
+ 不要忽略IT和用户审查会议

## 7.9 关键角色
设计维度模型的过程中的主要角色包括：
+ 数据模型师领导整个维度建模工作  
+ 业务分析员、超级用户和BI应用程序开发人员描述业务用户的分析性需求  
+ 数据管理员协助推动整个机构就维度模型的名称、定义和业务规则达成共识  
+ 源系统专家提供与运营系统相关的知识  
+ ETL团队需要了解源系统和目标系统 5  
+ IT部门中的有关人员和业务社区的主要代表审查设计模型并反馈信息  
+ 数据库和数据访问工具厂商要了解数据库设计的基本原则，以便优化他们的产品

## 8.1 制定标准
对于DW/BI系统中的各种组件，需要在全系统范围内制定标准  
有一些关于如何在数据仓库中创建表的系统范围内的决策值得借鉴，和所有的一般性决策一样，您可能会找到与该决策相背离的具体实例  
简要地记录一般决策，并且在每次以不同方式创建对象的时候都记录自己的决策，接下来将给出在设计和评估标准时应当牢记的细节内容  
### 8.1.1 遵守命名约定
在关系数据仓库中表和列的名称实际上应当与逻辑模型中的名称相同，如第7章所讨论，数据对象名称应当一致、有描述性，并且应当面向业务用户，记住，表和列的名称是用户界面的关键要素  
表和列的命名应该使用一致的风格，最保险的列命名风格时所有的字母小写，单词之间用下划线连接 5  
如果您的命名约定中有大写形式，就需要意识到数据设置中区分大小写的问题，开发服务器、测试服务器和产品服务器都应当遵循相同的数据库大小写规则，这样程序脚本才能在部署过程中保持一致    
### 8.1.2 为空还是不为空
在数据仓库维度表中应当鼻毛使用空字符串，它们常常会使业务用户混淆，尤其是当业务用户构建的查询对一个有时为空的属性进行筛选时更容易出错  
最好为其赋一个默认值如“N/A”、“Not known”,或者一个长度为零的字符串  
正如第6章所讨论，处理"空"日期列最好的办法时对日期维使用代理键，即不使用日期或者日期时间类型的列  
如果选择不将日期列映射到一个外键上，则可以将缺失的日期映射到一个具体的日期，如数据库日期可以取到的最小值或最大值，不管哪种方案都需要向业务用户传授如何对列进行筛选 5  
### 8.1.3 设置登台表
正如我们在第10章将讨论的，大多数ETL系统都需要一组登台表来支持ETL过程；此外，登台表命名方式要和他们所服务的过程或目标紧密结合起来  
### 8.1.4 制定文件位置标准
需要设计使用有关源代码、脚本、二进制文件和数据库文件位置的标准  
一种方式是基于主题域或者表（如账单、客户和产品等） 建立代码树的主干，而另一张方式是基于相似的功能（如数据定义语言、初始装载和日装载）建立顶层分枝  
对于应用程序文档（如选定ETL应用程序所需的数据模型和文档等）的位置也需要制定相应的标准  
### 8.1.5 对用户访问的表使用代用名或者视图
作为一种例外情况下的规则，如果业务用户和程序员都使用了相同的名称，那么我们建议所有授权用户访问表时都使用代用名，大多数RDBMS都支持这种功能 5  
代用名提供了一个额外的间接层次，可以极大地简化新版本数据表的部署工作  
一种供选的或者说补充的方法是通过视图来实现上述功能，视图要比代用名更加灵活，在是视图中可以重命名列并且进行简单计算  
如果物理设计需要雪花型维度—通常是因为BI访问工具的需要，我们将为这些用户创建一个单独的、物理反规范化的维度表；因为这些用户偶尔会需要直接访问数据  
警告！我们建议在数据仓库中使用多表视图时要有一定要限制，并且要对多表视图进行细致而合理的规划，比如10个表的联接运算要比包含同样数据的单个平面表处理起来慢得多  
尽管还有一种很有诱惑力的想法是通过在规范化的数据结构上定义简化视图来创建数据仓库，但是这种方法只有在数据存储量很小并且数据非常干净的情况下才能够使用 5  
### 8.1.6 主键
我们建议在关系数据仓库中就如何声明主键和外键制定一个通用的规定    
第6章提到过，维度表使用单列的整数代理键作为其主键：对每一行来说列都必定是唯一的，作为一个一般性规定，要将维度表的代理键声明为表的主键  
注意：有很多管理代理键分配的方法，您的ETL工具可能就提供了这样的功能  
接下来为键定义数据类型，为了获得良好的查询性能，键列最好使用联接运算效率最高的数据类型  
尽管很多人愿意将日期键暴力为日期类型，但是事实上它们也应当被简化为高效率的代理键 5  
对事实表来说，必须首先理解逻辑主键是什么，主键一般是维度表外键的一个子集，要保证唯一确定某一行，并不需要在事实表中参照每个外键    
很多事实表也有一个逻辑键，用于反向连接到源系统，用户维度建模的术语来说，逻辑键就是退化维度  
事实表的代理键一般对查询的意义不大，在少数情况下，为事实表中的每个行指派一个代理键会比较有益：  
+ 有时机构的业务规则规定存在多个相同的事实行是合法的
+ 更新事实行的特定ETL技术只有在使用代理键时才可行的 5
+ 按照顺序分配的代理键使得确定在ETL重启和恢复过程中装载工作暂停的位置变得简单
+ 代理键使DBA在进行调试或者达成特殊的管理目的时，可以容易确定单个事实表的记录
+ 在某些具有特殊结构的BI应用程序能够依赖事实表间的一对多关系，此时父事实表的代理键作为外键插入到子事实表中  

注意：通常，要将事实表主键定义为一个组合键，包括一组维度外键或者唯一定义事实表中一行的退化维  

### 8.1.7 外键
下一个要讨论的问题是强制声明事实表和其相关维之间的外键的必要性，声明外键关系可以杜绝违背数据库中参照完整性关系，确保事实表中任何一行在其相关的维度表中都能找到对应的行 5  
毫无疑问，系统实现过程的最优方法和起始点正是声明事实表和维度表之间的外键关系  
很多数据仓库团队都会在初始装载历史数据的过程中丢弃约束，随后在增量处理的过程中恢复和复查约束条件，并且将它们添加到恰当的位置  
如果认真地考查为了使装载过程能够在一个可以接受的时间段中完成而禁用外键约束的做法，您会发现还有一些问题需要考虑
还有一种比声明更好的方法是分割事实表，这和约束检查是不同的  
将当天的数据装载到一个空表或者分区中，强制进行约束检查，随后将该分区转换为较大的事实表 5  
这种方法的优点使对新一天的分区使用快速装载技术，并且只需要对那一天的数据进行参照完整性检查（假设每天都装载）  
管理分区事实表使系统的复杂性有所提高，并且一般需要一定量的脚本程序来实现自动化  
注意：要声明和强制执行事实表及维度表之间的外键约束。当且仅当无法在分配的时间内装载数据时，才能够考虑丢弃约束  
## 8.2 设计物理数据模型
现在已经建立起一些系统范围的规定和标准，该是设计物理数据模型的时候了  
物理模型和逻辑模型之间的主要区别时它能够对物理数据库的特征进行彻底而详细的描述，该描述从数据类型开始，经过表分割、表存储参数设定和磁盘分片等过程  
下面的讨论主要关注数据仓库管理员和OLTP管理员的区别，并且根据我们的经验指出一些最容易出错的地方  
### 8.2.1 设计物理数据结构
开始之前，确保逻辑模型应该已经完成，已经进行了一些重要的数据探查活动，确保了模型的可行性，此外，还建立了很多从源到目标的映射，并且注意到该如何生成每个表和列  
逻辑数据模型和其相关文档是物理模型的起始点，实际工作中，可能需要背离逻辑模型，从而支持访问工具的某种特殊需求，改善查询性能或者保持维护周期在可接受的时间段内 5  
在建立数据模型的过程中，逐表进行会收集到更多的信息，现在是收集这些今后要使用的信息的大好时机  
很多数据建模工具都有存储这类信息的空间，如果团队使用数据建模工具驾轻就熟，就应当尽一切可能使用该工具开发和注释物理模型  
我们发现使用电子表格这种低技术含量的方式往往能够提高团队的效率，而数据建模工具有时却会阻碍信息的收集和共享    
### 8.2.2 确定源到目标的映射
第7章所讲述的详细维度设计工作表是源到目标映射表的基础，物理设计团队应该添加其他表和列层次上的元数据来进一步修饰该映射表，对每个表都需要验证或者收集下述信息：
+ 确保表的名称和命名标准一致 5
+ 为用户访问的视图指定一个名称和一个显示名
+ 为业务用户描述表，很多数据访问工具都能够以工具提示的形式来显示这些描述
+ 明确表示事实表、维度表、元数据表还是登台表，在设计和实现ETL系统时可以设计其他登台表，要弄明白添加到模型中的每一个新表  
+ 详细说明源表，撰写一份有关数据转换的简要描述

此外，对每个列需要收集和描述以下一些信息： 5  
+ 确保列的名称和命名标准一致
+ 为列指定一个显示名称、业务描述或者技术描述，这些名称在各种包括OLAP多维数据集的BI应用程序中进行显示
+ 有选择性地为列指定一个分组名称  
    分组名在具有很多列的维度表中非常有用，是将相关列进行分组的一种方法  
    当设计物理模型时，要对所有的列都做深入考虑，而现在正是描述这些信息的时候
+ 确保每个列都有正确地数据类型、长度和精度  
    通常，列所拥有的数据类型基于存储数据的源系统，因为数据仓库一般都从多个源系统集成，所以可能需要对有些数据类型进行转换 5  
+ 指定列地数据长度时尽量不要比源系统所指定地还要短，如果将列定义为相对较短的类型，就需要在ETL过程中采取一些适当的安全措施来防止截断某些重要信息  
+ 规定每个维度属性如何处理变化值，可以使用第6章所阐述的缓慢变化维度技术  
+ 针对每个目标列详细描述源系统、表和列，包括数据类型转化的简要描述，例如“源ADDRESS.ZIPCODE从证书转换为字符”
+ 规定是否允许列取NULL值，如果不允许列取NULL值的话就为其指定一个默认值
+ 规定列是否是多层层次结构或者向下钻取路径的一部分，例如日-月-季度-年这样的层次结构，为该层次结构合理命名，如“财政日历” 5
+ 不要忘了添加用于维护和审计的列  
    例如如果维度表中的某一列是以类型2跟踪的，就应当为表添加生效日期和截止日期列，以便跟踪行处于活动状态的日期范围  
    可以使用一个标识符来指示行当前是否为活动的，或者也可以创建一个列来表明导致新的类型2的行被创建的是哪一个过程
+ 通过用户访问工具添加其他所需的列，例如，可以为日期维度表添加一个当前周期指示符，或者为各种属性添加排序列  

### 8.2.3 星型 VS 雪花型
逻辑数据模型常常被组织为星型模型，每个维都形成一个单独的表，如图8-2所示，相对于一组规范表来说， 我们更偏爱这种平面结构的表，这是因为：  
* 直接查询关系数据库的用户能够较好的操纵模型 5
* 关系查询一般都比这种结构的表现要好
* 在ELT过程中，由于平面表的数量相对较少，相关联的键也较少，因此更容易处理  
![avater](8-2反规范化星型维示例.png)  

另一方面，雪花型维度将每个等级层次都分割到其自己的表中，如图8-3所示，雪花结构支持下面的特性：
* 数据库引擎保证了层次之间的参照完整性，在本例中，每个产品都与一个并且只与一个子类别和类别相关联 5 
* 将来粒度较高的事实表可以挂靠到雪花结构的合适层次中  
    例如销售预测一般可以在产品类别或者子类别层完成，而不是在单个SKU层完成，包含销售预测数据的事实表应该指向维度表结构的产品子类别层  
* 每个层次都有其自身的代理键，当创建汇总维度表或者OLAP多维数据集时可能会用到  

如果基本的BI访问工具已经选定，应当修改模式来满足其需求，有些工具在维度表规范化为雪花模型时才可以较好地进行工作，而另一些则适应性更强  
大多数情况下，每个维度表（包括逻辑上的和物理上的）都应当反规范化或者平面化成一个单独的表结构  
如果数据访问工具需要使用雪花结构设计，应该先为其他BI工具的用户创建平面化版本的反规范化的维 5

### 8.2.4 使用数据建模工具
现在来谈谈如何使用数据建模工具开发物理数据模型，使用了数据建模工具之后，信息流可以流入元数据中，用户通过元数据可以了解到数据仓库中数据的情况，使用数据建模工具的好处有：
+ 将其他数据模型和数据仓库模型集成起来
+ 有助于确保命名和定义的一致性，为了最好地利用数据建模工具，团队必须设计一组标准并且在工作环境中遵循该标准  
+ 可以创建一组以多种有用的格式编制的好文档 5
+ 可以为大多数流行的RDBMS生成物理目标定义脚本
+ 提供了合理而直观的用户界面，便于输入有关表对象和列对象的注释

在前面的小节里，我们建议您收集大量有关表和列的信息，这些信息很可能本身并不适用于数据建模工具  
尽管您可能需要从电子表格开始建模过程，但是数据建模工具可以管理您收集的所有信息—也许是以许多注释或者表和列的扩展属性进行维护的  
一旦从电子表格环境转向了数据建模工具，电子表格本身价值就不高了 5

### 8.2.5 进行初步的规模估计
在物理设计的早期阶段，数据仓库团队常常不得不去估计仓库的规模，此时尽快对整个数据仓库的存储需求做一次估计是非常有用的  
尽管在理论上我们可能更希望使用建模工具来进行规模估计，但是在项目的当前阶段那些工具往往都不便于进行类似操作，我们通常被迫使用数据模型中的注释和简单的电子表格来存储考虑到的事项  
注意：在几乎所有的数据仓库中，维度表的大小和事实表的大小相比都是微不足道的，第二个要考虑的事项是事实表中索引的大小  
进行初步规模估计的相关工作包括：
+ 估计事实表和大型维度表中行的长度，要考虑到varchar和可为空的列中的内容一般都比列的最大长度要小，我们通常从源中抽样1000或者更多的行进行合理的估计 5  
+ 对每个表来说，为整个历史装载估计所需的行数（如果历史装载规模仍然有争议就按照历史信息的年份估计），并估计在产品阶段每次装载所增加的行数
+ 对于每个标准RDBMS，要考虑到索引所占的空间和基本数据所占的空间是差不多的，在设计初始索引方案之后要进行更为准确的估计  
+ 临时/排序空间，一般需要一个大小为索引两倍的临时空间来构建该索引，通过使用分区表就能降低临时空间需求  
+ 聚集表占用相当多的空间，这取决于数据库中存储的聚集的程度、数据的稀疏程度和层次结构的深度，平均来看，聚集和其索引占据的空间和原子层事实表所占的空间一样大  
+ 和聚集表一样，OLAP多维数据集也占据了相当大的空间，OLAP数据库技术的存储数据的效率上差异很大，大约是未加索引的关系数据库的40-300% 5  
+ OLAP数据库可能需要一定的额外空间来支持多维数据集处理，不同的OLAP厂商在如何进行处理过程的控制方面有所不同，因此在初始规模估计中这里只给出了一些一般性的指导原则  
+ 总而言之，最常用的一条经验规则就是，整个数据仓库占据的空间一般为原子星型模式中数据的3到4倍  

也有一些其他的数据库也是整个DW/BI系统的一部分，它们当中的大多数都非常小，但是有些数据库也可以增长到很庞大  
+ 为元数据表预留20MB的空间，这应该绰绰有余了  
+ 登台数据库和文件系统存储需求差别很大，这和如何进行ETL系统的设计有关，一定要在方案中考虑到历史信息的装载也需要存储空间，那要比运行过程中的增量装载所需的空间大得多 5
+ 可以开发一个日志记录数据库用于记录DW/BI系统的操作情况，要为日志记录数据库预留至少10GB的空间，如果要跟踪更多用户的详细活动情况，则需要的空间也会显著增加  
+ 很多BI工具需要将数据库作为其操作的一部分，数据库中保存了报表定义、用户和使用情况信息等，一啊不能用于调度和分发报表  
    通常如果没有报表快照进行存档的话，可以为BI工具数据库预留10GB或者更少空间  

## 8.3 创建开发数据库
在ETL工作开始之前需要适时建立一个开发数据库，一般要等到模型设计一到两天才变更一次时再开始考虑建立该数据库  
刚开始不要担心开发数据库的物理细节，因为初始开发阶段要处理的数据量相对较少  
指派专人作为开发数据库的管理员很有好处，他应该密切关注磁盘使用情况和共享开发数据库的一般运行情况  
在开发周期中需要对数据库结构的变更进行管理，最关键的是要确保每个团队成员都能够正确地参与沟通，并且能够提醒其他团队将要进行的变更  
当开发接近完成的时候，团队就要转向测试和部署方面的工作
## 8.4 设计处理数据存储
还要一些其他类型的表需要作为整个BI系统的一部分来进行设计和部署，主要包括：  
+ 支持ETL系统的登台表，登台表常常被设计用来为日常处理保留当前这一天的数据，为了今后能够有最大的灵活性，应该将登台表分割到不同的数据库或者模式中  
+ 用于ETL处理和数据质量的审计表，在第9章提到，构建一个ETL审计系统的方法很多，审计表通常位于数据仓库数据库模式中，这是因为事实表和维度表都包含有参考审计表的外键
+ 用于监视用户访问情况的访问监视表，第13章将会讲到，应当监视对DW/BI系统的访问情况，大多数机构都会创建一个日志记录数据库（或者模式）或者监视数据库（或者模式）  
+ 安全表  
    如果对每个业务用户的信息访问都进行筛选，使之仅能够访问一部分数据（例如只能访问某个特定销售区域或者某一组职员的信息），那么就需要在数据仓库中定义额外的数据结构来支持这种访问 5  
    预定义的报表可以阿静事实表和恰当的安全表联接起来，以便为每个用户仅返回他所需要的行  
    对即席访问来说，应当创建一个视图将安全表和事实表联接起来，以业务用户的ID作为过滤条件  
    OLAP数据库所含有的安全特征使得定义复杂的安全性角色相对简单一些，但即使依赖OLAP并且将其作为主要呈现服务器，也同样可能需要创建这里所说的安全表  
    
在开发周期中上述这些表经常会演化，甚至在部署阶段也是如此  
## 8.5 设计初始索引方案
要习惯于将这一阶段的项目开发中设计的索引方案作为初始索引方案，并做好准备在数据仓库生命周期中不断调整索引，因为用户使用风格改变之后，您需要对使用风格有搅好的了解并通过调整适应它  
警告：在雇佣索引方面的专家时，查证其在维度数据仓库方面的经验和业务知识尤为重要  
本节简要分析了主流RDBMS所使用的不同类型的所以和查询策略，随后较为详细地讨论了用于事实表和维度表的所以策略  
### 8.5.1 索引和查询策略综述
为了设计有效的索引方案，最重要是理解RDBMS查询优化器和索引是如何工作的，以及数据仓库和OLTP需求有哪些不同  
1. B-Tree索引  
    为RDBMS设计的第一种索引是经典的B-TREE索引，这种索引对高基数的列特别有效（order_number、customer_key) 5  
    B-Tree索引将可能值构建成一棵树，并且给出一个存放叶节点值的行的ID构成的列表，要找出含有给定值的行就要遍历树，用给定得值和每个分枝上的值进行比较，给定值较大，则查找上一分枝，反之亦然    
    B-Tree是大多数数据库默认的索引类型，有些RDBMS可以根据每个表声明的主键自动生成唯一的B-tree索引，通过该索引可以强制实施主键约束  
    为了便于查询，有些数据库引擎在一个表上仅使用一个B-tree索引，也有一些数据库在一个查询中使用多个B-tree索引  
2. 聚簇索引  
    有些数据库引擎使用聚簇索引，聚簇索引基于表的键值来排序和存储数据行，每个表只能由一个聚簇索引，聚簇索引很适合应用于高基数的列或列集合  
    事实上，聚簇索引通常是唯一的或者几乎是唯一的，表的主键索引一般都默认定义为聚簇索引，对其他索引来说，聚簇索引的键值通常是行标识符，聚簇索引对查询优化器的工作大有裨益 5  
3. 位图索引  
    很多数据库引擎与B-Tree索引相反，更适用于低基数的列，位图索引在具有星型模式的反规范化维中非常有用  
    最经典例子就是性别代码或者“是/否”指示符索引，如果行中含有位串所表示的值，则该位串中的每一位都设置为1，否则就设置为0  
    位图索引本质上是关于列的每个可能值的位串，同一列上的位图索引通常要比B-tree小很多  
    尽管数据库优化器能够在查询中使用多个位图索引，但是通常位图索引只能建立在单个列上，位图索引一般都用于维度表的属性和层次关系  
    提供位图索引的数据库引擎在技术上也在不断地改进，因此位图索引也逐渐可以应用于基数较高的列了 5  
4. 其他类型的索引  
    有些RDBMS使用其他一些专用的索引结构或者优化策略，这其中有很多都非常适用于数据仓库应用，事实上，可用的索引和查询优化技术以及它们在数据仓库应用环境中的适用性是RDBMS平台的选择过程中最重要的参考因素之一  
5. 星型模式优化  
    尽管关系数据库的效能有所不同，但是大多数关系数据库在提供星型模型优化时都会遇到星型查询所固有的n-way联接问题  
    您一定希望对维度星型模型进行规范化的查询，首先从包含查询条件的维度表开始，随后给出能够满足查询条件的一个维度组合键列表，接下来优化器可以根据该列表利用事实表键上的索引从事实表提取合适的行  
    这种方法实际上需要对不相关的维度表求笛卡尔积，对星型模式查询来说，求笛卡尔积要比通过事实表将各个维度表联接起来高效得多  
    有些支持星型模式优化的RDBMS需要特定的索引才能实现优化过程，也有一些RDBMS需要确定星型方法是否和查询的结构相适应 5  
    
### 8.5.2 为维度表建立索引
维度表应当有单独的一列作为主键，这样就可以在该键上建立唯一索引，如果RDBMS中支持位图索引，可以为哪些常用作筛选条件的维度属性每列添加一个位图索引  
如果不支持位图索引，应该评估一下在维度属性上创建一些B-tree索引是否有用，有些关系引擎查询优化器在同一查询中可能会用到多个B-tree索引    
为小型维度表添加过多的索引并不会有多少好处，如果业务需求按时了这些查询执行的频率很高，而且维度表又非常大，那么在最常用筛选条件的维度属性上建立一组索引就显得非常有价值了  
和事实表一样，有多种索引类型的直列型DBMS也有一些简单的所以规则：为每个能够用于联接条件、少选条件或者列分组的列都单独建立索引  
### 8.5.3 为事实表建立索引
事实表的首选索引是建立在主键上建B-tree索引或者聚簇索引，大多数数据仓库查询都是以日期作为约束的，日期键在主键索引中应当是第一位的 5  
除了主键索引之外，其他事实表所以的确定主要取决于RDBMS中支持的索引类型和优化策略  
如果选择的RDBMS不允许在一次查询中使用多个索引，这种情况下，必须在事实表列上创建多个组合索引，随着模式复杂程度提高，事实表键索引的数量也明显增加，一个八维模式可能有6个以上的键索引  
不同于单列的键索引，最初的索引方案通常没有事实表索引，一般来说，非键事实表所以都是单列索引，其使用频率较低，而且和键索引相比作用比较小  
如果维度关系数据库支持适度的查询——既可以是预先定义的报表，也可以是即席查询——则应该设计一个聚集策略，本章稍后会讲到这一点  
### 8.5.4 为装载数据(loads)建立索引
我们时常会担心索引对BI查询的通用些问题，然而大多数成功的数仓面临的压力事实上来自于如何保证系统的装载和维护周期尽可能简短 5  
如果一次装载量超过表规模的10%-20%，那么在插入、修改数据之前撤销索引并且在此后重建所有索引通常更为高效，在决定如何取舍之前最好对自己的环境进行简单测试  
### 8.5.5 为OLAP建立索引
将OLAP数据库作为可查询的数据存储已经越来越流行，创建OLAP的标准方法：创建一个维度关系数据库，并且使用它为OLAP提供输入  
一旦已经开始创建OLAP数据库，就要捕捉哪些在维度关系数据库上执行的用于维度处理、完全多维数据集处理和增量多维数据集处理的查询，然后为关系数据仓库添加索引，使之能够支持OLAP处理  
在少数情况下，业务用户需要直接访问OLAP数据库，而且OLAP中所有标准报表都作为数据来源，在这种工作环境下，可以为关系数据仓库添加的索引很少，一般需要的索引仅仅是为了支持ETL和OLAP处理  
和关系数据库一样，OLAP数据库引擎也需要使用索引，它们也有查询优化器，但OLAP技术对数据库管理员是不透明的，您一般很少能够控制 5  
### 8.5.6 在装载之后分析表和索引
有些数据库引擎在每次装载过程完成之后不会对表和索引自动进行统计和重新计算，有些甚至在重建一个完整的索引之后也不会进行验算  
这对任何装载和索引创建过程都是非常重要的步骤，因为查询优化器必须拥有关于表和索引大小的准确信息，以便设计出高效的查询方案  
## 8.6 设计OLAP数据库
OLAP数据库设计很大程度上依赖于选择的OLAP技术，OLAP市场仍然很不成熟，产品在功能和伸缩性上都存在很多变数  
所有的OLAP数据库都是强维度的，它们具有和维度关系数据仓库中的维度表和事实表相类似的维和事实，对OLAP实施方案来说，这些层次关系的重要性要远远高于关系数据库  
OLAP从业人员首要关注的就是OLAP服务器对规则层次结构的挑剔程度：层次结构中某一层次上的属性仅能够上卷到一个更高层次的某一个属性  
基本的OLAP事实和维度关系数据源也十分相似，您可以定义一些直接来自于维度关系数据库的基本事实  
### 8.6.1 OLAP数据粒度和深入钻取
OLAP数据设计中首要的关键决策是对数据粒度的确定，如果不以具体产品为例，则很难提供什么建议，因为OLAP数据库在伸缩性上差别很大 5  
将某种应用程序（如OLAP数据库）中的信息排除在外，而后有时用户又会用到这些信息，这种情况通常是不可避免的  
在所有OLAP技术中，都可以通过预定义深入钻取路径来使数据集显得比其原来更大，这种预定义使业务用户访问能力从多维数据集延伸到源数据库-通常是维度关系数据库  
如果OLAP多维数据集中实施深入钻取，就需要提供深入钻取查询的关系数据库被正确索引  
### 8.6.2 完善OLAP维度
尽管维度的基本结构从本质上来说是从关系维度表产生的，但是仍然可以加入更多的元数据，至少可以加入业务名称和描述性注释，这些几乎总是作为工具提示信息提供给业务用户  
在一个维度上可能定义多种层次结构，例如，许多日期维都包括一个日历版和一个财政版的层次结构 5  
有时您可能希望向OLAP中添加属性，这些添加的属性是从维度表中存储的属性计算而得到的，如果用户主要访问OLAP数据库，那么仅在OLAP实现属性的添加更为容易些  
一次性构建OLAP维度，加工这些维度并且仔细检查它们，维度对于已完成的DW/BI系统来说已经成为主要的用户界面要素，在构建维时，经常会发现一些数据问题，有时可能需要进行额外的数据清洗工作    
有些OLAP工具需要将维度数据源换成星型结构，有些则需要将维度层次规范化为雪花结构，但大多数OLAP工具都能接受这两种结构，创建额外的维度表副本来支持OLAP工具的结构性需求是非常简单的，而且也是非常必要的  
### 8.6.3 定义OLAP计算
基本多维数据集可能包括和事实表粒度一样的事实，或者也可以从某个简单的聚集（如求和）中创建  
设计OLAP数据库的下一个主要步骤是添加计算事实或度量，将计算迁移到OLAP数据库中完成了两件工作：使计算的定义一致，所有OLAP应用程序、用户都可以使用这些计算；将计算迁移到用于高性能计算的数据库服务器平台上 5  
对计算的定义是OLAP技术的另外一个优点，OLAP数据库都使用功能比SQL语言搞一个数量级的语言来定义复杂计算  
除了定义新的计算度量（如收益率），OLAP数据库还提供了一个场所对用于度量的默认聚集规则进行定义，有些事实仅在某个维度上可加，作为替代可以制定一些规则，如平均或者本期末  
## 8.7 建立测试数据库
现在，我们可以将注意力转移到一些与性能相关的主题上，为了设计聚集、建立索引和配置物理存储，需要可观的数据量、高质量硬件和有关使用风格的详细信息  
现在项目中又有了新的工作，要开始扩展测试服务器，装载和测试历史数据，针对实际数据开始测试增量装载过程，测试服务器应当尽可能在各个方面都和产品服务器相似  
## 8.8 设计聚集
在大型数据仓库中最能够提供性能的做法是提供一系列合适的聚集，或者给出一些可以与原子粒度事实共存的摘要记录，聚集导航器通常是数据库引擎的一部分，不过有些BI工具也具有这样的功能  
简单来说，聚集就是可以通过执行带有GROUP BY子句的SQL查询生成的汇总表  
### 8.8.1 确定如何聚集
我们已经将聚集简单地描述为汇总表，可以由执行GROUP BY的SQL查询来生成，但是还有很多其他可以用来实现和管理聚集的技术    
Oracle数据库的实施通常依赖于物化视图，物化视图这个名称很恰当，它们实际上是物化的视图定义或物理存储，它们产生出数据集    
通过在关系数据库中定义维度结构改善Oracle内置聚集导航的效能 5  
另一种实现关系聚集的方法是使用OLAP引擎，聚集管理和导航是OLAP的核心职能——它们早就被作为OLAP引擎的第一特征，并且引导OLAP引擎发展了数十年  
其他每一种数据库引擎都有自己管理聚集的一套机制，虽然技术细节不同，但是基本原理都是相同的  
另一种使用数据库引擎内置的聚集管理和导航功能的方法是BI查询和报表工具，这些工具又有聚集导航的功能    
我们认为对于呈现服务器的数据库平台，要将聚集管理和导航功能作为一条重要选择标准  
### 8.8.2 确定聚集的内容
好的聚集设计既不会聚集太多的内容，也不会聚集太少的内容 5
有关聚集规模的另一个极端是系统中创建的聚集过少，可能只有一些月度聚集和少了其他种类的聚集  
要确定创建什么样的聚集，最好的方法就是监视查询，设计出和查询风格相匹配的聚集  
聚集设计的一个小技巧就是将潜在聚集的规模（越小越好）和其他使用情况平衡起来，对聚集的使用既包括直接使用它的查询也包括间接使用到它的查询  
即使使用查询日志并且对业务用户的查询风格有了足够透彻的了解，确定聚集的最优集合也是一个非常困难的问题  
### 8.8.3 维护聚集
在创建聚集时，除了一开始的创建工作之外，还必须随着时间推移维护它们 5  
当新的日数据被装载到原子事实表中时，聚集也必须进行更新，您一定不希望看到对聚集的查询所返回的结果和统一查询对原子表所返回的结果有所不同  
如果手工维护汇总表，则ETL过程必须增加一个步骤，保证汇总表为最新
如果维度表对聚集中使用的属性进行类型1属性覆盖处理，就会出现问题，设想我们已经有一个存有产品SKU层日数据的简单事实表，且已经按月和产品子类别定义了聚集，如8-4所示  
![avater](8-4.png)
我们考虑两种情况：  
1. 假设产品子类别的一个描述从“男鞋”变为“鞋类”，对业务用来说，这种变化并不重要，他们仅想看到报表中最新的描述，而并不在意以前的描述，因此按类型1覆盖属性来对待，因此就可以将其作为类型1覆盖属性来对待 5  
    对这种变化的处理可以通过仅更新维度表来进行，不需要触及原子事实表或者聚集  
2. 如果现有SKU被重新分配到另一个不同的产品子类别“男士用品”，在细节层是没有问题的，但是原子数据和聚集数据现在就出现了不一致，因为聚集事实表中所有的历史行都讲SKU销售事实汇总成原先的子类别了  
    这种从细节到聚集的不一致是不可接受的，同样需要对以前聚集的事实行进行更新和重新计算，这样它们才能和产品子类别记录的详细事实所产生的即时汇总信息保持一致  
    
最小化上述这种裂痕的显著方法是避免在类型1属性上创建聚集，但是可能会导致无法达到最优的查询性能  
注意：类型1维度属性变化实际上会导致建立在该属性上的所有聚集都破裂，无论聚集的形式是汇总表、物化视图还是OLAP聚集 5  
### 8.8.4 完成索引
当在测试服务器上测试性能和设计聚集时，不可避免地要再次关注索引方案，可以参考第13章，了解如何进行性能测试，以及索引方案演进之后需要密切关注的问题  
## 8.9 设计和构建数据库实例
数据仓库的呈现层应当存在于硬件服务器上属于自己的RDBMS实例中，因此才能对系统汇总的很多参数（如内存）进行优化，使之达到数据仓库的需求  
有一些参数对数据仓库数据库的性能是非常重要的，要准确记录每个参数的情况，并且要注明当数据库增长和演进后，哪些参数最可能需要调整以及为什么要调整  
在数据库的各种参数中需要关注的包括内存设置和数据库恢复模型，而且要确保数据库使用的恢复模型与备份策略相匹配，数据量很大的情况下，要确保数据设置为能够支持大批量数据装载的状态  
### 8.9.1 内存
大容量内存对数据仓库查询性能很有好处，因为大多数查询都是多步处理过程  
有些数据仓库应用可以在内存中保留一个关于数据的工作集，很多用户都可以访问内存的速度来访问这些数据，它可能是最近一段时间的维度表，并且该数据集很小，适合装入内存，有助于改善查询性能 5  
### 8.9.2 块大小
有些RDBMS允许数据库管理员选择数据库的块大小，一般是从2KB到32KB，很多数据仓库的操作执行起来使用大块更快，例如数据装载和表扫描，其他操作，如常用的检索分散在各个事实表中的数据查询用大块效率很低  
这个问题的解决方法和那些数据仓库配置方面的问题一样，取决于系统的独特环境  
### 8.9.3 保存数据库构建脚本和参数文件
大多数RDBMS都运行DBA使用支持点击操作的图形用户界面（GUI) 来配置和建立数据库，但它们自带的文档记录功能一般并不是很好  
我们建议至少应当设定一个针对数据字典表的脚本程序，使之每天晚上运行，从实际的模式中生成表的定义脚本并且存档  
任何改变呈现服务器上物理表布局的团队成员都应该首先用数据建模工具对这些改变进行建模，并且最好使用工具来促成这些改变，而不是通过在数据库控制台输入或者使用鼠标点击图形界面来实现 5  
应当从数据模型开始，从模型生成表定义脚本，并且并按照团队的代码管理标准来编辑和保存该脚本  
在ETL处理管道触发器中过早改变物理表需要担负较为复杂的跟踪职责   
同样要强调一下在DW/BI环境中使用代码管理工具的好处  
## 8.10 设计物理存储结构
在我们已经讨论过的数据机构之下还有一个和物理模型相关的层次：块的存储结构、文件、磁盘、分区和表空间或者数据库  
### 8.10.1 计算表和索引的大小
正如本章早些时候提到的，我们建议在物理设计过程的开始阶段对磁盘需求进行一次粗略的估计  
现在逻辑设计和物理设计都结束了，是该将数据存入磁盘的时候了，进行准确的计算是非常必要的    
为规划存储需求的活动确定时间段，但是要记录并了解超过该时间段的增值性需求，一般来说，将初始数据装载规划为固定大小，并至少加上六个月的运行装载增长期  
### 8.10.2 设计分区方案
如果RDBMS支持分区表，且数据量很大的情况下，应当对事实表分区。实现分区表的最好方法是按照活动日期来分区 5  
划分事实表有几个方面的好处，分区最大的优点表现在表的装载和维护操作中  
装载时：表分区使得在增量处理过程中实现快速装载成为可能  
分区在装载过程中表现出来的优势是可以快速装载某个单独的分区  
维护时：有些数据仓库联机保留了一个数据的滚动窗，如3年到5年的数据，通常，一般每个月或者每天都要从表中删除最旧的数据，进行了分区之后，就可以简单而快速的删除最旧分区  
最后，大多数数据库引擎都提供了可以将某些分区声明为只读的选项，这对于进行快速增量备份非常重要，对于非常大的表来说，它们的可维护性就有了巨大的差别 5  
为了充分利用表的分区，一般必须声明标准，规定如何进行划分，以及如何对索引进行类似的划分  
现在，不同的RDBMS中已经有了好几种可用的分区方式，多列分区可以切实改善查询性能，但是相对于单列分区的处理来说更为复杂  
为了在增量装载中利用分区的优势，需要随着时间的不断推进，设计一些脚本来自动创建新的分区，并且将这些分区存放到适当的文件或者表空间中  
分区有助于解决很多棘手的问题，但是却大大增加了系统的复杂程度，如果有一个中等大小的单个事实表（几百GB以上），就应当对齐进行划分  
如果数据仓库相对较小，划分所带来的收益就可能抵不上额外开发、测试和许可等方面的工作所花费的成本 5  
分区能够显著提高数据仓库的性能和可维护性，并且从头开始实现分区并没有想象的那么困难  
### 8.10.3 设置存储
在何处存储数据以及如何存储数据会对系统性能和可管理性产生较大的影响  
DW/BI系统显然是数据密集型系统，因此需要大量的存储设备  
为了使系统性能最大化，数据应当尽可能分散到较多的磁盘上，这些磁盘应当由尽可能多的硬件控制器来进行管理  
1. 故障容错  
    强力建议您的磁盘系统具有故障容错能力，最好能够达到完全冗余存储，如果项目对价格很敏感，也可以为大多数数据库配置RAID5，而系统表和事务处理日志应当进行完整镜像处理 5  
    要记住容错配置不能代替常规的系统备份  
2. 存储区网络  
    越来越多的机构开始使用存储区网络（SAN）体系结构来进行企业数据存储，SAN可以提高存储效率，因为多个应用可以共享相同的存储池  
    有些机构考虑到性能方面的原因而不愿意使用SAN，尽管直接进行磁盘管理达到的最佳性能确实会比SAN的性能稍好一些，但是SAN便于管理的优势使其成为许多机构的必然选择，然后，在设置SAN时最重要的是要获得专家的协助  
3. 卷和驱动器配置  
    不考虑容错等级的问题，最便于管理的配置就是定义一个单独的阵列卷来保存所有的数据库表和索引，这种配置实现了非常好的查询性能，并且很适合于小型数据仓库  
    如果没有使用RAID技术，DBA和系统管理员需要充分了解访问的风格，以便推算出哪些表和索引会被同时访问 5  
    当数据和索引的总量超过了单个控制器的能力，就需要建立一个多卷的系统。在多卷系统中对数据分段是一种好办法，可以改善装载性能  
    请记住避免I/O瓶颈的最好方法是尽可能使用较多的磁盘控制器，显然，由少量控制器和大容量磁盘构成的系统成本相对较低，因此需要找出系统的成本、性能和可靠性之间的最佳平衡点  
    
## 8.11 小结
由于项目和项目、平台和平台之间的差异，物理实现的细节也会有所差异，项目的数据量是最主要的驱动因素之一，完成一个好几TB数据量的数据仓库远比设计一个50GB的数据库更加复杂  
团队所能利用的硬件、软件和工具既能够使物理实现方面的工作更为平稳地展开，也可能会使这项工作变得更为困难  
## 8.12 管理工作和降低风险
物理数据库设计过程中最大的风险就是团队一开始会有假定设计是固定的倾向，物理设计在开发周期中会不断演进，和性能相关的设计元素甚至会在系统产品化后继续演进  
进行最初数据库设计的工作简单明了，并且风险较低，尤其是在开发环境当中更是如此，当ETL系统设计开始后，这项额外的列和表的准确结构才会显现出来    
为了在ETL系统的开发开始之后最小化对数据模型的更改，要确保在创建物理数据库和表之前调查数据的内容  
如果系统进行工作特性的测试，那么数据的索引和聚集策略可能会改变，即使DW/BI系统已经成型，索引和聚集表设计仍然可能随时间推移而改变  
## 8.13 保证质量
为了提交高性能的系统，一旦ETL系统建立并且数据已经装载，就必须设计一套更改索引和聚集策略的规程，当获得更多有关业务用户实际上如何使用数据库的信息后，应当随时间推移不断地修改  
OLAP技术和其他前台BI工具都会具有帮助您构建性能更好的索引和聚集的功能，要确保提高性能的建议都是基于系统使用情况的，而不要仅从理论上去考虑    
## 8.14 关键角色
完成物理设计过程的主要角色包括：
+ 数据库管理员应当充任项目这一阶段主要驱动者  
+ 业务分析员、数据模型师、项目经理和业务项目领导者都必须支持数据库管理员的工作，为其提供建议并且在出现问题的时候积极协助解决
+ 安全经理在项目生命周期的早起阶段就应当参与进来，不要幻想可以在项目结束的时候再去考虑安全性方面的事项
+ OLAP数据库管理员和关系数据库管理员可能并非由同一人兼任

设计和实现物理数据库所需的工作量取决于项目的复杂程度

+ 如果事先没有规定统一的命名约定，那么物理设计就非常花时间，这些问题应当在维度建模的过程中加以解决，在更坏的情况下，必须立即解决，不然会在很长一段时间内给数据仓库带来麻烦  
+ 数据库规模和技术环境的复杂度对工作量起主要影响，当复杂程度增加时，物理设计过程复杂度呈乘法增长
+ DBA技能水平和对目标平台的经验同样也会影响时间表 5
+ 复杂的安全模型会增加完成物理设计所需时间

# 9 抽取、转换和装载介绍
![9](9.png)
在建立DW/BI环境的过程中，后台的抽取、转换和装载(etl)系统常常要花费70%的时间和工作量  
每个人都需要知道的三个字母：将数据从它的源系统中的位置取出来(E)，对数据进行一些转换(T)，并将数据转载(L)到最终的表集合中供用户查询  
设计和建立ETL系统的最优方法，取决于源系统、数据的限制条件、可用的脚本语言和ETL工具、职员的技能，而且还取决于BI工具  
随着成千上万个数据仓库的成功，大量经验的积累，已经出现了一些了ETL最优方法 5  
对这些最优方法进行认真的考虑之后，我们发现几乎所有的维度数据仓库后台都需要34个子系统，难怪ETL系统会消耗生命周期中的大部分资源  
ETL系统是DW/BI项目的基础，它可以使数据仓库项目成功，也可以使数据仓库项目中断  
本章讲述的ETL架构这一主题在数据仓库ETL工具箱一书中讲得更为详细  

## 9.1 归拢需求
从最困难的挑战之一出发确立ETL系统的架构：归拢需求  
ETL系统需求主要是一些您必须用到的约束，即必须使系统符合这些约束 5  
### 9.1.1 业务需求
从ETL设计角度来看，业务需求就是DW/BI系统用户的信息需求  
很多情况下，在维度建模的过程中进行初始业务用户访谈和数据源调查都无法完全揭示出数据的复杂度和限制，ETL团队有时会有重大的发现，会对业务需求能否按照最初希望的那样得以解决产生影响  
即使在ETL系统的技术后台开发中，也必须保持ETL团队、数据设计师、模型师、业务分析员和业务用户之间的对话  
### 9.1.2 合规性
法规和报表需求会迫使很多机构严格制定他们的报表，并且提供证明来说明报表中的数据是正确和完整的，并且没有经过任何篡改  
有些合规性问题会超出DW/BI系统的访问，但是大多数合规性问题都会在系统索要考虑的范围之内，一般数据仓库中需要谨慎处理的需求包括： 5  
+ 保存数据源和随后数据临时的存档副本  
+ 为改变任何数据结构的完整的事务处理流程提供证明  
+ 完整记录用于分配、调整和推导的算法  
+ 随着时间推移为数据副本的保密性提供证明，包括在线和离线两种  

### 9.1.3 数据质量
有三股极有影响的力量聚焦到数据质量问题上，它们促使总经理将数据质量问题排在几乎最靠前的位置上加以考虑： 5  
1. “如果我能够找到数据，那么就可以把自己的业务管理好”这样的理念是一种长期的文化趋势，这种想法在不断升级
2. 大多数机构都明白自己的数据源是完全分布式的，一般都分散在世界各地，因此需要获得对各种不同数据源进行有效集成的方法
3. 对合规性需求的急剧增长意味着那些粗心大意地对数据进行处理的方式已经不再被忽略和原谅

这些极有影响力的思潮汇聚到一起就讲数据质量问题突显出来了，在这种背景下，ETL团队需要更加灵活和主动——数据质量是无法由ETL系统独立解决的  
### 9.1.4 安全性
在过去的几年里，IT届的安全意识以及有了显著的增长，但是对于大多数DW/BI团队来说，很多时候都是在系统建成后才想起安全性问题，而此时安全性已经成为了一种负担 5  
数据仓库行业的主旋律对安全性的心态是矛盾的，数据仓库本身追求如何向决策制定者广泛发布数据，而安全性则要求对数据进行限制，只有那些需要了解的用户才能访问数据，而且，安全性必须扩展到物理备份中  
在归拢需求的过程中，DW/BI团队应当寻求高级管理人员的指导，了解数据仓库哪些方面对安全性特别敏感，合规性需求和安全性需求很可能是相互重叠的，最好将这两个主题综合起来考虑  
### 9.1.5 数据集成
数据集成是IT界的一项重大课题，因为数据集成的最终目标是要使所有的系统无缝联接、协调工作  
数据集成通常表现为数据仓库中的一致性维度和一致性事实  
一致性维度是指确立整个业务过程中公共维度属性，这样就可以使用这些属性生成“横向钻取”报表 5  
一致性事实是说对独立的公共业务指标达成一致，如关键绩效指标(KPI)等，这样的话，就可以通过计算差值和比率来比较这些数值了  
### 9.1.6 数据等待时间
数据等待时间描述了源系统数据通过DW/BI系统提交给业务用户的速度有多快  
但是有时，如果对数据等待时间的需求非常急迫，那么ETL系统的架构就必须从批处理方式转换为流处理方式了  
### 9.1.7 存档和沿袭
在前面有关合规性和安全性的内容里已经暗示了对存档和沿袭的需求，即使没有保存数据的法规需求，每个数据仓库都需要很多旧数据的备份，既是为了和新数据进行比较以生成变化捕捉报表，也是为了便于今后的再处理    
我们推荐：所有的登台数据都应当进行存档，除非有明确的决策说明某种特定的数据集今后不需要再恢复了 5  
当进行存档时，每一个临时/存档数据集都应当有相应的元数据描述其来源和产生该数据的处理步骤  
### 9.1.8 用户提交界面
ETL系统的最后一步是将数据移交给BI应用程序，我们对这个数据移交过程的立场是一定要坚决遵守规定，ETL要和建模团队协同工作，必须对数据的内容和结构负责，这会使BI应用程序的工作变得简单快捷  
ETL团队和数据模型师需要和BI应用程序开发人员紧密合作，确定出数据提交的准确需求  
### 9.1.9 可用的技能
ETL设计决策必须基于建立和管理系统时能获得的人力资源来制定  
需要考虑一下到底是手工编写代码来生成ETL工具还是使用厂商的开发包，是一项重要决策 5  
### 9.1.10 遗留许可证
最后，很多情况下在主要的设计决策指定过程中，高级管理人员会含蓄地坚持使用已有的遗留许可证  
在很多情况下，这种需求都可以接受的，因为这样做的优点是团队的每一个都对已有的环境非常熟悉，但是在有些情况下，在ETL开发过程中使用遗留许可证却是错误的做法  
## 9.2 ETL系统的34个子系统
尽管我们采用业界统一的称谓“ETL”来描述这些步骤，但是整个过程实际上应该分为四个主要步骤：
+ **抽取**  
    在ETL环境中从源系统收集原始数据并且在任何对数据的重要重构发生之前都将数据写入磁盘，子系统1-3都支持抽取过程  
+ **清洗和一致化**  
    通过ETL系统中的一系列处理步骤发送源数据，这些步骤提高了从源系统接收数据的质量，并且对两个或多个源系统中的数据进行了合并，创建并强化了一致性维度和一致性度量，子系统4-8描述了支持清洗和一致化过程所需的架构  
+ **提交**  
    从物理上将数据组织和装载到呈现服务器的模板维度模型中，子系统9-21 提供了若干种将数据提交到服务器上的功能  
+ **管理**  
    用一致的方式来管理ETL环境中的相关系统和过程，子系统22-34都描述了支持ETL系统现行管理所需的组件 5  

## 9.3 抽取数据
毫无疑问，ETL架构中最初几个子系统解决的是了解源数据和抽取数据的问题，并且将数据传输到数据仓库环境中，这样ETL系统就能够独立于工作系统对这些数据进行操作  
### 9.3.1 子系统1—数据探查
数据探查是对数据的技术性分析，对数据的内容、一致性和结构进描述，从某种意义上讲，任何时候您在某个数据库字段上执行select这种调查性查询时，就是在进行数据探查  
现在有很多专门设计用来对数据进行有效探查的工具，将经费花在购买专用的工具上可能要比花钱开发自己的探查工具好得多，因为主流工具可以探查到多种数据关系，并且用户界面操作简便  
数据探查担负着两种不同的任务：战略性的和战术性的  
一旦确定了某个候选数据源，就应当进行一次数据探查评估，针对早起的采纳/不采纳问题提供决策，较早地早出那些不合格的数据源是一个责任重大的步骤，即使带来的是坏消息，也将使您赢得团队其他成员的尊重 5  
一旦将某个数据源引入项目的基本战略决策已经定下来，就需要进行一系列战术性的数据探查工作来尽可能多地确定出各种问题，通常这一工作从数据建模过程就开始了，一直到ETL系统设计过程  
该子系统揭示出来的问题最终会产生两种详细说明：1）将数据送回原来的数据源，请求改善数据质量；2）构成了数据质量子系统的需求  
探查阶段为ETL团队提供了指导，告诉他们需要使用多少数据清洗机制，并且使他们不会因为创建处理脏数据的系统分散了注意力而遗漏项目的主要环节  
### 9.3.2 子系统2—变化数据捕捉系统
在数据仓库最初的历史装载过程中捕捉源数据内容变化并不重要，因为您只是装载某个时间点以前的所有数据  
大多数数据仓库表都很大，不能在每个ETL周期中都进行更新，因此系统必须具有增量更新的功能，仅在当前源数据和上次更新后的数据相比发生改变时，才传递相关的变化 5  
这种将最新的源数据分类出来也叫做变化数据捕捉（CDC），变化数据捕捉子系统的主要目标是：  
+ 可以将变化的源数据分离出来进行有选择的处理，而不是进行完全更新
+ 捕捉对源数据的所有变化（删除、编辑和插入），包括那些通过非标准接口对数据做出的改变
+ 使用原因代码来区别实际更新与错误的更正
+ 支持对其他元数据的合规性跟踪
+ 尽可能早地实施变化数据捕捉，最好在大量数据传到数据仓库之前就完成相关工作 5

捕捉数据变化并不是一件小事，必须针对每个数据源都仔细评估自己所采用的策略是否恰当，下面给出集中捕捉数据源变化的方式，每一种在适当的情形下都是有效的：
+ **审计列**  
    有些情况下，源系统可以追加审计列来存储某个记录被添加或者修改的日期和时间  
    可以通过源系统的应用程序取代数据库触发器来填充列，必须对他们的完整性加以特别的关注，对每一列都进行分析和测试，以保证该源系统是能够表明变化的可靠源，如果出现空值，则必须寻找探测变化的其他途径  
    如果借助于源应用程序填充字段并且DBA小组允许使用后台脚本修改数据，那么ETL系统就无法使用审计列，如果允许这样的情况发生，那么在增量数据装载的过程中就会面临遗漏变化数据的高风险  
+ **定时抽取**  
    借助定时抽取，一般可以选择所有创建或修改日期字段中的值都等于SYSDATE-1的行，这表示所有以前的记录，这种操作是错误的 5  
    单纯基于时间装载记录（根据修改日期值等于SYSDATE-1的行）非常不可靠，因为如果装载过程中断，那么重新开始装载就会装载重复行，如果由于运行失败而跳过一天，那么带来的风险就是遗漏的数据将不会再被装载到数据库中  
+ **完全“差异比较”**  
    完全差异比较保持了对以前数据的完全快照，并且将该快照和当前数据记录一条接一条的比较，找出存在变化的地方。  
    这种技术的好处在于，可以确保找到每一处变化；但是它显然也有缺陷，大多数情况下这种技术是资源密集型（非常消耗资源）  
    如果不得不进行完全差异比较，就尽量在源系统机器上做比较，就不需要将整个表或库传输到ETL环境中了  
    同样可以使用循环冗余校验算法来进行检查，如果某一条复杂记录发生改变则可以迅速发现 5  
+ **数据库日志刮取**  
    日志刮取能够在预定时间及时有效地给出数据库重做日志的快照，并且从中搜索出于哪些事务对ETL装载过程中所关注的表产生影响  
    刮取的变体就是嗅探，这种方式是对重做日志过程进行监视，即时捕捉事务，但刮取日志的事务可能是所有技术中最棘手的，由于事务日志已满而阻碍新事务处理运行的情况很常见，但是如果日志被清空的话，其中记录的所有事务都丢失了  
    如果您已经试遍了其他的方法，最终需要借助日志刮取来查找新的记录或者变化的记录，那么就要说服DBA创建一个专门的日志来满足您的需求  
+ **消息队列监视**  
    在基于消息的事务处理系统中，所有针对兴趣表的事务处理都需要对消息队列进行监视，流的内容和日志嗅探中获得的内容十分相似，这一过程的好处就是如果每天的消息队列都已经就绪，则开销较低  

### 9.3.3 子系统3—抽取系统
显然，从源系统中抽取数据是ETL架构中的一个基本组成部件，一般的情况是，每个源都可能是在不同的系统、环境或者DBMS中 5  
您可能希望自己的ETL系统从多种系统中抽取数据，而这些系统往往包含很多不同类型的数据，这将带来一些固有的难题：  
+ 那些需要从大型机环境中抽取数据通常会遭遇COBOLcopybook、EBCDIC到ASCII码变换、压缩十进制数、重定义、OCCURS字段和多种可变记录类型的问题
+ 而另一些机构可能需要从关系DBMS、平面文件、XML源、WEB日志或者复杂的ERP系统等形式存在的源中抽取数据  

每一种情况都会面临很多调整，有些源，特别是那些时间较久的遗留系统，它们所使用的过程语言可能和您的ETL系统所支持的或团队所熟悉的过程语言不同，这种情况下，应该请求源系统的所有者将数据抽取成为平面文件格式的形式    
注意：尽管XML格式的数据由很多优点，而且是自描述的，但是并不能指望使用它进行频繁的大数据量传输，一般XML格式的文件其有效载荷都要低于整个文件大小的10% 5  
从源系统中获取数据有两种基本方式，即以文件形式或者以流形式  
如果使用ETL工具，而且数据存储于数据库中，就可以将抽取设置为流的形式。注意我们不是指运行中的细流，而是指构建抽取过程的流，这样数据从源系统数据流出并经过转换引擎进入登台数据库将整个成为一个单个的过程  
相反，抽取到文件的方法则包括三到四个离散的步骤：将数据抽取到文件、将文件移动到ETL服务器、转换文件内容、将转换数据装载到登台数据库中  
注意：尽管流式抽取更吸引人，但是抽取到文件也是有一些优点的：
抽取到很容易从不同的点重新开始，只要保存了抽取文件就可以在不影响源系统的前提下重新运行装载。在使用网络传输数据之前可以很轻松地对数据加密和压缩。在传输之前或之后很容易通过比较文件行计数来验证是否所有地数据都以及正确传输 5  
如果使用公共网络传输数据，则数据加密非常重要，有时即使使用内部网也需要进行加密

数据的清洗和一致化是ETL系统的一项关键工作，它们是ETL系统为数据增值的步骤。此外这些子系统都可以设计用来创建能够对源系统中错误进行诊断的元数据，从而解决产生脏数据的根本原因，改进数据质量  
### 9.4.1 改进数据质量文化和过程
下游系统经常会抱怨原始数据源总是出现这样或者那样的错误，如果数据录入人员仔细一些就好了，可以设想在输入界面中，客户的社会安全编号经常为空或者填的是一些乱七八糟的东西  
看上去很小的数据质量问题实际上是业务过程走向失败的重要迹象，这一论述不但使我们将注意力正确地转移到数据质量问题的源头上，还为我们指明了解决的方法  
解决数据质量问题的技术性尝试只有成为源自机构高层的整个质量文化的一部分，才能够取得成功  
适用于所有要解决数据质量问题的九步法模板： 5  
+ 宣布高层对数据质量文化的委托事项
+ 在执行层驱动业务过程重构
+ 投资改善数据录入环境
+ 投资改善应用程序集成
+ 投资改变业务过程的工作方式
+ 提高团队的端到端质量意识
+ 提高各部门之间的协作
+ 公开宣传注重数据质量的好处
+ 不断度量并提高数据质量 3

在药店里，需要投资改善数据录入系统，为采购员的助手提供进行录入所需内容和选项。公司领导层需要让助手们明白他们的工作是非常重要的，而且绝对会影响很多决策制定者，勤奋工作的助手应当受到公开表扬和嘉奖  
最终目标是使团队具有端到端的质量意识，并且真正懂得业务价值是源自优质数据的 5  
### 9.4.2 子系统 4—数据清洗系统
ETL数据清洗过程通常用于修正脏数据，然而同时，当机构的生产系统捕捉到数据时还希望数据仓库能够提供该数据的准确描述，在这两个矛盾的目标之间找到恰当的平衡点是非常必要的  
解决方法时设计一个能够在装载数据时更正和除去错误的ETL系统，并且具有便于使用的结构、修正方式、标准、规则和优质的清洗功能的模块，这样系统就可以自行进行文档编制  
清洗系统的目标之一是提供一个用于数据清洗的综合架构，捕捉与数据质量相关的事件，同时在数据仓库中对数据质量进行度量并最终加以控制  
最先应当关注的：是尽可能从进行最好的数据探查分析开始入手，并且会帮助您确定自己的数据清洗系统需要达到什么样的复杂程度  
清洗子系统的目的是汇集多种技术提高数据质量，这一子系统的目标应当包括： 5  
+ 对数据质量问题的早期诊断和鉴别分类
+ 让源系统和集成工作提供更优质的数据
+ 具有为ETL过程中遇到的数据错误提供特定描述的功能
+ 具有能够随着事件推移捕捉所有数据质量错误和度量数据质量的框架
+ 对最终数据的质量可信度进行度量的附加功能 2


1. 质量筛选  
    ETL架构的核心就是一组质量筛选，它们在数据流管道中充当诊断过滤器，每一种质量筛选都是一种测试  
    测试如果失败，那么就必须将错误事件记录丢弃到错误事件模式，并且停止处理，将出错的数据发送到悬挂文件或者仅对数据进行标记  
    尽管所有的质量筛选在结构都是相似的，但是一般可以按照范围由小到大的顺序将它们分为三种类型：列筛选、结构筛选、业务规则筛选 5  
    列筛选对单个列中的数据进行测试，一般是比较简单和显而易见的测试，如测试一个列是否包含意外的空值，某个值是否落到了规定的范围之外，或者某个值没有按照规定的格式给出  
    结构筛选测试了各列数据的关系，可以对两个或多个字段进行测试来验证它们是否具有某种层次结构，例如一系列多对一关系，结构筛选还测试两个表中字段之间的外键/主键关系    
    业务规则筛选则进行更为复杂的测试，这些测试并不适合使用较为简单的列筛选或者结构筛选来实现，例如终身白金级的飞行员是指那些至少从业5年并且至少飞行了两百万英里的飞行员  
    业务规则筛选还包括聚集阈值数据质量检查，例如可以检查以下，为肘部扭伤这样比较轻微的诊断所开具的核磁共振诊察数目是否不大可能出现，如真的出现，那么达到了阈值时就会抛出错误  
2. 响应质量事件  
    我们已经说明了每个质量筛选都必须判定当错误抛出时会发生什么事件。选项有1)中断处理；2)向悬挂文件发送损坏的记录以便以后进行处理；3)仅标记数据并且将其传送到管道的下一步骤 5  
    到目前未知，只要有可能的话采用第三种选择是最好的，将那些有错误的数据标记出来通常时非常有效的  
    对错误的事实表数据可以在审计维度上进行标记，错误的维度数据也可以使用审计维度来标记。如果出现了数据丢失或者无用数据，可以在其自身的字段里使用唯一的错误值进行标记

### 9.4.5 子系统7—重复数据删除系统
通常维都由几个方面的来源，一种很常见的情况是由很多面向客户的源系统会创建和管理单独的客户主表，客户信息需要由多行业务信息和外部源中的信息合并而成  
即使进行了最可靠的匹配，数据中的其他行也还是可能会相互矛盾，您需要确定哪些数据应当继续保留    
很少由某个通用的列可以使合并操作更简单些，有时唯一可用的线索就是几个列的相似度，有时匹配是基于模糊标准的    
存活是指将一系列匹配的记录整合为统一的像，将匹配结果中质量最高的列整合成一个一致行 5

## 9.5 向呈现层交付数据
ETL系统的主要任务是在提交阶段传递维度表和事实表。由于这个缘故，交付子系统是ETL架构中的中枢子系统  
大多数这种子系统都集中关注维度表的处理，维度表是数据仓库的心脏，它们为事实表提供了上下文，因此也就为所有的度量提供了上下文  
对很多维度来说，基本装载计划是相对简单的：为了装载到目标呈现表中，对数据进行基本的转换操作并创建维度行  
制备事实表当然是非常重要的，因为事实表保存了用户希望看到的主要业务度量，然而，为事实表的呈现所作的准备工作一般都是十分简明的    
### 9.5.1 子系统9—缓慢变化维管理器
ETL架构中最重要的要素之一是能够实现缓慢变化维度SCD逻辑 5  
当数据仓库收到通知，得知某一维度一个已有的行发生了改变，有三种基本的SCD响应—类型1覆盖，类型2添加新行，类型3添加新列。图9-4显示了处理SCD代理键管理的整个流程  
子系统2中阐述的变化数据捕捉过程显然在将变化数据提交给SCD过程方面起到了非常重要的作用  
![avater](9-4.png)  
1. 类型 1：覆盖  
    类型1方法是在已有维度行中对一个或者多个属性进行简单的覆盖，也就是从变化数据捕捉系统中去除修正数据并且对维度表的内容进行覆盖。  
    类型1更新必须从早期永久存储的登台表一直传播到所有受影响的登台表，因此如果这些登台表中有某个参与了最终装载表的重建，则覆盖的影响也一并保留下来了 5  
    如果类型1更新导致了性能方面的问题，就应该考虑禁用数据库日志记录或者打破了的DBMS装载程序  
    类型1更新会使很多构建于变化列上的聚集失效，因此维度管理系统（子系统17）必须通知受影响的事实提供者（子系统18）删除和重建受影响的聚集  
2. 类型2：创建一个新行  
    类型2 SCD是精确跟踪维度变化的标准方法，可以将维度中的变化和事实表正确关联，支持类型2变化需要一个强大的变化数据捕捉系统，这样一旦变化出现就可以被探测到  
    类型2更新意味着对原先版本的维度行进行复制，并且创建一个带有新代理键的新维度行  
    如果ETL工具不能自动进行处理，类型2ETL过程还必须更新最新的代理键映射表，这些小到只有两列的表在装载事实表数据的过程中是非常重要的 5  
    在抽取过程中处理变化维度行所使用的查找和键分派逻辑如图9-4所示。一旦已经辨识出行发生了类型2属性的变化，就需要从键序列中生成一个新的代理键并且更新代理键映射表  
    当创建了一个新的类型2行后，至少需要一对时间戳，以及一个可选的变化描述字段。  
    对类型2 SCD行更复杂的处理包括添加和更新5种额外的ETL内务处理列，如图所示，需要类型2ETL过程来查找原先的有效行，并且对这内务处理列进行恰当的更新：  
    + 日历日期（变化的日期）
    + 行有效DateTime(变化的准确日期/时间戳）
    + 行终止DateTime(下一变化的准确日期/时间戳，默认为12/31/9999)
    + 改变列的原因或者捕捉列变化代码的文本列  
    + 当前标志（当前/到期） 5
    
   警告：有时后台脚本在事务处理数据库运行修改数据时并没有更新相应元数据字段，如last_modified_date，使用这些字段作为维度时机会产生不一致  
   通常可用系统日期或者“到……时”形式的日期来导出类型2有效时间戳  
   类型2过程并不像类型1那样改变历史信息，因此类型2变化并不需要对受影响的聚集表进行重建，只要变化是“当前”出现的并且在时间上并不向后追溯即可  
3. 类型3：添加一个新列  
    类型3方法用于支持属性的“软”变化，需要用户既考虑到属性原先的值，又考虑新值。  
    如果情况不像预期，则类型3方法需要ETL系统能够改变维度表，新模式添加一个新列。图9-5显示了一个类型3 SCD是如何实现的。 5  
    和类型1过程一样，类型3变化使得在变化列基础上构建的所有聚集都失效了，因此维度管理必须通知受影响的事实提供者，使他们删除并重建受影响的聚集

4. 混合方式：集中类型的组合  
    每个维度表都会用到一种、两种或者全部的方法来管理数据变化。这种情况下就可以使用维度中的两列，一列表示某个属性当前(类型1)视图，另一列表示同一属性的历史(类型2)版本  

### 9.5.2 子系统 10—代理键生成器
我们强烈建议对所有维度表都使用代理键，意味着ETL系统中需要有一种生成代理键的稳健机制  
尽管使用数据库触发器创建代理键的做法普遍，但是同时也会造成性能方面的瓶颈 5  
为了有效改进性能，要考虑使用ETL工具生成和维护代理键，但要避免将源系统的操作键（自然键）和某个日期/时间戳串联起来  
### 9.5.3 子系统 11—层次管理器
一个维度通常都会有多样的、同时存在的、嵌入式的层次结构。这样层次结构一般都作为维度属性并存在于同一维度中，其唯一要求就是每个属性在维度主键中出现的时候只能取单一值。  
部分不齐整的层次结构（如通信地址）通常建模为固定的层次结构。  
完全不齐整的维度一般其组织结构是不均衡的，而且深度也是不确定的。支持不齐整层次机构的数据模型设计和ETL解决方案都需要使用到一个包含组织图的桥接表  
在呈现层并不推荐使用雪花结构。然而在ETL登台区使用规范化设计可能比较恰当，因为它可以辅助维护用于填充和维护层次结构属性的ETL数据流 5  
### 9.5.4 子系统 12—专用维度管理器
专用维度管理器是总管各项事务的一个子系统，是ETL架构中支持机构的专门维度设计特性的一个代理。有些机构的ETL系统需要具有下面讨论的所有功能，而其他可能只关注其中某一些设计方法  
+ **日期/时间维**  
    日期和时间维都是唯一的，因为在数据仓库项目的一开始就对它们进行了完全的描述，而且它们并没有常规的来源。一般来说这些维度都是使用电子表格建立的      
+ **杂项维**  
    将事实表中的关键字段全部删除后，留下的文本和各种各样的标志就组成了杂项维。在ETL系统中有两种创建杂项维的方法。  
    如果维度的理论行数是固定且已知的，那么可以预先创建杂项维度。  
    另一种情况下，在处理事实行输入的时候，可以即时创建新的观察杂项维度行。如图9-6所示，这个过程需要装配杂项维属性并且将它们和已有的杂项维度行进行比较，看看该行是否已经存在了 5    
    ![avater](9-6.png)  
+ **微型维**  
    当类型2方法不可行的时候，微型维可以作为一种在大型维中跟踪维度属性变化的方法。这种方法最常用于大型的客户维。  
    从ETL的角度来看，微型维的创建和杂项维的处理相似，有两种供选方案：预先创建所有合法组合或者即时辨识和创建新的组合  
    杂项维度通常是从事实表输入中创建的，微型维则是从维度表输入中创建的。要记住，规模很大并且很复杂的客户维度可能需要几种微型维度  
+ **缩小维**  
    缩小维是一致性维度，是基本维的一个行子集和/或列子集，ETL数据流应当从基本维度中创建一致的缩小维 5  
+ **小型静态维**  
    一些维是完全由ETL系统创建的，并不需要实际的外部源参与。这些维通常是小型的查找维，它们将操作代码转换成文字。完全是由ETL团队直接创建，最终形式为一个关系表  
+ **用户维护的维度**  
    通常数据仓库需要创建全新的“主”维度表。这些维度没有正式的记录体系，相反，它们都是自定义描述、分组和由业务人员创建的用于报表和分析的层次结构  
    ETL团队通常最终要为这些维度负管理责任，但这样做也不是很成功，因为ETL团队无法意识到这些自定义分组所发生的变化，因此维度最终由于有失维护而失去作用  
    最好的方案是让适当的部门用户拥有维护这些属性的权限，数据仓库团队需要向这些用户提供用户界面来方便他们进行维护  
    
### 9.5.5 子系统13—事实表构建器
事实表构建器子系统主要关注ETL架构需求，能够有效创建三种基本类型的事实表：事务处理粒度、周期快照和累积快照事实表 5  
装载事实表时的一个重要要求是维护它和相关联的维度表的参照完整性关系，代理键管道（子系统14）就是用于协助支持这一需求的  
1. 事务处理粒度事实表装载器  
    事务处理粒度表示了在特定时刻定义的度量事件。  
    事实表是三种类型的事实表中最大且最详细的一种。最简单的情况是对大多数数据进行单纯的添加操作。  
    大多数情况，目标事实表应当按照日期进行分区，以便于管理和提高表的性能。此外，还应当含有审计键、时序ID或者日期/时间戳字段，以便装载工作能够进行备份或者重启 5  
   对于延迟到达的记录进行条件更困难些，需要子系统16中讲述的一些附加的处理功能。  
   最后，在必然会更新现有记录的事件中，处理过程应当分两步进行：第一步是不覆盖或者删除原来的行就插入更新的行，随后在第二步中删除原来的行  
2. 周期快照事实表装载器  
    周期快照事实表描述了某个经常重复的度量或者度量集合，如银行账户月报。周期快照粒度表中的事实必须真正符合粒度，应当仅描述那些适合周期定义的时间跨度的度量  
    周期快照和事务粒度事实表有相似的装载特性，对插入和更新操作的处理也相同。传统做法是在适当的周期结束的时候将周期快照全部装载  
    传统的周期末装载现在已经有所改变，有些机构使用热滚动周期快照。在滚动周期快照中，特定的行是当最新的结余在前一天有效时才进行装载 5  
    有些情况下，计算周期末结余的业务规则比较复杂，依赖于数据仓库之外的其他周期性处理，并且ETL系统中没有足够的可用信息来频繁地执行这些复杂计算  
3. 累积快照事实表装载器  
    累积快照粒度描述了某个限定了开始和结束的过程的当前进化状态。通常这些过程持续的时间都较短，因此并没有引入到周期快照中。订单处理是累积周期快照的经典例子  
    累积快照的设计和管理与前两种类型的事实表区别很大。所有的累积快照事实表都有一组日期，通常是4到8个，描述了一般处理流程。  
    每次当有事件发生时，累积快照事实表行就需要进行破坏性的修改，日期外键被覆盖，并且各种事实也被更新    
    很多RDBMS使用长度可变的行，因此对累积快照事实行的重复更新可能会引起行的增长，会影响到磁盘块的位置，偶尔在更新活动之后删除并且重新装载行可以提高性能 5  
    累积快照事实表是描述开始和结束都定义好的有限过程的一种有效方法。然而，按照定义来看累积快照应当是最新的视图  
    通常最好使用全部三种类型的事实表来满足多种需求。周期性历史信息可以使用周期性抽取来捕捉，并且在处理过程中，所有范围没有被限定的细节都被捕捉到一个相关联的事务粒度事实表中  

### 9.5.6 子系统 14—代理键管道
ETL系统必须包含一个在引入的事实表记录中使用恰当的维度代理键替换操作自然键的步骤。参照完整性(RI)是说对事实表中的每个外键，对应的维度表中都存在一条记录  
键查找过程应当为每个自然键或者默认值产生一个匹配项，在该事件里，查找过程中会出现一个未解决的参照完整性故障，需要将该故障反馈给可靠的ETL过程予以解决，这一过程如图9-7所示  
![avater5](9-7.png)  
事实表数据被处理完并且还没有装载到呈现服务器的时候，需要进行代理键查找将要引入的事实表记录中的操作自然键替换为恰当的代理键。为了保护参照完整性，我们总是首先完成维度表的更新。  
我们的方法是将当前的维度表作为和每个自然键相对应的代理键的最新值。  
每当需要当前代理键时，应当检查自然键等于预定值的维中所有记录，并且使用当前行指针或者起始有效期和终止有效期选定与事实行中历史上下文相对准的代理键。这种方法显然效率较低  
使用一个代理键映射查找表速度要快得多，在映射表中每个自然键恰好有一条记录。这些记录中含有和操作自然键相对应的代理键的唯一当前值。不要在事实记录中保留自然键——事实表只需要含有代理键  
正如图9-7底部所示的那样，在企图装载某个副本行时，代理键管道还需要处理键冲突，这将交给子系统4处理 5  
当键冲突被识别出来后，代理键管道需要判定停止过程，将冲突数据发送到悬挂文件，或者应用恰当的业务规则来确定能否修正冲突数据问题，装载该行，并且向错误事件模式中写入一行错误记录  
要注意，如果您在执行代理键查找时需要装载历史行，或者由于不想将最新值映射到某个历史事件上而保留了很多延迟到达的事实行，那么您将需要一个略有不同的处理过程  
当使用代理键代理事实表自然键之后，事实记录就已经做好了装载准备  
### 9.5.7 子系统 15—多值维度桥接表构建器
正如第6章所讲述的，有时事实表必须在最低粒度上支持带有多个值的维。如果无法改变事实表的粒度来直接支持该维，那么该多值维就必须通过一个桥接表连接到事实表  
ETL团队的挑战在于创建和维护桥接表。当遇到事实行的多值关系时，ETL系统既可以将每个观察值集合为一个单独的分组，也可以在出现了相同的观察值集合时重用分组 5  
第6章介绍的一种桥接表构造包含了一个加权因子，用于支持从桥接表进行加权报表生成  
### 9.5.8 子系统 16—延迟到达数据处理器
然后，由于多种原因，ETL系统可能需要处理延迟到达事实数据和维度数据  
在有些环境中，需要修改标准的处理过程来处理延迟到达事实 ，即到达数据仓库的时间向后延迟的事实记录。  
您不得不在数据仓库中向后查找历史信息，并判定该活动发生时哪些维度键是有效的。可能还需要调整后续事实行中的半可加性的账户结余。在合规性很高的环境中，还有必要同接口和合规性子系统相联接  
如果还没有完整的上下文环境事实度量就已经到达，将出现延迟到达维。如果可以允许常规的批量更新周期有一天或者多天的数据等待时间，通常就可以等待有关延迟维度的情况报告 5  
但是更多情况下，这种延迟是不可接受的，尤其是实时环境中，您无法将行悬挂起来等待维更新完成  
为了满足这种需求，ETL系统还需要额外的功能，首先要支持延迟到达类型2 维度更新。  
这种情况下，需要将修改后的客户记录添加到维度中，使用新的代理键，随后进入客户表中破坏性地修改后续事实行的外键。  您也需要为受影响的维度行重置生效日期。还需要在维度中向前扫描，看看后续是否还有有关与该客户的类型2行，并且在所有受影响的行中修改该列  
还有第二种情况，当您接受到某个带有一个合法的客户自然键的事实行，但是还没有将该客户装载到客户维中就产生了延迟到达 5  
如果您认为客户确实是新客户，但是尚未进行处理，那么就在新的客户维度记录中指定一个带有一组哑元属性值的新客户代理键。  
稍后当您获得关于该新客户的更完整的信息之后，就可以回到该哑元维记录，并且对其属性进行类型1覆盖，这一步避免了对事实表键的破坏性更改  
您不可避免会遇到一个维度“并非很正确”的临时性时期，但是这些维护步骤可以将对键和其他字段进行更新所带来的影响降到最低  
### 9.5.9 子系统 17—维度管理系统
维度管理员是一位核心职员，他制定一致性维度并且将其发布到数据仓库社区中。一致性维度必然是一种需要集中管理的资源，每个一致性维度必须有单一的、一致的来源。维度管理员职责主要包括下述ETL处理事项：  
+ 在维度设计过程中征求数据管理员和干系人的意见并落实公共描述性标签 5
+ 对新的源数据，要向一致性维度添加新行，生成新的代理键
+ 为已有维度记录的类型2变化添加新行，生成新的代理键
+ 适当地按照类型1变化和类型3变化修改行，但是并不改变代理键
+ 如果出现了任何类型1和类型3变化，更新维的版本号  
+ 同时为所有的事实表提供者复制修改后的维 5

在位于单一主机上单一表空间的DBMS中管理一致性维度要相对容易些，因为维度表只有一个副本。然后多表空间、多DBMS或者多主机的分布式环境中，管理一致性维度就变得非常困难了。
这种情况下，维度管理员必须认真管理新版本的发布，将新版本同时发送给每个事实提供者。这个版本号应当用于支持交叉钻取查询，保证所有的用户或那些从多个呈现服务器集成数据的BI应用程序都是用同一版本的维  
### 9.5.10 子系统 18—事实提供系统
事实提供系统需要管理一个或者多个事实表，并负责事实表的创建、维护和使用。事实提供系统的职责包括：  
+ 从维度管理员那里接受或者下载维度副本 5
+ 在一个无法直接复制而只能在本地更新维度的环境中，为了在代理键管道中更新当前键映射表，事实提供系统必须处理标记为最新和当前的维度记录，而且也要处理所有标记为最新但是延期的维度记录
+ 在使用正确的代理键替换自然键之后，将所有的记录添加到事实表中
+ 在错误更正、累积快照和延迟到达维度变化之后，对所有的事实表中的记录进行修改
+ 去除那些已经失效的聚集
+ 重新计算受影响的聚集，如果某个维度的新版本并没有更改版本号，聚集就北徐扩展为处理新装载的事实数据。如果维的版本号发生了改变，则可能以前全部的聚集都需要重新计算 5
+ 保证所有基本事实表和聚集事实表的质量，要确保聚集表是经过正确计算得到的
+ 在线更新事实表和维度表
+ 通知用户数据库已经更新，如果进行了最重要的更改还要告诉他们更改情况，包括维度版本变化、加入延期记录和改变历史聚集

### 9.5.11 子系统 19—聚集构建器
在大型数据仓库系统中，聚集是提高系统性能的一种最好方式  
从ETL的角度来看，聚集构建器需要填充和维护聚集事实表，并且在聚集事实表需要的时候缩减维度表。最快的更新策略是增量更新，但是对维度属性的重要变更需要删除和重建聚集。聚集必须始终和基础数据保持一致 5  
用户可以反馈那些运行速度很慢但却对设计聚集非常重要的查询。尽管在某种程度上您可以依赖信息反馈，但是仍然需要建立一个日志，记录那些需要经常访问但是却执行很慢的查询  
### 9.5.12 子系统 20—OLAP多维数据集构建器
OLAP多维数据集以一种直观的方式来表示维度数据，使分析用户能够对数据进行切片和切块  
如果在您的数据仓库架构中包含了OLAP多维数据集，那么应当将关系维度模式看作是OLAP多维数据集的基础，关系维度模式是OLAP多维数据集的首选来源。  
因为很多OLAP系统并不直接借鉴参照完整性或数据清洗方面的问题，最好在常规ETL过程结束的时候装载OLAP多维数据集  
您应该认识到有些OLAP工具对层次结构要比关系模式更为敏感。在装载OLAP多维数据集之前，强化维度中层次机构的完整性非常重要的。类型2SCD非常适用于OLAP系统，因为新代理键只是作为一个新的成员来处理 5  
类型1和类型3SCD对那些不适用于OLAP的历史信息进行了重述。覆盖某个属性值可能会引起该维在当前背景下重建，并且所有使用该维的多维数据集都会遭到破坏，需要删除  
### 9.5.13 子系统 21—数据传播管理器
在ETL过程中，数据传输管理器负责将一致的、集成的企业数据从数据仓库呈现服务器转换到其他环境用作特定用途  
这里所讲述的所有情况都需要需要从数据仓库呈现服务器上抽取数据，也可能会进行一些轻量级的转换，并最终装载到目标格式中  
这个过程也可以看做是另一种ETL过程。应当将数据传输看作是ETL系统的一部分，并且利用ETL工具来提供这种功能  
## 9.6 管理ETL环境
数据仓库的目标之一是能够为增强业务功能提供适时的、一致的和可靠的数据。为了实现这一目标，ETL系统必须按照下述三条标准不断地加以完善  
+ **可靠性**。必须始终如一地完成各项操作，及时提供在任何细节层次都可以信赖的数据
+ **可用性**。数据仓库必须满足其服务层协议。数据仓库应当像所承诺的那样具有相关功能，并且具有较好的可用性
+ **易管理性**。成功的数据仓库永远不会停止发展的脚步，它会随着业务过程的发展不断增长和变化，为了实现这一点，ETL过程需要进行适度的进化  

ETL管理子系统辅助实现可靠性、可用性和易管理性方面的目标 5  
### 9.6.1 子系统22—作业调度器
每个企业数据仓库都应当拥有一个健壮的ETL调度器。一般来说，整个企业ETL过程应当是可以通过一个单独的、元数据驱动的作业控制系统进行管理的    
调度要比仅按照时间表启动各项工作涵盖更多的内容。当然，需要一种可靠的机制来管理ETL的执行策略。此外，调度器还应当能够控制ETL工作之间的关联关系和依赖关系。  
如果需要进行实时处理，就需要使用一个能够支持选定实时架构的调度器。工作控制过程在执行时也必须根据ETL过程的步骤和统计情况来捕捉源数据。调度器应当完全支持自动化过程，包括当出现需要解决的问题时及时通知您对系统进行调整。  
管理作业调度的底层设施既可以像一组SQL存储过程，也可以像设计用来管理和协调多平台数据抽取和装载过程的集成工具。在任何情况下，都需要建立一个能够创建、管理和监视ETL工作流的工作环境  
作业控制服务需要包括： 5  
+ **作业定义**。创建业务操作过程的第一步是采用某种方式来将一系列步骤定义为作业，并且指定作业之间的一些关系。这里需要制定ETL过程的执行流程。  
+ **作业调度**。工作环境至少需要提供一些标准的功能，如基于时间和基于事件的调度，ETL过程通常基于上游系统事件，这还包括监视数据库标志的功能、检查文件是否存在的功能和对比创建日期的功能等  
+ **元数据捕捉**。负责运行装载的职员都需要使用监视系统来了解当前的运行状况。作业调度器需要捕捉当前装载在哪一步骤、装载过程何时开始以及装载过程会持续多久。每次ETL过程执行的时候ETL工具都应当捕捉这种数据    
+ **日志记录**。  
    这意味着收集关于整个ETL过程的信息，而不仅仅收集当前发生的信息。 一旦在工作执行过程中出现了错误，日志信息就可以支持过程恢复和重启，将日志写入文本文件是可接受的最低层次。  
    我们更倾向写入数据库的系统，因为利用这种结构的数据很容易创建图和报表。同样，它也使通过创建时间连续的调查来辅助装载过程的分析和优化成为可能 5  
+ **通告**。  
    一旦ETL过程设计完成并且进行了部署，就应当以非干涉方式工作。它应当能够不需要人的干涉就可以允许，而不会出现故障  
    
注意：有些人需要在装载过程中了解是否发生了某些未曾预见到的事件，尤其当应答对继续完成装载过程至关重要时更是如此  
### 9.6.2 子系统23—备份系统
数据仓库和其他计算机系统面临着相同的风险。备份系统的目标是使数据仓库能够在出现故障后回复到原先的工作状态  
这包括备份重启发送故障的ETL作业所需的中介登台数据。还要设计存档和检索过程使业务用户能够访问存储时间较久的数据，并将这些数据从数据仓库迁移到成本较低、性能也较低的存储介质上  
1. 备份  
    即使您的系统完全冗余，并且有通用的电源、完全RAID构成的磁盘和具有故障切换能力的并行处理器，最终系统仍然会出现一些危险，即使有最好的硬件，仍然会有人删错了表或数据库。一个完善的备份系统需要提供下列功能： 5  
    + 高性能，备份需要能够在分配时间段内完成。这可能包括不会显著影响性能的在线备份，如实时分区  
    + 简单管理，管理界面应当提供一些工具，便于您识别要备份的对象（包括表、表空间和重做日志）、创建时间表，并且为后续存储维护备份验证和日志
    + 自动化夜间操作，备份设备必须提供存储管理服务、自动调度、截止和设备处理、报表和通告等功能
    
   数据仓库备份通常是物理备份，也就是某个时间点对数据的一个镜像，包括索引和物理布局信息  
2. 存档和检索  
    要决定从数据仓库中迁移哪些数据也是一个成本-效益问题。保存数据也需要一定的开销—数据要占用磁盘空间，并且会降低装载和查询的时间。 5  
    另一方面，业务用户可能仅需要使用这些数据做一些评论性的历史分析。同样，审计师也可能会要求将存档数据作为合规步骤的一部分。解决方案就是不要将数据丢弃，而是将其存放在成本较低的介质上，但仍然是可以访问的。  
    截止本书成稿位置，联机磁盘存储的成本下降得很快，大部分的存档任务都可以直接通过写入磁盘来完成。您仅需要确保自己在以后的各种场合都能够解释这些数据  
    数据多长时间会变得陈旧取决于行业、业务和研究的特定数据  
    一旦决定要对某种数据进行存档，问题就会变为从长期来看存档数据意味着什么？显然，需要利用已有的机制在物理上将数据从当前介质转移到另一种介质中，而且要确保数据能够被恢复，还要为数据的访问和变更创建审计追踪  
    最终，您必须实现一个资料管理系统来进行数据存档，而且要经常性地刷新数据并将数据迁移到比较新的结构中，同时使用比较新的存储介质进行存储 5  
    
### 9.6.3 子系统24—恢复和重启系统
在系统系统开始工作后，将会有难以计数的原因造成系统故障，使您无法控制ETL过程。造成ETL工作故障的常见原因：  
+ 网络故障
+ 数据库故障
+ 磁盘故障
+ 内存故障
+ 数据质量引起的故障
+ 未通报的系统升级

为了使您避免这些故障，需要一个可靠的备份系统和一个配套的恢复和重启系统。在装载过程中难免会出现一些无法恢复的错误，因此必须制定相关的方案。首先，要寻找恰当的工具并且设计使故障影响最小化的过程  
当然，恢复和重启系统既可以用于重新开始某一项已经停止的工作，也可以用于现场恢复整个工作和重启该工作 5  
当故障发生后，最初的下意识反应就是抢救已经处理的内容，并且从该发生故障的点重启装载过程。这需要ETL工具具有健壮而可靠的检查点功能，这样就可以很好地确定出已经处理的内容，以及在当前时间点上哪些作业不需要重启  
ETL过程执行的时间越长，则您越要意识到可能会出现故障。  
仔细考虑何时通过写入磁盘的形式来从物理上对数据进行登台操作，同时要巧妙地选择恢复点和装载日期/时间戳或连续的事实表代理键，只有这样才能使您创建一套恰当的重启逻辑  
### 9.6.4 子系统25—版本控制系统
版本控制系统是用于存档和恢复ETL管道中所有逻辑和元数据的一种快照功能。注意，需要为整个ETL系统指定主版本号，就像软件发布版本号一样  
注意：除了整个ETL系统有一个主版本号之外，ETL系统的每一部分都要有一个主版本号。而且，如果已经证明当前版本中存在一个很大的错误，那么应当能够完全恢复以前版本的ETL元数据上下文 5  
### 9.6.5 子系统26—版本迁移系统
当ETL团队结束设计和开发ETL过程这一艰巨的任务，并且完成了创建装载数据仓库的工作后，该工作必须进行打包并且迁移到下一个工作环境中——从开发到测试再到最后的生产环境——按照机构所采用的生命周期进行  
大多数机构都将开发、测试和生产环境分离开来。生产系统中要做的任何操作都应当在开发过程中进行了设计，并且部署脚本也应当在测试环境中得以测试  
不管是部署新的模式、添加列、改变索引、改变聚焦设计、修改数据库参数、备份还是恢复，每一项后台操作都应当经过严格的脚本编写和测试  
### 9.6.6 子系统 27—工作流监视器
成功的数据仓库始终是可用并且可靠的，能够满足业务社区的需求。为了实现这一目标，ETL系统必须不断地对工作流进行监视，确保ETL过程高效运作，使数据仓库总能及时装载    
工作流监视器利用作业调度器捕捉到的元数据提供了一个仪表盘和报表系统，它考虑到了ETL系统中的方方面面 5  
关键性能度量包括处理记录的数量、错误摘要和采取的行动。大多数ETL工具捕捉都是为了测量ETL的性能而捕捉度量    
和作业调度器相结合后，工作流监视器也应当跟踪系统性能，并且捕捉底层部件的性能度量，包括CPU使用、内存分配与竞争、磁盘使用与竞争、数据库性能和服务器使用与竞争    
工作流监视器要比某些人所想象的那样具有更重要的战略角色，它是整个ETL管道中性能问题分析的起始点。ETL的性能瓶颈可能出现在很多地方，好的工作流监视器会告诉您瓶颈究竟发生在什么位置。下面给出一个瓶颈的重要性排序列表：  
+ 针对源系统或者中间表的索引极差的查询
+ SQL语法引起的错误的优化器选择
+ 随机存取存储器（RAM)不足造成的性能下降
+ RDBMS的排序操作
+ 过渡频繁的I/O操作
+ 不必要的跟有读操作的写入操作 5
+ 完全删除和重建聚集，而不是增量地建立聚集
+ 在管道中筛选（变化数据捕捉）操作应用得过于迟缓
+ 没有使用并行处理或者管道处理方式
+ 不必要的事务日志记录，尤其是当进行更新的时候
+ 网络通信量和文件传输开销 2

### 9.6.7 子系统 28—排序系统
某些常见的ETL过程需要将数据按照某种特定的顺序进行排序，例如聚集和加入平面文件源。因为排序是非常基础性的ETL处理功能，所以建立了一个单独的子系统来确保其得到足够重视，成为ETL架构中的部件之一  
关键是要选择最高效的排序方法来满足需求，最简单的做法是直接使用ETL工具的排序功能。但是有些情况下使用专用的排序包会更加高效  
### 9.6.8 子系统 29—沿袭和依赖分析器  
您的ETL系统中需要两个月来越重要的要素，即对数据仓库中数据的沿袭和依赖进行跟踪 5  
+ **沿袭**。从中间表或者用户报表中的特定数据元素开始，确定该数据元素的源、其他包含该元素和其源的上游中间表，以及该数据元素和其源所经受的所有转换  
+ **依赖**。从某个源表或者中间表中的特定数据元素开始，确定包含该数据元素和其他派生的所有下游的中间表和用户报表，以及应用于该数据元素和其派生元素的所有转换规则  

在高合规性的环境中，沿袭分析通常是重要组成部分之一，在这样的环境中您必须能够解释改变所有数据结构的完整处理流程    当评估源系统影响数据仓库和ETL系统的源系统和其下游时，依赖分析非常重要  
不论您使用ETL工具还是手工编写ETL系统，都要意识到它和其他的软件一样都需要编制文档。  
当您的数据仓库和ETL系统演化了以后，ETL团队需要能够快速了解和记述系统的细节内容。当出现了合规性问题并且需要关于某个特定数据集的变化跟踪信息时，也需要予以解决。这时，可靠的文档和元数据就会为您提供有力支持 5  
ETL团队应当建立一个设计模板并且将各种活动组织到一起。该模板应清晰地标识出和抽取、转换、查找、一致化、维度变化管理及最终目标表的交付相关的ETL元素。如果ETL工具支持您对ETL系统进行注释的话，就对系统做详细的注释  
### 9.6.9 子系统 30—问题自动调整系统
一般来说，在降ETL过程移交给负责每日系统操作的部门之前，ETL团队负责开发这些ETL过程，而质量保证团队负责对这些过程进行完整的测试。为了完成这些工作，ETL架构中需要带有一个预先设计好的问题自动调整系统  
一旦ETL过程完成了开发和测试，ETL系统要提供的第一个层面上的运行支持应当是用于监视生产应用程序的一组（操作）  
理想状况下，您完成了ETL过程的开发之后，应将它们打包成一个自动化的调度器，并且当ETL过程只需的时候要有健壮的工作流监视功能对它们进行监视  
如果出现问题了，ETL过程应当自动通知问题调整系统出现了一些意外情况，需要关注和解决。这种自动反馈可能会以较为简单的错误日志、操作员通知消息、监督者通知消息和系统开发人员消息等形式出现 。  
ETL系统可以根据情况的严重程度和涉及到的过程通知相关的个人和团队。ETL工具可以支持多种消息传递方式，包括电话、APP、短信、邮件等 5  
每个通知事件都应当写入到数据库以便了解出现问题的类型、状态和解决方案  
一般来说，ETL系统支持应当按照相对标准的支持结构  
第一层支持一般是一个帮助台，当用户发现某个错误的时候首先就会想到使用它。如果帮助台无法解决问题，则第二层支持就是通告。ETL经理是第三层支持，而且应当非常了解如何为ETL生产过程中出现的问题提供支持  
### 9.6.10 子系统 31—并行/管道系统
除了提供高质量的数据之外，ETL系统的目标还包括在分配的处理时间窗内将数据装载到数据仓库。  
并行/管道系统为ETL系统可以帮助ETL在规定的时间约束内完成装载任务。该系统目标是充分利用通常都能够获取到的多处理器或者网格计算资源 5  
应当验证源DBMS的并行处理是否正确，并且要避免产生冲突的过程。如果可能的话，选择某种ETL工具来自动管理中间转换过程的并行处理  
### 9.6.11 子系统 32—安全系统
安全性是ETL系统需要考虑的一项重要内容。但是严重的安全缺口通常都是来自于机构内部而并非从外部入侵的黑客  
我们建议基于角色管理ETL系统中数据和元数据的安全性。为了满足合规性需求，您需要证明ETL模块的某个版本没有改变或者给出哪些人员对模块做了修改  
您应当加强授权控制，使用户按照与个体和角色相应的权限来访问ETL数据和元数据、此外，您可能还需要维护所有个体和角色对ETL数据和元数据进行访问的历史记录。其他要注意的问题是大批数据的移动过程  
另一个需要考虑的后台安全问题使管理员对生产仓库服务器和软件的访问、显然，团队的很多成员都应当拥有访问开发环境的特权，但对生产仓库应当控制得严格一些。另外备份存储介质应当和在线系统具有相同的安全性 5  
### 9.6.12 子系统 33—合规性管理器
在高合规性的环境中，支持合规性需求是ETL团队的一项新的重要需求。数据仓库中的合规性简单来说就是维护一个关于数据的“保管链”  
同样行政部门也必须认真维护一个证据保管链，以便证明证据没有发生改变或者被篡改，从合规性敏感的数据进入数据仓库开始，数据仓库就必须小心谨慎的进行保护  
而且数据仓库必须经常及时地显示这一类数据的额抽取条件和内容，使之处于数据仓库的控制之下。数据仓库必须跟踪那些授权访问这里数据的人员  
最后，当疑心重重的审计员进行检查时，您还需要回过头看这类数据在最初接收时存档的、加有时间戳的版本，而这些信息远程存储在一个可信的第三方那里  
合规性需求意味着您实际上无法改变任何数据，无论是什么原因。如果数据必须进行更改，则必须将改变后的记录的新版本插入到数据库 5  
因此每条记录都必须具有开始时间戳和结束时间戳，这一对时间戳能够准确地描述记录“当前为真”的这一段时间的跨度  
在进行任何大的变动之前，一定要确保从合规性主管那里获得了公司的方针政策  
合规性的基础是前面讲述的几个子系统间的交互，以及一些关键技术和功能：  
+ **沿袭分析**。通过指出最后一个环境数据来自何处来证明初始源数据经过了包括存储过程和手工变化在内的多种转换。这需要所有转换的完整文档和技术功能，以便对初始数据再次进行转换操作  
+ **依赖分析**。指出那里用到了初始源数据元素 5
+ **版本控制**。有时会再次使用ETL系统对源数据进行处理，这就需要对任何给定的数据源都标识出ETL系统的抽取版本  
+ **备份和存储**。  
    当然所请求的数据也可能已经存档了很多年，由于审计的原因需要进行恢复。也可能需要证明存档数据没有经过修改  
    可以让某个可信的第三方保管散列码，然后当需要恢复原始数据的时候，再次对其进行散列编码，并且将其和可信的第三方保管的散列码进行比较，从而验证数据的正确性  
+ **安全性**。 指出哪些人访问或者修改了数据和转换，而要为用户分配角色和特权  
+ **审计维**。审计维将运行过程中元数据的上下文直接和数据联系起来，以便在装载时捕捉质量事件 5  

### 9.6.13 子系统 34—元数据知识库管理器
ETL系统负责大多数描述DW/BI环境的元数据的创建和使用。整个元数据策略中应当有一部分详细说明ETL元数据的捕捉，包括过程元数据、技术元数据和业务元数据  
最后，一定要指定DW/BI团队中某个人担任元数据管理员，负责创建和实施元数据策略  
## 9.7 实时的意义
实时处理是数据仓库技术中越来越常见的需求，您的DW/BI系统也极有可能有实时性方面的需求  
创建一个实时的DW/BI系统需要为实时数据收集高度概括性的真实业务需求，并且确定一个恰当的ETL架构，将多种技术整合到一个固化的平台中  
### 9.7.1 实时的分类
询问业务用户是否希望“实时”提交数据，业务用户的回答通常是没有价值可言的  
为了避免这种情形，我们建议将实时设计方面的挑战划分为即时、频繁、日常三个类别，当我们和业务用户沟通他们的需求时就使用这些术语，并且我们设计的数据交付管道根据这三个选项而有所不同  
即时意味着每一刻屏幕上显示出来的数据都代表着源事务处理系统的真实状态。即时实时系统通常通过企业信息集成(EII)解决方案实现，源系统自身负责支持远程用户屏幕的更新，并且响应查询请求 5  
EII技术提供了一定的轻量级数据清洗和转换服务，但是这些功能都必须以软件的形式执行，因为数据是连续不断地被传送到用户屏幕上的  
在对用户业务需求访谈中，应当仔细评估对某个即时实时解决方案的需求程度，牢记该解决方案中源应用程序进行的重要装载和即时更新数据的易失性，如存货状态跟踪就是一个很好的例子    
频繁意味着数据每天都更新很多次，但不能保证屏幕显示的就是当前的真实数据。在常规ETL架构中，频繁提交的数据通常是小批量处理的  
这意味着这些数据要经过变化数据捕捉、抽取、登台到ETL后台的文件存储区、清洗和错误检查、按照企业标准进行一致化、分配代理键和很多其他转换操作等，以便数据做好装载到呈现服务器的准备  
在EII解决方案中，这些步骤几乎都被省略或者进行了大幅度的削减 5  
频繁提交数据和日常提交数据的区别在于两个方面：变化数据捕捉和抽取  
为了每天能够多次从源系统中捕捉数据，数据仓库通常必须利用高带宽的通信渠道，如遗留应用程序间的信息队列通信、累计事务处理日志文件，或者事务处理系统中通过某些事件触发的低层数据库触发器  
日常意味着用户屏幕上显示的数据是在上一个工作日下载的批处理文件，或者说是和源系统中上一个工作日的数据相一致的  
日数据一般都是由运行于源系统的过程在每个工作日结束之前对原始数据进行修正。如果您的系统中也会有这样的情形，应当向业务用户解释，他们如果需要即时或频繁地更新数据可能会遇到哪些问题  
业务需求收集阶段对实时ETL设计过程至关重要，到底是采用即时更新还是采用频繁更新或者每日更新都是一项重大的决策，即时解决方案和另外两种方案有明显不同，而且您也不会中途改变策略 5  
### 9.7.2 实时的权衡
响应实时需求意味着您需要改变您的DW/BI架构来使数据能够更快地显示在用户屏幕上  
我们假定您在开始设计实时系统之前，像以前一样专注于数据质量、集成、安全性、合规性、备份、恢复和存档。这里给出一些在实现实时架构时需要权衡的问题  
+ 使用从消息队列或日志文件中读取数据的方式替代批处理文件抽取  
    源系统传输过来的批处理文件可能描述了源数据的一个干净、一致的视图  
    另一方面，消息队列和日志文件数据是原始即时数据，可能尚未经过任何修正处理的过程，或者也不符合源系统中业务规则的强制性规定 5  
    最坏的情况下，这种原始数据可能：1. 是不正确或者不完整的，因为其他一些事务可能会到达得晚一些；2. 包含DW/BI尚未处理过的未解析的外键；3. 需要一个面向ETL数据流程的并行批处理，全天候地修正甚至代替热实时数据  
    而且如果源系统中对消息队列或者日志如果应用了复杂的业务规则，那么您一定不想在ETL系统中再重复这些业务规则了  
+ 仅用于列筛选和简单编码查找的约束数据质量筛选  
    由于通过ETL管道处理数据时间减少了，因此有必要将那些比较耗时的数据质量筛选需要剔除掉，尤其是结果筛选和业务规则筛选。即使在大多数要求较高的实时应用中，也应当将大多数列筛选应该保留下来  
+ 允许提前到达的事实存放在维度的旧的副本里。如果实时系统不能等待维度被解析完毕，而且维度有可用旧副本，那么就应当使用这些旧副本，否则就生成一个空版本的维度来代替  
+ 除去数据登台  
    有些实时架构(尤其是EII系统)直接将数据从生产源系统传送到用户屏幕上，而在ETL管道中没有将数据写入到永久性存储介质中 5  
    如果DW/BI团队对这样的系统也有一部分责任，那么团队就应当认真地同高级管理人员谈一谈可能遇到的备份、恢复、存档和合规性等方面的责任，而不论那些任务现在是否是生产源系统所关心的问题  
    至少，数据流经过DW/BI系统时应当作为一个整体来捕捉  

### 9.7.3 呈现服务器上的实时分区
为了支持实时需求，数据仓库必须无缝地将其现有历史时间序列扩展到当前时刻  
解决这一难题地设计方案是创建一个实时分区，作为常规静态数据仓库的扩展。为了实现实时报表，您需要创建一个特殊的分区，它在物理上和管理上都是和数据仓库表相分离的  
理想情况下，实时分区是一个真正的数据库分区，其事实表按照活动日期进行分区。如果这样的分区不可行，那么实时分区就必须是一个单独的符合特定更新和查询规则的表 5  
不论哪一种情形，理想情况下的实时分区都应当满足下述严格的需求集合。它必须：  
+ 包含所有自静态数据仓库上一次更新以来发生的活动。假设现在静态表每天晚上都是在午夜更新的  
+ 尽可能无缝地连接到静态数据仓库事实表的粒度和内容，理想情况下可以作为事实表的一个真正的物理分区  
+ 要编制轻量级索引，以便到来的数据能够连续不断地缓缓流入。理性情况下，实时分区是完全没有索引的，不过着在某些RDBMS中是不可能做到的，因为他们所创建的索引都是不可分割的  
+ 通过将实时分区驻留于内存来支持高速应答的查询，即使没有索引 5  

实时分区可以有效地和事务粒度及周期快照粒度事实表一同使用。我们还没有发现需要在累积快照事实表中应用这种方法  
1. 事务粒度实时分区  
    如果静态数据仓库事实表具有事务粒度，那么从“已记录的历史信息”开始，该事实表为源系统中每个单独的事务保留恰好一条记录。实时分区和其底层的静态事实表应有相同的维度结构  
    如果实时分区是真实的表分区，那么所有的应用程序都能够很好地将事实表的静态部分和实时分区联系起来  
    如果实时分区是一个独立的物理表，那么就必须通过从静态事实表到实时分区的交叉钻取来将该实时分区和已有的应用程序关联起来  
    您的实时分区也可以偏向于保留快速装载的性能，但是同时也需要提供快速查询的性能 5  
    因为您向静态事实表和实时分区发送了相同的查询，所以您可以让聚集导航器整理一下，看看是否所有的表都支持聚集  
2. 周期快照实时分区  
    如果静态数据仓库具有周期性粒度如月度，那么实时分区可以看作是本章前面所讲述的热滚动月。实时分区只是当前进行的这个月的一个镜像。同样，该实时分区也应当驻留于内存中  
    和事实粒度相比，查询应用程序从静态事实表到实时分区进行交叉钻取所采用的逻辑稍有不同  
    尽管账户结余和其他强度策略都可以直接跨表操作，但是在当前滚动周期中半加性累计总数可能需要向上升级为一个整月相等价的形式，以便使结果看起来不会很不规则

### 10.3.1 步骤 1—制订高层计划

![10-1](../syn_pic/data_warehouse_life_cycle/10-1.png)  
这种单页计划是和管理层沟通项目复杂性的一种很好的方式 5  
图10-1中有意识地强调了同一时期的问题和未解决问题，该方案应当经常更新和分布  
### 10.3.2 步骤 2—选择ETL工具
在数据仓库市场上可以购买到大量的ETL工具  
ETL工具能够提供多种功能。有些ETL工具擅长从特定源系统中抽取信息（如COBOL大型机和其他一些非关系数据结构)  
ETL工具从多种源读取数据，包括平面文件、ODBC、OLEDB、以及大多数关系数据库的本地数据库驱动器  
该工具包含了为数据定义转换操作的功能，包括查找和其他类型的联接。能够将数据写成多种目标格式，并带有某种管理ETL系统整体逻辑流程的功能 5  
的确，手工编写代码的方式（通常基于SQL）仍然非常常见，然而，使用ETL工具是行业标准最优方法，下面给出了这样做的几点原因  
+ 使用图形工具自行编制文档  
    手工编码系统通常不可避免地会有大量的登台表、SQL脚本、存储过程和操作系统脚本  
    当有新的开发人员参加到项目中来-或者原来的开发人员一年之后又重新回顾这些代码-要想弄明白代码的含义以及为什么这样编写是非常困难的  
+ ETL工具提供了先进的转换逻辑，如模糊匹配算法、对姓名和地址进行不重复的集成访问以及数据挖掘算法  
+ 改进系统性能，降低了专业知识水平的门槛。只有很少的SQL开发人员是真正懂得如何使用关系数据库的专家，他们能够操作超大规模的数据量，同时保持很好的性能 5  

在数据仓库项目的第一阶段，您不应当奢望收回在ETL工具上的投资。事实上的好处在后续阶段才会显现出来，尤其是当以后对已有系统进行修改的时候  
在生产系统中，在可管理性、生产率和培训方面进行长期的改进是非常重要的  
### 10.3.3 步骤 3—制定默认策略
如果您已经了解到ETL工具的底层设施有哪些需求，就应当为ETL系统中的常见活动设计一套默认策略。这样活动包括：  
+ **从每个主要的源系统抽取数据**  
    我们在第9章讲述了从源系统中抽取数据的基本策略，在ETL系统设计过程的当前环境，需要确定从每个源系统中抽取数据所使用的默认方法  
+ **存档抽取到的数据**  
    在对抽取到的数据进行转换操作之前，您应当存档这些数据，至少要保留一个月。存档的目的是便于进行系统审计和回复。通常都是将存档数据存储到文件中，而不是将它们存储到数据库中 5  
+ **为维度和事实修正数据质量**  
    正如第9章所讲述的，在ETL过程中您应当对数据质量进行测量，而不要等到您的业务用户找出数据质量问题  
    要为如何检查数据质量设计一个一般的、标准化的过程，并且实现计划好如果数据不能通过质量控制应当采取哪些措施。我们在介绍ETL子系统4-8时讲述了一个度量和解决数据质量问题的综合架构    
+ **管理维度属性的变化**  
    在第9章，我们讲述了管理维度属性变化所需的逻辑，也就是类型1（覆盖历史属性）和类型2（跟踪历史属性）。您可能需要对其他一些维采用不同的策略，尤其是对大型维  
+ **确保数据仓库和ETL系统都能满足系统可用性的需求**  
    满足系统可用性需求的第一步是记录它们的可用性。您也应当记录数据源何时可用，何时会阻塞高层作业序列。如果怀疑您的环境系统可用性是较大的问题，则您必须对对架构供选方案进行评估  
+ **设计数据审计子系统**  
    您应当为数据仓库中的每一行都标记审计信息，这些信息描述了数据是如何进入系统的 5  
+ **组织ETL登台区**  
    大多数ETL系统在ETL过程中都至少进行一次或两次数据登台操作。当您设计ETL系统时，应当对登台区究竟需要多大有较为清楚的了解，还要知道您是主要依赖于文件系统还是数据库来保存登台数据。这些决策取决于您所选择的ETL工具  

这里讲述的默认策略将成为ETL系统面向表的特殊模块的一部分。当您继续编写详细的说明时，可以参考该总体方针。实际开发时，每个表的这一模块结构上相似，有利于系统的维护  
### 10.3.4 步骤 4—由目标表向下钻取
当您为通用ETL任务设计总体策略的时候，要深入考虑对数据仓库中每个需要输入数据的目标表进行的详细转换  
在离线情况下，在您完成物理模型之后同时也就完成了源到目标映射，但是现在您应当确保自己知道每个目标表中每一列数据都来自于何处  
在完成源到目标映射的同时，您也必须完成数据探查。ETL团队也必须进行更多的数据探查和调查，以便彻底地了解每个表和列必须要做的转换 5  
1. **确保齐整的层次结构**  
    调查维度数据中的层次机构是否足够齐整非常重要。假定某个产品维包含了一个从产品SKU向上到产品类别的层次结构。如图10-2所示  
    ![10-2](../syn_pic/data_warehouse_life_cycle/10-2.png)
    为了使这种层次结构体现最大价值，每个产品SKU向上连接的模型都必须有且只有一个，最后到达树形结构的顶层  
    齐整的多对一层次结构之所以非常重要是有多方面原因的。如果您计划创建一个预先计算好的聚集（这对关系数据库和OLAP数据库实现中的查询性能都很重要），或者要在较高粒度的层面上添加数据，那么保证层次结构的齐整就很重要了  
    如果您的数据量很大，就需要中间聚集来达到良好的查询性能，而只有当中间聚集定义于齐整的层次结构上时它才能最大程度地发挥出自己的作用 5  
    在企业DW/BI系统中另一种常见的设计是在数据仓库中存储不同粒度的数据，如在模型层设计销售预测而不是为每个SKU设计  
    销售预测事实表可能会使用产品维的一致性缩减子集，但是为了能够稳定地和实际销售数据联接，您需要使层次结构变得齐整  
    根据我们的经验，除非源系统中的层次结构管理得很好，否则您不可能有可靠的层次结构  
    如果源系统不是规范化——尤其是当您的层次结构来源于业务用户的个人电脑上的excel表格时——那么您既需要对其进行修正，也需要考虑聚合的数据不要使用该层次结构  
    修整层次结构（尤其是那些由业务用户开发的层次结构）是主数据管理（MDM）系统的重要任务。有时DW/BI团队通过创建一个标准版本的雪花体来解决这一问题 5  
    为了便于ETL而实现一个标准化的维度版本是没有错的，但是您不应当在呈现服务器上将其复杂性呈现给用户  
2. **设计详细的表示意图**  
    您也应当规划一下要使用哪些表，以及以什么样的顺序使用，还要考虑到在每个表上进行的转换操作的先后顺序。可以用图形来描述这一复杂的重构过程  
    ![图10-3](../syn_pic/data_warehouse_life_cycle/10-3.png)  
    图10-3显示了对特定表向下钻取非常有用的细节层次，该图针对的是前面在公用公司的例子中所提到的若干表中的一个  
    必须在键查找步骤之前为事实表处理所有的维度表。搞清楚它们之间的依赖关系非常重要，因为它们都是作业控制流程必然会经过的固定点 5  

### 10.3.5 设计ETL说明文档
现在该将所有物理设计和高层规划上的一些总体策略综合到一起为整个ETL系统设计一份详细的说明了  
到目前为止设计的所有文档——源到目标映射、数据探查报告、物理设计决策——都应当放到ETL说明的第一段进行介绍。然后，将我们在本章讨论的所有决策都记录下来：  
+ 从每个主要源系统中抽取数据的默认策略
+ 存档策略
+ 数据质量跟踪和元数据
+ 管理维度属性变化的默认策略
+ 系统可用性需求和策略
+ 数据审计子系统的设计
+ 存储区的位置 5

ETL说明文档接下来的一段讲述每个表的历史装载策略和增量装载策略。一份好的说明文档应该为每个表都提供2到10页的详细信息，并且记录下述信息和决策：  
+ 表设计，包括列名、数据类型、键和约束
+ 历史数据装载参数（月数）和数据量（行数）
+ 增加的数据量，用新行/装载周期和更新行/装载周期来度量
+ 为事实和维度处理延迟到到达数据
+ 装载频率
+ 处理每个维度属性的变化（类型1，2或者3）
+ 表分区，例如按月分区
+ 数据源综述，包括对特有的源特征的讨论，如某种特有的短时访问窗  
+ 详细的源到目标映射
+ 源数据探查，至少包括每个数值列的最大值和最小值，每个列的不同值得数目，以及空值出现频率
+ 源数据得抽取策略，如源系统API，对数据库直接查询或者转存到平面文件中  
+ 每个源表的变化数据捕捉逻辑详细设计了源系统之间达成了什么样的协议，以及数据仓库是如何完成的 5
+ 依赖关系，包括在每个表装载之前必须先处理哪些表（强依赖）
+ 转换逻辑，以伪代码或图表的形式存档
+ 避免错误条件的前提条件，例如，ETL系统必须在进行处理之前检查文件和数据库  
+ ETL管道中每个主要步骤恢复和重启的假设
+ 每个表进行存档的假设，包括保留数据的长度和描述表的内容的元数据
+ 清理步骤，例如删除工作文件
+ 对ETL系统这一部分实现起来的难度进行估计，包括容易、中等或困难

ETL说明的最后一部分讲述整个工作流程并且排定作业的先后顺序。创建一个依赖树，用于描述哪些表必须在其他表之前进行处理。不论是否进行并行处理，了解一下不能违背的逻辑依赖关系都是十分重要的  
注意：如果您将单页的高层流程图、数据模型和源到目标映射都整合到一起，那么对您计划要做的事情就有5页描述了，这样您就比大多数团队都有了个更好的开端。同样，计划和记录的越充分越好 5  
### 10.3.6 开发沙盒源系统
在ETL开发过程中，您需要深入地调查源系统的数据。在早起开发过程中，很容易围着源系统转悠却不担心可能出现的阻断性查询，这样的话您的项目就可能要终止了  
注意:有些源系统需要通过程序接口来抽取数据，这种情况下，需要向厂商进行查证，确保您的沙盒系统与他们的许可证一致  
创建沙盒源的最简单方法是从一个一致的时间点备份中恢复源数据库。如果您的源系统非常大，可以攫取一个事实数据的随机集合  
最好的方法是确定一个最大维度中成员的随机集合（通常是客户维度）。要确保您已经选定了一个客户维成员的随机集合；然后从完整的表集合中找出其他维的来源；最后，抽取一个和客户维度相对应的事实事务集合  
创建一个仅仅复制原始数据的沙盒系统非常简单。从积极的角度来看，一旦系统进入产品部署阶段，这样的沙盒可以成为培训材料和教程的基础 5  
## 10.4 设计一次性的历史装载处理
一旦完成了ETL说明文档的创建，我们一般都会重点关注用于历史数据的一次性装载的ETL过程  
历史装载过程和增量装载过程有很多共同点，并且取决于ETL工具，您可能会在一个过程中重用另一个过程中的重要功能  
正如我们所讲述的，但这两个过程是明显不同的，通常建立一个单独的ETL系统组件来同时提供这两项功能会经常出问题，这是非常划不来的    
一般来说，从最简单的维度表开始建立ETL系统。通常至少有一个维度表会完全作为类型1维度进行管理，也就是您不需要跟踪维的历史情况    
一旦您已经建立了简单维度表，可以通过将一个或者多个列按照类型2进行管理来解决维的装载问题 5  
通常，您为一个缓慢变化维度开发的方法或代码是可以被所有缓慢变化维度所重用的。您开发第二个和后续的缓慢变化维要比第一个容易得多  
当您已经成功地进行了维度表的装载之后，您可以设计和实施事实数据的历史装载  
小技巧：力争将每个表的ETL过程都分离到自己的文件、模块、包和脚本中  
有些ETL工具会允许您创建一个文件或者模块，其中包含了所有为整个数据仓库输入数据的逻辑。那通常不是一种好的做法，因为这降低了系统的灵活性和可维护性  
首先，如果不借鉴其他人的作业，很多开发人员都会觉得很难利用系统进行作业。但即使只有一个开发者，长期来看ETL系统模块化都便于建立一个更便于维护的应用程序 5  
对一次性历史装载ETL过程的总体流程有所了解之后，我们开始进行更深入的研究，从维度表开始  
### 10.4.1 步骤5—使用历史数据填充维度表
在本节，我们讨论ETL维度处理，从最简单的情况开始  
1. **填充类型1维度表**  
    最容易填充的表的类型是所有属性都按照类型1进行管理的维度表，它们也可以适当地更新。对于类型1维，您对源系统每个维度属性都抽取当前值  
    有时尽管数据理论上是很简单的，但是却以某种无法访问的方式存储的。只要一份报表是经过正确定义并且是可以预先安排好的，就是一种可以接受的抽取机制  
    当您定义报表或者抽取维度的源数据的查询时，要确保每个实体都占用一行：列或者您已经标识为源系统自然键的列事实上都被唯一标识为一个维度成员 5  
    验证唯一性的简单方法时验证行数是否和抽取流中不同成员的数目相等  
2. **维度转换**  
    即使最简单的维度表也需要基本的数据清洗，当然也需要进行代理键分配  
    1. 简单数据转换  
        最常见、最简单的数据转换形式是数据类型转换，所有的ETL工具都有很多进行数据类型转换的函数  
        对于维度表来说，我们建议使用某个默认值或者空字符串来代替空值，这一点在第8章就已经讲到了，这样做的主要原有是为了可用性，业务用户通常都会被数据库中NULL的用户搞糊涂  
    2. 联合各独立的源  
        通常维都来自多个源。客户信息可能需要通过将多个业务条目和外部源进行合并后得到。很难找到一个通用键来使该合并操作变得更容易些 5  
        原始数据和已有维度表数据可能需要在不同的时间存储于不同的字段中，以便于匹配。有时匹配使基于模糊的标准进行的，拼写差别较小的姓名和地址可能会被错误地匹配到一起  
        如果您对姓名和地址进行语法分析，将它们区分开来，那么大多数用于重要的客户维的合并和非重复复制工具即过程都能很好地完成作业  
        当您认为使用标点符号会导致无法精确匹配的时候，语法分析就非常重要了  
        在数据语法分析和清洗阶段之后，可以对来自多个源的数据进行对比  
        一旦某一对记录被确定为一个匹配，那么就在一个带有数据仓库代理键的映射表中添加一条新的记录，并且对多种不同源系统的键都进行匹配 5  
        接下来，您需要设计一个使用模糊逻辑的通道集合来解决拼写错误、输入错误和对等的拼写，如I.B.M.、IBM和International Business Machines  
        所有的模糊匹配算法都无法保证返回的是精确的匹配结果。您的业务用户必须提供一个有关匹配算法不精确程度的可接受水平的说明  
        在大多数机构中，使用以前的大型项目可以将已有客户信息进行合并。该项目中的数据分析和数据操作同等重要。这对主数据管理系统来说是非常有价值的作业  
    3. 生产代码的译码  
        在数据准备过程中的一项通用合并任务是查找和生产代码相等的文本。有些情况下，等价的文本来自于非正式的非生成源，如电子表格  
        确保您的ETL系统包含有创建默认译码等价文本的逻辑 5  
        通常创建一条规则来生产一个虚标签是很有意义的，稍后可以很容易进行更新。例如可以创建一个虚标签：“生产代码ABC：无可用描述”  
    4. 验证多对一关系和一对一关系  
        您最重要的维度不可避免地会有一个或者多个上卷路径，如从产品到产品模型，从子类别到类别，如图10-4所示。大多数情况您都希望这种层次结构的上卷是非常齐整的  
        ![10-4](../syn_pic/data_warehouse_life_cycle/10-4.png)          
        您可以通过对“多”属性进行排序来验证两个属性间地多对一关系，如产品到产品模型，也可以验证每个“一”属性上的值都是唯一的  
        也可以使用SQL语言的HAVING字句来查找背离情况。您可以执行一个查询来返回拥有多个产品模型的产品集 5  
        检查您的ETL工具的功能，看看这样的一般性测试可以得到哪些特点。数据探查工具应当可以使用简单的命令就能执行这一的测试  
        很多数据管理员都想通过在登台数据库中将数据装载到规范化的维度表来验证上述类型的关系。规范化维度表看起来形如图10-5所示的雪花维  
        ![10-5](../syn_pic/data_warehouse_life_cycle/10-5.png)  
        在登台区雪花结构是有一些价值的：它能够防止您装载违背多对一关系的数据。然而，一般情况下，您应该像我在前面所讲述的那样预先验证关系，这一就不会将异常数据装载到维度表中    
        对两个属性间一对一关系进行一致化是一个比较相似但是又更为严格的问题。如果需要这样的设计，就必须从两个方向对关系进行确认 5  
        同样可以使用我们处理多对一关系时所讲述的方法，对两个属性都进行测试。当然您最好按照呈现服务器数据库中的维度版本来确定规范化的水平  
    5. 维度代理键分配  
        一旦您确信自己的维度表版本中每个真正唯一的维度值都占一行，那么就可以开始分配代理键了  
        代理键一般都是按照整数分配的，每个新键增加1即可。
        一次性创建序列：  
        > create sequence dim1_seq cache=1000; -choose appropriate cache level  
        
        填充键映射表: 5  
        > insert into dim1 key map (production key_id, dim1_key)  
        select production key_id, dim1_seq.NEXT  
        from dim1_extract_table;  
        
        并非所有主流关系引擎都使用序列结构。另一种常见方法是将表的一个列指定为IDENTIFY列，SEQUENCE和IDENTITY方法少由不同，但是它们都能够达到相同的目的  
        另一方面，大多数ETL工具都自带了代理键生成器，并且在数据库外部作业  
3. 维度表装载  
    一旦数据都准备好了，装载过程就变得非常简单了。宁应当对大多数表的插入操作使用快速装载方法，而且从一开始就应当培养使用这类技术的习惯  
    每一种主流关系数据库引擎都具有快速将数据装载到数据库的技术手段，而且每一种主流引擎都有卓越的实用程序，您可以从操作系统中直接进行调用  
    大多数数据库也都具有类似的用户界面。有些对SQL语法进行了扩展，包含了一个BULK INSERT语句，还有一些发布了API将数据从流中装载到表中 5  
    批量装载对数据仓库技术非常重要，因为它能够快速插入数据，我们了解到使用快速批量装载方法的执行速度要比一行接一行的记录插入方式快10到100倍  
    批量装载程序和API都有很多的参数，具有一定的转换功能，对使用这两种方法还有下述一些建议：  
    + **关闭日志记录**  
        面向事务处理的数据库将对数据库所作的所有改变都记录到日志文件当中。然而，日志明显增加了开销，并且对装载数据仓库表没有什么意义  
        您应当为ETL系统设计一个或者多个可恢复点，这样，出现故障的时候您就可以重新启动处理过程  
    + **快速模式下的批量装载**  
        在批量装载的快速模式下，大多数据数据库引擎的批量装载程序或API都对目标表有一些严格的条件 5  
        这些条件都有所不同，但是它们的共同点是要求目标表是未建立索引，或者表必须是空的  
    + **对文件预排序**  
        按照主索引对文件进行预排须可以显著提高索引速度  
        当然，您应当经常对整个过程进行评估，评估的方式就是将预排序、装载数据和为表编制索引加起来所需要的时间和除去预排序步骤后所需的时间进行对比  
    + **小心谨慎地进行转换**  
        有些情况下，装载器会要求您进行数据转换、计算和字符串与日期/时间处理。要小心谨慎地使用这些基本功能，并且测试其性能。建议您使用ETL工具来执行大部分转换操作  
    + **依赖ETL工具**  
        大部分ETL工具都带有一些利用数据库引擎地批量装载器功能。您要确保自己了解如何在您的关系数据库上使用这些功能 5  
    + **在完全更新之前截断表**  
        Truncate table语句是删除表中所有行效率最高的一种方式。通常每天ETL处理刚开始将登台数据库的所有表都清理干净  
    
    注意:大多数关系引擎都有TRUNCATE TABLE命令，它在逻辑上和DELETE命令式等价的。通常TRUNCATE TABLE命令是无法撤销的     
4. 装载类型2维度表历史  
    大多数维度表都混合使用类型1和类型2两种属性  
    当您设计一个将初始历史数据填入维度表中的系统时，将需要考虑到包含类型2属性的维相关的作业  
    应当为以类型2方式进行管理的维度属性重建历史。通常重建维度属性历史是非常困难的，并且有时是完全不可能的。从操作层面上来看，历史维度属性通常都会被转移到单独的历史表或者离线存档文件中 5  
    您需要做一些调查作业来评估以下，看看哪些属性的历史是可以重建的，以及进行重建的难度如何  
    如果您已经识别出属性历史的来源，就需要构建整个维度行在过去每个时间点上的镜像  
    您知道每个维度属性的当前值，因此可以从当前时间点开始。沿着时间向后反推，每当属性变化发生时应当为维度行建立一个新副本。要确保行的生效日期/时间戳列是正确的  
    此时不要将创建好的行写入维度表，将它们写入登台表或者文件中，也可以将它们作为结构保留在内存中  
    您可能需要在维度表上开辟多个通道，通常为了处理不同类型2属性集合中的变化，需要一条从当前到过去的通道 5  
    此过程并不适合标准SQL处理，最好使用一个数据库游标结构，甚至还可以使用过程语言如VB、C或者JAVA来处理，这样效果更好  
    当您已经完全重建历史后，建立最后一个贯穿数据的通道来设定行终止日期/时间戳列。最重要的是要确保序列中没有间隙  
    您可能会被这样的事实搞糊涂：为了进行历史装载，每个维度成员都是按照年月日逆序排列，但在运行过程的增量装载中有时按照正常顺序进行的  
5. 填充日期维和其他静态维  
    每个数据仓库数据库都应当有一个日期维，通常粒度为每一天占一行  
    为了历史数据建立日期维是很简单的，因为您已经知道了要装载的历史事实数据的日期范围。大多数项目都手工建立日期维，一般都是用电子表格形式创建 5  
    如果已经创建了日期维，可以首先将其保存为文本，然后装载到数据仓库中，并且随后使用标准的日期装载方法来装载平面文件  
    您可以在本书的配套网站www.kimballgroup.com 上找到日期维店子表格示例  
    您可以用相同的方式来创建其他一些维。此外，您也可以直接使用INSERT语句，这些表通常都只有很少的行  
    
### 10.4.2 步骤6—执行事实表历史装载
事实表数据的一次性装载和运营过程中的增量处理明显不同。在事实表历史装载中最令人头痛的是数据量，通常要比日常增量装载大数千倍  
历史装载更像是一个数据集成的项目，而并非是简单的应用 5  
1. 历史事实表抽取  
    当您标识出那些应当列入抽取过程基本参数范围内的记录时，需要确保这些记录对数据仓库是有用的。因为很多事务处理系统在源系统中都保留了操作信息，从业务角度来看可能没有多少意义  
    在设计阶段创建的业务规则有助于确定这些事件是在数据仓库中如何描述的，以及在结果抽取过程中应当使用哪些筛选和求和计算  
    确定和验证这些规则是很令人讨厌的。在抽取过程中进行表面上的数据分析往往会忽略业务规则。跟踪这些细微的区别需要熟练的探测技能和无限的耐心  
    在这一阶段累计审计统计数据是不错的想法。当抽取过程创建了结果集后，通常就能捕捉多个分类小计、总计和行数  
    1. 审计统计数据  
        在ETL系统的设计阶段，您应当明确有关数据质量的多种度量。通常都是一些计算，如计数和总计等，您可以使用它们对数据仓库和源系统进行比较，进而对数据的集成进行交叉检查 5  
        如果还没有捕捉到这些数据质量统计数据，那么这一阶段就需要对数据进行遍历来计算那些度量。这些数值应当向后与日常业务报告相关，向前与数据仓库装载过程结果相关。应该将这些数据存放在为这类信息创建的表中的元数据类别中  
        向后和业务系统紧密相关是非常重要的，因为这确立了数据仓库的可信度。如果结果集没有和该来源后向相关，那么可能对业务规则的改变没有反映到抽取过程中。最好停止抽取过程并且立刻进行检查  
        注意：有些情况下数据仓库很难或者无法完全和源系统相对应。很多情况下，数据仓库抽取还包含那些已经应用于源系统的业务规则。如果无法准确地和源系统数据相对应，那么您就要能够解释它们之间的区别  
2. 事实表转换  
    很多项目中，事实数据都相对比较干净的，ETL系统开发人员花费大量的时间来改进维度表内容，对事实通常只需要相当有限的转换  
    对事实数据最一般的转换包括空值的转换、旋转数据或撤销对数据的旋转、预先计算好的导出计算结果，然后，所有事实数据都必须进入代理键管道，将自然键更换为数据仓库的维度代理键 5  
    1. 空事实表值  
        显然所有主流数据库引擎都支持空值。然后在很多源系统中，空值是用一些特定值来表示的，而且这些值都是合法事实。大多数情况下，此时您应当将其中的特定值替换为NULL。在事实表中使用空值进行数值度量是很常见也是很合理的  
        空值在对整个事实表进行总计和平均值计算时非常有用，只是在维度表中我们要坚决将空值剔除掉。然而，在事实表参考维度键的外键列中也不能有空值，这些列中的值都应当定义为NOT NULL    
        数据中的空值处理起来很棘手，因为本质上存在两种类型的空值  
        1. 数据中的空值可能意味着进行度量时，该值不存在或者不能存在。换句话说，在这里取任何值都是错误的   
        2. 反过来说，数据中的控制可能意味着度量过程无法提交数据，但是值在某个时间点是确定存在的。第二种情况使用一个估计值要比使事实记录无法进行分析好得多 5  
        
       如果您给忽视了我们再三提到的有关使用代理键的建议，并且设计了一个事实表使用日期作为参考数据维度表的外键，空日期就会造成某种特殊的问题  
       在事实表中不能使用带有空值的外键，因为在SQL中空值和自身是不相等的。您应当使用匿名的整数键来实现联接，并且随后在维度表中使用某条特殊记录来表示空日期  
    2. 改进事实表内容  
        最终事实表行的所有事实都必须以相同的粒度进行表达。如果抽取中出现了不同粒度的交叉，转换过程必须删去这些聚集，或将它们转移到适当的聚集表中  
        事实记录中也包含导出事实，但是很多情况下在视图或OLAP数据库中计算导出事实要比物理表中进行计算效率高得多  
        例如，一个包含收入和成本的事实记录中可能希望表示出净利润。如果用户可以查看物理表，或者他们经常将净利润作为筛选条件并且您又不想为其创建索引，那么最好将净利润预先计算好并且进行物理存储 5  
        类似地，如果有些事实需要同时以多种度量单位标识，也使用相同的方法  
    3. 重构事实数据  
        在ETL系统中改善数据的内容并不是事实表的转换。有时结构需要调整——对数据进行非规范化或者规范化以使其从业务的角度来看更加有用  
        事实非规范化的常见例子是一个财务模式，它在源中有大量关于事实表的类型字段，指示这些相关的数值是预算金额、实际金额还是预测金额，将这些记录旋转为单列，以简化变量的计算  
        您的ETL工具可能具有旋转和撤销旋转的转换操作，因此可以用它来执行该任务。或者，您也可以使用SQL语法给出的一个示例  
        > select  
        > account_number,  
        > sum(case when amt_type ='actual' then trxn_amt else 0) as actual_amt,  
        > sum(case when amt_type ='budget' then trxn_amt else 0) as budget_amt  
        > from sourc_trxn_table  
        > group by account_number 5  
        
        有些情况下您可以将这一过程反转过来，通过撤销旋转来对事实数据进行规范化。例如在图10-1中描述的使用系统一行之中13个月的数据对应13列    
    4. 维度代理键查找管道  
        维护事实表和维度表之间的参照完整性关系非常重要，您不能使事实行参考一个不存在的维度成员    
        在将数据装载到目标事实表时，代理键管道是最终事实处理过程的名称。代理键管道是将自然键改换为代理键的过程，也能够处理所有的参照完整性错误  
        维度表处理必须在事实表进入代理键管道之前完成  
        在用于历史装载的代理键管道中，当事务处理发生的必须将事实和受影响的每个类型2维度行关联起来，并进行参照完整性验证 5  
        在历史装载过程中确保参照完整性是比较简单的，因为您并不需要使这一过程自动化。一种比较好的方法是在进入代理键管道之前，先登台干净的事实数据，然后为历史数据每个维都创建一个自然键列表  
        接下来，要确认历史事实数据中的每个自然键就是一件简单的作业了。调查所有的问题并且修正维度表或者重新设计事实表的抽取，以便尽可能减少不合逻辑的行  
        现在您已经确信没有违背参照完整性的问题了，您可以设计一个没有任何参照完整性错误处理的历史代理键管道，如图所示  
        ![10-6](../syn_pic/data_warehouse_life_cycle/10-6.png)  
        历史装载是一个一次性的过程，而且通常比较容易暂时停下来手工修正错误，然后再继续进行。您不需要一个像增量处理那样的高度自动化的系统 5  
        还有一些方法可以将历史装载的代理键管道的性能设计得更加优良，而且您的设计取决于ETL工具的功能、所要处理的数据量以及您的维度设计等  
        理论上，您可以定义一个查询将事实登台表和每个与自然键相关的维度表连接起来，并返回每个维度表的事实和代理键  
        如果历史数据量并不大，而且假设您已经将事实数据登台到关系数据库中，并为维度表建立了支持该大型查询的索引，那么这种做法的效果就比较好。这种方法有以下一些好处：  
        + 利用了关系数据库的优势
        + 对所有维度进行并行的代理键查找 5
        + 简化了为类型2维找出正确维度键的问题，如图10-6所示，类型2维度的连接必须包含一个子句，说明表中事务处理日期在镜像维度成员的行生效日期和行终止日期之间  

       如果历史数据量很大，在几百GB到TB之间，那么就没有人愿意使用这种方法了。大多数大型事实表都会按照日期进行划分，通常是按月来划分。使用上述方法一次性处理一个月的数据还是比较合理的  
       到类型2维度表的复杂联接对系统有极高的要求。您应当使用这种关系方法对一次传输中所有的类型1维度进行代理键查找，单独处理类型2维  
       另外一种可以代替数据库联接方法的供选方案是使用ETL工具的查找操作程序。依据ETL工具性能不同，有的可能会很容易就实现并行查找，有的则相当困难，还有一些甚至无法实现  
       当所有的事实表生产键都被替换为代理键后事实记录就做好了装载准备 5  
   5. 指定审计维度键  
       企业类的数据仓库应当在每个事实行都包含一个审计键。审计键指向描述装载特征的审计维，而这些描述主要包括时限和数据质量方面的度量  
       审计维和审计子系统都很复杂，包含很多表，也包含很多数据质量的度量。最简单的情况下，审计系统包含至少一个审计维度来捕捉所有最基本的信息  
       在ETL工具中，您在一项作业、一个包或模块中首先要做的事情是建立向审计维添加新行的逻辑。在开始抽取和转换过程之前进行这一步作业，并且为刚刚插入的审计维度行指定键  
       在代理键管道之前或者之后立刻将审计维度键添加到事实表中。然后将事实数据插入数据仓库事实表中。最后在作业或装载历史数据模块完成之前，回到审计维度去更新信息，包含作业要花费多长时间和处理过程中装载了多少行等  
3. 事实表装载  
    在装载事实表时您关心的主要是装载过程能否能够顺利完成。当然，在装载全部历史数据之前，应当使用较小的数据集来测试ETL过程 5  
    无论您测试历史ETL装载过程时结果有多好，当您进行完全装载的时候都会遇到问题。有些ETL工具帮助您建立一个错误流，将在表插入过程中发生故障的行抽取出来  
    在历史装载过程中，您可能会直接将该错误流导入一个文件中，稍后在手工处理这些问题行  
    有些数据库技术支持指定批量大小的快速装载。看看您的快速装载技术的相应文档，了解一下如何进行参数设定  
    除了使用批量装载器和设定合适的批量大小（假设适用于您的数据库引擎）之外，改进历史装载性能的方法是将历史数据装载到某个分区表中，这样就可以并行将数据装载到多个分区中。向分区表进行装载的步骤是：  
    1. 在装载数据之前，使事实表和每个维度表之间的外键（参照完整性）约束失效。能够使装载性能提高一到两个数量级 5
    2. 删除事实表上的索引或使其失效。和装载带有索引的数据相比，先装载数据随后再建立索引的装载方式通常会更快些。查证一下数据库文档，如果在您的环境中可以对一个大小合理的数据子集进行测试就更好了  
    3. 使用快速装载方法装载数据，这和我们在本章前面提到的维的快速装载方法类似。如果事实表进行了划分，同时向多个分区装载数据会显著提高装载性能  
    4. 创建事实表索引或者使原来的索引有效  
    5. 必要的话，执行一些将多个表分区整合到一起的步骤。您应当查证一下数据库引擎的文档，确认一下到底应当在多个分区整合之前建立索引，还是在整合之后建立索引  
    6. 确认每个维度表的代理键列是否都只有唯一索引 5
    7. 将事实表和维度表之间的外键约束恢复为有效
4. 测试、测试、再测试  
    当历史事实数据最终完成装载过程之后，您可能要开始庆贺了。尽管这是一个重要的里程碑，但是只有当您确认所有数据都是准确的之后，才能宣布自己已经取得了成功  
    在向前推进到增量装载过程之前，确保您预留了足够的时间对历史数据的正确性进行充分测试。一旦验证完毕，要对数据库进行完全备份  
    
## 10.5 设计增量ETL处理过程
增量ETL过程最大问题是标识新行、变化行和删除行  
如果您已经得到了一个插入、修改和删除行的流，ETL系统会按照实际上和历史数据装载相同的业务规则来应用转换。根据特点不同，您也许可以重用历史装载程序中的某些组件  
维度和事实的历史装载很大程度上或者完全是插入操作。在增量处理时，主要也是进行插入操作，但是也不可避免地要对维度和某种类型的事实表进行更新操作  
将增量装载过程设计为一次处理一天的数据量是一种比较好的做法，您还可以通过参数来设定要处理哪一天的数据  
### 10.5.1 步骤7—维度表增量处理
增量ELT系统的开发从维度表开始。维度增量处理和前面所说的历史处理有很多相似之处 5  
1. 维度表抽取  
    很多情况下，多有一个客户主文件或者产品主文件可以作为维的单一来源。在另一些情况下，原始数据将将维度和事实数据混合到了一起。例如，客户订单可能包括客户信息、产品信息和来自订单详细条目的数值型事实  
    通常完整的将维度表的当前快照抽取出来是最简单的，并且可以通过转换步骤来确定哪些内容变化，以及是如何处理的。如果维度表很大，您必须使用我们在后面要讲到的适用于事实表的方法来标识变化的记录集  
    如果可能的话，仅抽取那些发生变化的行即可。如果源系统维护了用户标识变化类型的指示符，这就非常简单了  
    使用这样的指示符，您可以将新记录放到一个文件中，不需要进行查找操作就可以添加到维度表中。第二个文件中保存经过缓慢变化维过程后更新的记录  
    如果您不知道变化类型，那么就认为整个变化记录的结果都要经过缓慢变化维过程 5  
2. 标识新的维度行和变化的维度行  
    此时，ETL过程需要执行一个开销很大的对比操作来确定新行和变化行  
    在进行标识过程之前，维度数据应当是干净的并且是转换好的  
    原始数据都有操作自然键，它必须和当前维度行中相同的列匹配。记住，数据仓库维中的自然键是一个普通的维度属性，并不是维度的代理键  
    首先，查看一下ETL工具是否有某种转换功能或向导可以为您执行上述任务  
    通过执行一个从输入流到主维度的查找操作可以找到新的维度成员，对比它们的自然键即可。另一种标识新行的供选方案是使用输入数据和已有维度表之间的外部链接 5  
    如果维度包含类型2属性，就为系统出现的每个维度属性设定生效日期/时间戳列。如果您每天晚上处理数据的话，日期通常都是前一天  
    下一步是确定维度行是否变化了。最简单的方法是将输入数据和存储于主维度表中当前存储的对应成员一列列地进行对比  
    如果维度很大，有一百多万行，那么简单地进行列的比较就显得太慢了。另一种比较流行的方法是使用散列函数或者校验和函数来加速比较过程。该算法将一个非常长的字符串转换为一个非常短的字符串，而且转换后的字符串几乎是唯一的  
    此外，您的关系数据库引擎可能还有像EXCEPT这样的语法，在查找变化行的时候可以有很高的查询性能  
    作为一般规则，您不应当删除源系统中已经删除了的维度行。这些维度成员很可能会有一些与之相关的事实数据。很多源系统中都有类似于“软删除”的操作，即在行上添加一个标志，当该行被移动到历史表或已删除表时就更新该标志 5  
3. 维度属性的过程变化  
    ETL应用程序必须含有一些业务规则，以便确定如何处理那些和数据仓库中存储的值相比发生了变化的属性值。如果是修正后的描述是对以前信息的he且可信的更新，应当使用缓慢变化维方法  
    准备维度记录的第一步是判断您是否已经拥有了该条记录。带有自然键的原始数据必须和当前维度行的同一属性相匹配  
    如果所有输入的维度信息都和维度表中的对应行相匹配，那么就不需要进行进一步的处理了。如果维度信息已经变化了，将类型1和类型2变化应用于维  
    注意：对属性值的变化跟踪主要有三种方法。类型3需要维度表结构所有改变，创建一个新列集合来保存和属性的当前版本相对应的上一个版本。这种结构变化类型在ETL系统中一般都不是自动完成，而是将其作为数据模型的一次性变化来处理

### 10.5.2 步骤8—事实表增量处理
大多数数据仓库过于庞大，以至于无法在单个装载窗中对它们的核心事实表进行处理。一般都对新的事实行和更新的事实行进行增量处理  
注意：仅对自上次装载以来添加的和更新的记录进行增量装载是非常高效的  
用于事实表增量处理的ETL过程和历史装载是不同的。历史ETL过程不需要是完全自动化的，我们可以停止装载过程来检查数据，并且为下一阶段做准备 5  
您的系统应当具有数据质量检查点，这样就可以有计划性地判断是否需要停止、暂停和寻找操作源，或者继续装载数据  
1. 事实表抽取和数据质量检查点  
    一旦从源系统抽取了新的事实行和变化的事实行，接下来应将没有转换的数据做一个副本写到登台区中。同时，对抽取到的数据的数据质量进行度量。登台后的数据有三个用途：  
    + **存档以便于审计**。大多数机构都将这些登台数据存储一段时间，通常是几个月或者更长时间  
    + **在数据质量验证结束后提供一个新的起点**。当您计算了数据质量的度量后，需要确定抽取是否满足下一步作业的最小必要需求。数据在您做出决定的时候应当存储在磁盘上  
    + **为重启过程提供一个新的起点**。如果处理过程中出现灾难性的故障，您的系统应当能够重新启动。大多数情况下，最好从一个ETL系统的一个静态数据集重启，而不要从源系统的抽取开始 5  
   
   ETL系统下一阶段设计要考虑究竟是将抽取的数据登台到一个表中，还是将其登台到一个文件中  
2. 事实表转换和代理键管道  
    增量事实表的转换和您为历史装载建立的转换非常类似。您应当可以利用以前编写的代码  
    增量事实数据的代理键管道和历史数据的代理键管道很类似。最主要区别在于对违反参照完整性的错误处理必须是自动化的。处理违反参照完整性的情形可以使用下述方法：  
    + **停止装载**。尽管这是很多ETL工具的默认做法，但是通常都不是一个有用的解决方案  
    + **丢弃错误行**。有些情况下遗漏的维度值是一种信号，说明该数据和数据仓库的潜在业务需求是不相关的。您应当在ETL过程中早点设定筛选条件——可以在抽取逻辑中，但是应当在代理键管道中有实施该逻辑的设计原因 5  
    + **将错误行写入到文件或者表中，以便进行后续分析**  
        设计一套将正确的行移动到数据仓库的机制，而且ETL系统应当向合适的人员发送消息。尽管您自认为不会发生违背参照完整性的问题，但是这种方法不能有效保证这一点  
    + **通过在维度中创建一个虚行来修正错误行，并且将其代理键返回给管道**  
        在增量代理键管道中，对违反参照完整性的情况最优吸引力的错误处理方式是为未知的自然键创建一个虚维度行  
        随后可以将所有其他的属性都设置为默认值。查询并将新维度成员的代理键返回到错误行，对错误行进行合并后再将其返回代理键管道
    + **不推荐的方法：通过将错误行映射到每个维度中的某个未知成员上来修正错误行** 5  
        正如我们在第8章所讲述的，在每个维度表中插入一个单独的虚行是很常见的，也称作未知成员行。您可以指定一个特定的代理键值，如“-1”  
        这种方法的问题在于所有错误行都被映射到相同的维度成员，对在事实表抽取中自然键的任何值都是如此，因为没有在事实表中保留自然键，这就将以后修正错误行所需的信息丢掉了  

   对大多数系统来说，可以针对维度表子集的一个查询、视图或者物理表来执行代理键查找  
   如果维度表非常大，您可能会通过在事实数据上建立一个预处理通道来显著地改进性能。第一个通道为那些实际用于增量事实中的超大型维标识了自然键，并且创建了一个仅面向于这些键的查找表  
3. 延迟到达事实和代理键管道  
    在大多数数据仓库中，增量装载过程都在午夜后开始，处理前一天发生的所有事务 5  
    然而，有时一些事实会延迟到达，当数据源是涉及多台机器的分布式系统或者甚至是世界范围的分布式系统时，更有可能出现这种情况，连续性或者说等待时间造成了数据无法及时收集  
    如果您的维完全都是按照类型1覆盖方式管理的，延迟到达事实不会造成特殊的难题。延迟到达事实必须和当事实发生时有效的维度成员的版本相关联。对延迟到达事实的查找应当使用能够识别出正确代理键的联接条件  
    该复杂的联接条件要比标准情况中的等联接条件执行起来慢很多。您应当从延迟到达事实中分离出正常的事实，并且对量最大的数据使用等联接方法处理  
    您也可以对那些延迟一天的事实进行同样的操作，并且随后在处理延迟事实时，使用复杂联接条件进行查找  
4. 增量事实表装载  
    装载增量事实表行的物理过程和前面讨论的一样。对于历史事实装载，您应当使用数据库的批量装载技术来执行对事实表的插入操作 5  
    在历史事实装载中，使用快速装载方法确保数据装载是非常重要的。在大多数数据仓库中，增量装载都无法使用这些快速装载技术。对增量装载来说，通常使用非快速装载方法要比完全填充表或者为表建立索引快得多  
    如果您的事实表很大，就应当为了便于处理而对它们进行划分。如果您总是将增量数据装载到空的分区中，那么就应当使用快速装载方法  
    对于历史比较久的事实表来说分区就会很多了。为此，可以设想执行一个过程将日分区合并为周分区或月分区  
    最大的事实表通常都是事务处理。事务处理层的事实表都通过插入操作进行装载，很少进行事实数据的更新，有时甚至是禁止的  
    一个周期快照事实表通常都有月末快照。当前月的数据有时会每天进行更新。这种情况下，事实表的月分区很容易对当前月的数据进行重载，并且具有较高的性能 5  
    累计快照事实表保持对相对短期的过程进行追踪，例如填充订单。该表的维护开销较大，尤其当您的数据库引擎不支持类似于批量插入方法的批量更新功能更是如此  
    为了以最好的性能应用更新，将目标表的新镜像登台到一个登台表，然后使用SQL集合操作在单个事务处理中更新多个行即可  
5. 加速装载周期  
    仅处理变化增量是加速数据登台周期的一种方式。本节给出了几种可能对您很有价值的补充方法：  
    1. **更频繁的装载**  
        每夜处理意味着只处理每个月数据量的1/30。大多数数仓都采用夜间装载周期  
        如果您的夜间处理开销过大，就试着在白天对数据进行一些预处理。在白天，将数据转移到一个登台数据库或者ODS中，同时执行数据清洗作业 5  
    2. 并行处理  
        一种缩短装载时间的方法是使ETL过程并行化。这可以以两种方式出现：并行运行多步骤，或者单个步骤以并行方式运行  
        警告：将处理过程分解为并行的步骤既是一种好方法也是一种坏方法。实现并行化处理的一种简单方式是将所有的源数据都抽取到一起，然后装载并转换维，随后立刻检查事实表和所有维度表之间的参照完整性  
        这种方法并不比最简单的顺序装载方式快——可能还要慢一些，因为每一步都启动并行处理，对相同的系统资源如网络带宽、I/O和内存等都有竞争  
        + **多个装载步骤**。将ETL作业划分为几个需要同时提交的独立作业。主要目标是创建独立的作业。如果在一个装在运行成功之前需要另一个装在的结果，这两个装载在同一作业中就是顺序关系  
        + **并行执行**。数据库本身也可以确定某些它能够并行执行的任务。例如，创建索引一般都是并行化的，涉及到机器上多个可用的处理器。应该确保您的数据库可以进行并行活动，而且这种功能是可用的，监视您的作业并观察发生了什么 5  
    3. 并行结构  
        您可以在两个服务器上建立一个三次镜像或者聚簇配置来维持对数据仓库的连续装载，一个服务器管理装载，而另一个处理查询。每天将维护窗缩减到几分钟以便两个服务器所属磁盘进行交换  
        根据需求和预算不同，对表、分区和数据库有几种比较相似的实现方法。例如，您可以向一个离线分区或者表进行装载，并且在最小故障的前提下将其交换到作业装载。其他系统有两个数据仓库数据库，一个用来装载另一个用来查询  

### 10.5.3 步骤9—聚集表和OLAP装载
聚集表从逻辑上是很容易建立的，简单点说就是将一个实际的大型聚集查询的结果集存储为表  
我们认为聚集表也面临着增量和完全刷新的而选择，其原因和基本表也是一样的。幸而，事实表的增量装载是聚集表增量装载的一个非常好的来源  
如果聚集表含有沿着日期维进行的聚集（假设粒度为月），聚集维护过程就更为复杂 5  
如果聚集表定义在按照类型1覆盖管理的维度属性上，也会发生类似的问题。维度属性上任何类型1变化都会影响到该属性上定义的所有事实表聚集  
聚集管理系统要将聚集和底层事实数据保持同步，这一点是非常重要的    
大多数的数据仓库都不直接管理聚集表。它们使用关系数据库索引视图或物化事务、OLAP数据存储，或者第三方软件来管理整个过程。聚集管理技术应当能够处理我们刚刚提到的更新操作    
大多数聚集管理技术在处理简单问题上都很在行，例如向已有聚集表添加新的事实行  
从我们的经验来看，这些技术对稍难一些的问题就没有这么好的执行效果了，比如说从聚集的某一部分返回数据并且添加到另一部分中。这些问题中难度最大的是和事实表更新相关或者类型1维度属性中的变化相关问题 5  
聚集管理技术是风险厌恶型的，这就意味着为了处理复杂的更新，聚集管理需要对聚集表的重要部分（或全部）重新计算  
聚集管理技术可以帮助您控制已经建立的聚集。明智的数据仓库经理在类型1维度属性上定义聚集会非常谨慎，并且很可能会对于接受到很多更新的事实表相关的聚集进行数量限制    
### 10.5.4 步骤10—ETL系统操作和自动化  
理想的ETL操作能够快速地运行常规装载过程，而不需要人工干预。当然，您要在架构计划中明确给出在您的数据仓库中所需的操作功能。最后，我们将罗列出常用的操作功能，随后对一些实现它们的方法简单进行介绍：  
1. 调度作业  
    调度作业通常比较简单。您的ETL工具应当具有在某一时间调度某项作业并开始执行的功能。大多数ETL工具都具有在第一项任务成功完成之后继续执行第二项任务的功能  
    如果您的ETL工具不具有这样的功能，或者您不喜欢它的作业调度方式，那么就可以写一个脚本程序来实现这种类型的作业控制 5  
    有些机构有公司标准作业控制管理工具。使用作业控制工具的功能可以使操作系统调用和启动ETL工具  
2. 自动处理预测到的异常和错误  
    尽管启动作业是非常简单的，但是要使其运行完整却是一件比较困难的工作，需要认真处理数据错误和异常。好的错误处理应当一开始就加入到ETL作业中。债胡搜啊，应该确认行数是否合理    
    自动化ETL系统应当涉及一次到多次数据质量检查，并且您应当内置处理这些问题的方法  
3. 恰当地处理不可预测地错误  
    有些处理是可以预测，例如某个准备要进行填充的字段接受到了某些提前到达的事实或者NULL值。其他一些错误都是无法预测的，从数据接收混乱到处理过程中的断电故障，有很多种  
    有些ETL工具可以将多个从异常情况中恢复的步骤组合，这其中也包括数据库操作步骤。一个可靠而灵活的系统设计会在处理过程中登台数据以便为系统重启提供一个中间点 5  
    您的ETL设计师要能够胜任设计ETL系统的重任，要考到作业冲突后的系统重启和其他一些重要的错误。您必须决定将哪些数据操作整合为一个事务，并且建立能够使您的系统稳定重启的功能  
4. 维护数据库对象  
    很多数据库管理任务在ETL周期中都能稳定工作。这些任务包括数据库备份、周期性索引维护、表分区管理和表空间增长等。例如，在每次夜间装载完成之后对数据库都进行备份是一种很好的做法。将备份整合到ETL工作序列中是很合理的  
    对那些已经真正成为生产系统的数据仓库来说，您需要以某种形式来提供上述功能。一开始这些功能可以仅仅是初步的，但是您应当在稍后立刻开发出更为稳健的工具  
5. 开发和设计ETL自动操作  
    在项目中，增量处理的ETL操作自动化设计必须早点动手，并且要和整个ETL系统有机地整合到一起。在ETL过程中，您也应当花点时间评估数据质量问题，并且在在数据没有达到质量标准时，应当自动停止装载过程   
    在13章，我们在整个DW/BI系统的部署和维护背景下讨论有关ETL应用程序开发和测试的主题 5  
    
## 10.6 小结
开发ETL系统是DW/BI项目中的最大挑战之一。无论您的访谈和分析过程是如何完成的，当您开始建立ETL系统的时候都会遇到数据质量问题  
抽取逻辑是非常困难的一部。您需要和源系统编程人员密切合作来设计一个能够生成高质量数据抽取过程，同时不能给事务处理系统留下过于沉重的负担，而且抽取逻辑还需要能够整合到自动化ETL系统的处理流程中  
我们尽力去解决维度设计实现过程中的问题，强调使用代理键的重要性并且详细讲述如何处理键分配的问题  
任何问题的解决都有多种方法。最难的问题需要耐心和坚持不懈的意志  
## 10.7 管理工作和降低风险
ETL系统开发过程最大风险是跳过了设计阶段  
第二个最大的风险是数据中有些突如其来地雷会导致ETL系统代价更高，甚至导致ETL系统无法实现  
在ETL系统开发项目中降低风险的最佳途径是说服源系统应用所有者承担为ETL系统交付数据的责任，也就是负责数据抽取这一步的工作。到目前为止用于抽取干净数据的最高效方法是为元数据本身添加一些处理过程  
抽取、转换和装载过程的性能通常是至关重要的，有时需要一些创造性思想来克服遇到的困难    
强烈建议您使用ETL工具，在ETL系统的首次迭代开发中，ETL工具可能不会对生产率有多少改进，但随着时间的推移，它能不断降低维护、修正和扩展的成本，因此它是物有所值的  
## 10.8 保证质量
ETL系统必须确保数据仓库中数据的准确性  
当数据流经ETL系统时，您可以改善其质量。在过程的所有环节中，数据的质量都必须得到业务用户的认可  
## 10.9 关键角色
ETL系统开发过程中的关键角色包括：  
+ ETL小组负责这一部分总体工作。ETL小组必须经常和项目团队的其他成员进行沟通  
+ 项目经理必须积极主动，和ETL小组紧密配合。项目经理必须对全局情况有所把握，协调他们之间的利益  
+ 质量保证分析员和数据管理员在ETL开发过程中承担积极的角色。质量保证过程需要按照他们的指导进行设计。此外，当数据装载完毕后，必须进行验证  
+ 数据库管理员继续积极参与到整个开发活动中，负责建立合适的备份、恢复和存档过程  

## 10.10 关键交付内容
ETL系统开发过程中的主要交付内容包括：
