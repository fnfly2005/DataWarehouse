## 6.1 使用维度建模的场合
在深入研究设计维度模型的具体指导方针之前，我们将其与事务处理源系统中遇到的规范化模型先做个比较  
### 6.1.1 什么是维度建模
维度建模一种将数据结构化的逻辑设计方法，因此他们对业务用户来说很直观，并且能够提供较快的查询性能    
在数据库设计中个，维度建模具有不可抗拒的魅力，当设计人员将可理解性和性能作为最高目标时，维度方法总是首选  
维度建模将客观实际划分为度量(measurement)和上下文(context)  
度量是由机构的业务过程和支持它们的业务源系统来捕捉的，常常以数值形式出现，我们将它们称作事实 5  
事实由大量文本形式的上下文包围着，这些文本形式的上下文只有当事实被记录时才为真，这种上下文被直观地分割成多个独立的逻辑块，我们将其称作“维”  
机构的每一个业务过程都可以用维度模型来描述，维度模型由一系列含有数值型独立的事实表组成，而事实表中的数值型度量则被一系列带有文本形式上下文的维度表所环绕，如图6-1所示  
![6-1](6-1.png)  
关系数据库平台中存储的维度模型一般都被称作星型模式，存储在多维联机分析处理（OLAP）的结果中的维度模型被称为多维数据集  
### 6.1.2 怎样进行规范化建模
第三范式3NF建模和维度建模差异很大 5  
3NF建模是一种想法设法消除数据冗余的设计，数据被分成很多离散的实体，每一个实体在关系数据库中都对应一个数据表，这种方式下事务的加载和更新会比较简单和迅速  
因为规范化建模捕捉了数据元素之间的微观关系，所以即使简单的订单业务过程也会产生好几十个表，这些表通过令人困惑的、蜘蛛网式的联接关系相互联系到一起，一般来说，这些实体中的每一个都会转换成一个单独的物理表  
两种模型都有同样的信息和同样的数据关系，只不过是结构不同罢了 
业界有时将符合第三范式模型称作ER模型，第三范式模型和维度模型都可以用ER图的形式来描述，所以我们尽量避免将符合第三范式的模型称作ER模型，而是将其称作规范化模型来防止混淆  
尽管产生的表结构不同，无论使用哪一种方法，表都具有同样的关系数据库属性 5  
从规范化结构转换到维度模型的过程相对比较简单，维度表常常使用很多低基数描述符将符合第二范式的表组合到一起  
维度建模方法要比规范化建模方法出现得更早，在20世纪60年代，在由General Mills 和Dartmouth大学牵头的联合研究项目中就提出了一组含有“维度”和“事实”的词汇  
### 6.1.3 维度建模的好处
可理解性是维度模型得以广泛应用的一个主要原因，这也促使维度模型成为将提交到商业智能层面上的数据进行组织和结构化的最优方法，建模过程中要让所有的事项尽可能简单，但是有不能简单过了头  
查询性能是维度建模的第二个主要驱动因素
除了易于使用和查询性能之外，还有一个优势就是每个维都是一个等效的事实表入口点，这种对称性使维度模型能够承受查询模式中的意外变化 5  
那些产品具有优势的ETL和BI工具厂商早已将维度智能集成到自己的产品中了，如果预先建立ETL向导来辅助用户使用标准的维度功能（处理缓慢变化维）  
最后，对维度模型进行扩展也比较容易，这样可以容纳那些事先未预料到的新数据  
每隔一段时间，业界总会有一些人想要绕开有关维度建模和数据仓库数据库可行性的问题而直接去查询和分析业务数据  
厂商们声称自己的中间件能够屏蔽源系统复杂性，事实上，它并不能解决用户对性能有不同需求这一固有的问题，也无法跟踪维度属性随时间推移变化的情况  
## 6.2 维度建模入门
在简单介绍了维度建模的基本情况后，我们进一步来探究维度建模规定中的一些标准术语  
### 6.2.1 事实表
事实表存储了从机构业务活动或者事件中提炼出来的性能度量，术语“事实”其实就是指每种性能度量，事实的价值一般都是连续的，也就是说事实在相当广泛的范围内都具有有效值  
事实表一般都和业务过程度量事件相对应  
几乎每一个事实都是数值型的，大多数有用的事实既是数值型的又是具有可加性的  
然而以数值型数据表示的某一时点的强度度量，如库存量或者账户结余的事实就是半可加的 5  
有些数值型度量是完全不可加的，例如某个单位价格或者温度等，因为所有维度中的这里事实是独立的  
大多数文本型的上下文都是从一个离散的值清单中抽取出来的，这样的上下文应当存储到维度表中，而不是存储到某个文本事实中  
在重点关注一致性维度时，如果事实表的定义是相同的，则这些事实表也是一致的，一致的事实在不同的表中应有相同的标准名称，如果事实并不一致，那么必须为事实的不同解释赋予不同的名称  
并非所有的数值都存放在事实表，一个比较好的经验性法则是看看属性是否是离散取值的，以及是否用于过滤和标记，满足这些条件就应当放在维度表中，如果属性有许多值可选，并且可用于计算，则应该将其放入事实表  
事实表一般都很大，有几百万行甚至几十亿行的记录，但是使用事实表仍然有很高的效率，事实表通常存储维度模型中90%以上的数据，我们对于最小化事实表中的冗余数据比较谨慎 5  
1. 事实表键  
    事实表是由一个多方键标识的，许多方键是由来自同一业务过程中若干相交维度表的外键所组成的，事实表描述的是多对多关系    
    事实表中的每个外键必须和对应维度表中的唯一主键相匹配，事实表的外键应当不为空，为空则违背了参照完整性  
    事实表的主键一般是维度表外键的子集，有时事实表中唯一标识某一行的主键可能包含退化维  
2. 事实表粒度  
    事实表的粒度是产生事实行的度量事件的业务定义，粒度确定了事实表的主键，而粒度本身是以业务术语表达的，要预先检查是否存在可能出现的外键    
    事实表中的所有表都应当有统一的粒度，粒度就是您的锚点：维度设计中必须包含足够的维度上下文来实现粒度，而且您也必须将那些未被该粒度的维度上下文排除在外 5  
    事实表应当包含从业务过程中捕捉最底层的、最细节的原子粒度，粒度原子事实表不受变化影响，具有良好的扩充性能，可以通过添加最近获得的事实及维度属性进行扩充，也可以通过添加整个新维来实现扩充（假设粒度不会被新的维所改变）  
    提交的维度模型如果具有最细节的数据，那么就可能使模型具有最大的灵活性和可扩展性  
    
### 6.2.2 维度表
维度表不具有健壮性和完整性，它们当中充满了“大而笨重”的描述性字段
维度表的属性一般有两个主要用途:查询约束/过滤和标记查询结果集  
您可以通过“按照”(by)这个词来区分维度或其属性（如By年、By产品、By区域等），在请求查询或者报表时，业务人士经常使用这一的字眼 5  
维度属性一般都是文本字段，或者是作用和文本相同的离散数字，维度属性应当具有以下特点：
+ 是冗长的（名称由完整的词汇组成）
+ 描述性的  
+ 是完整的（没有遗漏值）2  
+ 是离散取值的（维度表中的每一行仅取一个值）  
+ 质量是有保证的（没有错误拼写、不可能的取值、舍弃的值和同一属性的不同版本） 

维度属性列中的空值常常会使业务用户感到困惑，因此推荐使用描述性字符串来代替空值，如“未知”、“不适用”等   
尽管维度表并不一定总有几十个属性，但是这种情况也是很常见的，可以使用代码和缩写形式作为维度属性，最明智的做法是使用相应的描述性字段 5  
存储为维度属性的操作代码常常具有嵌入式的含义，应当为每个嵌套于其中的部分都建立单独而详细的维度属性，这样它们就可以支持过滤和标记  
维度表几乎必然会用于表示层次关系，在一个单独的维里面解析几个多对一的层次关系的情况很常见，这也是反规范化的必然结果  
维度表由那些联系紧密的属性块组成，这些属性块分布描述主要业务对象，如产品、客户、职员和设备等，那你可以将它们看作是数据仓库的名词  
大多数维度模型只有8-15个维度表，储存事务处理细节的事实表构成了维度表的大部分维，而维度表还有少数维用于捕捉周期性的快照  
1. 维度表键  
    尽管事实表有一个多方键，但是维度行却是由单一键字段唯一标识的，建议维度表的主键应指定为从1开始的简单证书序列 5  
    大多数情况下一个4字节的证书可以表示大量的代理键，使用代理维度来取代运营自然键有以下好处：
    + 性能  
        代理键排列紧密而高效，紧凑的代理键带来较高的性能  
    + 运营键管理实践提供的缓冲  
        在运营系统中，数据常常会保留一个相对较短的周期，保存数据的能力很有效，运营键可能会在一个休眠周期后被重用或者重新指派，对于数仓来说就会出大问题  
    + 通过映射集成不同数据源  
        在一些运行环境中，同样的实体被不同的源系统赋予了不同的自然键  
    + 处理未知或者不适用的条件  
        事实表的外键不能为空，使用代理键代替运营自然键可以使我们很容易将某个代理键值分配给那些描述这里空条件的行 5
    + 跟踪维度属性值的变化  
        捕捉维度变化最长久和最合规的一种方法是创建一个新的维度行以表达新的属性特征，并且指派一个可用的代理键作为其主键  
    
   使用代理键的代价，指派和管理维度表的代理键，以及适当地使用代理键替代事实行中的运营自然键都给ETL系统带来了一定的负担  
2. 一致性维度  
    规范化的一致性维度是所有精心设计的数据仓库都在追求的目标，有时也将其称为主维度或者公共参照维度，为整个企业数据仓库环境所共享，其中加入了代表了不同业务过程的多个事实表  
    一致性维度有两种情况，最基本的情形是，如果两个一致性维度是完全相同的，那么它们有相同的键、属性名、属性定义和域值，不管它们联接了什么样的事实表  
    当一个维度是另一个更为详细的粒度维度表的完全子集时，这两个维度也是一致的 5  
    ![6-2](6-2.png)
    如图6-2所示，虽然整个详细产品维和缩小的品牌维并不完全相同，但是两个维度中的公共属性应当在每个维度表中都是用相同的标识、定义和取值  
    注意：缩小的一致性维度表是为了描述事实表，这些事实表既可以较为容易地捕捉较高细节层次的度量，又可以捕捉那些聚集成较小粒度的事实，这是视性能而定的  
    一致性维度对于DW/BI系统的企业极为重要，因为它们具有以下优点：  
    + 一致性，一致性维度确保了每个事实表都是用一致的过滤条件，并且产生的查询结果集合也被唯一地标识 5  
    + 集成，严格遵守一致性维度的做法使DW/BI环境形成了一个综合而完整的系统，一致性维度使查询可以横向钻取描述不同业务过程的事实表  
    + 缩减了开发周期，使系统得以快速上市，如果机构的某些投资用来定义和管理一致性维度，一旦这些维度建立起来，它们就能极大地缩减项目的开发时间，因为已经有了可用的公共维就不必再重复创建了  
    
   近来，很多行业都对主数据管理很感兴趣，一致性维度就是DW/BI环境中的描述性主数据  
   尽管试图使面向不同行业的部门在所有描述其业务的属性上达成共识无疑是在蛮干，但是即使在跨多种行业的综合性大公司内，一般也有少数核心产品或者客户属性应当在整个企业内部保持一致  

### 6.2.3 四步维度设计过程
在理解了主要词汇和核心概念之后，现在最重要的就是讲述我们是如何将那些凌乱的维度和事实拼凑到一起的，维度模型的逻辑设计是由四个主要决策所驱动的 5  
1. 第一步:选择业务过程  
    设计过程的第一步是确定要建模的业务过程或者度量事件，这一步一般都发生在同高级业务管理人员一起制定业务优先级的过程中  
2. 第二步:声明粒度  
    一旦业务过程被确定下来，设计团队就必须声明事实表的粒度，清楚地定义事实表的行到底代表什么在提出业务过程维度模型的过程中至关重要  
    在业务协商过程中定义事实表粒度时要非常仔细，不要跳过这一步，粒度就是从业务角度对”准确地来说,什么是事实表行”这一问题作出的回答    
    一般的，事实表粒度的选择应当尽可能是最小的原子单元，在最小粒度的数据之上设计出来的事实表是最健壮的设计  
3. 第三步:确定维度  
    一旦事实表的粒度已经稳固地确立下来，对维的选择就相当简单了，一般来说，粒度本身就能确定一个基本的或者最小的维度集合，设计过程就是在此基础上添加其他维 5    
4. 第四步:确定事实  
    四步设计过程的最后一步是仔细选择适用于业务过程的事实或指标，事实都必须是实际存在的，不要将那些和明确声明粒度不相匹配的其他时间段里的事实或者其他细节层次上的事实混杂进来

## 6.3 企业数据仓库总线架构
对DW/BI整体环境的构建进行规划是一项极为重要的活动，从历史角度来看，争议热点主要集中在两种都不怎么有效的方式上  
到底应该从建立一个集中式的、规划好的架构的角度为整个企业建立数据仓库，还是为每个具体的业务单位建立小型的独立解决方案  
面向部门的度量解决方案常常被称作数据集市，鉴于对这一术语的定义差异很大，理解也不一致，我们将这一术语去掉了  
### 6.3.1 规划危机
设计企业DW/BI数据架构的人物令人望而生畏，大型企业往往面临两个巨大的、看起来似乎并不相关的挑战  
一方面，项目群经理必须了解企业最复杂的资产-源数据有关内容，并且要知道这些源数据的具体位置 5  
另一方面，项目群经理必须知道哪些事项让管理人员夜里都睡不好觉，因为人们总是希望DW/BI系统提供的数据能够解决当前亟待解决的问题    
为了缓解这方面的压力，DW/BI项目群经理可以先将解决方案一小块部署到某个部门，并且尽量将其完善  
遗憾的是，在很多仅需要建立独立点解决方案而并非一定要建立一致的企业数据仓库的情况下，人们往往忽略应该设计一个框架将那些独立的系统集成起来形成一个耦合的整体  
无效有效地将孤立的、单独的数据烟囱式的解决方案集成到一起已经成为DW/BI系统的致命伤，这比错失数据分析的时机要严重得多，烟囱式解决方案的存在严重阻碍了集成式的企业数据仓库开发进程  
如果一次性创建数据仓库过于困难，需要将整个目标划分为若干独立小块予以实现，那么应该做哪些工作呢？ 5

通过总线矩阵开发数据架构是一种理性的方法，可以将企业数据仓库设计过程中那些令人生畏的工作分解开来

### 6.3.3 价值链的意义
很多业务都通过一系列业务过程步骤来监督内部活动逻辑流程  
在制造业，一个产品的流动过程是从获取原材料到加工、再到最终交付给客户的整个过程  
从原始起点到消费终点的流程被称作是”供应链管理“有一系列的运营系统和该链上的逻辑步骤相对应：  
+ 原材料购买
+ 原材料交货
+ 原材料库存
+ 材料账单
+ 生产制造 5
+ 将产品运输到仓库
+ 制成品库存
+ 客户订单
+ 为客户送货
+ 货品计价

### 6.3.4 通用矩阵的常见问题
当拟定数据仓库总线矩阵的时候，人们有时会为每个行和列所表达的细节参差而头痛，矩阵行的常见问题一般可以分为以下两类：
+ 涉及整个部门的行和包含内容过多的行  
    公司组织结构图中的方框标识职能单位，而非业务过程，矩阵行不应当和它们相对应  
+ 以报表为中心的行和定义过窄的行  
    这里问题与上述情况恰好相反，总线矩阵也不应当类似于被请求报表的细目清单  

在定义矩阵时，设计师很自然地会受到类似的限制，即定义的列既不能过宽也不能过窄，矩阵列的场景问题如下：  
+ 过于一般化的列  
    一个总线矩阵中person列可能会涉及到多种类型的人员，这些群体之间实际上是不重叠的，将它们都汇总到一个一般性的维中会造成混淆 5  
+ 为层次结构图中的每一个层次都分配单独的列，总线矩阵的列应当表示维的最细节层次 
    和创建独立的矩阵列（如产品、产品分组和营业范围等）不同，我们提倡在产品层次结构的每个层次上都建立一个单独的产品列  
    因为矩阵中阴影部分的矩阵单元显示出列和某个业务过程相交，所以可以在该单元格内指明细节层次（如果不是在最小粒度的层次上）

### 6.4.1 日期和时间
在每个数据仓库中日期维都占有特殊的地位，这是因为几乎每个事实表都是一个观测值的时间序列，事实表总是有一个或者多个日期维  
图6-6显示了一个以天为粒度的标准日期维度表，该表抱哈了很多用于导航、过滤和描述日程的属性  
![6-6](6-6.png) 5  
1. 代理日期键  
    使用代理键这项规定对日期维来说也同样适用，使用SQL日期格式的字段作为日期维度表和事实表之间的联接键是错误的做法  
    对大多数维来说，代理键可能是毫无含义的，对于日期维，我们建议您按照时间顺序为代理键分配序号  
    设计人员有时希望使用带有更多含义的整数作为日期维度表的主键，通常是将整数按照YYYYMMDD的形式组织来代替代理键的序号  
    如果使用YYYYMMDD形式的键的动机是为了避开日期维并直接将其作为事实表中的一个可识别字段进行过滤，那么这种做法就是坏主意  
    如果使用YYYYMMDD形式的键是为了确立和维护基于日期的事实表分区，那么只要保证所有查询都仍然可以利用日期维度表，我们就可以将其称作为智能键 5  
2. 时刻(Time of Day)  
    对单个事务的精确跟踪可能会要求数据具有更明确的日期/时间戳含义，一般可以精确到分甚至是秒，日期分析常常和一天内的时间段分析区别很大，因此我们将它们在事实表中区分为两个不同的维    
    只有当一天的时间段由具有一定含义的文本描述时才能将时刻作为一个维，例如午饭时间、第三个班次等，也可以按照层次分组  
3. 日期/时间戳  
    如果业务中需要计算和分析两个特定日期/时间戳之间的差值,那么这些计算应当在ETL系统中执行,并且可以作为事实供用户进行简单而快捷的访问  
4. 多时区的情形   
    对于那些跨越时区的业务来说，可能需要用当地时间来表达日期和时刻，也可能同时使用全球标准时间来表达，例如世界协调时间（UTC)  
    可以通过为每个当地时区和UTC建立单独的日期和时间维来解决这一问题，如图6-7所示 5

### 6.4.2 退化维
很多维度设计都在绕着事务处理控制文档转圈，如订单、发票、提货单或者飞机票等  
这些编号正是设计父-子关系的事务处理系统时所用到的父标题键，不要忽略这些标题编号，而是将它们直接添加到事实表中  
标题票号可以作为维度键，但是如果像事务日期和客户维那样将标题记录信息分解成独立的维度，那么这个维度就没有其他属性可，我们将其称作退化维 5
退化维常常出现在事务事实表的设计过程中，这种表本身就具有一种父-子关系的数据结构  
退化维很常见也很有用，退化维的键应当是实际的事务编号，它本身就是事实而不需加入任何东西，退化键可以作为一条连接到运营系统的纽带  
### 6.4.3 缓慢变化维
根据定义来看，维度表属性变化相对来说并不频繁，总的来说，共有三种处理缓慢变化维（SCD）的基本方法，这三种方法可以配合使用形成更多混合方法  
注意：在一个单独的维度表中，可以根据业务需求的不同找出几种跟踪属性变化的方法  
1. 类型1：覆盖维度属性  
    使用类型1缓慢变化维解决方案时，一旦属性值发生变化，则仅使用最新的值来更新或者覆盖旧值 5  
    尽管类型1方法实施起来比较简单，但是类型1的属性变化会造成无法预料的复杂性  
    不要因为感觉到类型1方法执行起来很简单就立即将其作为默认的处理方式  
2. 类型2：添加一个新的维度行  
    缓慢变化维的第2种解决方案是准确跟踪维度属性值变化最常用和最有效的方法  
    使用类型2方法后，维度表中奖会插入一个带有新代理主键的行来反映新的属性值，如图6-8所示  
    类型2 方法在维度发生明显变化时使用，而且这种方式的优点在于可以通过改变属性来划分历史记录，每个代理键都对应维度行在某个时间跨度内的唯一版本 5  
    类型2方法需要使用一个代理键来唯一标识新的信息，一些涉及人员会建议您将生效日期和自然键连起来唯一标识变化的维度行，我们强力反对这种行为

3. 类型3：添加一个新的维度属性  
    类型3缓慢变化维度属性的处理方案相对来说并不常用  
    使用类型3方法时，应该向维度表添加一个新列，这种方式仅适用于少数这样的软变化 5  
    如果一个维度属性按照某种可预测的规律进行变化（如按照年度来变化），同时业务不仅要考虑历史值准确性和当前值情况，还需要基于属性的所有历史值来汇总事实，你们就可以在维中建立一系列类型3属性  
4. 微型维：添加一个新维  
    类型2是处理缓慢变化维最主要的方法，但它对于跟踪大型维度表（具有数百万行）中的属性变化是不适用的  
    遗憾的是，大型客户维似乎比中型的产品维更容易变化，这将您至于两难境地  
    解决困境的方法是将那些经常变化的客户属性分解开来放到独立的维度表中，这里将其称作微型维度，如图6-9所示  
    ![6-9_5](6-9.png)
    尽管微型维方法可以跟踪一个大型维度表中相对常见的变化，但是采用微型维也需要作出一些妥协：
    + 分带值，客户统计属性会集中到一个离散值的带状区域内，这就对数据的特征进行了限定    
    + 受限的增长，微型维自身无法增长太多，一条常用的经验是微型维含有的行不应当超过二十万，解决这一问题的方法是建立一个次级微型维
    + 与核心维度分离  
        微型维度在物理上与那些较稳定的客户描述符是分离开的，这使其浏览数据非常困难，仿佛是单独的维一样  
        如果需要，可以在客户维里为微型维属性的当前值增加类型1的列，这样不需要经过事实表就可以对客户信息库进行独立的统计分析 5   
    
   最后，这种方法仅在事实表行出现时才允许您将统计信息和客户信息关联起来，如果在上述过程中遗漏与统计信息的关联，可以使用一个无事实的事实表随时间推移不断捕捉客户维和微型维直接的关系，每当出现一个关系变化就装载一行  
5. 混合缓慢变化维方法  
    上述三种基本的缓慢变化维方法对于大多数情形都足够用了，然而，有时还需要将这三种方式组合起来使用  
    如果业务用户有时想要保留以前和某个事实相关联的准确维度属性，同时有希望能够基于当前维度特征来积累历史事实，那么这三种基本缓慢变化维都无法单独满足整个需求  
    一种解决方案是通过为基本维度表添加一个行来捕捉类型2的属性变化，此外，所有当前行和以前的类型2行都有一个“当前”属性被覆盖（类型1），如图6-10所示  
    不要将历史属性和当前属性放到同一个物理表中，当前属性可以加入到一个同维度自然键或者维度表的永久标识符相联接的支架表中 5  
    最后，如果您的维度表有上百万行，并且需要对很多属性进行历史跟踪和当前跟踪，那么除了类型2跟踪所使用的代理键之外，还应该考虑将维度自然键或者永久标识符作为事实表的外键  
    这种方法的本质是通过事实将两个维度表联系起来，当事实行装载的时候，类型2维具有基于有效值过滤或分组的历史准确属性，维的自然键联接到一个具有当前类型1值的表中  
    您可以使用该维度表的属性来汇总或者过滤基于当前配置文件的事实，而不必考虑维度行载入的时候这些值是否真正有效

### 6.4.4 角色扮演维
当同样的物理维度表在维度模型中扮演不同的逻辑角色时，我们将其称作“角色扮演维”，最常见的例子就是日期维  
事实表中的每个日期外键都指向某个特定的日期维度表，如图6-11所示，不能将这些外键都联接到同一个表中，业务SQL语句会认为，通过同时联接到同一个表中的两条路径进行查询所得到的日期也是相同的，而实际上却并不是这样  
![6-11](6-11.png)  
您需要骗过SQL语句，让它认为每个独立的日期维度表都具有唯一的列名称，而不要出现两次联接  
即使确实无法使用如图6-11所示的单个日期维，也可以在后台创建和管理单独的物理日期表 5  
根据行业不同，除了日期维之外，角色扮演还可以出现在多种类型的维度上，例如：
+ 客户（收货人、付款人和订货人）
+ 工厂或者港口（起点、终点）
+ 供应商（参照、执行）
+ 代表（卖主、服务商）3

注意：单独的维在一个事实表中可以同时扮演多种角色  
### 6.4.5 杂项维
在确定了明显的业务维后，常常会有大量乱七八糟的标记和文本属性无法在现有维度表找到安身之所，这种情况在对一个复杂的事务处理业务过程进行建模时表现得更为明显 5    
面对这种情形，设计人员常常会提供一些拙劣的供选方案，应该尽量避免使用如下的方案：
+ 将它们留在事实表中，将标记和文本属性留到事实表中是一种糟糕的做法，因为这样做可能会造成事实表行惊人地膨胀
+ 将他们构建成独立维，将每个标记和属性都构建成单独的维也不是一种好方法，这样会导致事实表中维度外键的数量膨胀
+ 清除它们，尽管将标记和指示符完全清除掉是一种吸引人的做法，但是如果最终证明业务是与这些标记和指示符相关的，那么就可能会出现错误  

解决这一问题的关键在于仔细研究标记和文本属性，并且将它们放到一个或多个杂项维中，如图6-12所示，而不是将它们乱堆到事实表中 5

### 6.4.6 雪花型和支架
当冗余属性和译码都被转移到单独的表，并且通过人工键连接到原始表时，维就变成雪花型了，图6-13给出了一个实例产品的雪花型维度表  
![6-13](6-13.png)
一般情况不建议将维度模型构建成雪花型,因为雪花型总是会让用户感到十分复杂，很不容易弄清楚  
任何认为维度模型的雪花化可以辅助对维度模型进行维护的观点都是站不住脚的，维护工作应当在ETL系统中进行，而且要在被装载到呈现服务器上的维度模式之前进行  
有些设计人员采用雪花型的维是为了节约存储空间，这种微不足道的节约却需要额外花费更多的精力额开销来创建和使用人工键的过程进行管理，而且叉属性的流量几乎必然会使性能降低 5  
尽管上述的一般性建议是和雪花型相冲突的，但是有些情况下，我们也承认构建一个类似初始雪花型的“支架”维可能会比较有好处，如图6-14所示，如果底层维非常大，这种方式就真的能够节省空间了  
![6-14](6-14.png)  
另一个经常会遇到的支架是从基本维分类出来的日期维，如客户身日或者产品引进日期，如果确实需要日期支架，那么要注意支架日期应当在标准日期维度表中所粗出的日期范围之内  
尽管支架是雪花型的一种允许的形式，但是维度模型中不应当到处都是支架，因为这样做可能会为模型的易读性和查询的执行性能带来负面影响  
### 6.4.7 处理层次结构
层次结构是维度模型中不可避免的，本节我们回顾一下常见的固定层次关系和深度可变层次结构 5  
1. 固定层次结构  
    作为维度表反规范化的结果，维度表中简单的多对一层次结构比较常见，而且是以维度中的一组属性来表现的  
    层次机构常常和数据中的钻取路径相关，向下钻取是DW/BI系统中用户使用的最古老也是最受推荐的策略  
2. 通过桥接表实现深度可变层次结构  
    遗憾的是，并非所有的数据层次都是简单和固定的

## 6.5  更多关于事实的讨论
### 6.5.1 三个基本粒度
不论行业和与其相关的业务过程如何，所有的度量事件都可以被刻画为事务、周期快照或累积快照  
1. 事务事实表  
    大多数基本事实表和公共事实表都是面向事务的，其粒度是每一行对应一个事务，或者一行对应事务中的一个条目  
    事务数据在其最低层次上一般都有数目巨大的维与之相关联，如图6-18所示  
    仅当活动发生时，事务事实表中才会插入行  
    正如本章早些时候讲述的那样，细节事务模式是非常强大的，分钟度量使您能够监视大多数粒度的业务活动 5  
2. 周期快照事实表  
    第二种最常见的事实表类型是周期快照，这种方法给了设计人员很大的自由度，在该时间跨度上的所有描述活动的事实都是可以获取的  
    有些业务过程天生就是和被描述为周期快照的形式，例如，银行账户的值可以很自然地看作是每天或者每月的快照，图6-18举例说明了汽车贷款的一个月度周期快照  
    ![6-18](6-18.png)  
    不言而喻，周期快照是细节事务事实的补充，而并非替代品  
    和事务事实表类似，一旦周期快照行被装载到事实表中后，一般都不会再次对它们进行访问 5   
3. 累积快照事实表  
     第三种类型的事实表是最不常见的，然而，它却在特定应用程序中发挥着非常重要的作用  
     周期快照按照规定时间捕捉度量，累积快照与此不同，它用于描述业务过程（有明确定义的开端和结尾）中某个不确定时间跨度里的活动，如图6-19  
     ![6-19](6-19.png)  
     和前两种类型的事实表相比，累积快照事实表具有一些独特的特征：累积快照包含多个日期外键，很多日期值在行首次装载时是未知的  
     除了典型的数量和总额之外，累积快照常常还包括另外另种类型的事实：首先是里程碑计数器，其次是累积快照常常用延迟事实来描述两个里程碑事件之间的时间差 5  
     当然，累积快照事实表和上述两种事实表的主要区别在于它的事实行是要更新的  
     累积快照事实表不能取代基本的事务事实表，因为在累积的过程中，不可避免地会丢失一些维度细节  
     事务和快照是数据仓库的阴阳两面，事务为我们展现了细节行为最完整的视图，而快照使我们迅速而容易地监视总体执行情况  
     
### 6.5.2 不同粒度的事实及其分配
有时单一的业务过程也会产生不同细节层次上的度量  
一个单独事实表中所有的度量都必须对表规定的粒度而言是合法的 5  
那么在面对度量原本就存在于不同粒度上的情形时，维度模型师应当如何去做？  
第一种方法是试着将所有事实都强制转化到最低粒度层次上，这个过程也被看作是将较较高层次的事实分配到较低的粒度上，和其他捕捉到的度量对齐，如图6-20所示  
![6-20](6-20.png)  
在分配之前，数据管理员需要定义分配算法，如果业务人士无法就某一种分配方式达成共识，可能最终要加入两种唯一标识的事实  
分配的妙处是所有由一个业务过程生成的度量限制可以共存于同一个事实表中了，每个维度属性都可以平等地作为过滤和分组度量地候选项 5  
标题事实和运营值的精确匹配是至关重要的，因为运算过程中不可避免地会遇到舍入误差，从而影响分配

### 6.5.4 无事实的事实表
有时公司的业务过程中虽然有事件发生，却没有创建或者生成可以量化的度量，这些没有事实的事件产生了“无事实的事实表”  
可以针对捕捉学生出勤率的事件设想一个无事实的事实表，如图6-24所示 5  
![6-24](6-24.png)  
我们一般使用一个哑元计数器事实，它的值总是为1，您甚至不需要以物理形式存储该事实  
另一种事实表的粒度是每个学生每天注册（不是出席）一门课程占一行，而且学生参加课程则计数器为1，否则为0，这种情况下，1或者0这样的取值也是一种度量  
无事实的事实表在数据仓库中并不是记录事件的最佳方法，因为在该表中没有和事件相关的自然的数值型度量

到目前为止，本章的重点是维度模型，大多数风险都是因为在ETL实现过程中一次的工作量过大而造成的 5

## 6.6 有关维度建模的错误和误区
尽管业界已经接受了维度建模的方法，但是整个行业对此的理解仍然还有一些错觉和误区  
### 6.6.1 将关注点集中在部门报表上导致的错误观点
第一类错误观点是将关注点集中到部门报表和分析结果而没有将设计工作集中到度量过程上  
将全部精力投入到报表或者查询模板的需求分析上很容易在数据建模过程中产生一份具体的报告，但是却无法捕捉到相关维度和主要度量  
以报表为中心方法的一个漫画式的例子是这样的，一家公司建立了一个一百多个维度模型，使用星型模式和OLAP多维数据集进行组合，其目的是支持订单管理分析  
和集中精力于具体的报表需求或者部门需求不同，我们建议维度设计的关注点应当放在物理性能度量过程中，才能够避免下述的错误观点： 5  
1. **错误观点**：建立维度模型是为了提交一份具体的业务报表或应用程序  
    **真实情况**：维度模型应当围绕物理度量过程或者事件来建立  
2. **错误观点**：认为维度模型是部门解决方案，使得多个维度模型从同一个源中重复抽取数据  
    **真实情况**：维度模型不应当被限定于部门范围之内，面向部门的方法很容易造成不一致、非集成的解决方案
3. **错误观点**：如果不重建原有星型模式或者创建独立的事实表，就无法合并新的数据源 5  
    **真实情况**：如果新的数据源是呈现服务器上已有度量过程的另一个捕获系统，在粒度相同的前提下，不需要改变已有报表应用程序，新数据就可以很好地同原有数据相合并  
4. **错误观点**：使用维度建模，事实表只能是单一粒度的，是不可改变的  
    **真实情况**：规定创建事实表时采用单一细节层次确保了度量指标不会被错误地双重计数  
    如果度量本来就是不同粒度的，最简单的方法是为每个细节层次都建立一个事实表  

### 6.6.2 提前汇总导致的错误观点
下面要讲述的一组错误观点是由于在维度模型中混淆了恰当细节的层次而产生的  
1. **错误观点**：维度模型对业务问题进行了假设，因此是固定不变的，当需求变化时，模型也必须进行修改 5  
    **真实情况**：当您提前汇总信息时，已经对业务问题进行了假设  
    然而具有原子信息的维度模型和业务问题是独立的，这些维度模型可以应答新的、以前没有指定过的问题，且不需要改变数据库  
2. **错误观点**：维度模型仅在可预知的使用方式下才是适用的，维度模型不适用于探索性查询  
    **真实情况**：规范化模型和维度模型都包含相同的信息和数据关系，尽管二者的差异很大，但是都能够解决相同的问题  
3. **错误观点**：维度模型不是可伸缩的，如果将细节数据存储到维度模型种，性能将会下降 5  
    **真实情况**：星型模型的伸缩性能非常好，维度建模并不禁止存储质量好的历史信息  
4. **错误观点**：维度模型不可扩展，无法满足数据仓库今后的需求  
    **真实情况**：在最低细节层次上表示数据的维度模型具有最大的灵活性和可扩展性  
5. **错误观点**：维度模型不支持复杂查询，它消除了实体间的多对多关系，仅允许多对一关系  
    **真实情况**：维度模型和规范化模型的逻辑内容是相同的，一种模型里的所有数据关系在另一种模型中都能够进行准确表达 5  

### 6.6.3 过于重视规范化导致的错误观点
有些人认为规范化能够解决数据集成问题，事实上将数据规范化对集成并没有多少好处  
数据集成是一个不同于任何具体建模方法的过程，它需要标识出公司内部使用的不兼容的名称和度量，然后在企业范围内就建立和管理公共的名称和度量达成一致  
1. **错误观点**：一致性维度这样的维度建模概念为ETL增加了过多的负担  
    **真实情况**：数据集成取决于标准的名称、值和定义  
    在整个公司范围内达成共识，以及随后实现对应的ETL系统规则都是较为困难的规则，但是无论您使用规范化模型还是维度模型，都无法避免做大量的工作 5  
2. **错误观点**：当单独的源系统超过2个时，维度模型就不适用了，因为从多个数据源集成数据是非常复杂的  
    **真实情况**：数据集成所面临的挑战和建模方法没有什么关系  
3. **错误观点**：对维度属性的改变仅仅是维度模型中的问题  
    **真实情况**：每个数据仓库都必须处理时间差异问题，但一个实体的特征变化时，需要一种系统的方法来记录这种变化    
4. **错误观点**：多个数据集市无法进行集成，它们都是自下而上构建的，针对的是单个部门的需求，而不是整个企业的需求，因此混乱不可避免 5    
    **真实情况**：一点也不错，要集成面向部门建立的维度模型是一项艰巨的任务，这些独立的解决方案没有构建一致性维度  
    要在整个企业范围内对定义、业务规则和采用的实施方法达成一致，克服组织结构上的差异和文化障碍是不可避免的

在进入建模过程之前，花时间考虑一下DW/BI环境的运行管理和数据管理的意义  
分立的业务单位并不一定要放弃他们的私有属性和名称，但是他们需要和其他业务单元就企业的一个一致属性和事实集合达成共识，而该集合是数据集成的基础，如果有必要，该企业集合完全可以是崭新的 5  
除了前面已经讨论的角色，在第9章还描述了在数据仓库环境的运行管理中ETL团队所需的两个特殊职位：维度管理员和事实提供者  
维度管理员负责定义、创建和发布一个或者多个一致性维度  
同时，当接收到维度管理员传过来的维后，事实提供者负责设计、建立和发布一个具体的事实表  
尽管过多的人员参与设计过程会增加设计过程延期的风险，但是这样做会使设计过程的内容更加充实和完备，是值得花费额外开销的  
### 7.2.2 回顾需求
在建模开始之前，建模团队自身必须熟悉业务需求 5  
在第3章，我们阐述了详细的项目层业务需求文档——该文档主要关注高优先级的业务过程和相关分析结果，还阐述了业务分析员有兴趣解决的问题和难题  
此时，最大的诱惑是跳过业务需求审查过程而直接进入到设计阶段，请不要跳过业务需求审查过程，设计一个业务驱动的数据模型比源驱动的数据模型要难得多  
业务基于属性和状态得组合定义了有关数据得新分类，要在维度模型中揭示和包含这些隐含的业务规则，而不要仅限于满足业务用户的需求，这是提交一个超出用户需求的工作平台的秘诀  
在深入到设计维度模型的工作中之前，必须详细了解业务需求

### 7.2.4 确定命名约定
在创建维度数据模型的过程中，不可避免地会出现命名约定问题  
设计维度模型的过程中有一项工作就是公共定义和公共名称达成共识 5  
不过，确立命名约定并不一定很花时间，大多数大型机构都具有某种功能（如数据管理），专门负责命名约定的问题  
如果您的机构尚未建立一组能够适用于DW/BI的命名约定，就需要在设计维度模型的时候确立命名约定  
一种常见方法是使用一个由三部分构成的列名标准：*原始词_0个或多个限定词_类词*  
原始词就是一个分类词，一般和列所属的实体相对应，有些情况下不必要使用限定词，例如：销售事实表中描述销售量的字段可以命名为sales_dollar_amount  
接下来，核心建模团队要和数据管理员一起设计一个初步的名称合集，并给出相应的理由 5  
在最终模型审查会议上，和主要干系人一起就最后的命名约定达成一致  
一旦就最终的名称集合达成了共识，要立刻认真地编制文档，并且将其应用于最后的数据模型

在整个建模过程中，建模人员需要不断增进对源系统的理解，了解其结构、内容、关系和导出规则  
数据探查使用查询和报表工具来探查被调查系统中的内容和关系 5  
### 7.2.6 获取场所和用品
安排合适的场所并且获取一系列所需的用品都是准备工作的重要内容，而这两点经常被忽视  
除了开会的地方，团队成员还需要基本的用品，如活动挂纸  
## 7.3 再论四步建模过程
四步过程设计维度数据模型不会出现在单个设计会议中，它是一个全局策略，指导几个星期的设计工作，涉及到一系列会议  
### 7.3.1 第1步：选择业务过程
第一步是选择要建模的业务过程  
注意：一种测试是否正确选择了业务过程的方法是转到第2步，为业务过程中捕捉的最细节的数据声明事实表粒度  
业务过程是维度数据仓库的基石，理想情况下，您选择的由初始业务过程所捕捉的核心数据都来自于单一的源系统或模块  
### 7.3.2 第2步：声明粒度
创建一个高层维度模型的第二步工作是声明粒度，或者说是为所选择的业务过程声明事实表的细节层次，声明粒度的例子有： 5    
+ 当使用扫描设备度量时，零售票据上的单个条目
+ 保险单中的单笔交易
+ 从医生那里收到的账单上的条目
+ 航空公司的登记证
+ 每周每个商店里每件产品的库存度量 2  

注意，这些粒度都以业务术语声明，第3步和第4步所作出的设计决策取决于如何正确地将粒度可视化  
最常见的设计错误是没有在设计过程的一开始就声明事实表的粒度  
选择粒度时既要考虑到用什么来满足业务需求，又要考虑基于源系统收集的数据可以得到什么 5  
一般都能够将原子数据综合到一起生成更高层的汇总数据，而如果没有细节数据可用的话，将无法从汇总层向下钻取  
在强调原子数据的好处时，我们也是现实主义者，某些行业中一些过程的原子数据代表着巨大的数据量，因此个别案例中必须使用细节程度较低的数据  
尽管设计团队仔细考虑和声明了事实表的粒度，但是设计团队在设计过程的某一点上改变对粒度的看法也是合理的  
例如，如果粒度声明为配货中心单个存货条目的日库存量，而设计团队发现遗留系统能够正确跟踪批号，那么他们可能决定转而跟踪批号，向上钻取到存货条目  
在设计过程的任何节点上，设计团队都必须能够清楚地描述目标事实表的粒度 5  
### 7.3.3 第3步：识别维度
第3步是在规定的粒度层次上确定应用到事实表的维度，总线矩阵中的所有维度都应当针对粒度进行测试  
有时事实表粒度以主要维的形式给出  
粒度声明可以使您创造性地思考如何将维添加到一个显然在源数据中不存在的事实表  
事实表的主键几乎从不会声明为完全由维度外键构成的集合  
正是在这一步，现实世界的复杂事物开始逐步从建模过程中浮现出来，找出这些隐藏的维是很有难度的，因为它们迫使人们对数据和数据的用途重新进行思考 5  
选择维同样也会使您对粒度的声明重新进行思考，如果一个提议的维度无法和事实表粒度相匹配，既可以把维度放一边，改变事实表的粒度，也可以考虑使用多值的设计方案  
检查维度并确保它们都是有意义的  
### 7.3.4 第4步：识别事实
建模过程的最后一步是识别业务过程中的事实（或者称做度量），事实必须适合于粒度，要坚决抵制匆匆忙忙将与事实表粒度不匹配的事实添加到事实表的做法  
大多数面向事务的业务过程只有少了的基础事实，如数量和总额，但是却有很多从这些基础事实导出的组合和计算结果（一般用于监视业务过程）  
## 7.4 设计维度模型
和所有数据建模工作一样，维度模型的开发也是一个迭代的过程  
构建维度模型的过程一般要经过三个阶段，如图7-1所示：高层设计时期，详细模型开发时期，进行模型的审查、再设计和验证等一系列工作  
对单独的业务过程(如销售订单) 或者几个紧密相关的业务过程（如卫生保健索赔，它将职业索赔、设备索赔和药物索赔等索赔包含到一组相互区别但是又紧密关联的事实表中）来说，整个建模过程一般要花费3到4个星期  
建议核心团队在模型开发期间安排一系列设计会议，在这段时间里数据模型师会不断更新设计文档，以保证每次会议的文档都是最新的  
### 7.4.1 建立高层维度模型
设计会议最初的任务是为目标业务过程创建一个高层维度模型图 5  
记住一点非常重要——大多数DW/BI项目任务都有双重目标：既要完成工作也要通过项目锻炼队伍  
1. 召开初始设计会议  
    初始设计会议应着急所有的核心建模成员，并且还要邀请受认可的外部团队成员或者源系统和ETL团队中的相关人员参加，初始设计会议的主要目的有两个：  
    第一，是为了将设计团队的所有成员首次召集起来，确立对项目目标和设计过程的共同理解，并且树立起团体协作精神  
    第二，通过初始设计会议创建最初的高层设计，而且这个会议应当由主数据模型师来主持  
    直接目标是创建高层维度模型图，它是对业务过程中的维度表和事实表的图形描述 5  
    开发初始维度模型最有效方法是使用活动挂图，要记住大多数初始建模都需要采用头脑风暴法，即让大家自由发表意见  
    当您一步步完成初始模型的设计过程时，团队成员很可能会发现很多问题和挑战，设计过程的推动者需要立刻明确问题并将它们记录到问题列表中  
    这时应该立刻展开高层设计，后面还有很多机会去讨论细节内容和相关问题  
2. 为高层模型图编制文档  
    设计初稿应当汇总起来使用图形表示，形成一份可以称之为高层模型图的提交材料，如图7-3所示，高层模型图是一个实体层上的数据模型  
    高层模型显示了应用于一个给定业务过程的具体事实表和维度表，这种图形描述有以下几个用途： 5  
    + 当和干系人沟通有关项目、范围和内容方面的事项时，它是一份随手可得的设计介绍  
    + 它方便了设计团队和业务伙伴之间的讨论，它是沟通详细设计的切入点，也为设计审查提供而文档支持，保证了对业务需求的满足  
    + 它是一种有用的培训工具，对数据模型进行了介绍，甚至是最不认真的用户也可以理解  
    
   ![7-3](7-3.png)
    为了帮助理解，在高层模型图中保持一致性也很有用的 5  
    每个业务过程可能有不同的维数，当某个维没有应用于具体事实表时，我们偏向与完全将其取出，并且为它们留下位置，确保视觉上的连续性  
3. 识别属性和度量  
    当高层维度模型完成后，设计团队就应该开始关注维度属性和事实度量的初始集合  
    这个过程中提交的材料叫做初始属性和度量清单，使用图7-2所示的电子表，为模型中每个表创建一个电子表格，并且填充其属性或者度量  
    这时就要进行集体讨论了，写下想法并且将注意事项和有争议的问题做笔记，该文档有助于记住所有与模型相关的难题的细微之处，也有助于您记住自己是打算如何解决这些问题的  

### 7.4.2 开发详细的维度模型
完成高层设计之后，现在要开始关注细节了，此后，核心团毒应当在非常正规的基础上工作，一个表接一个表，一个列接一个列地定义一个详细的维度模型 5  
详细的维度模型建模过程是为高层模型填补缺失的信息，解决涉及问题，并且不断测试模型能够满足业务需求，确保模型的完备性，您也需要确定信息来源的位置、定义，确立有关这些属性和度量如何填入模型的初步业务规则  
1. 识别数据源  
    建模过程的一个重要组成部分是找出最适用于填充目标设计模型的源数据  
    详细数据分析是设计工作的一个重要部分：首先必须立即支持业务需求所需的数据，其次必须对每个候选数据源都有足够深入的了解  
    从业务需求来看，可以识别出正式数据源和非正式的数据源，尽管非正式数据并不存储于公共的数据存储系统中，但是业务用户需要将正式数据和非正式数据综合起来使用  
    一般可以通过开发一个新的IT过程来获取非正式数据，从而使非正式数据源也包含于数据仓库中，数据必须经过严格的ETL过程，要像将其他数据源导入数据时那样对这些数据进行清洗和操作 5  
    注意：不要让用户简单地直接将数据导入到数据仓库中  
2. 了解候选数据源  
    当您创建了高层设计模型之后，要想在维度表和事实表设计的过程中取得长足进步，就需要为每个表明确数据源，图7-4给出了数据源定义的示例

3. 数据探查和选定数据源  
    有了候选数据源列表之后，设计团队需要评估每个数据源并且确定最好的数据源  
    收集和审查所有可用的源系统文档：数据模型、文件定义、记录格式、书面文档和源系统程序 5  
    研究源系统数据时，对工作本身的期望和实际情况是有一点出入的，和源系统的所有者以及核心业务用户沟通会有很多好处，核心业务用户一直使用候选数据源中的数据，他们会给您一定的启示  
    数据探查是对数据源内容的系统分析，其所涉及的范围从计算字节数、检查基数到检查数据是否满足DW/BI的要求等更为深入的分析，每个数据源都应当进行详细的数据探查  
    探查分析被分成一系列测试：开始于单个字段，结束于包括扩展数据库在内的整个表集  
    查看到底有多少行赋有空值或者背离定义域的内容是非常有意义的工作  
    数据探查可使用结构筛选来调查同一个表中字段间的关系，可以将数据表中作为键的字段显示出来，和高层的多对多关系一起形成层级关系 5  
    检查键值的唯一性也非常有用，如果违背了这一规则（键字段有重复的实例）就会出现非常严重的错误，反映了某个业务规则没有在源数据中强制执行  
    在数据探查阶段还要检查表和表之间的关系，包括检查外键和主键之间的关系以及是否出现了没有子类的父类等  
    最后，数据探查可以检查企业中特有的复杂业务规则，这些最后的测试可以称作业务规则筛选  
    数据探查过程主要有以下结果：  
    + 经过考虑后对候选数据源做出基本的"可行/不可行"决策 5  
    + 在项目继续进行之前，源系统数据质量问题必须得以修正，这些修正产生了大量的外部依赖关系，必须对这些关系进行恰当的处理    
    + 当从源系统抽取数据之后，在ETL处理流程中就可以修正数据质量问题，但是请不要期望有很高的数据质量，因为ETL系统只能修正相对较少的数据质量错误  
    + 未预料到的业务规则、层次结构和外键-主键关系，在细节层次理解数据有助于找出那些需要合并到维度数据模型中的问题  
    
   基于上述分析结果，可以选择用于填充维度设计的最优源系统，对基本数据源的选择常常由业务驱动，考虑的时候可以参考的标准包括：  
    + 数据可访问性，可访问性越强越合适 5  
    + 数据供给的持久性，通常，在处理流程中数据仓库的数据流随时随地地可能会中断，数据源持久性越强越合适  
    + 数据准确性，当数据在系统之间传递时，不可避免地会有一些更改，如果从这些下游系统提供数据，用户会质疑其准确性，此时，为了确保数据准确性，最好返回最原始的数据源  
    
   您应该将所有考虑到却没有选择的数据源都记录下来，当数月之后重新提出选择最合适的数据源问题时，这个文档就有用了  
   在此期间，核心设计团队要缠住源系统专家向他们请教相关问题，使用数据探查功能增进对源数据的理解，并且确定将该数据源引入模型中是否适合  
4. 确立一致性维度  
    在这些详细设计会议中，还应当定义关键一致性维度，在促使企业各部门就表及属性的命名、描述和定义等达成共识的过程中，数据管理员和业务分析员是主要的人力资源 5  
    毫无疑问，决策者（如业务发起人/驱动者或者管理委员会）必须参与解决一致性维度的定义和命名问题  
    有些维度属性信息可能包含有事务事件数据，不过所占的量一般非常少，而且常常是以代码的形式出现的  
    有些维度表可以有多个数据源，尤其是客户维，这些数据源中所含有的客户可能是不同的，每个客户的属性也可能是不同的，那些公有信息很有可能不相匹配，应当努力将这些数据源联合并集成为一个单独的一致性维度  
5. 确定基本事实和导出事实  
    数据探查工作能够确定由度量事件的源系统生成的计数和数量，然而可能有很多业务用户想要分析的度量是从这些基本事实中推导而来的，您也需要将这些导出事实整理存档  
    有两种类型的导出事实，具有可加性的导出事实完全可以由同一事实表行的其他事实计算生成 5  
    第二种类型的导出事实是非可加性的事实，如比率和累积事实，这些事实一般都在与基本事实不同的细节层次上进行描述  
    定义度量的整个范围是一个极具挑战性的行政过程，要指定数据管理员来协助这个过程  
    尽管在ETL过程设计之前，导出事实不需要是完备的，但是确保要考虑到所有基本事实很重要，直到支持导出事实计算所需的所有基本事实都明确下来，维度模型才是完备的    
    当业务用户对自己所需要的东西有了明确的想法之后，就可以召开一个专门的设计会议来确定导出事实，如果没有，您仍然需要设计一个导出事实的初始集合  
    通常，直到设计团队准备开发BI工具的时候，维度设计中才会开始考虑所需的导出事实

如果已经对模型有了足够的信息，就需要开始图7-1所示的模型审查阶段和验证阶段，至少，设计团队应当计划与下述三个群体进行讨论：

# 8.物理数据库设计与性能规划
本章面向的是数据设计师和数据库管理员  
本章讲述把逻辑设计转变为物理数据库所需的步骤  
__物理数据模型和数据库实现依赖于项目的个体因素:__
* 逻辑数据模型
* 关系数据库管理系统平台
* 数据量
* 使用方式
* 访问工具

__应注意的中心主题和指导原则：__
+ 制定计划并且将物理实现计划和整体项目规划紧密联系起来
+ 不要经不起诱惑将一些简单(但是错误)的方法作为临时度量，这样很可能会导致项目停滞不前
+ 制定并遵守标准
+ 使用可用的工具

![avater](8-高层物理设计过程.png)

应当在该生命周期阶段的早起对物理数据模型进行迭代

## 8.1 制定标准
### 8.1.1 遵守命名约定
数据对象名称应当一致、有描述性，并且应当面向业务用户  
表和列的名称是用户界面的关键要素  
### 8.1.2 为空还是不为空
> 疑问点  
> 处理"空"日期列最好的方法是对日期维使用代理键，即不使用日期或者日期时间类型的列  
> 如果选择不将日期列映射到一个外键上，则可以将缺失的日期映射到一个具体的日期，如数据库日期可以取到的最小值或最大值  
> 不管哪种方案都需要向业务用户传授如何对列进行筛选  

### 8.1.3 设置登台表
即ETL临时中间表的意思  
命名方式要和他们所服务的过程或模板紧密结合起来
### 8.1.4 制定文件位置标准
需要设计使用有关源代码、脚本、二进制文件和数据库文件位置的标准  
如果不能并行作业的话就采用顺序方式  
使用一个具有检入、检出、区分和合并等功能的源码控制系统  
__开发大量代码前要决定如何组织代码库和脚本库:__
* 基于主题域或者表建立代码树的主干
* 基于相似的功能建立顶层分枝

### 8.1.5 对用户访问的表使用代用名或者视图
建议所有授权用户访问表时都使用别名   
别名提供了一个额外的间接层次  
另一种补充的方法是视图  
警告！多表视图一定要限制和规划，因为不合理的多表连接会导致性能下降

### 8.1.6 主键
如何声明主键和外键应制定通用的规定  
维度表使用单列的整数代理键作为其主键  
为键定义数据类型，以获得良好的查询性能  
事实表的逻辑键往往指的是退化维度  
__少数情况下为事实表建立一个代理键是有用的：__  
+ 有时机构的业务规则规定存在多个相同的事实行是合法的情况
+ 更新事实行的特定ETL技术只有在使用代理键时才可行的情况
+ 按照顺序分配的代理键使得确定在ETL重启和恢复过程中装载工作暂停的位置变得简单
+ 代理键使DBA在进行调试或者达成特殊的管理目的时，可以容易确定单个事实表的记录
+ 在某些父子关系事实表中，退化维度不可用时，可以将父事实表的代理键作为外键插入到子事实表中

### 8.1.7 外键
声明外键关系可以杜绝违背数据库中参照完整性关系  
还有一种比声明更好的方法是分割事实表：
1. 将当天的数据装载到一个空表或者分区中
2. 强制进行约束检查
3. 随后将该分区转换为较大的事实表

## 8.2 设计物理数据模型
### 8.2.1 设计物理数据结构
开始之前，确保逻辑模型应该已经完成  
* 并且进行了一些重要的数据探查活动，确保了模型的可行性  
* 此外还建立了很多从源到目标的映射，了解该如何生成每个表和列

在创建模型过程中，是收集如数据类型这些信息的大好时机  
使用电子表格这种方式往往能够提高团队的效率  
### 8.2.2 确定源到目标的映射
__对每个表都需要验证或者收集下述信息：__
+ 确保表的名称和命名标准一致
+ 为用户访问的视图指定一个名称和一个显示名
+ 为业务用户描述表
+ 明确表示事实表、维度表、元数据表还是临时表
+ 详细说明源表，撰写一份有关数据转换的简要描述

__对每个列需要收集和描述以下一些信息：__
+ 确保列的名称和命名标准一致
+ 为列指定一个显示名称、业务描述或者技术描述
+ 有选择性地为列指定一个分组名称
+ 确保每个列都有正确地数据类型、长度和精度
+ 指定列地数据长度时尽量不要比源系统所指定地还要短，如果将列定义为相对较短的类型就需要采取一些适当的安全措施
+ 针对每个目标列详细描述源系统、表和列，包括数据类型转化的简要描述
+ 规定是否允许列取NULL值，以及默认值
+ 规定列是是时多层层次结果或者向下钻取路径的一部分，为该层次结果合理命名
+ 添加用户维护和审计的列，如类型2跟踪的生效日期和截止日期列
+ 通过用户访问工具添加其他所需的列，如为各种属性添加排序列

### 8.2.3 星型 VS 雪花型
星型模型的优点：
* 直接查询关系数据库的用户能够较好的操纵模型
* 关系查询一般都比这种结构的表现要好
* 在ELT过程中，由于平面表的数量相对较少，相关联的键也较少，因此更容易处理

### 8.2.5 进行初步的规模估计
在物理设计的早期阶段，尽快对数据仓库的存储需求做一次估计是非常有用的  
初步规模估计的相关工作包括：
+ 估计事实表和大型维度表中行的长度，抽样调查
+ 对每个表来说，为整个历史装载估计所需的行数，并估计在产品阶段每次装载所增加的行数
+ 对于每个标准RDBMS，要考虑到所有所占的空间和基本数据所占的空间是差不多的
+ 临时/排序空间，一般需要一个大小为索引两倍的临时空间来构建该索引
+ 聚集表占用相当多的空间，一般情况下聚集和其索引占据的空间和原子层事实表所占的空间一样大
+ OLAP多维数据集也占据了相当大的空间
+ OLAP数据库可能需要一定的额外空间来支持多维数据集处理
+ 整个数据仓库占据的空间一般为原子星型模式中数据的3到4倍
+ 为元数据表预留20MB的空间
+ 临时数据库和文件系统存储需求差别很大
+ 可以开发一个日志记录数据库用于记录DW/BI系统的操作情况，要预留10GB的空间
+ 很多BI工具需要将数据库作为其操作的一部分，可以为BI工具数据库预留10GB或者更少空间

## 8.5 设计初始索引方案
数仓生命周期需要不断调整索引，因为用户使用风格会改变
### 8.5.1 索引和查询策略综述
最重要是理解RDBMS查询优化器和索引是如何工作的，以及数据仓库和OLTP需求有哪些不同
1. B-Tree索引  
经典B-TREE索引，构建于多个列之上  
这种索引对高基数的列特别有效（order_number、customer_key)  
将可能值构建成一棵树，给出一个存放叶节点值得行得ID构成得列表  
要找出含有给定值得行就要遍历树，用给定得值和每个分枝上的值进行比较  
给定值较大，则查找上一分枝，反之亦然    
2. 聚簇索引  
基于表的简直来排序和存储数据行，每个表只能由一个聚簇索引  
适用于高基数的列或列集合  
表的主键索引一般都默认定义为聚簇索引，通常是行标识符  
3. 位图索引  
与B-Tree索引相反，更适用于低基数的列，在星型模式的反规范化维中非常有用  
经典例子就是性别代码或者“是/否”指示符索引  
本质上是关于列的每个可能值的位串，同一列上的位图索引通常要比B-tree小很多  
一般都用于维度表的属性和层次关系
4. 其他类型的索引  
有些RDBMS使用其他一些专用的索引结构或者优化策略
5. 星型模式优化  
大多数关系数据库在提供星型模型优化时都会遇到n-way链接问题

### 8.5.2 为维度表建立索引
在维度表的主键建立唯一索引  
可以为维度属性每列建立位图索引  
如果不支持位图索引可以考虑B-tree索引  
小型维度表不建议添加过多索引  
为每个能够用于联接条件、少选条件或者列分组的列都单独建立索引
### 8.5.3 为事实表建立索引
首选索引是在主键上建B-tree索引或者聚簇索引  
日期键在主键索引中应当是第一位的  
随着模式复杂程度提高，事实表键索引的数量也明显增加，一个八维模式可能有6个以上的键索引  
初始索引方案通常没有事实表索引，需要根据业务需求来添加  
### 8.5.4 为装载数据(loads)建立索引
如果一次装载量超过表规模的10%-20%，那么在修改数据之前撤销索引并且在此后重建所有所有通常更为高效
### 8.5.5 为OLAP建立索引
创建OLAP的标准方法：创建一个维度关系数据库，并且使用它为OLAP提供输入，捕捉哪些在维度关系数据库上执行的用于维度处理、完全多维数据集处理和增量多维数据集处理的查询，然后为关系数据仓库添加索引，使之能够支持OLAP处理  
### 8.5.6 在装载之后分析表和索引
装载过程完成之后对表和索引进行统计和重新计算是非常重要的步骤，因为查询优化器必须拥有关于表和索引大小的准确信息


## 8.6 设计OLAP数据库
依赖于选择的OLAP技术  
所有的OLAP数据库都是强维度的  
层次维度关系对OLAP实施方案来说非常重要  
### 8.6.1 OLAP数据粒度和深入钻取
首要关键决策是对数据粒度的确定，因为OLAP数据库在伸缩性上差别很大  
所有OLAP技术中，都可以通过预定义深入钻取路径来使数据集显得比其原来更大  
这种预定义使业务用户访问能力从多维数据集延伸到源数据库-通常是维度关系数据库  
如果OLAP多维数据集中实施深入钻取，就需要提供深入钻取查询的关系数据库被正确索引  
### 8.6.2 完善OLAP维度
给关系维度表添加更多的元数据，业务名称和描述性注释  
在一个维度上可能定义多种层次结构  
在OLAP中实现属性的添加要更为容易些  
一次性构建OLAP维度，加工这些维度并且仔细检查它们  
创建额外的维度表副本来支持OLAP工具的结构性需求是非常必要的
### 8.6.3 定义OLAP计算
设计OLAP数据库的下一个主要步骤是添加计算事实或度量  
将计算迁移到OLAP数据库的好处：
+ 使计算的定义一致
+ 将计算迁移到用于高性能计算的数据库服务器平台上
+ OLAP数据库都使用功能比SQL语言搞一个数量级的语言来定义复杂计算

有些事实仅在某个维度上可加，作为替代可以制定一些规则，如平均或者本期末

## 8.9 设计和构建数据库实例
有一些参数对数据仓库数据库的性能是非常重要的，要准确记录每个参数的情况  
要注明当数据库增长和演进后，哪些参数最可能需要调整以及为什么要调整

确保数据库使用的恢复模型与备份策略相匹配

大容量内存对数据仓库查询性能很有好处，因为大多数查询都是多步处理过程  
有些数据仓库应用可以在内存中保留一个关于数据的工作集，很多用户都可以访问内存的速度来访问这些数据

### 8.9.2 块大小
有些RDBMS允许数据库管理员选择数据库的块大小

解决方法取决于系统的独特环境

## 8.10 设计物理存储结构
### 8.10.1 计算表和索引的大小
逻辑设计和物理设计完毕后，将数据存入磁盘之前，应对磁盘需求再次进行准确的计算  
一般来说，将初始数据装载规划为固定大小，并至少加上六个月的运行装载增长期  
### 8.10.2 设计分区方案
如果RDBMS支持分区表，且数据量很大的情况下，应当对事实表分区  
实现分区表的最好方法是按照活动日期来分区  
分区最大的优点表现在表的装载和维护操作中  
装载时：表分区使得在增量处理过程中实现快速装载成为可能  
维护时：有些数据仓库联机保留了一个数据的滚动窗，一般每个月或者每天都要从表中删除最旧的数据，有了分区后，可以简单而快速的删除旧分区  
但分区增加了系统的复杂程度，如果数据仓库相对较小，划分所带来的收益就可能抵不上额外所花费的成本
### 8.10.3 设置存储
何处存储数据以及如何存储数据会对系统性能和可管理性产生较大的影响  
为了使系统性能最大化，数据应当尽可能分散到较多的磁盘上  
这些磁盘应当由尽可能多的硬件控制器来进行管理
1. 故障容错  
    磁盘系统应具有故障容错能力，容错配置不能代替常规的系统备份
2. 存储区网络
    SAN可以提高存储效率，但设置SAN时最重要的是要获得专家的协助
3. 卷和驱动器配置
    最便于管理的配置就是定义一个单独的阵列卷来保存所有的数据库表和索引，适合于小型数据仓库  
    当数据和索引的总量超过了单个控制器的能力，就需要建立一个多卷的系统，对数据分段可以改善装载性能   
    避免I/O瓶颈的最好方法是尽可能使用较多的磁盘控制器，需要找出系统的成本、性能和可靠性之间的平衡点

# 9 抽取、转换和装载介绍
每个人都需要知道的三个字母：
+ 将数据从它的源系统中的位置取出来(E)
+ 对数据进行一些转换(T)
+ 并将数据转载(L)到最终的表集合中供用户查询

所有的维度数据仓库后台都需要34个子系统

## 9.1 归拢需求
从最困难的挑战之一出发确立ETL系统的架构：归拢需求
### 9.1.1 业务需求
从ETL设计角度来看，就是DW/BI系统用户的信息需求  
ETL团队有时会有重大的发现，会对业务需求能否按照最初希望的那样得以解决产生影响  
所以即使在ETL系统的技术后台开发中，也必须保持ETL团队、数据设计师、模型师、业务分析员和业务用户之间的对话
### 9.1.2 合规性
一般数据仓库中需要谨慎处理的需求包括：
+ 保存数据源和随后数据临时的存档副本  
+ 为改变任何数据结构的完整的事务处理流程提供证明  
+ 完整记录用于分配、调整和推导的算法  
+ 随着时间推移为数据副本的保密性提供证明，包括在线和离线两种

### 9.1.3 数据质量
三个极有影响的力量聚焦到数据质量问题：
1. “如果我能找到数据，那么就可以把自己的业务管理好”，这种想法在不断升级
2. 大多数结构都明白自己的数据源是完全分布式的，因此需要获得对各种不同数据源进行有效集成的方法
3. 对合规性需求的急剧增长意味着那些粗心大意地对数据进行处理的方式已经不再被忽略和原谅

ETL团队需要更加灵活和主动——数据质量是无法由ETL系统独立解决的
### 9.1.4 安全性
数据仓库本身追求如何向决策制定者广泛发布数据，而安全性则要求对数据进行限制  
安全性必须扩展到物理备份中  
团队应当寻求高级管理人员的指导，了解数据仓库哪些方面对安全性特别敏感  
合规性需求和安全性需求很可能是相互重叠的，最好将这两个主题综合起来考虑
### 9.1.5 数据集成
数据集成是IT界的一项重大课题  
数据集成通常表现为数据仓库中的一致性维度和一致性事实  
一致性维度可以生成“横向钻取”报表  
一致性事实是说对独立的公共业务指标达成一致，如KPI等，这样就可以通过计算差值和比率来比较这些数值了
### 9.1.6 数据等待时间
描述了源系统数据通过DW/BI系统提交给业务用户的速度有多快  
有时对数据等待时间非常急迫，那么ETL系统的架构就必须从批处理方式转换为流处理方式了
### 9.1.7 存档和沿袭
每个数据仓库都需要很多旧数据的备份  
所有的临时数据都应当进行存档，除非有明确的决策说明某种特定的数据集今后不需要再恢复了  
当进行存档时，每一个临时/存档数据集都应当有相应的元数据描述其来源和产生该数据的处理步骤
### 9.1.8 用户提交界面
ETL要和建模团队协同工作，必须对数据的内容和结构负责  
ETL团队和数据模型师需要和BI应用程序开发人员紧密合作，确定出数据提交的准确需求
### 9.1.9 可用的技能
ETL设计决策必须基于建立和管理系统时能获得的人力资源来制定  
到底是手工编写代码来生成ETL工具还是使用厂商的开发包，是一项重要决策
### 9.1.10 遗留许可证
是否坚持使用已有的遗留许可证，有些情况下需要有充分的理由为自己的工作下赌注

## 9.2 ETL系统的34个子系统
+ 抽取 子系统1-3 从源系统收集原始数据并将数据写入磁盘  
+ 清洗和一致化 子系统4-8 通过ETL系统中的一系列处理步骤发送源数据，这些步骤提高了数据质量，并对多个源系统中的数据进行了合并，创建并强化了一致性维度和一致性度量
+ 提交 子系统9-21 从物理上将数据组织和装载到呈现服务器的目标维度模型中
+ 管理 子系统22-34 用一致的方式来管理ETL环境中相关系统和过程

## 9.3 抽取数据
### 9.3.1 子系统1——数据探查
任何时候您在某个数据库字段上执行select这种调查性查询时，就是在进行数据探查  
现在有很多专门设计用来对数据进行有效探查的工具，开发时可以做一些调研  
数据探查担负着两种不同的任务：
+ 战略性  
一旦确定了某个候选数据源，就应当进行一次数据探查评估，针对早起的采纳/不采纳问题提供决策
+ 战术性  
进行一系列战术性的数据探查工作来尽可能多地确定出各种问题  
该系统揭示出来的问题最终会产生两种详细说明：  
    + 将数据送回原来的数据源，请求改善数据质量
    + 构成了数据质量子系统的需求

### 9.3.2 子系统2——变化数据捕捉系统
系统必须具有增量更新的功能，仅在当前元数据和上次更新后的数据相比发生改变时，才传递相关的变化  
CDC子系统的主要目标是：
+ 可以将变化的源数据分离出来进行有选择的处理，而不是进行完全更新
+ 捕捉对源数据的所有变化（删除、编辑和插入）
+ 使用原因代码来区别实际更新与错误的更正
+ 支持对其他元数据的合规性跟踪
+ 尽可能早地实施变化数据捕捉

下面给出集中捕捉源数据变化的方式：
+ 审计列：一般是指添加时间和修改时间，必须对他们的完整性进行关注，如果出现空值则必须寻找探测变化的其他途径  
如果借助于源应用程序填充字段并且DBA小组允许使用后台脚本修改数据，那么ETL系统就无法使用审计列
+ 定时抽取：单纯基于时间装载记录（根据修改日期值等于SYSDATE-1的行）非常不可靠  
因为如果装载过程中断，那么重新开始装载就会装载重复行  
如果由于运行失败而跳过一天，那么带来的风险就是遗漏的数据将不会再被装载到数据库中
+ 完全“差异比较”
保持了对以前数据的完全快照，并且将该快照和当前数据记录一条接一条的比较，找出存在变化的地方，可以确保找到每一处变化  
缺点是大多数情况下这种技术是资源密集型  
如果不得不进行完全差异比较，就尽量在源系统机器上做比较，就不需要将整个表或库传输到ETL环境中
+ 数据库日志刮取  
日志刮取能够在预定时间及时有效地给出数据库
重做日志的快照，从中搜索出于哪些事务对ETL装载过程中所关注的表产生影响  
刮取的变体就是嗅探，这种方式是对重做日志过程进行监视，即时捕捉事务  
但刮取日志的事务可能是所有技术中最麻烦的额，业务事务日志已满而阻碍新事务处理运行的情况很常见，但清空的话其中记录的所有事务都丢失了  
如果已经试遍了其他的方法最终需要借助日志刮取来查找新的记录或者变化的记录，那么就要说服DBA创建一个专门的日志来满足您的需求
+ 消息队列监视  
流的内容和日志嗅探中获得的内容十分相似，这一过程的好处就是如果每天的消息队列都已经就绪，则开销较低

### 9.3.3 子系统3——抽取系统
面临的问题：
+ 从大型机环境中抽取数据通常会遭遇COBOLcopybook、EBCDIC到ASCII码变换、压缩十进制数、重定义、OCCURS字段和多种可变记录类型的问题
+ 从关系DBMS、平面文件、XML源、WEB日志或者复杂的ERP系统等形式存在的源中抽取数据，特别时时间较久的遗留系统

注意：尽管XML格式的数据由很多优点，但是并不能指望使用它进行频繁的大数据量传输，因为文件其有效载荷都要低于整个文件大小的10%  
两种基本获取数据的方式，文件形式或者流形式  
抽取到文件的方法：将数据抽取到文件、将文件移动到ETL服务器、转换文件内容、将转换数据装载到临时数据库,有以下优点：
+ 很容易从不同的点重新开始,只要保存了抽取文件就可以在不影响源系统的前提下重新运行装载
+ 并且使用网络传输数据之前可以很轻松地对数据加密和压缩
+ 在传输之前或之后很容易通过比较文件行计数来验证是否所有地数据都以及正确传输

## 9.4 数据的清洗和一致化
是ETL系统为业务机构提高数据价值的步骤，并可以创建能够对源系统中错误进行诊断的元数据，解决产生脏数据的根本原因，改进数据质量
### 9.4.1 改进数据质量文化和过程
看上去很小的数据质量问题实际上是业务过程走向失败的重要迹象  
解决数据质量问题的技术性尝试只有称为源自机构高层的整个质量文化的一部分，才能够成功  
解决数据质量问题的九步法模板：
+ 宣布高层对数据质量文化的委托事项
+ 在执行层驱动业务过程重构
+ 投资改善数据录入环境
+ 投资改善应用程序集成
+ 投资改变业务过程的工作方式
+ 提高团队的端到端质量意识
+ 提高各部门之间的协作
+ 公开宣传注重数据质量的好处
+ 不断度量并提高数据质量

### 9.4.2 子系统 4——数据清洗系统
ETL数据清洗过程通常用于修正脏数据，同时机构的生产系统捕捉到数据时还希望数据仓库能够提供该数据的准确描述，需要找到恰当的平衡点  
解决方法时设计一个能够在装载数据时更正和除去错误的ETL系统，并且具有便于使用的结构、修正方式、标准、规则和优质的清洗功能  
目标之一是提供一个用于数据清洗的综合架构，捕捉与数据质量相关的事件，实现这种架构的的关键是从进行最好的数据探查分析开始入手，帮助您确定自己的数据清洗系统需要达到什么样的复杂程度  
这一子系统的目标应当包括：
+ 对数据质量问题的早期诊断和鉴别分类
+ 让源系统和集成工作提供更优质的数据
+ 具有为ETL过程中遇到的数据错误提供特定描述的功能
+ 具有能够随着事件推移捕捉所有数据质量错误和度量数据质量的框架
+ 对最终数据的质量可信度进行度量的附加功能  


1. 质量筛选  
ETL架构的核心就是一组质量筛选，每一种质量筛选都是一种测试  
测试如果失败，就必须将错误事件记录丢弃到错误事件模式，并且停止处理，将出错的数据发送到悬挂文件  
数据质量筛选分为三种：列筛选、结构筛选、业务规则筛选：
    + 列筛选对单个列中的数据进行测试，是否按有空值，取值范围，值的格式
    + 结构筛选测试了各列数据的关系，对两个或多个字段测试验证它们的层次结构，表之间的外键/主键关系
    + 业务规则筛选：如对客户个人信息进行复杂的时间相关的业务规则测试，对聚集阈值数据质量检查
2. 响应质量事件  
每个质量筛选都必须判定当错误抛出时会发生什么事件：
    1. 中断处理
    2. 向悬挂文件发送损坏的记录以便以后进行处理
    3. 仅标记数据并且将其传送到管道的下一步骤

只要有可能的话采用第三种选择是最好的，对错误的事实表数据可以在审计维度上进行标记，错误的维度数据也可以使用审计维度来标记，可以在其自身的字段里使用唯一的错误值进行标记

### 9.4.3 子系统5——错误事件模式
错误事件模式是一个集中式的维度模式，用途是记录质量筛选所抛出的每一个错误事件

批量维度可以一般化为一个数据是流式而不是批量式的处理步骤

错误事件模式还包含一个较低粒度上的次级错误详细事实表  
这个表中的每条记录都确定了和错误相关的数据记录中的某个字段  
错误事件事实表确定了表、记录、字段和准确的错误条件

## 9.5 向呈现层交付数据
ETL系统的主要任务是在提交阶段传递维度表和事实表  
交付子系统是ETL架构中的中枢子系统  
大多数这种子系统都集中关注维度表的处理，维度表是数据仓库的心脏  
### 9.5.1 子系统9——缓慢变化维管理器
架构要素之一是能够实现缓慢变化维度SCD逻辑  
当数据仓库收到通知，得知某一维度一个已有的行发生了改变，有三种基本的SCD响应——类型1覆盖，类型2添加新行，类型3添加新列
![avater](9-4.png)
1. 类型 1：覆盖  
类型1更新必须从早期永久存储的临时表一直传播到所有受影响的临时表   
类型1更新会使很多构建于变化列上的聚集失效，因此维度管理系统必须通知受影响的事实提供者删除和重建受影响的聚集  
2. 类型2：创建一个新行  
类型2 SCD是精确跟踪维度变化的标准方法  
类型2变化需要一个强大的变化数据捕捉系统  
类型2更新意味着对原先版本的维度行进行复制，并且创建一个带有新代理键的新维度行  
类型2ETL过程还必须更新最新的代理键映射表，这些小到只有两列的表在装载事实表数据的过程中是非常重要的   
在抽取过程中处理变化维度行所使用的查找和键分派逻辑如图9-4所示  
对类型2 SCD行更复杂的处理包括添加和更新5种额外的ETL内务处理列：  
    + 日历日期（变化的日期）
    + 行有效DateTime(变化的准确日期/时间戳）
    + 行终止DateTime(下一变化的准确日期/时间戳，默认为12/31/9999)
    + 改变列的原因或者捕捉列变化代码的文本列  
    + 当前标志（当前/到期）
    
   注意有时后台脚本在事务处理数据库运行修改数据时并没有更新相应元数据字段，如last_modified_date，使用这些字段作为维度时机会产生不一致，通常可用系统日期来替代
3. 类型3：添加一个新列  
用于支持属性的“软”变化，需要用户既考虑到属性原先的值，又考虑新值  
需要ETL系统能够改变维度表，新模式添加一个新列  
和类型1过程一样，类型3变化使得在变化列基础上构建的所有聚集都失效了
![avater](9-5.png)
4. 混合方式：集中类型的组合  
每个维度表都会用到一种、两种或者全部的方法来管理数据变化  
有些情况下，可以使用维度中的两列，一列标识某个属性当前类型1视图，另一列标识同一属性的历史（类型2）版本

### 9.5.2 子系统 10——代理键生成器
ETL系统中需要有一种生成代理键的稳健机制   
使用数据库触发器创建代理键的做法普遍但会造成性能瓶颈  
为了改进性能，考虑使用ETL工具生成和维护代理键，但要避免将源系统的操作键和某个日期串联起来做代理键  
### 9.5.3 子系统 11——层次管理器
一个维度通常都会有多样的、同时存在的、嵌入式的层次结构  
层次结构一般都作为维度属性并存在于同一维度中，其唯一要求就是每个属性在维度主键中出现的时候只能取单一值  
部分不齐整的层次结构通常建模为固定的层次结构  
完全不齐整的维度一般其组织结构是不均衡的，深度是不确定的，这种情况都需要使用桥接表  
在呈现层并不推荐使用雪花结构，然而在ETL临时区使用规范化设计，可以辅助维护用于填充和维护层次结构属性的ETL数据流
### 9.5.4 子系统 12——专用维度管理器
+ 日期/时间维  
    对于财务报表周期和多种文化日期也会具有挑战性  
+ 杂项维  
    将事实表中的关键字删除，留下的文本和各种各样的标志就组成了杂项维  
    如果维度的理论行数是固定且已知的，那么可以预先创建杂项维度  
    另一种情况下，在处理事实行输入的时候，可以即时创建新的观察杂项维度行  
    ![avater](9-6.png)
+ 微型维  
    当类型2方法不可行的时候，微型维度可以作为一种在大型维度中跟踪维度属性变化的方法，最常用于大型的客户维度  
    创建微型维度有两种供选方案：预先创建所有合法组合或者即时辨识和创建新的组合  
    杂项维度通常是从事实表输入中创建的，微型维则是从维度表输入中创建的  
    规模很大并且很复杂的客户维度可能需要几种微型维度
+ 缩小维  
    缩小维度是一致性维度，是基本维度的一个行子集或列子集，应当从基本维度中创建一致的缩小维度
+ 小型静态维  
    查找维度完全是由ETL团队直接创建，最终形式为一个关系表
+ 用户维护的维度  
    ETL团队通常最终要为这些维度负管理责任，但这样做也不是很成功  
    因为ETL团队无法意识到这些自定义分组所发生的变化  
    最好的方案是让适当的部门用户拥有维护这些属性的权限，数据仓库团队需要向这些用户提供用户界面来方便他们进行维护

### 9.5.5 子系统13——事实表构建器
事实表构建器子系统主要关注ETL架构需求，能够有效创建三种基本类型的事实表
1. 事务处理粒度事实表装载器  
    事务处理粒度事实表是三种类型的事实表中最大且最详细的一种  
    最简单的情况是对大多数数据进行单纯的添加操作：
    + 大多数情况是按照日期进行分区，以便于管理和提高表的性能  
    + 此外应当含有审计键、时序ID或者日期/时间戳  
   
   更新现有记录的处理过程：
    + 第一步是不覆盖或者删除原来的行就插入更新的行
    + 随后在第二步中删除原来的行
2. 周期快照事实表装载器  
    描述了某个经常重复的度量或者度量集合，如银行账户月报  
    事实必须真正符合粒度，应当仅描述那些适合周期定义的时间跨度的度量  
    传统做法是在适当的周期结束的时候将周期快照全部装载  
    有些机构使用热滚动周期快照，特定的行是当最新的结余在前一天有效时才进行装载
3. 累积快照事实表装载器  
    描述了某个限定了开始和结束的过程的当前进化状态，通常这些过程持续的时间都较短，订单处理是经典例子   
    累积快照事实表都有一组日期，通常是4到8个，描述了一般处理流程  
    每次当有事件发生时，累积快照事实表行就需要进行破坏性的修改  
    
### 9.5.6 子系统 14——代理键管道
ETL系统必须包含一个在引入的事实表记录中使用恰当的维度代理键替换操作自然键的步骤  
参照完整性(RI)是说对事实表中的每个外键，对应的维度表中都存在一条记录  
键查找过程应当为每个自然键或者默认值产生一个匹配项  
![avater](9-7.png)
为了保护参照完整性，我们总是首先完成维度表的更新  
效率较低的方法：
+ 将当前的而维度表和每个自然键相对于的代理键的最新值
+ 每当需要当前代理键时，应当检查自然键等于预定值的维中所有记录，并且使用当前行指针或者起始有效期和终止有效期选定与事实行中历史上下文相对准的代理键

改进方法：使用一个代理键映射查找表速度要快得多，在映射表中每个自然键恰好有一条记录  
不要在事实记录中保留自然键——事实表只需要含有代理键  
在企图装载某个副本行时，代理键管道需要处理键冲突  
当键冲突被识别出来后，代理键管道需要判定停止过程，应用恰当的业务规则来确定能否修正冲突数据问题，装载改行，并且向错误事件模式中写入一行错误记录

### 9.5.7 子系统 15——多值维度桥接表构建器
挑战在于创建和维护桥接表，因为既可以将每个观察值集合为一个单独的分组，也可以在出现了相同的观察值集合时重用分组


### 9.5.8 子系统 16——延迟到达数据处理器
有些环境中，需要修改标准的处理过程来处理延迟到达事实  
如果还没有完整的上下文环境事实度量就已经到达，将出现延迟到达维  
更多情况下，这种延迟是不可接受的，尤其是实时环境中  
首先要支持延迟到达类型2 维度更新：
+ 需要将修改后的客户记录添加到维度中
+ 使用新的代理键，随后进入客户表中破坏性地修改后续事实行的外键
+ 需要为受影响的维度行重置生效日期  
+ 还需要在维度中向前扫描，看看后续是否还有有关与该客户的类型2行，并且在所有受影响的行中修改该列

不可避免会遇到一个维度“并非很正确”的临时性时期，但是这些维护步骤可以将对键和其他字段进行更新所带来的影响降到最低

### 9.5.9 子系统 17——维度管理系统
一致性维度必然是一种需要集中管理的资源，每个一致性维度必须有单一的、一致的来源  
维度管理员职责主要包括下述ETL处理事项：
+ 在维度设计过程中征求数据管理员和干系人的意见并落实公共描述性标签  
+ 对新的元数据，要向一致性维度添加新行，生成新的代理键
+ 为已有维度记录的类型2变化添加新行，生成新的代理键
+ 适当地按照类型1变化和类型3变化修改行，但是并不改变代理键
+ 如果出现了任何类型1和类型3变化，更新维度的版本号  
+ 同时为所有的事实表提供者复制修改后的维

### 9.5.10 子系统 18——事实提供系统
需要管理一个或者多个事实表，事实提供系统的职责包括：
+ 从维度管理员那里接受或者下载维度副本
+ 在一个无法直接复制而只能在本地更新维度的环境中，为了在代理键管道中更新当前键映射表，事实提供系统必须处理标记为最新和当前的维度记录，而且也要处理所有标记为最新但是延期的维度记录
+ 在使用正确的代理键替换自然键之后，将所有的记录添加到事实表中
+ 在错误更正、累积快照和延迟到达维度变化之后，对所有的事实表中的记录进行修改
+ 去重那些已经失效的聚集
+ 重新计算受影响的聚集，如果维度的版本号发生了改变，则可能以前全部的聚集都需要重新计算
+ 保证所有基本事实表和聚集事实表的质量，要确保聚集表是经过正确计算
+ 在线更新事实表和维度表
+ 通知用户数据库已经更新，如果进行了最重要的更改还要告诉他们更改情况，包括维度版本变化、加入延期记录和改变历史聚集

### 9.5.11 子系统 19——聚集构建器
聚集构建器需要填充和维护聚集事实表  
最快的更新策略是增量更新，但是对维度属性的重要变更需要删除和重建聚集  
聚集必须始终和基础数据保持一致  
### 9.5.12 子系统 20——OLAP多维数据集构建器
关系维度模式是OLAP多维数据集的首先来源，因为很多OLAP系统并不直接借鉴参照完整性或数据清洗方面的问题  
OLAP工具对层次结构要比关系模式更为敏感，强化维度中层次机构的完整性非常重要  
类型2SCD非常适用于OLAP系统，因为新代理键只是作为一个新的成员来处理  
覆盖某个属性值可能会引起该维在当前背景下重建
### 9.5.13 子系统 21——数据传播管理器
负责将一致的、集成的企业数据从数据仓库呈现服务器转换到其他环境用作特定用途  
需要从数据仓库呈现服务器上抽取数据，这个过程也可以看做是另一种ETL过程

## 9.6 管理ETL环境
数据仓库的目标之一是能够为增强业务功能提供适时的、一致的和可靠的数据  
为了实现这一目标，ETL系统必须按照下述三条标准不断地加以完善
+ 可靠性，必须始终如一地完成各项操作，及时提供在任何细节层次都可以信赖的数据
+ 可用性，数据仓库必须满足其服务层协议
+ 易管理性，ETL过程需要进行适度的进化

### 9.6.1 子系统22——作业调度器
每个企业数据仓库都应当拥有一个健壮的ETL调度器  
需要一种可靠的机制来管理ETL的执行策略  
应当能够控制ETL工作之间的关联关系和依赖关系  
如果需要进行实时处理，就需要使用一个能够支持选定实时架构的调度器  
工作控制过程在执行时也必须根据ETL过程的步骤和统计情况来捕捉源数据  
调度器应当完全支持自动化过程，包括当出现需要解决  
需要建立一个能够创建、管理和监视ETL工作流的工作环境   
作业控制服务需要包括：
+ 作业定义：采用某种方式来将一系列步骤定义为作业，并且制定作业之间的一些关系，这里需要制定ETL过程的执行流程
+ 作业调度：需要提供一些标准的功能，如基于时间和基于事件的调度  
+ 元数据捕捉：需要使用监视系统来了解当前的运行状况，需要捕捉当前装载在哪一步骤、装载过程何时开始以及持续多久
+ 日志记录：收集关于整个ETL过程的信息，支持过程回复和重启，写入数据库的系统以创建图和报表来辅助分析和优化装载过程

### 9.6.2 子系统23——备份系统
备份系统的目标是使数据仓库能够在出现故障后回复到原先的工作状态  
还要设计存档和检索过程使业务用户能够访问存储时间较久的数据，并将这些数据从数据仓库迁移到成本较低、性能也较低的存储介质上  
1. 备份  
    在系统完全冗余、通用电源、完全RAID构成的磁盘和具有故障切换能力的并行处理器，最终系统仍然会出现一些危险，如删错了表或数据库  
    一个完善的比分系统需要提供下列功能：
    + 高性能，备份需要能够在分配时间段内完成
    + 简单管理，管理界面应当提供一些工具，便于您识别要备份的对象、创建时间表，并且为后续存储维护备份验证和日志
    + 自动化夜间操作，备份设备必须提供存储管理服务、自动调度、截止和设备处理、报表和通告等功能
    
   数据仓库备份通常是物理备份，也就是某个时间点对数据的一个镜像，包括索引和物理布局信息
2. 存档和检索  
    决定从数据仓库中迁移哪些数据也是一个成本-效益问题  
    保存数据需要一定的开销，业务用户可能仅需要使用这些数据做一些评论性的历史分析  
    审计师也可能会要求将存档数据作为合规步骤的一部分  
    解决方案就是不要将数据丢弃，而是将其存放在成本较低的介质上，但仍然是可以访问的  
    数据多长时间会变得陈旧取决于行业、业务和研究的特定数据  
    一旦决定要对某种数据进行存档，需要利用已有的机制在物理上将数据从当前介质转移到另一种介质中，而且要确保数据能够被恢复，还要为数据的访问和变更创建审计追踪  
    最终，您必须实现一个资料管理系统来进行数据存档
    
    
### 9.6.3 子系统24——恢复和重启系统
造成ETL工作故障的常见原因：2
+ 网络故障
+ 数据库故障
+ 磁盘故障
+ 内存故障
+ 数据质量引起的故障
+ 未通报的系统升级

为了避免故障，需要一个可靠的备份系统和一个配套的恢复和重启系统  
装载过程中难免会出现一些无法恢复的错误，必须制定相关方案  
恢复和重启系统既可以用于重新开始某一项已经停止的工作，也可以用于现场恢复整个工作和重启该工作  
当故障发生后，最初的下意识反应就是抢救已经处理的内容，并且从该发生故障的点重启装载过程  
需要ETL工具具有健壮而可靠的检查点功能  
仔细考虑何时通过写入磁盘的形式来从物理上对数据进行临时操作  
巧妙地选择恢复点和装载日期/时间戳或连续的事实表代理键
### 9.6.4 子系统25——版本控制系统
用于存档和恢复ETL管道中所有逻辑和元数据的一种快照功能  
需要为整个ETL系统指定主版本号  
如果已经证明当前版本中存在一个很大的错误，那么能够完全恢复以前版本的ETL元数据上下文
### 9.6.5 子系统26——版本迁移系统
当结束设计和开发ETL过程，完成了创建装载数据仓库的工作后  
该工作必须进行打包并且迁移到下一个工作环境中——从开发到测试再到最后的生产环境  
生产系统中要做的任何操作都应当在开发过程中进行了设计，并且部署脚本也应当在测试环境中得以测试  
每一项后台操作都应当经过严格的脚本编写和测试  
### 9.6.6 子系统 27——工作流监视器
ETL系统必须不断地对工作流进行监视  
工作流监视器利用作业调度器捕捉到的元数据提供了一个仪表盘和报表系统  
关键性能度量包括处理记录的数量、错误摘要和采取的行动  
工作流监视器也应当跟踪系统性能，并且捕捉底层部件的性能度量  
底层部件的性能度量包括CPU使用、内存分配与竞争、磁盘使用与竞争、数据库性能和服务器使用与竞争  
好的工作流监视器会告诉您瓶颈究竟发生在什么位置，下面给出一个瓶颈的优先级列表：
+ 针对源系统或者中间表的索引极差的查询
+ SQL语法引起的错误的优化器选择
+ 随机存取存储器（RAM)不足造成的性能下降
+ RDBMS的排序操作
+ 过渡频繁的I/O操作
+ 不必要的跟有读操作的写入操作
+ 完全删除和重建聚集，而不是增量地建立聚集
+ 在管道中筛选（变化数据捕捉）操作应用得过于迟缓
+ 没有使用并行处理或者管道处理方式
+ 不必要的事务日志记录，尤其是当进行更新的时候
+ 网络通信量和文件传输开销

### 9.6.7 子系统 28——排序系统
排序是非常基础性的ETL处理功能，建立一个单独的子系统来确保其得到足够重视  
要选择最高效的排序方法来满足需求，有些情况下使用专用的排序包会更加高效
### 9.6.8 子系统 29——沿袭和依赖分析器  
ETL系统中沿袭和依赖是两个逐渐重要的要素
+ 沿袭 从中间表或者用户报表中的特定数据元素开始，确定该数据元素的源、其他包含该元素和其源的上游中间表
+ 依赖 从某个源表或者中间表中的特定数据元素开始，确定包含该数据元素和其他派生的所有下游的中间表和用户报表

在高合规性的环境中，沿袭分析是重要组成部分之一  
当评估源系统影响数据仓库和ETL系统的源系统和其下游时，依赖分析非常重要  
ETL团队应当建立一个设计模板并且将各种活动组织到一起，该模板应清洗地标识出和抽取、转换、查找、一致化、维度变化管理及最终目标表的交付相关的ETL元素  
如果ETL工具支持您对ETL系统进行注释的化，就对系统做详细的注释


### 9.6.9 子系统 30——问题自动调整系统
当出现问题时，ETL过程应当自动通知问题调整系统出现了一些意外情况，需要关注和解决  
这种自动反馈可能会以较为简单的错误日志、操作员通知消息、监督者通知消息和系统开发人员消息等形式出现  
ETL系统可以根据情况的严重程度和涉及到的过程通知相关的个人和团队  
ETL工具可以支持多种消息传递方式，包括电话、APP、短信、邮件等  
每个通知事件都应当写入到数据库以便了解出现问题的类型、状态和解决方案  
ETL系统支持应当按照相对标准的支持结构  
第一层支持是一个帮助台，第二层就是通道，第三层是ETL经理
### 9.6.10 子系统 31——并行/管道系统
并行/管道系统为ETL系统可以帮助ETL在规定的时间约束内完成装载任务  
该系统目标是充分利用通常都能够获取到的多处理器或者网格计算资源  
应当验证源DBMS的并行处理是否正确，并且要避免产生冲突的过程
### 9.6.11 子系统 32——安全系统
严重的安全缺口通常都是来自于机构内部而并非从外部入侵的黑客  
建议基于角色管理ETL系统中数据和元数据的安全性  
为了满足合规，您需要证明ETL模块的某个版本没有改变或者给出哪些人员对模块做了修改  
应当加强授权控制，使用户按照与个体和角色相应的权限来访问ETL数据和元数据  
还需要维护所有个体和角色对ETL数据和元数据进行访问的历史记录  
另一个需要考虑的后台安全问题使管理员对生产仓库服务器和软件的访问  
团队成员都应当拥有访问开发环境的特权，但对生产仓库应当控制得严格一些  
另外备份存储介质应当和在线系统具有相同的安全性  
### 9.6.12 子系统 33——合规性管理器
简单来说就是维护一个关于数据的“保管链”  
从合规性敏感的数据进入数据仓库开始，小心谨慎的进行保护  
必须经常及时地显示这一类数据的额抽取条件和内容，使之处于数据仓库的控制之下  
必须跟踪那些授权访问这里数据的人员  
还需要回过头看这类数据在最初接收时存档的、加有时间戳的版本，而这些信息远程存储在一个可信的第三方那里  
合规性需求意味着您实际上无法改变任何数据  
如果数据必须进行更改，则必须将改变后的记录的新版本插入到数据库  
因此每条记录都必须具有开始时间戳和结束时间戳  
合规性的基础是前面讲述的几个子系统间的交互，以及一些关键技术和功能：
+ 沿袭分析：通过指出最后一个环境数据来自何处来证明初始源数据经过了包括存储过程和手工变化在内的多种转换  
+ 依赖分析：指出那里用到了初始源数据元素
+ 版本控制：需要对任何给定的数据源都标识出ETL系统的抽取版本
+ 备份和存储：可能需要恢复存档，也可能需要证明存档数据没有经过修改，可以让某个可信的第三方保管散列码  
+ 安全性： 指出哪些人访问或者修改了数据和转换  
+ 审计维：将运行过程中元数据的上下文直接和数据联系起来

### 9.6.13 子系统 34——元数据知识库管理器
一定要指定DW/BI团队中某个人担任元数据管理员，负责创建和实施元数据策略

## 9.7 实时的意义
### 9.7.1 实时的分类
实时设计应划分为即时、频繁、日常三个类别，和业务用户沟通他们的需求时就使用这些术语  
即时实时系统通常通过企业信息集成解决方案实现，源系统自身负责支持远程用户屏幕的更新，并且响应查询请求:  
+ EII技术提供了一定的轻量级数据清洗和转换服务  
+ 在用户业务需求访谈中，应当仔细评估对某个即时实时解决方案的需求程度，牢记该解决方案中源应用程序进行的重要装载和即时更新数据的易失性，如存货状态  

频繁意味着数据每天都更新很多次，但不能保证屏幕显示的就是当前的真实数据：
+ 在常规ETL架构中，频繁提交的数据通常是小批量处理的  
+ 数据要经过变化数据捕捉、抽取、登台到ETL后台的文件存储区、清洗和错误检查、按照企业标准进行一致化、分配代理键和很多其他转换操作  
+ 数据仓库通常必须利用高带宽的通信渠道，如遗留应用程序间的信息队列通信、累计事务处理日志文件

日常意味着用户屏幕上显示的数据是在上一个工作日下载的批处理文件： 
+ 应当向业务用户解释，他们如果需要即时或频繁地更新数据可能会遇到哪些问题  

业务需求收集阶段对实时ETL设计过程至关重要，到底是采用即时更新还是采用频繁更新或者每日更新都是一项重大的决策

### 9.7.2 实时的权衡
在实现实时架构时需要权衡的问题
+ 使用从消息队列或日志方式替代批处理文件抽取   
    这些原始即时数据尚未经过任何修正处理
    + 是不正确或者不完整的
    + 包含DW/BI尚未处理过的未解析的外键
    + 需要一个面向ETL数据流程的并行批处理，因为源系统中对消息队列或者日志如果应用了复杂的业务规则，ETL系统需要与之保持一致
+ 仅用于列筛选和简单编码查找的约束数据质量筛选  
    由于ETL管道处理数据时间减少，耗时的数据质量筛选需要剔除掉，但列筛选应该保留下来  
+ 允许提前到达的事实存放在维度的旧的副本里  
    如果实时系统不能等待维度被解析完毕，而且维度有可用旧副本，就应当使用  
    否则就生成一个空版本的维度来代替  
+ 去除数据临时中间表  
    有些实时架构EII系统 直接将数据仓生产源系统传送到用户屏幕，没有将数据写入到永久性存储介质  
    团队对这一的系统也有一部分责任，数据流经过DW/BI系统时应当作为一个整体来捕捉  

### 9.7.3 呈现服务器上的实时分区
数据仓库必须无缝地将其现有历史时间序列扩展到当前时刻  
解决这一难题地设计方案是创建一个实时分区，作为常规静态数据仓库的扩展
这个特殊的分区在物理上和管理上都是和数据仓库表分离的  
理想情况下，实时分区是一个真正的数据库分区，其事实表按照活动日期进行分区  
实时分区应当满足下述严格的需求集合：
+ 包含所有自静态数据仓库上一次更新以来发生的活动
+ 尽可能无缝地连接到静态数据仓库事实表的粒度和内容，理想情况下可以作为事实表的一个真正的物理分区  
+ 要编制轻量级索引，理性情况下，实时分区是完全没有索引的
+ 通过将实时分区驻留于内存来支持高速应答的查询

实时分区可以有效地和事务粒度及周期快照粒度事实表一同使用  
1. 事务粒度实时分区  
    实时分区和其底层的静态事实表应有相同的维度结构  
    如果实时分区是真实的表分区，那么所有的应用程序都能够很好地将事实表的静态部分和实时分区联系起来  
    如果实时分区是一个独立的物理表，那么就必须通过从静态事实表到实时分区的交叉钻取来将该实时分区和已有的应用程序关联起来  
    整理聚集导航器，看看是否所有的表都支持聚集，而实时分区是否可以驻留于内存中  
2. 周期快照实时分区
    如果静态数据仓库具有周期性粒度如月度，实时分区可以看作是本章前面所讲述的热滚动月  
    同样实时分区也应当驻留于内存中  
    和事实粒度相比，查询应用程序所采用的逻辑，在当前滚动周期中半加性累计总数可能需要向上升级为一个整月相等价的形式  
    
## 9.8 小结
本章介绍了ETL系统的关键砌块  
按照描述ETL过程的4个主要领域进行了归类：抽取、清洗与一致化、提交、管理

## 10.4 设计一次性的历史装载处理
历史装载过程和增量装载过程有很多共同点，但这两个过程是明显不同的  
一般来说，从最简单的维度表开始建立ETL系统，通常至少有一个维度表会完全作为类型1维度进行管理  
一旦您已经建立了简单维度表，可以通过将一个或者多个列按照类型2进行管理来解决维的装载问题  
通常您为一个缓慢变化维度开发的方法或代码是可以被所有缓慢变化维度所重用的  
当您已经成功地进行了维度表的装载之后，您可以设计和实施事实数据的历史装载  
力争将每个表的ETL过程都分离到自己的文件、模块、包和脚本中  
因为如果不借鉴其他人的作业，很多开发人员都会觉得很难利用系统进行作业  
即使只有一个开发者，ETL系统模块化都便于建立一个更便于维护的应用程序  
### 10.4.1 步骤5——使用历史数据填充维度表
1. 填充类型1维度表  
    对源系统每个维度属性都抽取当前值  
    当您定义报表或者抽取维度的源数据的查询时，要确保每个实体都占用一行  
    验证唯一性的简单方法时验证行数是否和抽取流中不同成员的数目相等  
2. 维度转换  
    即使最简单的维度表也需要基本的数据清洗和代理键分配  
    1. 简单数据转换  
        最常见、最简单的数据转换形式是数据类型转换  
        对于维度表来说，我们建议使用某个默认值或者空字符串来代替空值  
    2. 联合各独立的源  
        通常维度都来自多个源，客户信息可能需要通过将多个业务条目和外部源进行合并后得到  
        很难找到一个通用键来使该合并操作变得更容易些  
        原始数据和已有维度表数据可能需要在不同的时间存储于不同的字段中，以便于匹配  
        有时匹配使基于模糊的标准进行的，拼写差别较小的姓名和地址可能会被错误地匹配到一起  
        当您认为使用标点符号会导致无法精确匹配的时候，语法分析就非常重要了  
        在数据语法分析和清洗阶段之后，可以对来自多个源的数据进行对比  
        一旦某一对记录被确定为一个匹配，那么就在一个带有数据仓库代理键的映射表中添加一条新的记录，并且对多种不同源系统的键都进行匹配  
        您需要设计一个使用模糊逻辑的通道集合来解决拼写错误、输入错误和对等的拼写  
        所有的模糊匹配算法都无法保证返回的是精确的匹配结果  
        您的业务用户必须提供一个有关匹配算法不精确程度的可接受水平的说明  
    3. 生产代码的译码  
        在数据准备过程中的一项通用合并任务是查找和生产代码相等的文本  
        有些情况下，等价的文本来自于非生成源，如电子表格  
        确保您的ETL系统包含有创建默认译码等价文本的逻辑  
        通常创建一条规则来生产一个虚标签是很有意义的  
    4. 验证多对一关系和一对一关系
        ![10-4](10-4.png)
        多数情况您都希望这种层次结构的上卷是非常齐整的  
        可以执行一个查询来返回拥有多个产品模型的产品集来验证多个属性的多对一情况  
        在登台区雪花结构是有一些价值的：它能够防止您装载违背多对一关系的数据![10-5](10-5.png)  
        对两个属性间一对一关系进行一致化是一个比较相似但是又更为严格的问题  
        如果需要这一的设计，就必须从两个方向对关系进行确认  
    5. 维度代理键分配  
        一旦您确信自己的维度表版本中每个真正唯一的维度值都占一行，那么就可以开始分配代理键了  
        一次性创建序列：
        > create sequence dim1_seq cache=1000; -choose appropriate cache level
        
        填充键映射表:
        > insert into dim1 key map (production key_id, dim1_key)  
        select production key_id, dim1_seq.NEXT  
        from dim1_extract_table;
        
        并非所有主流关系引擎都使用序列结构，另一种常见方法是将表的一个列指定为IDENTIFY列，也能达到相同的目的
3. 维度表装载  
    应当对大多数表的插入操作使用快速装载方法  
    批量装载对数据仓库技术非常重要，比一行接一行的记录插入方式快10到100倍  
    使用批量装载程序和API的一些建议：
    + 关闭日志记录，日志增加了开销并且对装载数据仓库表没有什么意义
    + 快速模式下的批量装载，通常要求目标表是未建立索引或者空的
    + 对文件预排序，按照主索引对文件进行预排须可以显著提高索引速度
    + 小心谨慎地进行转换，这些基本功能使用要慎重，并且测试其性能，建议使用ETL工具来执行大部分转换操作
    + 依赖ETL工具，大部分ETL工具都带有一些利用数据库引擎地批量装载器
    + 在完全更新之前截断表，Truncate table语句是删除表中所有行效率最高地一种方式，通常每天ETL处理刚开始将登台数据库地所有表都清理干净，通常该命令是无法撤销的
    
4. 装载类型2维度表历史  
    应当未以类型2方式进行管理的维度属性重建历史  
    通常重建维度属性历史是非常困难的，有时是完全不可能的  
    需要做一些调查作业来评估以下，看看哪些属性的历史是可以重建的，以及重建的难度如何  
    如果您以及识别出属性历史的来源，就需要构建整个维度行在过去每个时间点上的镜像  
    可以从每个维度属性的当前时间点开始，沿着时间向后反推，每当属性变化发生时应当为维度行建立一个新副本，此时不要将创建好的行写入维度表，将它们写入登台表或者文件中  
    此过程并不适合标准SQL处理，最好使用一个数据库游标结构，或者过程语言JAVA  
    当完全重建历史后，建立最后一个贯穿数据的通道来设定行终止日期/时间戳列，确保序列中没有间隙  
    为了进行历史装载，每个维度成员都是按照年月日逆序排列，但在运行过程的增量装载中有时按照正常顺序进行的  
5. 填充日期维和其他静态维  
    大多数项目都手工建立日期维，一般都用电子表格创建  
    可以用相同的方式来创建其他一些维度，也可以直接使用INSERT语句，这些表通常都只有很少的行  
    
### 10.4.2 步骤6——执行事实表历史装载
事实表历史装载中最令人头痛的是数据量  
历史装载更像是一个数据集成的项目而并非是简单的应用  
1. 历史事实表抽取  
    需要确保列入抽取过程基本参数范围的记录对数据仓库是有用的  
    因为很多事务处理系统在源系统中都保留了操作信息，从业务角度来看可能没有多少意义  
    设计阶段创建的业务规则有助于确定这些事件是在数据仓库中如何描述的，以及结果抽取过程应当使用哪些筛选和求和  
    在这一阶段累计审计统计数据是不错的想法，因为创建了结果集后，通常就能捕捉多个分类小计、总计和行数  
    1. 审计统计数据  
        设计阶段应当明确有关数据质量的多种度量，如计数和总计  
        可以使用它们对数据仓库和源系统进行比较，交叉检查  
        这些数值应当向后与日常业务报告相关，向前与数据仓库装载过程结果相关  
        将这些数据存放在为这类信息创建的表中的元数据类别中  
        向后与日常业务报告相关确立了数据仓库的可信度    
2. 事实表转换  
    对事实数据通常的转换包括空值转换、旋转数据或撤销对数据的旋转、预先计算好的导出计算结果  
    所有事实数据都必须进入代理键管道
    1. 空事实表值  
        事实表中使用空值进行数值度量是很常见也是很合理的  
        空值在对整个事实表进行总计和平均值计算时非常有用  
        然而，在事实表参考维度键的外键列中不能有空值  
        存在两种类型的空值
        1. 数据中的空值可能意味着进行度量时，该值不存在或者不能存在  
        2. 第二种情况使用一个估计值要比使事实记录无法进行分析好得多-数据挖掘  
    2. 改进事实表内容  
        如果抽取中出现了不同粒度的交叉，转换过程必须删去这些聚集，或将它们转移到适当的聚集表中  
        很多情况下在视图或OLAP数据库中计算导出事实要比物理表中进行计算效率高得多  
        如果用户经常将某个事实度量作为筛选条件而您又不想为其创建索引，那么最好将度量预先计算好并且进行物理存储  
    3. 重构事实数据  
        ETL系统有时要对事实表数据进行非规范化或者规范化  
        如财务模式，指示这些相关的数值是预算金额、实际金额还是预测金额,将这些记录旋转为单列  
        > select  
        > account_number,  
        > sum(case when amt_type ='actual' then trxn_amt else 0) as actual_amt,  
        > sum(case when amt_type ='budget' then trxn_amt else 0) as budget_amt  
        > from sourc_trxn_table  
        > group by account_number
    4. 维度代理键查找管道  
        必须要维护事实表和维度表之间的参照完整性关系  
        代理键管道是最终事实处理过程，具体是将自然键改换为代理键  
        维度表必须在事实表进入代理键管道之前完成  
        历史装载的代理键管道中，必须将事实和受影响的每个类型2维度行关联起来，并进行参照完整性验证  
        在进入管道前，先登台干净的事实数据，然后为历史数据每个维度都创建一个自然键列表5  
        调查每个自然键的问题并且修正维度表或者重新设计事实表的抽取  
        可以设计一个没有任何参照完整性错误处理的历史代理键管道，如图所示![10-6](10-6.png)  
        历史装载是一个一次性的过程，不需要处理成一个高度自动化的系统  
        数据量不大的情况下，定义查询将事实登台表和每个与自然键相关的维度表连接，返回每个维度表的事实和代理键5  
        这种方法有以下好处：
        + 利用了关系数据库的优势
        + 对所有维度进行并行的代理键查找
        + 简化了为类型2维找出正确维度键的问题，如图10-6所示，类型2维度的连接必须包含一个子句来判断生效窗口        
       
       如果数据量很大，按月进行一次性处理比较合理  
       应当使用关系方法对一次传输中所有的类型1维度进行代理键查找，单独处理类型2维度5  
       替代方法是使用ETL工具的查找操作程序，根据ETL工具性能不同，具体效果也不同
       
   5. 指定审计维度键  
       每个事实都包含审计键，指向描述装载特征的审计维度，包括时限和数据质量  
       审计系统包含至少一个审计维度来捕捉所有基本信息  
       在ETL工具开始抽取和转换过程之前，要建立向审计维度添加新行的逻辑  
       最后在作业或装载历史数据模块完成之前，回到审计维度去更新信息5  
3. 事实表装载  
    主要问题是装载过程能否顺利完成，可以在装载全部历史数据之前，使用较小的数据集来测试ETL过程  
    有些ETL工具帮您建立一个错误流，把表插入过程中发生故障的行抽取出来  
    有些数据库技术支持指定批量大小的快速装载，可以看看相应文档来进行参数设定  
    改进历史装载性能的方法是将历史数据装载到某个分区表中，具体装载步骤是：
    1. 使事实表和每个维度表之间的外键约束失效，提高装载性能5  
    2. 删除事实表上的索引或使其失效，可以查看数据库文档，或者进行测试观察效果  
    3. 使用快速装载方法装载数据
    4. 创建事实表索引或者使原来的索引有效
    5. 必要的话，执行一些将多个表分区整合到一起的步骤，在整合之前或之后建立索引（参考数据库文档）
    6. 确认每个维度表的代理键列是否都只有唯一索引5
    7. 将事实表和维度表之间的外键约束恢复为有效
4. 测试、测试、再测试  
    只有确认所有数据都是准确后，才能宣布成功  
    一旦验证完毕，要对数据库进行完全备份
    
## 10.5 设计增量ETL处理过程
增量ETL过程最大问题是标识新行、变化行和删除行  
如果已经由一个插入、修改和删除行的流，根据特点不同，也许可以重用历史装载程序中的某些组件  
增量处理时，除了插入，还有更新和删除操作  
一次处理一天的方法是一种比较好的做法  
### 10.5.1 步骤7——维度表增量处理
1. 维度表抽取  
    维度来源有两种情况，一种是单一来源，一种是将维度和事实数据混到一起的事实来源5  
    小型维表可完整的将当前快照抽取出来，通过转换步骤来确定哪些内容变化；大型维表则需要用到事实表的方法  
    如果源系统维护了变化类型的指示符，可以根据这样的指示符进行添加和更新；如果没有，则整个变化记录都要经过缓慢变化维过程  
2. 标识新的维度行和变化的维度行  
    ETL过程需要一个开销很大的对比操作来确定新行和变化行  
    进行标识过程前，维度数据应当是干净的并且是转换好的  
    原始数据都有操作自然键，它必须和当前维度行中相同的列匹配5  
    可以看下ETL工具是否有某种功能可以为您执行上述任务  
    执行一个从输入流到主维度的查找，对比它们的自然键，或者采用将输入数据和已有维度表之间的外部链接  
    如果维度包含类型2属性，为系统出现的每个维度属性设定生效日期，通常是前一天  
    如果维度很大，可以使用散列函数或者校验和函数来加速比较过程  
    不应当删除源系统中已经删除了的唯独行，可以通过添加标注来执行软删除5
3. 维度属性的过程变化  
    对于变化的属性值应有业务规则判断，如果是修正后且可信的更新，应当使用缓慢变化维方法  
    将带有自然键的原始数据和当前维度行的同一属性相匹配，如果维度信息变化，将类型1和类型2应用于维  
    类型3需要维度表结构改变，一般不是自动完成，而是数据模型的一次性变化  
    处理变化维度记录的查找逻辑和键分配逻辑如图10-7所示![10-7](10-7.png)5  
    很多ETL工具都有在增量处理中处理维度属性变化的转换和向导
    
### 10.5.2 步骤8——事实表增量处理
大多数数据仓库过于庞大，一般都对新的事实行和更新的事实行进行增量处理  
事实表增量处理必须是自动化的，系统应当具有数据质量检查点  
1. 事实表抽取和数据质量检查点  
    从源系统抽取完事实行后，应将没有转换的数据做一个副本写到登台区，并对抽取到的数据的数据质量进行度量  
    登台后的数据有三个用途：
    + 存档以便于审计，通常这些登台数据需要存储几个月或者更长时间5
    + 在数据质量验证结束后提供一个新的起点，计算了数据质量的度量后，需要确定抽取是否满足下一步作业  
    + 为重启过程提供一个新的起点，处理过程出现灾难性故障，可以从ETL系统的一个静态数据集重启  
   
   ETL系统下一阶段设计要考虑是抽取数据登台到一个表中，还是登台到一个文件中
2. 事实表转换和代理键管道  
    增量事实表的转换及代理键管道都和历史数据类似，最主要区别在于对违反参照完整性的错误处理是自动化的  
    处理违反参照完整性的情形可以使用下述方法：5
    + 停止装载，通常是默认方法，但不太有用
    + 丢弃错误行，当该数据和业务需求不相关时，可以早点设定筛选条件，应当在代理键管道中有实施该逻辑的设计原因
    + 将错误行写入到文件或者表中，以便进行后续分析，这种方法不能有效保证违背参照完整性
    + 通过在维度中创建一个虚行来修正错误行，并且将其代理键返回给管道，为未知的自然键创建一个虚维度行    
    + 不推荐的方法：通过将错误行映射到每个维度中的某个未知成员上来修正错误行，这种方法的问题在于所有错误行都被映射到相同的维度成员，因为没有在事实表中保留自然键，会造成修正错误行所需信息丢失5
    
   大多数系统来说，可以针对维度表子集的一个查询、视图或者物理表来执行代理键查找  
   如果维表较大，可在事实数据上建立预处理通道改进性能，这个通道标识并创建了仅面向实际用于增量事实自然键的查找表  
3. 延迟到达事实和代理键管道  
    有时一些事实会延迟到达，连续性或者说等待时间造成了数据无法及时收集  
    类型2属性中，延迟到达事实必须和当事实发生时有效的维度成员的版本相关联，并识别出正确代理键的联接条件  
    应当从延迟到达实施中分离出正常的事实，并且对量最大的数据使用等联接方法处理，随后在处理延迟事实时使用复杂联接条件进行查找5
4. 增量事实表装载  
    和历史事实装载一样，应当使用批量装载技术来执行对事实表的插入，但通常增量事实使用非快速装载方法要快得多  
    如果事实表很大，可以使用分区和快速装载方法，历史比较久的事实表可以执行一个过程将日分区合并为周分区或月分区  
    最大的事实表通常都是事务处理，很少进行更新，采用月分区很容易对当前月的数据进行重载，且有较高性能  
    累计快照事实表维护开销较大，可将目标表的新镜像登台到一个登台表，然后使用SQL集合操作更新
5. 加速装载周期
    除了处理变化增量加速数据登台周期的补充方法：
    1. 更频繁的装载5  
        每夜处理相对每月每周缩短了装载时间，大多数数仓都采用夜间装载周期  
        可以试着白天对数据进行一些预处理来减轻夜间处理开销  
    2. 并行处理  
        并行运行多步骤，或者单个步骤并行  
        有时并行并不比顺序装载快，需要考虑系统资源问题
        + 多个装载步骤 将ETL作业划分为几个需要同时提交的独立作业5  
        + 并行执行 应该确保您的数据库可以进行并行活动，而且时可用的，监视您的作业并观察发生了说明
    3. 并行结构  
        可以建立三次镜像或者聚簇配置来维持对数据仓库的连续装载，每天将维护窗缩减到几分钟以便两个服务器所属磁盘进行交换  
        根据需求和预算不同，有几种类似方法，如两个数据仓库数据库，一个用来装载另一个用来查询

### 10.5.3 步骤9——聚集表和OLAP装载
聚集表就是将一个实际的大型聚集查询的结果集存储为表  
事实表的增量装载是聚集表增量装载的一个非常好的来源  
维度属性上任何类型1变化都会影响到该属性上定义的所有事实表聚集  
聚集管理系统要将聚集和底层事实数据保持同步  
大多数数据仓库使用关系数据库索引视图或物化事务、OLAP数据存储等聚集管理技术来处理过程5  
大多数聚集管理技术不适合处理稍难一些的问题，特别是事实表更新或者类型1维度属性中的变化相关问题  
聚集管理技术处理复杂的更新需要对聚集表全部重新计算，不是一个很好的解决方案  
需要对类型1维度属性上定义聚集更加谨慎  
### 10.5.4 步骤10——ETL系统操作和自动化
在架构计划中明确给出您的数据仓库中所需的自动操作功能：
1. 调度作业  
    按时间触发调度任务或者按事件触发调度任务5  
    可以写一个脚本程序来实现自定义作业控制  
    可以使用作业控制管理工具来调用和启动ETL工具 
2. 自动处理预测到的异常和错误  
    自动化ETL系统应当涉及一次到多次数据质量检查，并内置处理问题的方法  
3. 恰当地处理不可预测地错误  
    有些错误无法预测，如数据接收混乱到处理过程中的断电  
    ETL工具可以将多个从异常情况中恢复的步骤组合，也可以在处理过程中登台数据以便为系统重启一个中间点5  
    设计时应慎重考虑将哪些数据操作整合能够保证系统稳定重启
4. 维护数据库对象  
    数据库备份、周期性索引维护、表分区管理和表空间增长任务可以整合到ETL工作序列中  
    对于生产系统的数据仓库来说，工具的稳健性、可用性尤为重要  
5. 开发和设计ETL自动操作  
    在ETL过程中，也需要花事件评估数据质量，在数据没有达到质量标准是采取相应手段
    
## 10.6 小结
建立ETL系统必然会遇到数据质量问题  
需要和源系统开发合作生成自动化高质量数据抽取过程  
维度设计中代理键的重要性及如何分配的问题  
最难的问题需要耐心和坚持不懈  
## 10.7 管理工作和降低风险
ETL系统开发过程最大风险是跳过了设计阶段  
数据中有些突如其来地雷会导致ETL系统代价更高或无法实现  
降低风险的最佳途径是说服源系统应用所有者负责数据抽取这一步的工作，确保抽取干净的数据  
ETL工具能不断降低维护、修正和扩展的成本  
## 10.8 保证质量
ETL系统必须确保数据仓库中数据的准确性  
当数据流经ETL系统时，您可以改善其质量，最终必须得到业务用户的认可
## 10.9 关键角色
ETL小组负责这一部分总体工作  
项目经理必须对全局情况有所把握并进行协调  
质量保证分析员和数据管理员在ETL开发过程中承担积极的角色  
数据库管理员参与到整个开发活动中
