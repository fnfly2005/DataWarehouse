# 13 DW/BI 系统的部署和支持
![13](../syn_pic/data_warehouse_life_cycle/13.png)
部署阶段是生命周期的开发中最激动人心的阶段，也是最终将数据仓库及其BI前端提交给业务机构的时候。部署要提供按照业务用户需要的形式进行信息访问所需的所有功能部件。所有这些运营服务都需要在帷幕第一次拉开之前准备就绪    
测试和部署系统的技术任务不像用户所看到的系统任务那么有趣。它是一种面向细节的工作，将代码从开发环境中移到测试系统中，冲锋测试系统的功能和性能，然后将系统部署到生产环境  
这些技术性的部署任务是非常具有挑战性的，因为您的部署会引起已有生产系统的增量变化    
一旦系统进入生产阶段，还必须完成很多的运营任务来保持系统运行良好。这些任务包括对业务用户的操作情况进行运行监视、性能调整、索引维护、系统备份和其他同样令人感兴趣的主题 5  
本章可以分成三部分。第一部分集中讨论系统的部署过程，第二部分突出讲述了确保一个配置的成功所需要的其他比较关键但是和技术的关系并不密切的活动。第三部分讲述了运营任务和维护任务  
项目经理应该仔细阅读本章内容，与业务用户直接打交道的团队成员，包括桌面管理人员、培训人员以及用户支持代表，也会从本章中找到有用的经验和心得  
## 13.1 系统部署
DW/BI系统是一个复杂的软件项目，在部署之前需要进行广泛的测试  
很多人觉得测试是单调的、冗长乏味的，因此很容易出现在测试上投资过低的情况。如果在部署之前不进行端到端的测试，可以肯定在生产环境中系统运行会出问题。如果不严格的检查数据质量，就把数据质量问题留给了业务用户  
如果在部署之前已经对系统进行了测试、调整和校准，那么系统转化成产品地过程将会相当简单 5  
### 13.1.1 部署前测试
要顺利地部署一个新的DW/BI系统，或者改变某个已有系统，就要提前做好计划并反复测试直到已经确认达到了以下要求：  
+ **测试规程**：系统的测试规程必须十分严格，使您能够向用户或审计员证明系统已经经过彻底测试
+ **数据质量保证测试**：对于历史装载和运行过程中的增量装载，数据都应该使准确的和完整的
+ **操作过程测试**：数据库的装载和处理过程是正确的
+ **性能测试**：系统执行装载和查询与报表的性能都很好 5  
+ **可用性测试**：业务用户可以找到他们需要的信息，并能够完成必要的任务  
+ **部署测试**：部署脚本是可靠的，并且已经演练过  
+ **用户桌面计算机准备就绪**：业务用户的桌面计算机有足够强的性能来支持新的工具，并且计算机中已经安装了这些工具  

1. 系统测试步骤  
    除了能够支持业务工作之外，数据仓库系统还是一个软件项目。作为部署的一部分，必须测试系统的工作情况，包括整体的运行情况和各部分的运行情况。如果还没有使用过任何测试方法，那么应该采纳或者开发一种测试方法来对该项目进行测试  
    测试方法包含至少四个步骤：测试定义;设计一个基本测试数据集;设计用于测试异常情况的数据集;最后，运行这些测试程序并将结果计入日志 5  
    如果您所在的机构做软件开发，那么可能已经有了一个现成的测试团队。在部署阶段，应该尽量为项目指定一个内部测试专家。在一个小型的或者资源受限的团队中，我们会看到前台团队可以成为后台处理称职的测试人员  
    1. 像软件开发经理那样考虑问题  
        在整个系统开发、测试和发布的过程中，如果能够像专业软件开发经理那样思考问题是非常有用的。优秀的软件开发经理都懂得下面这些经验教训:    
        + 当开发人员提交一个可用的应用演示版本时，他们仅完成了该项目25%的任务。从某位自豪的开发人员手中得到的第一个测试版本是一个重要的里程碑，这也是您盼望已久的事情  
            当完成第二个25%任务时，才完成了整个应用程序的全部系统测试任务，也就是说所有运行单元都能工作  
            第三个25%是在某个模拟产品环境中验证和调试已经完成的系统。最后25%项目任务是文档工作，以及交付系统使其进入生产状态 5  
        + 不要相信有些开发人员所说的他们的代码有多漂亮可以不用生成文档这样的话。每个开发都必须在项目上停留足够长的时间，直到提交出完整的、可读性好的、高质量的文档时为止  
        + 使用一个BUG跟踪系统。应该在BUG跟踪系统或者问题报表系统中建立一个分支程序来捕获所有的系统奔溃情况、不正确的结果和所有的建议。应该有个管理器每天监视系统，并为出现的各种问题分配优先级  
        + 应该积极主动地从用户和测试人员那里收集错误报表。对报表里的每个bug都要积极处理，允许用户和测试人员看到为其报表分配的优先级别以及针对这些报表的解决方案的状态，然后修正所有会重复出现的、优先级高的错误  
        + 应该为测试和错误报表设立专门的高额奖赏。应该建立查找错误奖励制度，让高层管理人员表扬在这方面付出努力的人员  
    2. 定义测试  
        在开始测试过程的时候，首先应该列一个测试清单，列出您计划测试的内容和期望的结果。定义的测试要小，同时将测试分解为更小的任务，直到任务变得非常简单，很容易进行检验 5  
        为每个测试编制文档描述测试步骤和检验结果步骤。最好的测试设置是将将其设置为自动，写一个脚本同时运行测试和检验结果  
    3. 设计测试数据集  
        通过定义针对一组已知数据的测试来开始增量装载的测试过程。您可以利用已经构造好的数据集测试大部分系统性能。一个预定义的、静态的数据集的优势在于您可以很容易地检查结果  
        只要在正常运营中已经设计了识别和处理这些数据的过程，那么在最初的测试数据集中含有不完全的数据是很常见的    
        您可能期望反复运营测试，因此，应该使恢复测试数据库到测试之前的状态的这一过程实现自动化。可以开发一个应用程序，该应用程序能够将所有测试的最终结果与已知的正确结果相比较  
        然后在每次运行完测试组以后，找出所有偏离正确结果的测试结果。这种反复校验的方式称为回归测试，如果您始终带着“最后一次”的想法，那么这种方法对于揭示错误是没有多少价值的 5  
        必须完全确定在最终发布ETL和BI系统之前必须已经重新运行了回顾测试用例集  
        在运营测试过程中，正如本章后面将会讨论的那样，您需要对系统运营的控制流程进行重新检查  
        测试错误数据与测试正确数据一样重要，尤其是当您定义了会导致进程终止的数据质量条件时。您应当创建供选数据集来测试影响控制流的错误条件  
        当接近于部署系统的时候，在测试过程的其他环节还需要使用实际数据进行测试。现在，应该将控制流错误条件测试从各种系统的连通性测试和综合测试中分离出来  
    4. 运行测试并将结果计入日志  
        当开始实际运行测试时，必须记录用的是哪个系统和数据，产生的结果是什么。与其他任何软件和系统测试一样，最重要的事情是运行测试并且将结果计入日志。需要收集的关键信息是测试的状况和来自测试人员的观察报表的所有明确注释 5  
2. 数据质量保证测试  
    千万不要在数据质量测试上敷衍了事。提交错误数据比根本不提交数据的结果更坏  
    数据质量测试本质上由运行来自源系统的查询或报表、运行来自数据仓库的相应查询或报表以及相比较结果组成。复杂性源于要了解对应的查询和报表。要进行多系统匹配可能需要相当多的业务知识  
    要确保您复制现有的报表时细致入微。有时候我们会发现已有的报表在很早之前就已经有错误了。这使得要检验新的报表变得更加困难，但是毫无疑问，必须对这些报表进行审查并记录存在的任何差异  
    人们常常要求我们向他们推荐工具帮助他们测试数据质量。但至少，数据质量测试报表应包括行计数、总计和各个主要维及层次上的小计，包括时间  
    您的BI应用可能包括一些复杂计算，这些也很难验证。应该让某个人从外部环境中测试计算——通常在excel中——来确认这些计算是正确的 5  
    报表的定义有时候也包括计算。检查所有计算，在一个现实预算、实际情况和变化图表的预算变更报表中，要确认变更确实是两个列之间的差别  
    在测试和部署过程中，至少进行三次常规的数据质量测试：  
    + **测试历史装载**。历史装载测试是主要的数据质量保证过程  
    + **测试基本测试数据集的运行结果**。这个静态数据集通常很小并且最容易彻底地进行检查  
    + **测试真实数据**。一旦开始通过测试系统运营实时数据，就应该测试数据的有效性，我们下一节会讲到相关内容 5  
    
   为部署阶段开发的任何自动化数据质量测试最终都应该整合到ETL系统中。应该开发进程，将运行数据质量测试的结果计入日志，并向业务社区发布数据质量报表   
3. 操作过程测试  
    开发者每次都重点关注DW/BI系统的一个组件。经验丰富的开发团队从小范围的运营开始，并且从系统的初期阶段就开始进行端到端的测试，一旦两个组件就绪，就检查它们是怎样配合的  
    现在到了将所有部件整合起来并且确认整个系统是否能够按照设计方案运行的时候了。首先，利用这一章前面描述的基本静态数据集检验端到端运营  
    基本的测试数据集应该从开始（从源系统中抽取数据）的时候开始运行，一直运行到结束（运行和提交报表吧），并且不会触发任何导致过程中止的错误  
    注意：正如本章前面讲述的那样，通常基本测试数据集里应当包含在正常运行中可能会出现的错误类型示例 5  
    操作过程测试应该检验ETL作业是否在它们应该启动的时候启动，并且能够在正常的运营下正确地运行  
    检查增量ETL过程中的所有分支程序 。检验事务处理系统和ETL系统之间的通信是否像预期的那样工作   
    如第10章中讨论的那样，ETL系统应该检查行数和校验和。利用相关数据集和配置来测试这些程序  
    尽管在运营测试过程中有很多潜在的与基础结构相关的问题会浮现出来，但是您还是几乎不可避免地会被安全问题绊倒  
    在委托方的办公场所我们曾经被耽误过非常长的时间，等待系统账号的简历和授权。在大型机构中，您应当为摆平这件事计划好几个星期或者更长时间（这取决于您认识谁） 5  
4. 现场测试  
    一旦确信历史数据是正确的并且运行过程中无不良情况出现，就可以开始测试真实数据。在开始现场测试之前，在历史装载数测试完毕之后将数据库备份——尽管您希望一切都顺利，但是仍然要做好最坏的打算    
    设置测试系统，让它指向您在生产环境中使用的运营源系统。需要运行足够长时间的现场测试，以便观察数据环境和工作环境中的真实情况  
    进行现场测试主要基于以下原因：
    + **系统集成**。不管测试数据集和测试系统考虑得有多么周全和真实，您仍然还没有真正连接到生产系统。您的测试系统应该与生产系统尽可能地相似，尤其是关于哪些机器上应当运行哪些服务的问题  
    + **数据质量测试**。在增量处理之后需要对数据进行数据质量测试 5  
    + **性能测试** 性能测试对真实装载最有效，至少可以最终确认系统的性能调整已经生效  
5. 性能测试  
    在降系统部署成生产系统之前，应该做一些性能测试。系统越大，越复杂，则用户越多——尤其是即席用户，进行严格的性能测试也就越重要  
    进行性能测试的目标有很多，最常见目标如下：  
    + **系统调整**：怎样才能够将系统调整到它的最好性能？
    + **服务水平确认**：系统能够满足故障时间和查询性能方面的要求吗？ 5  
    + **净空（headroom）分析**：随着数据仓库的持续增长，已有的系统和硬件在什么时间之后不能再满足需求？   
   
   为了能够考虑到和系统相关的各种事项，性能测试过程一开始应当进行规划。您需要明确指定测试过程的目标，并制订测试方案来实现这些目标  
   1. 系统调整  
       在开始系统调整之前首先要问的第一个问题是您将度量什么？显然您需要度量时限，但是，如果希望提高性能，您还需要更多的信息  
       稍后，本章会讨论在系统操作过程中需要记录到日志中的各种类型信息  
       应该从运行一个基线测试开始，基线测试基于您所能想象的最佳系统配置度量性能。这种基线测试应该度量所能达到的系统最优配置下的性能和操作特征 5  
       接下来，一次只测试一种变化。改变一个索引，要进行测试；改变一个参数，也要进行测试。用一个电子表格跟踪这些测试和结果  
       应该使用这种“一次只进行一个改变”的方法来开发一个改进的系统配置，此后还要针对该新配置，进行第二轮“一次一个改变”的改进，直到您对系统性能满意为止  
       对即将完成的系统从总体上进行系统性能调整测试是至关重要的。您遇到的问题很可能是系统各单元竞争造成的资源短缺——通常是内存  
   2. 服务等级确认  
       逐渐地，DW/BI团队应该能够和用户社区在服务等级协议上（SLA）达成共识。这些协议包括数据等待时间、用户故障时间、用户查询性能  
       如果SLA就绪，那么毫无疑问必须进行测试，看系统是否能够符合协议的要求 5  
       服务等级协议(SLA)是将管理层的注意力集中到重要问题上的一种很有价值的工具。永远都不要做出比客户要求得更多的承诺，但是始终都要设法提交比客户所能想象到的更多的服务  
       在商谈包括即席查询性能度量在内的服务等级协议时要非常小心。不要承诺即席查询时间的绝对上限，例如所有查询服务都在10秒内完成。预期那样还不如同意90%的查询在5秒内完成  
       SLA中要很清楚地说明一些重要术语的含义，例如查询完成。这意味着是在服务器端查询完成，还是包括传输（通过一个带宽可能很低的网络）到客户端？   
   3. ETL处理性能：装入数据  
       很显然需要对ETL系统进行高水平的性能测试。在本章前面讲述的现场测试是处理性能测试的基础  
       要确定某一时间内的数据载入量很容易，而严格识别ETL过程中存在的瓶颈和改进时机则可能会更困难一些 5  
       建立处理系统最简单的方法是将主要组件串行化。这样一个串行化系统是很容易进行性能测试和争端的，因为工作单元都是独立的，您可以分别测试和调整每个单元  
       您也许想设计一个并行的处理系统。通过并行处理一些活动可以使总的装载时间有很大改进，但是要设计和挑战这样一个并行处理系统非常困难  
       性能测试必须在集中式处理系统上运行，DW/BI系统的所有部分都会竞争以下资源：ETL、建立索引及服务查询相关的数据库、用于维度和多维数据集处理的OLAP数据库，以及用于报表呈现的BI工具  
       需要考虑的另一个问题是要确保为了改进系统某一部分性能做出的改变不会对系统的其他部分产生消极的影响。典型的问题是索引和聚集的设计  
   4. 查询性能：取出数据  
       测试查询性能要比测试处理性能更为困难，尤其是测试即席查询性能。基本问题在于您并不知道用户将会查询什么 5  
       标准的报表应该是预先运行并且存储在缓存中的，或者是按用户要求运行的  
       对预运行标准报表的彻底性能测试应该仿真预计的预运行报表使用模式，如有500个人在上午8点到上午9点30分之间的时间段内访问报表  
       按要求生成标准报表需要做很多工作。第一个要求是根据关系数据库或OLAP数据库来为报表提供基本数据  
       然后，BI工具基于结果集绘制报表。最后，报表是分布式的，通常通过网络发布到用户的浏览器窗口中  
       按要求生成报表的方法通常用于参数化的报表，或者无法预先执行并且存储结果的报表，这些报表通常很少被访问 5    
       好的按要求生成报表的性能测试包括对哪些人在运行报表、何时运行报表和使用什么参数运行报表进行实际评估  
       最后，即席查询性能的实验室测试是及其困难。当然，如果测试的系统已经进入生产阶段，您应当从系统日志中获取大量的查询  
       一个好的即席查询性能测试和可伸缩性测试必须由多个用户多次独立完成的查询的集合所组成  
       OLAP数据库引擎、BI工具和关系数据库引擎都具有高速缓存，这些高速缓存都可以提供类似的查询服务  
       这些高速缓存很有价值，因为它们都能够极大地改进我们刚刚讨论过的这些常见行为的性能：关于某个相关主题的一系列查询 5  
       我们还没有看到市场上有哪种查询测试工具能够充分地解决行为模式的问题  
6. 可用性测试  
    除非您已经开发了面向用户的自定义软件并且将其作为BI解决方案的一部分，否则可用性测试并不会成为巨大的负担  
    数据库的结构、报表和维度内容应该是清洁的并且应该是对业务用户有意义的，这一点非常重要  
    应该和真正的用户一起进行一些可用性测试。让一些人漫游BI门户和报表，看他们到底是哪个环节卡住了  
    在新系统进入生产之前，必须已经实现了用户级安全、角色级安全、报表级安全和数据库级安全。如果用户没有进行正确的安全权限设置，那么从他们的角度来看，系统就是无法使用的 5  
7. 桌面准备和配置  
    在技术部署过程中最后的任务是确保业务用户的桌面电脑足以支持工具运行和分析  
    基于前台终端工具、实际返回的数据量和BI应用程序的复杂性，可以生成一个最低桌面配置  
    当您深入到用户社区中的时候，要考虑以下问题：  
    + **连通性**  
        在当今的高度网络化的机构中，连通性通常不是什么问题  
        但是要从机构的一个部门连接到另一个部门时，就可能会出现问题。例如，一个远程野外办事处可能没有能够连接到数据仓库服务器的网络配置 5  
    + **安装**  
        有些前台终端BI工具基于桌面电脑的，并且需要在用户的机器上安装所需的软件。现在大多数前台终端BI工具基于浏览器的，而且不需要在本地安装软件  
        如果安装过程并非是在DW/BI站点上单击一个按钮那么简单，则您要确保在安装说明书中进行详细的介绍  
        
   说到版本问题，即使是完全基于浏览器客户端也可能与旧的因特网浏览器版本之间存在严重冲突，而且可能只支持少数浏览器产品   

### 13.1.2 部署
在新硬件平台上，新系统的第一次部署实际上是自由的  
可以把部署过程想象成一种简单的情形，就当它只不过是一封电子邮件，当系统已经通过所有的测试并且准备进入生产的时候，剩下的只是发送一个消息而已 5  
第一次部署之后，工作变得更加困难了。任何对系统的更改（常常会有更改），都应该对业务用户社区产生最小的影响。要做到这一点，唯一的方法是：  
+ 在一个尽可能接近生产系统的测试系统上进行测试
+ 有可能的话，使用脚本而不要通过在用户界面上单击操作的方式  
+ 开发一个部署过程方案，准确描述要做什么和按照什么顺序去做  
+ 在进入生产之前，应该在测试系统上测试这个方案 5  


1. 关系数据库部署  
    将关系数据库从测试过程部署到生产过程有多种方式。当然，最简单是通过电子邮件宣布通过测试进入生产阶段（电子邮件部署）  
    如果您要为一个全新的系统部署整个数据库，那么可以使用文件备份和恢复。如果您的测试服务器配置不同，那么您可能需要更改恢复脚本将数据文件存放到生产服务器中正确的位置上  
    如果您正在通过添加一个新的主题范围来修改某个已有的关系数据仓库，则过程基本是一样的  
    如果这些都是真正添加进来的，那么在生产系统上创建新的空对象所产生的影响是零（但是首先要对添加的对象进行测试）  
    随着时间的推移，您或多或少都会更改已有的表格。这是一个棘手的命题，因为如果您ALTER一个表格，那么就会破坏已有的ETL过程的风险 5  
    当然您可以通过在测试系统上进行全面的测试来降低这种风险  
    在某次重要改变的开发过程中，开发中可能会为一个表创建若干个ALTER脚本。有一些很好的工具可以用来帮助管理这些对数据库结构的改变。如果您不使用工具，那么就需要开发程序跟踪和整合这些ALTER脚本    
    提示：您可以通过在网上搜索SQL Compare来查找辅助管理数据库结构变化的工具  
    在生成CREATE或ALTER生成脚本之后，您可能需要更改数据文件位置。如果系统的配置不同，那么在生产运行过程中，应该尽可能多地测试脚本，并且将存在的区别和需要验证地内容编制成文档  
    
    
    
# 编辑点
    
    建议建立一个视图层来抽象访问关系表，能够为逐步完善生产系统提供宝贵的灵活性  
    您如果要将数据从测试阶段带到生产，可以创建一个ETL模块，为表的子集将数据从测试环节复制到生产环节中  
    提示：部署一个新的关系数据库或更改某个已有数据库的部署方案应该包括以下几点：
    + 对SQL脚本的任何修改，以参数形式描述这些变化更加安全，但会变得复杂 5  
    + 在新数据库中创建所有静态数据，包括一个新的小型维度或一些元数据  
    + 如果直接在脚本中修改，提供一些机制来验证这些修改是正确的  
    + 运行命令执行所有脚本，包括在脚本执行中用到的所有参数  
    + 能够将数据转载到部分或所有表中的ETL程序  
    + 验证所有已完成事项正确性的脚本或命令 5  
2. ETL部署  
    测试ETL系统的部署过程非常重要，需要测试所有复制到目标服务器上的必备组件，确保所有连接关系进行了正确的设置  
    部署一个ETL模块的部署方案应该包括以下方面：
    + 复制到生产服务器上的ETL系统文件的位置
    + 需要将ETL文件复制到那里
    + 安装ETL程序文件的说明，应该包括所有的对话框选择
    + 创建所有ETL模块里用到的操作系统全局变量的说明 5
    + 检验所有已完成内容正确性的脚本或说明
3. OLAP数据库部署  
    部署一个OLAP数据库的具体方法取决于您选择的OLAP技术  
    OLAP的缺陷是对多维数据集结构的修改可能会导致重新处理，部署增量变化可以考虑在一个服务器上处理数据库，然后复制到生产服务器  
4. 报表部署  
    部署一个新的报表相对于变更数据库或变更ETL模块来说比较容易  
    修改和创建报表的各种要比修改底层数据库频繁得多，发布给用户前测试，确保定义准确 5  
    大多数公司都是在生产报表服务器得某个私有区域开发和测试新报表  
    部署报表的困难在于行政方面，需要制订能够未业务提供新报表和分析结果的策略和规程，应为报表开发一个质量保证过程和规程  
    底层数据库改变会要修改已有报表，在数据库变更进入生产之前，对标准报表集进行测试很重要，应该建立一个用户报表迁移项目来辅助重写关键用户报表  
    
## 13.2 文档和培训
用户需要一些新特性和新资源方面的指导，大部分文档其实都是披着华丽外衣的元数据
### 13.2.1 核心文档
首先对数据编制文档，描述业务过程的主题范围，包括事实、维度、表、列、计算，还包括构成这些主题范围的其他规则  
1. 业务过程维度模型描述  
    BI文档出发点是业务过程维度模型，应该为系统中每个维度模型写一个清晰而简洁的描述  
    如果总线矩阵的初始行选择是订单处理，编制一个描述订单主题范围的文档，可以通过回答以下问题完成：
    + 在该数据中捕捉到的业务过程的本质是什么？ 5 
    + 主要的业务规则是什么
    + 事实表的粒度是什么
    + 包含在事实表中的数据的数据范围是什么？
    + 哪些数据被忽略和为什么要忽略？
    + 哪些维度参与了这次业务过程？很多维度都要有自己的描述性文档，而且本文档能够链接到这些文档 5
   
   这个文档应该有一些能够以图形形式显示、一些示例值和少量证明该文档能够解决业务问题类型的演示报表  
2. 表描述和列描述  
    表名、列名、描述、样本值、注释都会帮助用户理解该表格的内容  
    如果在一个数据库存储了大量有关表和列的元数据，应该建几个报表以便用户浏览这些信息  
    大部分OLAP数据库都能够自行编制文档，尤其如果您已经特意填充了所有描述性字段
3. 报表描述  
    每个报表都必须有一套基本的描述性信息，如报表的标题和描述 5  

### 13.2.2 补充文档
最有价值的补充文档是以在线教程和服务指南形式出现的  
服务指南能告诉您的业务用户在他们有问题的时候应该找谁  
### 13.2.3 用户培训
BI应用程序的一个主要目的是向80%-90%不会学习如何直接查询数据的结构提供对数据仓库中数据的访问  
培训高级用户直接查询数据仓库分为两个阶段：开发阶段和提交阶段，设计课程和编写培训教材、教材和练习的实际课堂演示    
您必须指定大部分培训教材，因为培训过程中最重要的是您的数据特征和环境特征 5   
当数据库稳定下来，在实际部署规则全面铺开，能够创建并测试一套可靠的课程教材之前就开始着手业务用户培训   
1. 设计与方法  
    写教材之前，仔细考虑受众群体及其技能水平，确定如何授课，通常需要开设两个班：即席查询及短期BI应用程序班，中后期提供单独以数据为中心的班  
    设计过程包括拟定各个班的教学提纲，要讲授的内容以及讲授时间，提纲也是不断演化和充实的   
    ![13-1](13-1.png)  
2. 开发培训教材  
    应利用经典沟通原则：每个培训模块应该不超过成年人平均注意力时间跨度45分钟-1小时，各个模块有相同的结构，模块主题应使用一个相关的业务问题作为案例 5   
    这些模块应逐步变得复杂，每个模块快要结束时要进行额外联系来巩固初学者掌握的内容
3. 建立培训数据库  
    针对生产数据仓库的数据库来设计课程是不错的主意，通常也可以创建一个培训数据库  
    确保将培训数据库中内容的讲授作为第一个模块的一部分，如果培训是非常具体的业务过程，就在开班之前增加一个完整的数据内容模块  
4. 工作量水平计划  
    开设好的培训课程要花很多功夫，应该至少花8小时来创建1小时的上课教材

## 13.3 维护与支持
让DW/BI系统保持在最佳状态并不难，单需要一开始就计划并构建一个可维护的系统  
应规划前后台维护活动：前台要指定与用户沟通、处理用户疑问和安全管理规程和机制，后台应制定合理的计划来运行ETL作业、监视资源、进行运行性能调整以及备份和恢复重要数据  
成功的数据仓库是始终变化的，因此经常需要进行一些开发规则，其次即使没有辩护啊，人员从后台开发转移到前台支持  
### 13.3.1 管理前台
1. 提供用户支持  
    DW/BI团队需要向用户社区提供三层支持方法：第一层是网站和自服务支持，第二层是您业务分组的超级用户，第三层是DW/BI团队的前端人员  
    真正的帮助，您需要让业务用户与那些了解业务问题、数据内容和BI工具及应用程序的人员沟通 5  
    + 第一层：网站，具有丰富的内容和用于查找它们的工具是通过网站提供支持的基础  
    + 第二层：超级用户，如果某个用户需要使用某个当前不存在的报表，那么他们需要寻找有专业技能的人以获得帮助
    + 第三层：DW/BI团队，当网站和本地专家也不能解决问题时，DW/BI团队应提供最后一种支持资源，一个8人的DW/BI团队至少要有4个人负责前台工作  
   
   将DW/BI团队中的BI部分从数据仓库人力资源中分离除去会降低生产率  
   BI团队需要监视BI应用程序的使用情况，当数据和数据结构发生改变后要对它们进行维护，也需要向用户提供途径以便他们反馈信息和请求新的应用程序 5  
   BI团队要把创建自定义报表包括在其责任清单里，并确保有足够的人手来满足最重要的请求  
2. 维护BI门户  
    门户上应该出现的操作集和维护信息集如下：
    + 数据仓库状态：每个业务过程最近几天的数据是说明
    + 计划停机的时间安排
    + 任何非计划停机的诚实状况和完成状况  
    + 明确警告用户系统存在的问题，如数据质量问题，还包括诚实和完成状况的评估问题何时可以解决 5  
    + 系统当前的运行状况，其中包括：
        + 上一小时回复了多少查询
        + 已经生成了多少报表
        + 当前的活动用户的数量
        + 按照活动查询生成的报表，包括这些报表运行了多久以及由谁来运行它们
    + 关于系统容量的证明点，用于培养用户对系统的信心：
        + 上一次数据加载花了多长时间 5  
        + 最大数据装载的数据量由多大，以及装载要花多长时间
        + 允许同时存在的最大用户数  
    
   每12-18个月就应该审查一下整个DW/BI系统，评估需要进行哪些变更才能很好地为用户工作  
3. 管理安全性  
    安全措施应该在数据库层面上实现，部分安全管理员运行职责是确保报表和其他BI应用程序受到恰当地保护    
    大多数报表工具都由相当直观的用户界面来管理角色和特权 5  
    在数据库中实现的安全措施通常由DW/BI安全管理员处理，一个相对简单的自定义工具能在安全管理中极大地降低负担和潜在的错误  
4. 监视器的使用  
    DW/BI系统的使用状况对其性能有巨大的影响，应该开发一个监视系统，收集查询运行数和应用程序类型返回的行数  
    也要收集一些查询样本，样本有助于根据新的所有或聚集来调整系统性能  
    一些机构收集所有提交给系统的查询文本，这实际上会给系统带来巨大的性能负担，供选方案是偶尔在日志中记录查询文本  
    您应该记录有哪些用户登录到了系统，包括BI门户、BI应用程序以及数据库的用户验证登陆程序，查询文本是非常有价值的，单存储所有的查询文本的代价就太大了 5  
5. 关于使用情况的报表  
    BI门户网站应该划出少量屏幕区域来生成有关系统使用情况的报表，特别是分部门的使用情况  
    还应该知道哪些人正在对系统做什么操作、进行了多少操作，以及何时进行的操作，对系统性能调整至关重要  

### 13.3.2 管理后台
后台系统管理由数据调谐支持、ETL系统执行和监视、资源监视、数据增长和磁盘空间管理、性能调整、备份和恢复组成  
DW/BI系统的管理越自动化越好，越小的团队越应该做自动化  
1. 支持数据调谐  
    数据质量验证和调谐必须在所有用户访问之前就进行 5  
    尤其当发现以前的报表信息不正确时，可能需要进行一些挖掘来帮助用户充分理解数年来运营系统使用的信息是不准确的  
2. 执行和监视ETL系统  
    ETL系统应该自动防范攻击和自动修复可预测的错误  
    可以设计一个错误事件表，您必须制定和遵守一定的规程来惊恐研究和修正错误行  
3. 监视资源  
    您需要实现理解两类监视：
    1. 您的系统要发生什么？正在使用什么资源？何时使用以及由哪个进程使用?  
    2. 哪些人在访问系统？何时访问系统？他们在运行什么样的查询以及这些查询用了多长事件？ 5  
    
   DW/BI供应商提供了很多监视工具，监视DW/BI系统每个组件的资源使用情况  
   除了DW/BI部件的集成信息，还需要组合信息，甚至是同一个组件内的组合信息以便于使用和分析  
   运营计划的一部分应该建立一个系统基线性能配置文件  
   跟踪性能基线最重要的独立：内存、处理器以及磁盘使用，制定计划15分钟为间隔将监视信息存储在数据库  
4. 管理磁盘空间  
    磁盘空间用户是ETL作业最常见也最容易预防的 5  
    操作系统要设置一个报警器，两个警报：将一个月内用完磁盘空间发出警报，一周内用完发出警报  
    重点关注您所积累的大文件来检查剩余空间：
    + 用于事实表抽取的登台文件，按照每月数据量考虑，做以恶搞保守的估计，在运营计划上进行记录以便定期重新检查  
    + 关系数据库文件分组，包括临时空间，监视数据库文件存放的磁盘
    + OLAP数据库，包括临时空间  
    
   每个月您所使用的增量磁盘空间都会变成事实数据存储 5  
5. 性能调整  
    应该把性能调整作为系统进入生产环境时部署和测试过程的一部分  
    制定一项计划，按季度、半年或年度来安排运行性能调整，一份有关关键统计数据的简单报表有助于评估系统性能
6. 备份和恢复  
    不论系统的可用性需求如何，都需要一份备份和恢复计划  
    测试这些规程也同样重要，至少应当练习一次数据库恢复，做大量笔记  
    您要为每日或每月的装载周期制定故障计划 5  
    关系数据库在每次装载后要备份以下数据集：
    + 关系数据仓库数据库
    + 登台数据库
    + 文件系统中的登台数据
    + 元数据数据库  
    
   需要为数据仓库日志数据库建立常规的备份调度 5  
    关系数据库通常很大，在夜间装载周期末尾运行备份通常有挑战性，这里有几种替代方案  
    + 把数据仓库存储到一个存储区域网络（SAN)并使用SAM软件来执行备份 
    + 对大事实数据表进行分区，并将老分区设置为只读，主要依靠各个分区不同的备份策略
   
   日志数据库被不断写入，更像事务处理数据库而不是数据仓库数据库，需要制定严格的备份策略和恢复策略  
   登台数据库的备份策略和恢复策略有很多不同，多数仅包含最近加载的数据 5  
   元数据数据库也应像一个事务处理数据库那样使用完全恢复模式和日志备份  
   您的ETL工具、OLAP数据库、报表工具或者作业调度程序可能都会将信息存储到数据库，这些工具信息应该进行一致的备份  
   对于OLAP数据库的备份，您需要一个基于文件的备份机制或使用基于SAN的策略
7. 长期存档  
    许多企业都有一些数据由于法律和制度的原因必须存档很长事件  
    我们推荐一个进行长期存档的两部分策略：
    1. 不要把赌注押在任何一种物理媒介技术上，应该制定一个定期的迁移和刷新程序 5  
    2. 在虚拟机上封装当前应用运行环境，存档改虚拟机镜像，并提交迁移和刷新周期
    
## 13.4 小结
成功的DW/BI系统部署和运营需要在开发工作完成之前进行充分的规划和协调  
首先要关注的技术是：测试应用程序，并且将组件从开发阶段转到测试阶段然后投入生产阶段  
一个可靠的测试流程包括不同种类的测试，只有完成系统测试后才能准备部署  
您还需要关注桌面安装、集中式的用户培训和支持，一旦系统投入生产环境，就需要组建支持团队  
## 13.5 管理工作和降低风险  
最大风险是投入的工作量不足，因为多数CIO很难相信在运行操作过程中需要的团队应该与开始部署系统时具有一样的规模  
新DW/BI系统的最大部署风险是未能诚实地评估您地部署准备情况  
健康的系统不可避免会发生变化，要在实际事件之前部署变更，计划、记录和测试整个过程  
面向用户的部署难题是低风险的，可以通过使用团队以外的资源来提升教材质量  
前台运营需要一个规模较大的团队，CIO和业务发起人要确保安排好所需的各种人员，后台维护应尽量自动化  
## 13.6 质量保证
开发和质量保证团队必须进行比较平稳的部署，高质量的培训和文档教材展示了DW/BI系统的专业形象
## 13.7 关键角色
+ 主测试员负责制定和监督测试计划，制定测试方法，选择测试工具，并与所有其他测试员进行协调  
+ 质量保证分析员负责大多数预部署测试，包括功能测试，部署流程测试以及重要的数据质量测试  
+ 数据库管理员负责性能测试和调整，要在部署之前和持续运营过程中进行
+ 数据仓库施训人员的职责是开发和提交业务用户培训课程
+ 项目经理和业务项目指导者基于管理层和项目发起人的指示或者反馈意见来确定用户支持机构的结构和职员

## 13.8 关键提交内容
+ 测试计划：测试方法，测试工具，测试脚本和测试数据集
+ 测试系统部署方案：要详细说明如何把新系统从测试环境迁移到生产环境
+ 业务用户文档，利用整个生命周期中开发的文档或元数据
+ 培训教材和数据库
+ 运营计划：数据库维护活动和时间表、备份策略和恢复策略
+ 前台支持计划：服务台计划配备的职员，DW/BI团队中面向用户的BI分析员以及业务用户社区的超级用户

## 13.9 工作量估计
预部署测试阶段通常会持续四到八周  
有了好的元数据，就可以花几天来写总价、开发元数据报表并把他们集成到BI门户中  
如果您还没有编制单独的文档来描述表、列或者预定义报表，那么您的各种还需要花费好几周  
需要花8天事件来为为期一天的培训会议做准备  
前台运行支持需要相当大规模团队，这个团队应当与最初的DW/BI开发团队规模一样大

# 14 扩展 DW/BI 系统
您需要平衡新的开发工作于运行维护和支持之间的关系  
## 14.1 对已有环境的管理
DW/BI系统永远不会真正地完成，不能简单地将数据仓库地运行维护和保养工作移交给某个单独地维护机构去做  
让DW/BI系统的维护团队和开发团队一起工作会取得最好效果  
理想情况下，ETL系统已经实现自动化，但所有关注如何解决ETL系统中难题的人员都将对如何改进系统或者数据由深刻的见解  
当您不断向前推进时，开发人员可以转移到前台，数据仓库团队大小最好保持不变，这是数仓区别于大多数系统开发项目的地方 5
### 14.1.1 赢得业务用户  
业务用户使用您的DW/BI系统，那么系统必将失败，要确保前台团队能够持续地为业务用户提供支持并且对业务用户进行培训  
应该花几个小时的时间呆在正在使用DW/BI系统的部门，看一看人们与BI门户、数据存取工具及数据之间时怎样交互的  
### 14.1.2 向上管理  
您应该建立一个现行指导委员会或咨询部门，定期讨论设置目标、确定优先次序，寻找具备下面这些条件的委员会成员：  
+ 由远见卓识的领导能力，要懂得如何决策，必须能够预见到信息和分析结构对商业机构实现自己目标的潜在影响  
+ 以企业为中心，必须能够为企业的整体利益着想 5
+ 预算权威人士或者能够影响资金决策的人员
+ 来自IT部门和业务部门的指导委员会成员

通过实行指导委员会来为系统下一步的发展范围提供指导，提高整个企业范围内的问题的可见度  
### 14.1.3 度量并推销您的成功
根据项目开始时协商的成功标准来度量DW/BI系统的性能，如果系统性能不好，建议在开始新项目前找出原因并采取修正措施  
1. 评估投资回报率  
    在任务完成之后您应该对投资回报率进行再评估，评估投资回报率是一个持续的过程 5  
    系统刚刚部署完之后开始总结和分析系统开发成本是个很好的时机，包括DW/BI核心团队的人力成本、购买硬件和软件的花费，业务用户贡献的时间，以及量化运行维护成本  
    当系统以及称为成品时估计回报要容易得多，收益体现在以下几个方面：
    + 提供了业务用户生产力  
    + 减少了事务处理系统的负担和花费
    + 改进了商务决策，商务决策时基于来自新系统的信息和分析结果而制定的，应该将DW/BI项目作为财务影响力的一部分 5   
    + 改进的操作，当DW/BI系统成熟起来并且在系统中建立了更多的闭循环应用程序和可用的操作型BI时，您将会处在受到充分信任的重要岗位上  
   
   建议用户保存一份有关全部业务影响力成功事例的书面日志，记录与每个事例相关的人员、内容、时间、成功方式和影响程度  
2. 监视成功情况和服务度量  
    部署完一个月内要开始进行一次彻底的评价，关注的是项目和系统成功的软度量  
    评估项目的开发周期，回顾计划和进度表偏离了多少以及为什么会偏离，制定计划使得下一轮开发做的更好  
    评估部署阶段，不充分的测试和部署步骤是导致在最后时刻出现延误和劣质数据的常见原因 5  
    要从业务用户的角度出发来评估系统的有效性和性能，建议从业务用户那里收集可以量化的项目成功的度量标准  
    要确保系统能够满足在运行时间和查询性能的服务层协议  
    定期按照成功标准和服务度量对性能进行跟踪，这一点很重要  
    除了定期使用系统的程序获得大量度量数据，少数面向操作的度量必须直接从业务用户那里搜集  
3. 积极推销数据仓库  
    对于数据仓库，业务用户是有一定可选择余地的，不是做决策的刚需 5  
    数据仓库营销在以下两方面都很重要：
    + 一是为当前进行的维护工作提供资金支持
    + 二是为增长倡议提供资金支持
    
   项目经理应为数据仓库代言，设计演示版本，讲述数仓任务、成果、可用数据、当前用户和使用趋势以及未来计划  
   DW/BI业务影响样例日志可以作为市场营销计划的相关材料，适当的话应在整个组织和业界宣传这些成功事例  

### 14.1.4 经常进行沟通  
在项目的整个计划、开发和部署阶段，沟通是一种重要的习惯 5  
应经常与以下重要委托人沟通：
+ 业务发起人和业务驱动者，可通过维护定期的预测报告来维持其地位，与用户沟通取得的成功和存在的问题及改正问题的计划  
+ 业务用户，沟通存在的问题，通知用户将要进行的设计和开发的工作量
+ 一般业务社区，可以采用每季度发行单页形式的更新信息或者在公司通讯中开辟一个专栏等形式发布更新
+ IT管理部门，IT管理部门应该收到全部更新情况 
+ 数据仓库团队的更新， 内部保持沟通很重要，要为团队的沟通确定优先级 5

## 14.2 为增长与演化作准备
成功的DW/BI系统是不断演化和增长的，变化是系统成功的标志  
影响数仓的最初设计因素：发起人、用户及需求、技术架构和源数据都在迅速的发展变化  
在考虑增长的时机之前应该先对当前环境进行评估，要开发一套程序来确定得到这些增长时机的优先次序  
### 14.2.1 评估当前环境
本书提出的建议是基于一种环境假设，但事实并非如此的话，应该进行如下所示的自查测试：  
+ 该项目是否有一个有效的指导委员会？ 5
+ DW/BI团队是否了解业务用户的需求？业务用户是否相信DW/BI团队了解他们的需要？
+ 是否对数据按照维度进行组织，使之能够满足查询性能方面的要求？或者是否对数据进行合理构建使之满足要求的操作性能？  
+ 业务用户是否喜欢他们每天使用的那些工具？能够使他们很容易地进行数据转移吗？
+ 数据是否准确？数据是否与现有地操作型报告相符？如果不相符数仓团队是否充分了解数据不一致情况？是否已通知业务用户？
+ 数据是否及时更新？ 5
+ 用户是否对查询性能感到满意？查询性能是否是一致地和可预测的？
+ 数据仓库团队是否建立了BI应用？用户是否能够很容易地离开原先创建的报告和应用，并开始创建他们自己的用户分析？
+ 用户是否接受数据、应用模板、数据访问工具等方面的基本培训？
+ 用户是否知道数据的具体内容以及数据的来源？他们是否能够访问表和列的定义？是否知道具有代表性的值是什么样的？
+ 业务用户是否知道当DW/BI系统出现问题时应该找谁？是否有人能够及时地响应业务用户的请求？ 5  
+ 在部署完以后是否还能得到持续的培训？

如果这个测试失败，也就是说大多数答案都是否定的——那么应该在开始新项目开发之前先处理这些存在的问题  
一旦用户觉得满意了，那么就有牢固的基础，可以创建扩展的系统  
### 14.2.2 优化增长时机
指导委员会的职责之一就是创造优化增长与演化的时机  
通常有两种用于增长处理和演化时机优化的组织过程：一个用于相对较小的项目，另一种用于主要的数据仓库计划 5  
1. 区分次要增强功能优先次序  
    次要增强功能包括附加维度属性或计算好的事实，基于已有数据的其他BI应用程序以及对已有应用程序所做的更改  
    对次要的DW/BI增强功能的优化通常采用一个适合的过程来处理变化服务请求  
    通常这些次要增强功能由DW/BI团队处理，团队每个季度都向指导委员会汇报新近的变化  
    不要因为请求变化很小而认为不值得去做  
2. 区分主要提案的优先级  
    主要的增强功能可能会设计到来自某个业务处理过程的数据，而该业务过程在数据仓库还没呈现出来 5  
    项目业务发起人或项目经理应该进行时机的可行性和业务合理性分析，应该符合第2章中所描述的准则  

### 14.2.3 管理迭代增长
在此阶段应该着手准备下一个DW/BI项目，但仍然要面向业务  
在准备新项目的同时，也要巩固已经完成的项目基础，但都应该重新返回到生命周期最初阶段  
+ 项目计划和管理，每个新计划都应该被确立为一个定义好的项目，如第二章所建议的那样
+ 业务需求定义，用户业务需求在任何适合都应该具有驱动作用，参考第三章 5  
+ 技术架构策略与底层设施，根据需要增强和调整现有的技术架构
+ 维度建模，使企业数据仓库总线架构和业务维度保持一致，仍然需要整个企业的维度和信息的标准化，如第6章和第7章所述
+ ETL，ETL系统是开发任务的坚实基础，通常第二阶段的项目会有比最初项目更复杂的ETL系统，参考第9章和第10章
+ 应用程序设计与开发，参考第11和12章，一个DW/BI系统不但能够提交数据给用户，还应该是一个用于询问为什么和要是。。。又会怎样的问题平台，DW/BI团队应该检查这些决策点，询问哪些额外的数据源会对业务用户有用  
+ 部署计划、培训和支持，您可能希望加强现有的部署、培训和支持过程，参考第13章，如果BI工具环境有较大的变化，要确保提供额外的用户培训 5

## 14.3 小结
确保DW/BI系统始终保持生机、健壮和平稳地增长  
在继续关注技术环境的情况下，永远将业务用户放在第一位
## 14.4 管理工作和降低风险
首先评估最初或刚完成的DW/BI系统除去投资后是否还有利润，是否满足业务用户需要，这样可以降低增长风险  
时常地、定期地监视系统地进展情况，并且积极主动地沟通  
挡在生命周期迭代过程中需要合并其他业务过程中地数据时，可能需要解决比初始阶段更难以解决的问题，应该指定计划来提高已有职员的技能  
## 14.5 保证质量
劝说管理层增加DW/BI团队的人数来支持运行操作和新的增长  
扩展DW/BI系统来支持新的用户和创建新的BI应用程序可以从某个最优方法集开始，确保为那些已经成功的方法编制文档
## 14.6 关键角色
+ DW/BI主管或项目经理 负责确定运行策略、进行成功评估、进行营销活动和上下组织交叉等方面沟通
+ 项目经理或业务项目领导人驱动业务定义活动  
+ 执行指导委员会负责指导如何对新的主要工作区分优先次序  
+ 参与新项目的整个开发团队
+ 确保业务人士也能参与进来，评估系统是否能满足预期要求

## 14.7 关键提交内容
+ 现行市场营销和沟通计划，包括定期评估投资回报率
+ 指导委员会章程，包括角色和职责
+ 指导委员会定期更新回报文稿
+ 时机优先级防范
+ 项目章程、范围和时间表

## 14.8 工作量估计
DW/BI系统的开发工作永无止境，后续项目应该像第2章讲述的那样进行人员配备、计划制定和管理，现在回头开始，重新再做一遍
