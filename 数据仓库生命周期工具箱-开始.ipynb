# 1 Kimball 生命周期导论
Kimball生命周期提供了一个总体框架，它与数据仓库和商业智能实现过程中的各种活动密切相关
我们将引入生命周期图，讲述在项目中有效地使用生命周期所设计到地主要任务和一般准则  
本章关注的时整个森林，后面章节关注的是森林中的单个树木  
## 1.1 生命周期的历史
最先出现在基于局域网技术的产品上，包括一个关系数据库服务器和客户端图形用户界面  
25年前分析人员都使用这些产品来构建查询，并将结果以电子表格形式存放或以图表形式表示出来 5  
工作初期，业务领域没有什么最优方法，但决策支持发展方向和现在一致：提取、查询、分析和表现  
成功地使用数据仓库技术来管理数据取决于以下三个基本原则：  
1. 将关注点放在业务层面上
2. 按照维度组织数据，将数据以即席查询或者数据报表的形式提交给用户
3. 开发整个数仓环境时应采用逐次迭代的方法，每次的生命周期增量都应当是可处理的 5

Kimball生命周期是本书的核心理念  
Kimball生命周期的基本架构经得起时间考验，印证了“技术源于实践，结论来自验证”  
## 1.2 生命周期里程碑
DW/BI系统的成功实现有赖于多项工作和多个部件的合理集成  
Kimball生命周期确保了项目每个组成部分都在正确的世界按照正确的顺序组合到一起

### 1.2.1 项目/项目群规划
项目指的是Kimball生命周期的一个单循环  
项目群是指多个项目的资源运行协调、底层设施、活动时间表及项目间沟通，会不断自我更新    
项目和项目群两者并不需要确定唯一的方法或顺序  
项目规划过程突出的问题是DW/BI项目业务范围的界定，必须对业务需求有基本了解，项目规划是项目运行管理的基础  
### 1.2.2 项目/项目群管理
管理关注点在于项目状态监控、问题跟踪和为了保持业务范围边界而进行的变更控制 5   
现行管理也包括综合沟通计划的拟定，具有连续性的沟通对实现DW/BI目标至关重要  
### 1.2.3 业务需求定义
彻底了解业务用户和他们的需求，可以增加DW/BI项目提案成功可能性  
必须了解驱动业务的主要因素，进而将业务需求成功转换为设计过程所考虑的事项  
### 1.2.4 技术路线
业务需求定义后是三条并行路线，分别是技术、数据和商业智能应用三个方面  
技术路线主要是第4章说明“要做什么”，第5章说明“应当如何去做” 5  
1. 技术架构设计  
    确立了一个总体的架构框架和设想，必须同时考虑三种因素：业务需求、当前技术环境、方案设计的战略技术方向  
2. 产品选择与安装  
    技术架构设计是一个虚拟的清单，包括系统所需的功能和特定的架构部件，一旦选定产品就需要对它们进行安装和测试  

### 1.2.5 数据路线
数据路线从目标维度模型的设计到模型的物理实例化，再到达重装设备  
1. 维度建模  
    在收集需求过程中，需要将需求确定下来并且保存在一个初步的企业数据仓库总线矩阵  
    设计支持业务报表和分析性需求的维度模型是一种新的方法 5  
    本书第6章介绍了维度建模的概念呢，第7张讲述了开发维度模型的推荐方法和步骤  
2. 物理设计  
    重点关注物理结构的定义，设置数据库环境和制定适当的安全策略  
3. ETL设计和开发  
    有70%的风险来自于数据提取、转换和装载系统的设计和开发这个阶段  
    第9章论述了ETL系统的整体架构及其34个子系统模块，第10章关注实际应用，探讨ETL设计与开发过程中的细节和任务

### 1.2.6 商业智能应用路线
第11章讲述了一般概念和基本原理，第12章则介绍了设计与开发的最优方法 5  
1. BI应用设计  
    在业务需求定义完成后，就可以确立商业智能应用的候选方案，设计适当的导航界面组织各种功能  
    BI应用程序是从DW/BI解决方案向外传递价值的业务信息的工具  
2. BI应用程序开发  
    除了导航门户外，还包括配置业务元数据和工具，以及构建和验证专门的分析性应用程序和运营BI应用程序  

### 1.2.7 部署
上述三条平行的路线（技术、数据、BI应用）在部署环节又汇聚到一起  
在这里需要进行更为详尽的规划，精心组织部署是十分关键的，所有细节部分都应该在发布前做好准备 5  
### 1.2.8 维护
DW/BI系统产品化之后需要一些技术性的维护操作来确保系统最佳状态  
这些维护工作包括状况监视、系统性能调整、索引维护以及系统备份等，第13章会具体讨论  
同时需要继续关注业务用户，为他们提供日常技术支持和培训，并保持沟通  
### 1.2.9 增长
DW/BI系统在开发过程中的变更往往被看作是成功的迹象  
为了满足业务运行的要求，必须确立各项工作的优先顺序，然后回到生命周期的初始阶段，在14章会详细介绍 5

## 1.5 生命周期相关术语简介
DW/BI行业里充斥着大量术语，本书中将尽量保证其含义明确、语义一致  
### 1.5.1 数据仓库与商业智能
Kimball集团将提供信息以支持商务决策的整个过程都叫做数据仓库技术  
提交端到端完整解决方案是基本原则，包括对源系统的数据抽取、以及与业务用户进行交互的查询命令和应用程序  
商业智能主要是指对数仓中存储的数据进行记录和分析的过程  
业界内有将数仓分为数据库和商业智能两个子层，还有一些将商业智能看成顶层，而数仓则指数据存储的核心基础设施 5  
我们将数据仓库和商业智能两个术语的首字母放在一起进一步强化了它们之间的依赖关系  
DW/BI系统中可查询的数据都称作企业数据仓库，增值分析称作商业智能应用，数据仓库是商业智能的基础  
### 1.5.2 ETL系统  
ETL作为DW/BI环境中的厨房，在投入运营前，应对工作区和各个组成部分的蓝图进行细致的设计  
系统要设计得高效，同时保证质量高、种类全，加工能力也十分重要  
让业务用户进入后台系统很不安全，而且对数据得加工多样，用户并不需要了解所有的情况 5  
ETL工作区必须在进行源数据的提取之前就进行布局和构建，吞吐量和质量都尽可能高  
在数据进行质量检验并且确保能够满足需求之前，业务用户和BI应用程序都不能进入ETL系统  
本书第二版中扩展了ETL系统最优方法的覆盖面，通过34个子系统提供了一个内容详尽的清单  
我们使用数据登台来表示数据清洗和预处理过程  
### 1.5.3 业务过程维度模型
DW/BI系统前台必须按照业务用户的需求来设计和管理，维度模型是前台提交的主要内容 5  
维度建模是一套优化的设计原则，关注业务用户使用的便利性和BI查询性能这两个方面的目标  
维度模型的两个基本结构是事实表和维度表，事实表的度量源自业务过程或度量事件  
我们围绕着业务处理以及相关数据源来构造维度模型，这种理念可以为所有业务人员设计出相同、一致的数据视图  
为了获得最大程度的适应性和延展性，应该在分析业务过程所捕捉的最低的、最原子层面上精确声明事实表粒度  
维度表含有与特定的、实际的业务度量相关的属性描述和特征描述，在查询中用于约束、分组和标记等操作 5
在关系数据库中，维度模型需要从物理上实例化，会涉及到星型模型  
数据集市这个术语被人们边缘化，用来表示支持部门应用的、独立的非架构数据集  
### 1.5.4 商业智能应用程序
如果能够使用预定义的应用程序来查询、分析和表现来自维度模型的信息，那么业务社区多数用户都会非常满意  
BI应用程序既可以是简单的报表，也可以是直接和运营事务处理系统交互的分析性应用程序  
BI应用程序必须能够体现客户的需求，并且应该提供用于交互的“菜单”以描述可用的功能 5
## 1.6 小结
Kimball生命周期提供了一个整体框架，将成功实现DW/BI系统所需的多种任务组织到一起

## 2.1 确定项目
从何处开始入手DW/BI项目的开发？先花时间评估一下您所在机构对DW/BI项目的准备就绪情况
### 2.1.1 评估 DW/BI项目的准备就绪情况
为保证DW/BI项目成功，必须确保三个关键因素就绪  
如果您还不能自信的给所在机构打分，那么建议先减慢速度并重新考虑风险和障碍  
1. 强力型高级业务管理发起人  
    是否能够得到来自高级业务管理层的有力支持和帮助是至关重要的因素  
    强力型的业务发起人是所在组织内有影响的领导者，往往机敏干练并且交际广泛 5    
    理想的业务发起人要求严格，但是又比较现实，会为其他成员提高强有力的支持  
    得力的发起人能够冷静面对短期出现的难题和挫折，因为他们更关注长期成功  
    成功的DW/BI团队一般会在机构内部培训好几位强力型的业务发起人  
2. 强制性业务动机  
    那些能够成功实施DW/BI解决方案的结构通常都对由一个或多个强制性业务动机所引起的信息访问改进需求具有强烈的紧迫感  
    + 外部环境中的竞争和变化所激发的因素 5
    + 内部环境也能够推动机构进步，如并购带来的集成环境需求
    + 潜在市场机遇相关的战略构想也会成为压倒性因素  

   当DW/BI系统和战略性业务动机紧密结合在一起时，成功的把握往往会比较大
3. 可行性  
    这里讨论的是DW/BI准备过程中和数据本身相关的可行性：比如数据质量很差、甚至有些数据根本没有收集  
    数据探查是一种数据分析技术，主要描述数据的内容、一致性和结构 5  
    当评估处理候选数据源的准备情况时，应当迅速进行一次数据探查评估，将探查结果作为早期的合格整合或不合格证明
4. 不作为准备过程中合作破裂点的要素
    技术在这一阶段不是要紧问题，不需要为之分散精力  
    IT和业务之间保持健康而密切的关联以及良好的合作关系，能够使项目群的实现过程更为顺利，但缺乏积极关系也不一定就会造成合作破裂  
    大多数业务方都认为他们的工作中数据是实现各项功能的关键需求，但如果您所在机构仍然只是依据直觉而不是分析，那就要把分析文化问题放到DW/BI项目群准备过程中来考虑  

### 2.1.2 弥补不足并确定下步工作
强有力的业务管理发起人相对来说更为重要，强制性业务动机和可行性的重要性位居其次 5  
业务发起人地位至关重要的原因：
1. 发起人能够对必要的人物规划以及该人物的影响进行详细描述
2. DW/BI系统的开发维护费用很高，发起人强有力支持是获得和维持项目资金必需条件
3. 发起人在当前项目与其他重要项目出现矛盾时也能合理支配资源

如果项目其他一切条件具备，唯独缺少一位强力型业务发起人，那么项目群实施会面临很大的困难  
到现在为止，您需要开始评估整体的准备就绪情况了，此外还应制定一些风险缩减策略和意外故障计划来应对可能存在的缺陷 5  
下面给出几种常见情形和确保系统稳定运行的建议方案：
1. 强力型发起人、强制性业务需求和高质量的数据  可绕过剩下的缺陷内容直接进入项目初步设计
2. 低质量的数据  
    这种情况下项目不应当继续进行，而是为业务发起人确定另一个业务价值高而且数据可行性障碍较少的需求  
3. 能力弱的业务发起人和仅懂IT的发起人  
    如果项目任命的业务发起人缺乏组织向心力，不能预见到系统为机构带来什么效益或者不能够全身心投入，那么就需要在机构中物色新的业务发起人      
    第3章详细讲述了收集业务需求的方法和策略，现在我们将注意力放到高层业务问题和业务价值影响上  
    需要与业务管理部门共同找出以下几个和高层需求分析相关的问题：5  
    1. 理解业务管理部门的战略性商业计划
    2. 为他们想要监控或施加影响的核心业务过程确定主要的绩效度量或者成功度量标准
    3. 确定业务信息访问的改进对其绩效度量的潜在影响
    
   如果您的发起人只是对潜在业务价值缺乏清晰的认识，那么就需要帮助他们对可能出现结果有一个大概的印象  
   概念验证是用来论证潜在价值的，可以完全被舍弃，开发过程应当重点关注强制性业务动机 5  
   概念验证会破坏合理的期望设定，因为该系统往往几天就能建成，管理部门会很难理解为什么要花费数月来开发DW/BI
4. 多个业务发起人提出过多的要求  
    在机构中能有这样的热情是非常好的现象，但是来自机构中多个业务单位的需求也会引发很多难题  
    必须和业务机构共同协作来区分这些需求的优先次序，业务和技术共同建立业务影响和业务可行性矩阵  
5. 善意却过于冒进的业务发起人  
    具体表现为不断超出项目范围从范围蔓延到范围飞驰，IT人员必须制定下述原则，每引入一个新的主要数据源就将DW/BI的开发周期延长6个月  
6. 表现不佳的遗留系统，孤立的数据“竖井”  
    很多业务机构都已经建立了数据仓库/集市，这些数仓有者相似却不同的业务规则，却没有任何顶层框架或架构 5  
    由于纯底层的设施项目很难得到资金支持，需要依据已有的需求和业务发起人一同找到恰当的时机，可以在业务价值实现过程中得到架构和底层设施的副产品 

### 2.1.3 确定初步范围和章程
您需要开始划定项目的范围，保证该范围合理，事实上这两项规则是紧密关联的  
项目范围的确定应该由业务需求来驱动，生命周期图中，这种关系由 项目/项目群规划和业务需求定义的双向箭头来表示  
先讨论范围界定的对象十分重要，在首次向业务用户提交项目信息时，需要同时提交范围界定和合理性证明  
1. 聚焦于一个单独的业务过程  
    最重要的一项业务边界确定工作就是将关注点聚焦于单独的业务过程，因为单个业务过程常常由单个主要源系统模块支持5  
    早期的每次迭代中都应当将数据的来源限定到单个业务过程中，因为多个业务过程下，ETL工作量会成指数增长  
    在这里给出确定DW/BI项目初始范围的一些补充指导原则：
    + DW/BI项目的范围应当有可感知的业务价值，建议启动能够快速上市的小型项目  
    + 确定范围需要IT代表和业务代表的共同努力
    + 有时需要按照某个预定交付日期来确定范围，并设法在规定期限内提交内容，这种情况下提交内容也必须体现业务价值  
    + 一旦范围确定应当立即确立项目成功标准，了解管理层对DW/BI项目的期望极为重要 5
    
   仪表板给项目带来的弱点就是可能会导致项目超出预期，因为仪表板一般都包含多个业务过程的关键绩效度量  
   多个过程所需的详细数据会导致实施过程中出现时间不足地情况，团队常常用半手工化作业应急，但并不能长久维持  
   仅向高级主管提交汇总后的结果是不够的，因为他们的经理和分析员需要更为详细和透彻的业务分析，影响系统性能的真正原因都是隐藏在细节中  
   提交细节数据一种更可靠的方法是一次仅提供一个业务过程的数据，刚开始缺乏行政上的吸引力，但更为稳妥、持久，避免返工  
2. 快速应用程序开发的角色定位
    如果项目实际的欠交付率很高，那么就要关注敏捷开发 5  
    快速应用程序开发的原则和Kimball方法论有共同之处：
    + 将精力集中于所要交付的业务价值的主要目标上
    + 干系人（业务代表）和开发团队之间的价值协作
    + 强调同业务干系人进行面对面沟通、反馈和确定优先级等行动的重要性
    + 尽快适应进化后的需求，而且变更是不可避免地
    + 以迭代、增量地方式处理可重用软件地开发问题，多项任务并发、重叠 5  
    
   当创建一个DW/BI系统时，可以考虑使用敏捷方法，而且敏捷开发天生就适合BI应用层  
   有些开发团队容易掉入陷阱:凭空创建分析方法或报告方案，仅面向少量的用户，解决特定问题，最终开发出来数据烟囱  
3. 编制项目群范围/章程文档  
    一旦项目范围达成一致，就应当编制相关文档(项目群章程)，确保参与项目的每个人都对范围界定有一致的理解  
    项目群章程说明了项目的关注点和动机性业务需求、目标、方法、预期数据和目标用户、涉及到的团体和干系人、成功的标准、假定和风险  
    确立适当的项目边界和相关文档速度越快，需要更正的错误理解就越少 5

### 2.1.4 建立商业报告和合理性证明
一旦确立了项目的初始范围，就需要为项目创建业务合理性证明，该证明依赖于IT机构和业务机构之间紧密合作  
合理性证明意味着确定与项目群相关的、随时间而变化的各项项目预期成本和收益  
多数机构会采用某种标准化财务模型，被用于分析项目的合理性，重点关注那些数值的来源或者价值估计，以便输入到财务模型中  
1. 确定财政投资和成本  
    早期阶段项目本身充满了很多不确定因素，一般来说成本主要包括如下几个方面： 5  
    + 硬件成本,所有硬件组成部分及其升级费用，一定要考虑到已有的基础设施是否可以被利用  
    + 软件成本，购买或者升级软件的成本，包括数据探查、ETL、BI查询、报表和分析、元数据管理以及操作系统平台和数据库管理软件
    + 内部人员配备，是项目成本的主要组成部分  
    + 外部资源，在其他厂商的技术支持和外部人员提供的培训服务上面分配最少的预算
    
   还需要考虑运行成本，除了需要维护当前状态，还有要在性能提升和项目增长方法的投入，这里给出需要考虑的成本： 5
   + 运行维护开销，大多数硬件和软件还需要一定的运行维护开支  
   + 运行支持,DW/BI系统对业务用户的支持并不会随着时间的推移而下降，还需要具有运行监视和功能调整的
   + 支持增长所需的开销，还需要考虑到支持业务需求变更的运行开支，用户数量膨胀后带来的开支、新版本升级所需的开支  
2. 确定财务效益和收益  
    现在应该将注意力转移到商业收益上来，项目收益一般分为有形收入和无形收入  
    有形收入更为重要，因为它们是可计量的，下列清单列出了收入增加或成本降低对利润率的潜在影响： 5  
    + 由新客户带来新增销售而增加的收入
    + 由销售较为昂贵的产品或者销售外的产品而增加的收入
    + 通过改进对目标地址的命中率而提高邮件响应率
    + 通过改进分类规划提高每份邮件的平均销售额
    + 淘汰低利润的产品 5
    + 由于改进供应商管理和进行商务谈判而降低原材料成本
    + 在商业活动中通过适时调整而降低促销开着或之间市场营销成本
    + 降低了骗赔支出
    + 减少了客户流转和反覆
    + 降低了故障率 5
    + 提高了首次联系的终止率
    + 降低了存货费用和生产过剩
    + 缩短了新产品上市的时间
    + 基于更为完整的收入信息和成本信息改进价格政策
    + 削减了编制遗留系统报表的成本 5
    + 缩短了收集数据和生成分析结果的时间 
    
   无形收益或软收益也能够为机构带来价值，但无法用财务独立标准去独立，不能适用于ROI模型： 
   + 更及时的信息访问
   + 基于事实改进决策制定水准而不是盲目决断
   + 机构内部的更高程度的信息标准化和信息一致性
   + 对数据的更高的信任程度，并且减少了在数据校正方面的工作 5
   + 通过提供单一版本的数据来消除低效率
   + 对条件变更的响应速度更快
   + 提高了分析供选方案的能力
   + 提高了员工的满意度和干劲
   + 提高了客户的满意度 5
   + DW/BI是保持企业竞争力的战略必需品
   
   查找有形业务收益的关键是不断进行探寻和追问，从表面看不出问题的收益报告找出为什么、要是...怎样、...多少钱的陈述  
   可以追问：如果有...将会对业务有什么影响？您能否....？....价值多少？，从而推断业务收益的各部分数据   
   假设市场人员认为目标市场定向较好的邮购活动能使响应率提高10%：
   + 这份报告本身并没有包含足够信息，无法从中单独推算出财务影响  
   + 所以必须继续询问一系列问题，以高层财务影响预测得到以美元和美分计算的具体结果 5 
   + 可以询问从每次直接邮购活动中产生的平均收入是多少？和每年举行多少次邮购活动？就可以获得响应率增加10%推导的必须数据  

3. 利用投资和回报计算ROI  
    此时可以将投资和回报的有关数据输入到一个财务模型中计算投资回报率（ROI）
    也可以从不做任何工作所付出的机会成本的角度来考察财务合理性证明，比如当由缺乏足够的信息而无法做出决策时，就会存在失去商机而产生的成本  
    合理性证明仅用于测试机构中的人员是否能够承担自己的任务，是对机构承诺的事项的一种度量  
    除了ROI之外 有的时候还应该在合理性证明中对竞争性或市场条件变更等外部压力予以权衡 5  
    建议在解决方案实施前对投资回报率进行预测，在解决方案实施后对投资回报率进行监视  
    商务决策的改进取决于人和决策过程两种因素，DW/BI正是这样一种技术，好的决策建立在DW/BI信息和分析基础之上  
    需要记录各业务部门在使用DW/BI系统后做出的决策，在记录这些决策之后，那么计算ROI就变得简单了  
    随着这些贡献不断累积，那么逐渐营造出一种文化，从这里就可以明白商务决策支持需要战略投资
    
## 2.2 项目规划
定义好一个初步项目并且获得批准和资金支持之后，要做的事情就是进行详细的项目规划  
### 2.2.1确立项目标识
项目需要一个特色鲜明且不容易被忘记的名称，可以简化为首字母缩写  
一些机构还为DW/BI系统设计LOGO，成为系统高品质的一种标志
### 2.2.2 项目人员配备
项目团队最高层由项目拥有者中的全体决策人员和总经理组成  
发起人和驱动者负责确定总体方向和负担项目费用 5  
决策者在指导者的协助下建立常规的项目启动团队，并进行日常领导  
项目发起人和项目经理都是指导者，管理负责开发DW/BI系统的核心项目团队  
DW/BI实施过程中的决策者、指导者、常规团队和专门团队等角色，大多数以项目为中心，有些还需要担负项目群层面职责  
需要明确的是角色和人员之间通常都不是一一对应的关系，可能会由身兼多职的情况  
项目团队的规模要易于管理，一般大约由3-7个全职成员 5  
1. 决策人员:发起人和业务驱动者 不参与日常基础工作  
    1. 业务发起人/业务驱动者
        是项目业务所有者并且掌管项目财务权，协助确定业务优先级和进行业务范围方面的决策  
        有些情况下高级业务主管人员组成的筹划指导委员会会扮演发起人的角色  
        发起人通常在团队有较高地位，会指定一位业务驱动者代替他，业务驱动者应当对业务有深刻的理解  
    2. DW/BI 主管/项目群经理  
        IT机构的DW/BI主管和发起人协同工作，是业务发起人的辅助角色，是业务发起人和DW/BI团队之间的联系纽带 5  
        DW/BI主管负责整个项目计划和环境的全局领导和指导，有时被称为项目群经理，承担着横跨多个项目的职责  
        这个角色多数情况下被称为CIO，但这一角色还具有一定的业务职能  
2. 指导者:项目经理和项目领导者  
    指导者类似教练，DW/BI项目中参与项目指导的指导者和决策者都是为了项目团队能够取得成功  
    1. 项目经理  
        主教练，负责对人员协调、状态跟踪、项目进展和相关问题的沟通进行细致的管理  
        项目经理能够在问题变得严重之前识别、反应并解决问题，还能使参与项目的每个人都向同一个方向努力 5    
        项目经理是一个需要专职投入的职位，有时既要负责领导整个项目，也会在工作中亲自完成一些相对次要的活  
    2. 业务项目领导者  
        是业务社区的代表，需要真正了解各种业务需求，业务驱动者和业务项目领导者可能由同一人担任  
3. 主力阵容:核心项目团队  
    在需要反复迭代的DW/BI项目周期中,DW/BI团队承担着系统开发和操作的双重职责  
    在一些领域，设计师具有监督和协调的职能，也充当开发者的角色  
    1. 业务分析员  
        负责业务需求定义，将需求描述为技术架构、维度模型、BI应用，以用户为中心考虑问题，非常熟悉业务 5  
    2. 数据管理员/质量保证分析员  
        负责推动整个机构在定义、业务规则和数据仓库中数据的允许域值上达成一致,然后发布和补充这些定义和规则    
        数据管理员必须是受人尊重的领导者，能够尽心尽力解决那些不可避免地功能交叉问题并得到高层支持  
        有时需要同质量保证QA分析员协同工作，质量保证分析员负责确保载入数据仓库的数据是准确而完整的  
        质量保证分析员在进行初始数据装载地过程中工作量很大  
    3. 数据设计师/数据模型师/数据库管理员  
        数据设计师负责设计DW/BI系统的全局数据架构策略，确保系统的可重用性、集成性和优化 5  
        数据模型师负责进行详细的数据分析和开发维度数据模型 ,他对项目的成功有巨大影响  
        数据库管理员将维度模型转化为物理表结构，要负责整体的物理设计、包括磁盘布局、分区和初始索引设计  
        数据库管理员一般还负责数据库的日常操作和维护，确保数据的完整性、可用性和性能  
    4. 元数据管理员  
        负责监督收集什么样的元数据、将元数据保存在何处以及以何种方式发布给业务社区  
    5. ETL设计师/ETL开发人员  
        ETL设计师负责数据抽取、转换和装载这一过程的端到端设计 5  
        ETL设计师参与需求收集过程，和BI应用程序团队建立重要的工作关系  
        ETL开发人员在ETL设计师指导下进行数据抽取、转换、装载等过程的创建、单元测试和自动化，必须相当熟悉ETL工具  
    6. BI设计师/BI应用程序开发人员/BI门户开发人员  
        BI设计师对DW/BI系统的前台负由总体责任，要确保BI环境能够最优化地表达业务需求  
        BI应用程序开发人员负责创建和维护BI应用程序，也负责配置BI工具的语义层，能够理解业务需求和底层数据  
        BI门户开发人员设计BI门户，在此以后确保入口不断更新 5  
        有些组织将其核心团队划分为DW团队和BI团队，那么各部分之间的协作和沟通就变得非常重要了  
4. 专门团队  
    在一些专门、限定的范围内为项目做贡献，也可以由那些已经成为核心团队成员的人来担任  
    1. 技术设计师/技术支持专家  
        负责技术架构和DW/BI环境中地底层设施的设计，需要具有全局综合能力  
        依据开发环境不同，可能会需要平台专家来参与设计的早期过程从而进行资源和功能的规划  
    2. 安全管理员  
        DW/BI系统由很多安全漏洞，必须设置安全管理员来负责所有后台和前台安全、以及外部网络安全 5  
    3. 首席测试员  
        这一角色在测试和部署阶段很重要，负责提出测试的方法，确保测试能够覆盖数据、过程和系统的各个角落  
    4. 数据挖掘/统计学专家  
        负责开发数据挖掘模型并与BI应用程序开发人员合作设计使用含有数据挖掘功能的业务应用程序  
        应该非常熟悉业务并且通常还有统计学背景，数据挖掘应用程序属于数据仓库中比较成熟的客户端  
    5. 施训人员  
        负责对用户进行关于数据内容、BI应用程序以及专用数据访问工具等方面的培训，也可由项目团队成员承担  
5. 自由代理人  
    自由代理人即顾问，能够在项目团队中填补关键的技术空白 5    
    在使用外来顾问时应当注意如下要求：
    + 不要让顾问充任核心项目团队的任何关键角色，保留对项目的所有权
    + 要求您的顾问传授广播的知识，并为员工提供指导
    + 要求顾问给出相应的书面文档、以免自己对他所做的事一头雾水
    + 了解清楚您是应该雇佣一位具有专门技术的顾问还是仅需要增加一名在工作中边学边干的普通员工
    + 不要被顾问说服采用以数据为中心或以技术为中心的开发方法，不要抛开业务分析需求这个大前提 5
6. 将单个人才组织成团队  
    等团队人事安排到位，指导者需要将注意力转向团队的凝聚力和发展上，团队组建的各项活动对确保团队凝集力非常重要  
    有些团队成员没有成功的经验，投入时间通过课程使团队对一般性概念、方法、维度建模、ETL和技术专题烂熟于心  
    在培训上进行投资比以后由于缺乏知识进行返工要合算得多  

### 2.2.3 制定项目计划
DW/BI项目需要一份详细的、综合的项目计划,应当从项目任务和参加人员两个方面来考虑项目的复杂性  
细节跟踪水平是DW/BI项目计划中的常见问题，任务应该更具体，提供足够多的细节，以便跟踪关键任务进展 5  
有效项目管理的关键是选择真正要使用的各种工具，花一些时间用于项目计划的输入和维护工作（包括详细的依赖关系）  
建议为每个项目任务跟踪以下信息：
+ 人员:负责完成该任务的个体，每个任务应该只有一个主要负责人
+ 原来估算的工作量:即天数，永远也不应修改
+ 原来估算的开始日期:任务开始日期
+ 原来估算的完成日期:永远也不应修改 5 
+ 状态:指当前的状态标签（如未来的任务、正在进行中或已经完成）
+ 更新后的开始日期:实际开始日期
+ 更新后的开始日期:实际开始日期
+ 完成工作量:实际工作日的数量
+ 延迟天数：使用估算完成日期与当前日期进行比较 5  
+ 完成百分比:当前状态与当前估计要完成的整个任务的百分比  
+ 依赖关系:确定在开始执行该任务之前，必须完成的其他任务  

本书同步网站www.kimballgroup.com 上有一个项目计划任务列表的例子，一般来说任何一个变量都无法单独决定创建DW/BI系统所需的时间总量  
![2-2](2-2.png)  
建议确定出和整个项目生命周期相关的时间安排、人员或所需工作量等任务，可以仅列出，当项目可见度更高时再完善 5  
制定项目计划时要对未知的事项进行充分估计，因为大多数源数据都存在一些不可预料的数据真实性问题  
ETL开发时是一项经典的软件开发任务，它分为五个阶段：开发、单元测试、系统测试、验收测试和最终文档发布  
还需要在计划中充分估计到文化问题带来的影响，需要将这种文化落实到项目计划中  
项目经理一般起草一个项目计划纲要，随后和生命周期中每个主要活动的关键代表会谈确定他们的任务进展情况，这些代表负责为自己活动制定估计工作量和初步时间表 5    
当项目经理制定一份极其详细的项目计划后，千万不要推迟6个星期才启动该项目，这不仅仅是为了项目管理 

### 2.2.4 制定沟通计划
项目经理应当起草一份沟通计划来解决项目支持者的需求，经常性沟通可以确定和管理期望，是项目成功的一个重要因素  
沟通计划应当概略描述大致的消息内容、通信媒介和与各个支持者群体沟通的频率，如图2-3所示  
![2-3](2-3.png)  
提供过多信息并不比没有提供足够信息好多少，建议简化发送各个支持者群体的消息 5  
本节介绍了具有代表性的DW/BI各个支持者群体进行沟通的一些建议：  
1. 项目团队  
    本章介绍的如项目范围文档、状态分析会议、状态报告和项目计划都可以促进项目团队的沟通，但对于团队之外并不适用 
2. 发起人和驱动者情况介绍会  
    您需要使业务机构、IT机构的发起人和驱动者的意见保持一致，精心组织情况介绍会，控制在30分钟之内  
    要和您的发起人面对面的沟通，不要仅依赖项目状态会议记录和其他一些标准项目文档  
    有必要和发起人一同制定一个常规沟通方案，特别使您的项目遭遇停顿时。诚实是真正的良策，问题和难题应该早点汇报  
3. 与业务用户社区的沟通  
    沟通有助于业务用户持续参与到项目中来，他们的参与又可以使他们认同和使用DW/BI系统的可能性增大 5  
    通过与项目业务用户启动会议进行定期的、运行过程中的沟通非常重要  
    业务用户应当阅读项目状态报告首页的标题内容而不必了解计划细节  
    与用户沟通应当关注性能、未考虑到的事项和时间期限，应当提醒业务用户DW/BI开发过程的特点——迭代性  
    与业务用户沟通可以采用多种方式，所传播的应该是简短、中肯的信息，并且应该避免使用任何技术性词汇  
4. 同其他有关方进行沟通  
    1. 企业行政管理层  
        机构的主管团队一般对高可见性项目提议进行总结，这些情况介绍是得到业务机构接受和认可的开始 5  
    2. 没有参与DW/BI项目的IT人员  
        项目经理常常按月向其他IT管理人员提供项目状态情况介绍，有助于确保将DW/BI系统与其他开发活动集成到一起  
    3. 一般业务社区  
        一条简短的时事通讯或者WEB页，即可将DW/BI最新消息发布给其他机构，而不是让公司内部传闻充当信息源  

## 2.3 项目管理
本节我们将讨论确保DW/BI项目按计划实施的方法，包括项目的开始、团队工作的启动，经过状态监控、范围管理，最后到期望管理和警戒标识的辨识  
DW/BI项目还有一些特征需要用下述概念进行补充讨论：  
+ 交叉功能实施团队：该团队在DW/BI项目中承担不同的职责
+ 迭代开发周期：导致在开发过程中需要更多沟通来保证人员同步，需要对问题/变化进行跟踪以及更详细的文档
+ 不可避免的数据问题会严重损害精心制订的计划
+ 高可见度，必须进行主动沟通来保证业务机构的期望在掌控之中 5  

您应该熟悉基本的项目管理概念，我们会在较高层次上回顾这些关键活动  
### 2.3.1 召开项目团队启动会议
DW/BI项目正式启动始于项目团队启动会议，会议目的是使得整个项目团队对项目的当前状况及未来预期的状况达成共识  
如图2-4日程安排表所示，项目团队启动会议首先由业务发起人进行简要介绍，对项目整体目标和业务相关事项进行说明  
此后由项目经理来负责主持，对项目团队角色和职责进行讨论，在较高层次进行，不会估计工作量，只对一般项目管理期望进行审查，确定接下来进行的各个步骤   
项目经理通常会编辑一个项目启动信息包，并在会议上进行分发，包括项目范围文档、附带有相应职责的团队人员联系方式、初步项目计划和状态会议日程安排、状态报告、问题日志以及变化控制文档的样例  
![2-4](2-4.png) 5  
### 2.3.2 监控项目状态
定期对DW/BI项目状态进行监督，监控项目状态对实现这一过程至关重要  
项目处于困境时，恰恰是最需要对状态进行严密监控的时候  
1.  项目状态会议  
    应当定期对DW/BI项目召开状态会议，一般会议应该历时一个小时，并且安排在每周同一时间和同一地点  
    项目状态会议除了应当重点讨论项目当前完成情况和工作进度之外，特别要关注关键路径上已经延期的任务  
    ![2-5](2-5.png) 5  
    项目经理在确立状态会议的基调方面起重要作用，要排除过多偏离主题的讨论以确保会议效率，营造开放而自由的氛围，避免超时  
2. 项目状态报告  
    状态报告提供了项目进度的高层快照，报告的提交和定期安排的状态会议应当同步  
    状态报告文档记录了取得的成果、重大决策、下一时期的计划活动以及存在的主要问题，保持内容简练，分发给项目团队和领导层  

### 2.3.3 维护项目计划
整个项目计划应当每周更新一次以便能够准确地反映项目进展情况，随后还应当同核心项目团队共享更新后的计划  
如果项目中关键路径上的任务落后于预定安排，那么您需要评估自己所选择的处理方法，与业务发起人一同工作  
最重要的是要勇气向您的管理部门和发起人承认您已经落后了  
### 2.3.4 整理项目文档
DW/BI项目具有不断发展的特性，健全的项目文档有助于让团队的新成员尽快熟悉项目情况  
当项目提交尚未达到预期时，项目文档和决策点制定是很有用的 5   
一定要避免因时间压力导致取消文档编制的陷阱  
项目文档包括所有的项目沟通情况和需要提交的主要项目资料的副本  
### 2.3.5 范围管理  
DW/BI项目必然会发生变化，项目经理需要管理项目范围变更  
对于面对未预先定义的用户请求时，项目经理有如下几种选择：
+ 对这些请求说不 5  
+ 以零和假设为前提调整项目范围，工作总量不变，对范围内外的内容进行调整
+ 对项目范围进行扩展，需要强制性地延长项目期限，并且适当增加预算  

对于这3种方法需要依据特定的情况来进行正确的选择，不能凭空确定项目范围   
接下来我们将讨论问题跟踪及变更请求相关的若干技术  
1. 问题跟踪  
    DW/BI项目中集成了大量的资源、数据和技术可能会产生很多问题 5  
    问题跟踪对确保不遗漏任何事项十分重要，保证了每个人所关心的事项都得到了关注，也保证了解决问题的基本原理得以收录以供以后参考  
    确定问题或找出阻碍问题解决的因素是解决问题的第一步，允许并优先考虑将许多问题简化  
    DW/BI项目有两类问题：一类是影响项目的重大问题，另一类是完成某个重要任务时所必须解决的与该任务相关的问题  
    可以使用一种简单的日志形式来记录上述两个层次的问题，您想了解和需要记录的细节包括如下方面：
    + 问题和问题描述 
    + 问题识别日期
    + 呈报方 5
    + 归属方
    + 状态
    + 优先级别
    + 预定解决日期
    + 结束日期 2
    
   项目问题日志应当在每次项目状态会议上都进行审查，任务问题应当放在主要设计会议或工作会议上进行审查  
2. 变更控制  
    正式承认项目的变更对确保项目管理成功十分重要，所有对项目时间表、预算或范围产生影响的解决方案都应当考虑到变更  
    次要变更的累积作用常常会对真个项目产生巨大的影响，有太多的较小请求，那么问题规模就会扩大 5  
    和刚才介绍的问题日志类似，场次由项目经理负责管理变更控制日志，应当记录的信息包括：
    + 变更请求控制和相关描述
    + 请求日期
    + 请求呈递方 
    + 优先级别
    + 归属方
    + 估计工作量
    + 估计成本
    + 状态
    + 结束日期 4
    
   可以用变更控制日志记这些增加的请求以便日后发布，确定这些请求的业务价值对优先级排序十分有用 5    

### 2.3.6 期望管理
执行沟通计划是管理机构内部各部门对DW/BI项目计划期望的基本方法
另一种管理业务用户期望的方法：寻找机会让他们通过BI应用程序说明、数据模型设计审查、管理和可能的BI工具评估等活动参与进来  
项目计划中应当包含业务用户对生命周期中每次提交的主要内容的验收情况  
项目团队的期望管理常被忽视，对于DW/BI开发来说，返工是不可避免的事实，是一条变更业务和实现项目增长的必由之路  
### 2.3.7 辨识项目陷入困境的征兆
在谈到项目管理这一主题时我们想明确一下一般项目中的警示性的征兆，很多项目团队陷入到下面的陷阱中：5  
+ 没有从高级业务机构引入有影响力的设计人员
+ 认为参与项目的人员能够学习工作中所需的所有知识
+ 一次性处理的任务过多，而不是针对单个业务过程采用易于处理的迭代开发过程
+ 一心专注于技术层面而没有集中注意力于业务目标和需求
+ 认为在整个项目生命周期中不需要业务机构的参与就可以进行DW/BI项目开发，包括从数据分析和BI应用程序设计到管理的工作 5
+ 在明知元数据质量较差甚至很糟的情况下，还没有认真研究数据能否支持开发任务就承诺向前推进项目的开发进程 
+ 认为DW/BI数据库应当设计成标准化的事务处理数据库
+ 低估了数据清洗和质量保证的工作量
+ 过于关注ETL的操作性能和是否易于开发，没有关注BI查询性能和是否易于使用
+ 遗忘了在BI应用程序中为业务机构授权的需求 5
+ 没有认识到DW/BI项目的成功直接与用户的验收信息相关  

## 2.4 项目群管理
机构开始项目群开发时一般首先进行单个过程的开发，一旦第一个项目部署成功，需求将急速增长，需要进行单个项目之间的协调和管理  
项目群管理提升了对单个产品发布或项目成果的认识。当DW/BI环境在单个项目的基础上不断扩展时，整个项目群的开发就可以向那些离散的迭代过程吸收和学习  
除了决策人员中的DW/BI主管或项目群经理角色，设计师和数据管理员常常也在项目群层面工作  
### 2.4.1 确立管理职责和管理过程
项目群经理希望确立一个更为持久和明朗的管理结构，一般是行政层面的指导委员会或顾问团  
指导委员会由多个业务单位的代表组成，定期举行会议确定目标、业务优先级排序，高效的管理委员会有如下特征：5  
+ 目光长远的领导  
    需要对造成潜在影响的因素积极调查或者主动避让，对DW/BI的看法必须带着发展的眼光  
+ 基于企业层需求的优先级排序  
    必须坚持不断地激励自己所属的专门团队，也需要考虑到全局利益，组织年度规划过程以便确立主要的业务目标和目的  
+ 支持企业数据集成  
    必须关注实现一致的、集成的数据，必须清楚地了解集成对业务的好处，表明支持立场，因为会面临更多行政上的挑战  
+ 做好长期投资准备  
    支持和维护一个业务分析环境不便宜，需要为今后所能预见到的项目状况提供运行过程中的财政支持  
+ 确保IT机构和业务机构紧密配合  
    初始阶段IT和业务常常会有比较紧密的联系，但是这种关系很容易变得淡化，必须将业务接受程度作为项目群开发成功与否的一项重要工作标准 5  

在这样的行政管理结构下，常常会设立一个工作委员会来关注那些影响整个项目群的战术层面的问题  
### 2.4.2 将数据管理员的地位提升到企业层
管理企业数据资源的一种常用方法其实就是相互避让，数据管理和部署都由各个部门负责而并非在企业层  
由于每个部门的使用的定义和解释都会有些不同，因此没有哪个部门的数据能够和另一部门建立起关联关系，最终的结果就是一片混乱  
建立企业级的数据架构是项目群层所应该关注的重要问题之一，您需要使上述的混乱情况变得有序起来，需要有行政权力、财政支持和改变这种状态的意愿  
数据管理员应当为企业内部的所有交叉机构的信息确立通用的定义和业务转换规则，机构描述性主数据应当进行集中处理，分发给需要相关信息的项目，我们将在第6章详细介绍相关内容 5  
### 2.4.3 利用高效的方法和架构最优方法  
跨项目来说，项目管理层应当负责定义、记录和发起最优方法  
这些项目群层次上的资源被称作能力中心或优势中心，对DW/BI资产的利用起到了集中协调作用，并且推动了部门之间的协作以便实现系统的最优化  
这些资源有时就像一位内部顾问，借助它们可以利用已有的经验，避免很多重复性工作，有助于在确保一致性和高质量的前提下，缩短解决方案的提交时间  
这些资源的功能对于实现项目群层的优化十分重要，需要对这些资源的角色和功能进行明确的定义，避免混淆  
### 2.4.4 进行定期评估
如果管理层关心项目群资金，那定期评估就十分重要，可以确保DW/BI的交付能够满足业务需求 5  
鼓励项目群经理对DW/BI环境的健康程度进行评估，走出去和业务代表们进行沟通，调查过程中要对发现的问题进行标识并确定优先次序  
### 2.4.5 沟通、沟通、沟通
项目群管理层应当经常痛他人进行沟通或者为其他人员之间的沟通提供便利，无论是向上级推荐、对外宣传、向下级协调  
## 2.5 小结
DW/BI提案方面的定义会对下游的工作有深远的影响，最终影响到项目群的成功与否  
一旦确立DW/BI天，项目经理要关注人员安排、详细的项目任务以及时间安排   
由于交叉功能的项目团队需要针对高可见度的提案进行工作，因此需要加强在状态监督、文档编制、范围控制方面的要求  
赞助、管理和项目群层最优方法之间的协调对于最大化DW/BI系统投资的潜在业务回报来说至关重要  
## 2.6 管理工作和降低风险
首先评估您的机构是否准备好了，可以有效降低风险  
对业务赞助、问题集、人员的熟练程度和潜在的数据可行性进行客观的评价能够确保项目从一开始就稳步前进  
在和业务机构密切合作的基础上，随着开发的逐步进行，项目群进度一点点地推进，风险也在一点点地降低  
经常性的项目监督、主动沟通以及业务用户的参与都会提高向业务机构按时交付所需功能的可能性，每个生命周期方框都应当以业务用户对任务的验收情况作为结束  
## 2.7 质量保证
有成功记录并且经验丰富的人员是最理想的，如果团队没有经验，那么要准备好在重要决策点上花费一点财力对人员进行教育  
## 2.8 关键角色  
+ 项目经理和业务项目领导者驱动项目的定义活动，业务机构和IT机构的发起人也都要全身心参与
+ 项目规划也由项目经理和业务项目领导者进行指导，还需要其他所有团队成员积极参与其中
+ 项目运行管理的职责主要落在项目经理的肩上，同时也需要业务项目领导者协助
+ 现行项目群管理一般由DW/BI项目群主管或者项目群经理负责，其相关工作需要和行政管理指导委员会协力完成
+ 业务发起人不负责任何具体的项目任务，但是他们的参与极为重要，贯穿项目/项目群规划和管理的各项活动，特别是在业务优先级排序、问题解决方案的提出和机构文化转换等方面显得尤为重要  

## 2.9 关键提交内容
+ 项目/项目群范围和章程
+ 项目/项目群业务合理性证明
+ 项目/项目群沟通计划
+ 项目/项目群计划和任务清单
+ 项目/项目群状态会议日程和会议记录
+ 为业务发起人、行政管理委员会、IT人员和业务机构等做的项目/项目群进度介绍
+ 项目/项目群问题和变更控制日志

## 2.10 工作量估计  
确定DW/BI项目所需工作量很大程度上依赖于各项关注的准备就绪情况和具体环境  
如果找出了准备就绪事项中的重大缺陷，则需要几个月时间和精力来进行高层业务需求分析以及概念验证  
初步的项目计划应当在两周之内就设计完成，应当在具体执行的各个阶段不断地修订  
管理项目的工作量直接和项目的大小及范围相关，项目经理应当专职负责项目提案的确立，项目群管理永远在进行

# 3 收集业务需求
![3](3.png)
业务用户及其需求几乎影响到DW/BI系统设计和实施过程中的每一个决策，在DW/BI项目群中处于核心地位  
![3-1](3-1.png)  
业务需求分析在宏观层面上，需要从项目群角度来理解业务需求和优先权排序，可能涉及到整个企业或几个业务过程    
业务需求分析在微观层面上，需要针对某个单独的、范围相对较窄的项目来深入研究用户的需求和期望  
对于第一个DW/BI提案来说，需要做好两件事： 5  
1. 首先收集高层项目群需求来通盘考虑整个项目群的全景并为高级主管指明方向  
2. 随后针对特定项目需求进行细致而扎实的工作，在迭代过程中，深入到较低层次的项目需求上  

无论是项目群还是项目，业务需求收集的一般过程都是相似的  
以预准备阶段开始，随后同业务部门在交互式、面对面的会议上进行沟通，并在会后对讨论结果进行整理，形成下一阶段的工作文档和决心  
项目群和项目需求收集活动的不同指出在于您是在宏观层面进行操作还是在微观层面进行操作：包括参与人员、信息收集的广度和深度和由此产生的文档 5  
本章从宏观项目群层面出发，讲述了整个业务需求定义过程中涉及到的方法和技巧，随后回过头来进行项目需求的修正  
本章对于项目/项目群经理和业务分析员来说是必读内容

## 3.1 需求定义的各种方法  
了解业务需求，应当从业务用户的交谈入手，必须以用户的需求作为出发点，学会从业务用户的视角看问题  
除了询问用户希望数仓中有哪些数据，应谈谈他们的工作、目标和面临的挑战，了解他们现在在做什么、将来要做什么，如何进行决策  
从业务用户那里收集到需求后，应该与主要IT人员访谈过程中加入有关数据实际情况的内容，必须将业务需求和数据的可用性联系在一起考虑  
进行数据审计方面的访谈可以帮助您了解是否有能够支持业务请求的数据，以及数据是否是完整而可靠的  
### 3.1.1 个别访谈VS集体促谈会
同业务机构的干系人进行面对面的会谈对于准确理解他们的需求十分重要，交互式会谈可以帮助您建立比较明确的工作关系，极大地影响到他们对最终提交内容的验收态度 5  
有两种基本交互方法用来收集业务需求，个别访谈和促谈会：  
1. 个别访谈适用于个人或人数较少的小组的一种方法，只需要根据受访者日常安排确定相对较少的一段时间，受访者也能更多地参与到讨论中来，从中得到大量的细节数据  
2. 另一种方法是由某位发起人安排一个较大规模的小组会议，需要每位参与人员都投入较多时间，但总体来看缩减了收集信息所需的总时间量  
3. 还有另一种面对面沟通方法，可以让业务分析员深入到用户中去，让他们变成一个真正的业务用户  

建议您根据机构的专长、面临的挑战以及业务用户的影响因素选择恰当的方法 5  
上述两种方法可以混合使用，适用于绝大多数情况，个别访谈用于收集特定细节信息，而促谈会则用于在下步工作和业务优先级确定方面达成共识，首先采用个别访谈可以确保小组促谈会更有成效  
### 3.1.2 收集业务需求应避免使用的方法
有些机构试图使用其他方法收集业务需求，而不打算使用刚刚介绍的面对面沟通方法，接下来将对这些并不推荐使用的方法进行讨论  
发放调查问卷不是一种行之有效的需求收集方法，因为参与调查问卷的人员都是单个业务用户不是中高级管理人员，预先设计好的调查问题是静态的无法深究  
项目小组有时会翻出所有已有的业务报表样例，虽然简单快捷但这样收集是错的  
尽管分析已有的业务报表有一定价值，但如果没有进行业务层面上的面对面沟通，直接通过报表去理解业务中的是什么和为什么往往会产生误解 5  
您要建立的分析环境是针对当前业务问题的，同时也应包括对今后业务发展的一些考虑，不能仅徘徊在当前的报表功能当中  
另一种可能使用的需求收集方法是向业务部门方法一个数据元素清单，然后接受咨询的业务用户通常都只会说:我需要所有元素，这种以数据为中心的方法对建模人员比较方便，但对业务需求收集来说，会令您对业务部门想要实现什么一无所知  
## 3.2 访谈准备
在收集业务需求之前需要做一些必要的计划和安排，有助于您尽可能地消除未知因素  
### 3.2.1 确定访谈小组  
进行业务用户需求访谈和参加IT数据审计会议的应该是同一个小组，这个小组应当由下列成员所组成：
1. 主要访谈人  
    职责是把握提问的导向，通常是业务分析员，必须具备扎实的业务知识和良好的口头沟通能力，对业务有一定的兴趣和好奇心  
    这个角色时一个要求全面的角色，需求分析过程取决于主要访谈人和受访者之间能否进行真诚而自由的沟通
    如果打算雇佣一个咨询机构，那么要确保自己的项目人员参加到这一过程，以便实现这项技能向团队的迁移并且保证项目的连续性 5  
2. 记录员  
    主要职责是访谈过程中记录访谈内容，尽可能捕捉会谈的细节内容，这些记录可以作为整个DW/BI团队的基础文档，因为团队的多数成员并不熟悉访谈的主题  
    记录员还要搞清楚主要访谈人和受访者之间是否存在无法沟通或者误解的情况，波卓到主要访谈人所有遗漏的任何要点  
    应当能够意识到并且捕捉到繁琐的细节内容，必须要有过硬的书面表达能力  
    一般不鼓励使用录音机来代替记录员，用户会因为自己的言辞被录音而感到不舒服，多人情况下可以咨询其他与会人员  
    可以将录音机作为记录员的补充设备，这取决于记录员的工作经验和笔录的自如程度，而且应当在会谈介绍过程中告诉受访者该录音带仅供访谈小组成员使用 5  
    一些记录员使用便携式电脑来直接捕捉会谈内容并记录成电子形式，但有些受访者会使用白板来画出他们这个机构或业务过程的工作流程图，在纸上抄写这项描述业务演讲过程的图表要比使用电子方式方便得多  
3. 观察员  
    是可选得组成部分，数据仓库团队中的其他成员，数据建模人员、数据库管理员、BI应用程序开发人员或者参与工具评估的技术支持专家会列席这一的访谈，一次访谈中不应当有两个以上的观察员  

### 3.2.2 研究业务机构
在开始需求收集过程之前，做一些准备工作十分重要  
在企业/项目群层面收集需求时，从阅读公司的年度报告开始着手，找到有关公司战略动机方面的线索，应当关注有关业务运营方面的综述信息  
如果有可能，尽量拿到最新的内部业务/IT战略与规划会议的存档文件副本 5  
访谈小组还应当上网查询公司的站点，以便了解公司发布的公共讯息，也可以借助网络了解主要竞争对手的情况  
在需求收集会谈做准备的过程中，应该尽可能了解业务知识、顾客和市场竞争形势，而不是直接从业务人员那里了解  
忘记过去的人注定会重蹈覆辙，数仓方面的早期产品并不叫作数据仓库或商业智能，它可能被称作是营销信息系统或者是面向业务的报表和分析系统  
最好带着极大的热情去弄明白这项项目当中哪些是成功的，哪些是失败的，以及具体的原因，以及哪些人参与了这项早期的项目  
### 3.2.3 选择受访者
需要选择来自多个业务部门的代表作为受访者，您需要了解一下那些人比较有影响，哪些人比较有远见，哪些人是部门的主导者和推动者，哪些人是支持者，哪些人是反对者 5  
1. 业务受访者  
    在企业项目群层面上，显然需要收集横跨整个机构所有部门的需求，帮助团队从宏观层面理解机构的需求，确立一个全局的DW/BI项目群路线图  
    理解关键绩效业务过程和公共的描述性信息可以使您确立一个逐步展开并最终完成项目群的蓝图  
    企业数据仓库总线架构和相关矩阵可以用作包括架构规划、数据集成协调和组织沟通  
    在项目群需求收集中，我们几乎会同高级管理层上所有人相遇  
    除了主管人员，最好也对中层管理人员稍加了解，获得更深入的需求描述和更详细的内容十分有益 5  
    对中层管理人员的访谈有助于在设想中加入更加现实的成分，他们懂得如何将高层战略转化为具体业务策略  
    考虑项目群需求的时候，不需要花费过多的时间同单个分析员打交道，只需要同那些最资深优秀的交谈即可  
2. IT受访者的合规性/安全性受访者  
    对数据审计人员的访谈，由于使进行较为宽泛的项目群层需求收集，设法评估使用可用业务源系统支持业务需求的可行性即可  
    此外我们还经常在访谈中会见高级IT管理人员，IT管理层通常能够考虑到如何确保竞争优势，并确保DW/BI和总体方向及战略保持一致  
    因为项目群需求会影响到DW/BI的全局战略，因此在适当的时候与负责系统合规性和安全性的高级职员沟通也很有帮助 5  

### 3.2.4 设计访谈问卷 
进行访谈之前，主要访谈人必须设计一份访谈问卷，另外访谈过程中的提问会随着工作的职能和层次不同有所改变  
例如，与IT方面的领导们坐下来交谈之前，主要访谈人应当先列出一个系统的、面向数据的问题清单  
访谈问卷应当按照既定的访谈流程来组织内容，而且问卷最好设计为单页形式，应当将调查问卷作为一种应变的手段，不要完全按照上面的内容来进行提问

### 3.2.5 确定访谈时间表
在安排访谈时间表时，建议找一位行政助理来协助工作，可能需要提前两三周就和某位受访者约定访谈时间  5
1. 确定访谈顺序  
    需求定义过程可以从与业务发起人的会谈开始，最佳情况下，您应当先同业务驱动者进行访谈，随后再对业务发起人进行访谈  
    先从机构组织结构的中间层进行访谈，而不是从高层或底层开始  
    从底层开始访谈是破坏性的，因为会掉入细节陷阱，从高层入手适合理解项目群的整体设想，但需要一定的业务知识背景、信心、把握  
    项目群层面上收集需求一般需要同时访谈多个机构代表，应当将访谈活动安排得分散一些，可以做一些交叉沟通  
    因为就同一个主题，经常会出现一个群体的看法和另一群体的看法差异很大的情况，一旦随后出现需求矛盾，就很难再回头重新找这些人商量解决办法了 5  
2. 确定访谈的时间和地点  
    您可以安排对行政主管人员进行私人性质的访谈，也可以考虑组成一个两到三人的访谈小组对他们进行单独访谈  
    一般再业务用户自己的处所进行需求访谈效果最好，这样将使访谈被延期的可能性最小化，而且并于从业务用户那取一些证实问卷或者单据  
    访谈应当安排在受访者的办公室里或者受访者所在部门的会议室，避免他们原先准备要使用的关键材料或许有没有准备好  
    确保访谈地点足够容纳访谈小组、受访者、包括观察员，没有人愿意像沙丁鱼那样挤到一块  
    对非行政主管业务用户可以使单独访谈也可以是小组访谈，如果是和一组人会谈，那么这个受访小组的人员构成应当是同质的，应当由相同功能的部门或岗位上的人员充当受访小组的成员 5    
    进行小组访谈时，一次访谈的人数最好不要超过四个，他们所代表的组织结构层次不应当超过两个，另外和老板的老板参加同一个会议足可以压制大多数业务用户发言的热情  
    访谈的执行过程一般需要三十分钟，我们建议您安排一个小时，因为主管人员经常会迟到，单个受访者最少安排一个小时，受访小组安排一个半小时  
    两次访谈之间至少留出三十分钟时间，利用这段间隙听取汇报，补全会议记录  
    业务需求和数据审计方面的访谈经常会令人疲惫不堪，不应在一天内安排超过三到四场的访谈会  
    尽量将访谈安排在一个较短的时间段内，记住应该为会谈内容的吸收、分析和存档预留足够的时间 5  
    访谈时一条至关重要的途径，应该在一个合理的时间段内安排并完成所有的访谈，及早进行DW/BI实现过程的下一阶段工作  

### 3.2.6 通知受访者做好准备
在开始安排访谈之前，向受访者适当简述访谈事项以便他为参加访谈做些准备，这种做法是十分重要的  
至少应当以电子邮件形式给所有参与访谈的人员发一封信函，告诉他们访谈会的过程以及他们的参与和贡献对项目群的重要性  
这封介绍基本情况的邮件（参见图3-3中示例）应当时受访者了解访谈过程中的基本事项  
![3-3](3-3.png) 5  
应当向受访者解释，访谈的目标是为了了解他们的工作职责和业务目标，访谈的成果随后将转换成实现他们这一业务流程的信息和分析报告  
应该请求受访者带一份常规报表的副本或者电子表格形式的分析报表，提醒他们会谈的开始时间、整个会谈持续多长以及会谈的地点  
这种信函应当由高级别的发起人签名，可以是指导委员会的发起人或CIO，应当以受访者所尊敬的人的名义发送  
如果聘请了外部的咨询机构参加需求定义过程，则必须确保情况介绍信是由一位高层的内部人员签名的  
不要在信的后面附上50个想要咨询的问题，他们几乎必然不会为您给出的问题清单准备答案，有些受访者还可能因为您给出的问题难以解答而退出 5  
### 3.2.7 访谈中的基本规则综述
1. 牢记自己在会谈中的角色  
    第一条基本规则就是要遵守自己在访谈小组中所承担的角色和职责，主要访谈人询问各种不偏离正题的问题，记录员应当记录  
    主要访谈人应该问一些能够自由回答的问题，不要对用户就当前系统提出的批评进行辩解，不要相互争论冷落受访人  
    访谈人应当端坐一旁，而不随意发言，需要很强的聆听别人的技能，受访者一般都指导您要在需求收集过程中寻找哪些信息  
2. 让自己显得求知若渴  
    在进行访谈前，确保自己由足够好地心态，不能装作自己对什么都明白了  
    太复杂的问题会导致受访者难以理解，问一些简单明了地问题，更有利于理解复杂的概念 5  
3. 核实交谈内容  
    一定要确保自己真正理解了受访者所谈到的内容，将内容用自己的话进行阐述是保证理解访谈内容的一种方法  
    在访谈过程中提出自己的疑惑也很重要，应当有勇气承认自己的困惑，并在会谈中消除自己的错误认识  
    另一个比较重要的问题是不要想当然地用词，同一个单词会有多个不同的含义，而不同地单词则可能具有相同的含义  
    应当在访谈过程中将口径不一致的问题标记和注释出来，但是不要试图在访谈过程中解决这项问题，一位定义公共词汇还有更多的工作要做  
4. 健谈一些  
    对高层主管人员进行访谈会感到很有压力，访谈小组应当试着忽略行政级别上的差别，就当自己同受访者是平等的，要充分展现自己的信心，也不要以高高在上的态度同分析人员沟通 5  
5. 灵活安排访谈日程  
    在过程中，受访者人数一般会增加10%-20%，应将他们看作是意外的收获  
    记住对业务用户进行访谈非常消耗精力，一天之内进行三四次访谈那，访谈人可能会疲惫不堪，要灵活安排访谈日程  
    在访谈过程中，如果开始听到明显重复的内容，应该明白访谈过程差不多要结束了  
6. 持续管理期望  
    最后一条基本规则是在访谈过程中管理用户的期望，不能对那些无法提交的功能进行吹嘘，应该告诉业务受访者DW/BI系统是不断发展的  
    需要为他们设定这一的期望：数据仓库项目群的设计和开发需要一个迭代过程，在DW/BI环境中需求收集和功能发布是反复迭代进行的过程，即使到了下一阶段，也还可以回过头来收集更多的需求 5  

## 3.3 进行访谈
每次访谈开始时，都要向用户简短地介绍自己以及为什么要来参加访谈  
+ 可以由项目经理或者业务项目领导人介绍高层次的项目信息，由谁负责介绍应当预先就安排好  
+ 还应该对项目的目标和访谈的目的进行简要叙述，在访谈过程中介绍出席的团队成员和他们在访谈过程中的角色也比较恰当  

应当在项目介绍中谈谈整个需求过程，让受访者知道您会对访谈内容进行记录，并且会将记录的内容提供给受访者以供他们审查和提出反馈意见  
访谈前两分钟至关重要，可以奠定整个会谈的基调，建议您在会谈的这段时间里使用事先组织好的笔记，并且预先排练一下  
谈话过程中使用受访者的词汇，而且应当告诉他这并不是一次技术性的会谈 5  
应当明确地告诉受访者您对他们的尊敬，充分准备、准时到场、手机静音  
介绍完毕之后，主要访谈人应当立即进行实质性的会谈，最初的问题主要集中在用户的工作以及他们适合完成的工作，应该非常简单不需要让受访者进行深入的思考  
对于项目群需求分析应当多提一些顶层的问题，而不要深入细枝末节，这个层次的提问对主管来说比较容易回答，然而对业务部门的分析员进行访谈时就要将他们引入更为细节的问题  
一旦初始项目或项目群的重点区域已经确定下来，就应该回过头来同中层管理和单个分析员一起进行详细的项目需求分析  
需要像剥洋葱那样，先提出高层次的问题，随后按照受访者的引导来探究其中的细节情况，能够按照受访者的引导进行思考并且随后提出恰当的问题是区别平庸访谈人和优秀访谈人的一项指标 5  
不要期望在访谈过程中提出所有已经准备好的问题，也不要按照某种特定的顺序来提问，应当根据访谈流程灵活安排  
主要访谈人应当主要负责确保访谈的导向，不能打乱仗，小组成员不能随意地发言  
### 3.3.1 项目群层面的业务访谈  
该访谈的目的是为了对机构的情况有总体的了解，您可能会向某位主管提出如下这些较高层次上的问题：  
_您这个机构的工作目标是什么？您准备尽力去完成哪些工作？_  
通过以上问题可以了解到业务部门正在监控并想要施加影响的关键业务提案和潜在业务过程 5  
_对工作成功与否如何衡量？如何知道您已经取得了比较好的绩效？您每隔多长时间对自己的工作评估一次？_  
了解机构工作成功与否的标准极为重要，因为这是在DW/BI系统中必须要追踪到的核心数据要素,每隔多长时间对自己的工作评估一次这一问题的回答中隐含了数据等待时间和数据仓库的更新策略  
_您当前面临的主要业务问题是什么？现在用什么方法来解决？它们对机构的影响有哪些？_  
这些问题类似于“是什么让您在晚上难以入睡”这一的典型问题  
_您如何识别问题或者说您如何知道自己可能已经遇上麻烦，如何发现业务流程中出现了意外情况？改进对信息的访问会对您的业务产生哪些较大的影响？财务方面的影响有哪些？如果您能够...将对业务有什么意义？_ 5  
这些问题对DW/BI提案提供财务方面的合理性证明更是如此  
_您认为如何才能更好地利用本机构的信息？您是否期望自己的员工可以同信息直接交互?_  
通过这些问题，您可以对他们的想法有所了解，借此评估出他们是否具有足够的行政影响力来促进和推动基于事实的决策制定，还可以了解受访者对于员工使用数据的期望程度  
### 3.3.2 项目群层面上的IT访谈  
项目群数据审计访谈的目标是为了评估已有的数据是否能够支持从业务访谈中了解到的那些事项  
这些IT访谈会是为了提供事实验证，应当将这些数据审计会谈看作刨开表层的抓手，因为您需要借此来评估各种业务过程的可行性 5  
同时您也是在寻找有关可用的关键源系统的一个综合评述，包括更新频率、历史数据可用性、已有数据的情况、限制条件和已知的质量问题  
### 3.3.3 项目群合规性/安全性访谈
了解机构中是否存在安全、隐私和合规性方面的问题会对DW/BI项目群产生影响也是十分重要的，第一步是要了解机构中是否有专人负责安全、隐私和合规性方面的问题    
机构中有什么明确的指导方针、标准或是规章制度需要了解吗？应认真询问主管合规的职员，数据仓库向干系人提供的收益率和报表是否也是符合合规性要求的  
也可以问他数据保留需求方面的情况，即数据一般保留多长时间，以及是否由于法律或知识产权方面的问题而必须将某些特定种类的数据定期清楚  
也需要确认一下是否可以找到一位公司主管，使其在进行有关数据安全性、保密性和合规性方面的决策时提供一定的支持和保护 5  
## 3.4 总结访谈  
访谈的最后异步就是如何结束访谈了，主要访谈人应当看好时间，至少要在会谈结束之前留够5分钟来为访谈做有条理的总结  
### 3.4.1 确定项目群成功的标准  
现在可以谈一谈项目群成功的标准了，现在就是更好地了解受访者对项目群的态度和期望的最好时机  
主要访谈人应当问:_要使项目群取得成功必须首先完成哪些工作?_  
应该尽量让受访者说出可度量的、可量化的项目群成功标准,“使用方便”和“处理速度快”这两点还不够  
如果用户的成功标准不切合实际，最好早点了解他们的想法，这样您就有反应的时机并且还可以重新设定他们的期望 5  
在项目群层面上，成功度量一般都是高层因素，如查询数量或活动用户数，而非一些更为详细的运行指标（如数据库/服务器故障时间等）  
如果高级管理人员成功标准包括每个人使用一样的数据或单一版本的信息，或者提到“集成的数据”这样的内容，这时需要进行更为深入的研究，因为数据集成是机构所面临的一项挑战  
业务影响指标包括与节约成本或增加收入相关的财务影响，这些财务影响一般是DW/BI系统中最重要的成功度量  
最后要确立成功的度量可能会集中关注当前DW/BI项目群和早期项目群在基本性能上的对比，这样的例子可以在性能上形成强力的对比，往往非常有用而且会给人留下深刻的印象  
### 3.4.2 致谢并告辞
在对访谈进行总结之后，应当对受访者能够参与访谈并且提供有价值的见解表示感谢，最好征得受访者同意，提出保持联系的请求，以便进一步弄清楚会谈中的要点 5  
应该告诉受访者以后还会提交给他哪些材料并且告诉他反馈意见的方式  
告诉他们在某个日期他们将会收到一份存档的访谈记录，并且需要他们审查和评阅之后在另一个日期将意见反馈回来  
按照约定的日期行事十分重要，这样在与用户的每一次接触过程中都会增强相互间的信任  
在每一次访谈结束时都应当拒绝承诺一些事项，以此来提醒受访者在第一个提交版本中委托的事项并非是一成不变的，还要告诉受访者确定业务需求优先级的过程和大体时间表    
## 3.5 审查访谈结果  
团队应当在每次访谈之后立刻正式地互相听取汇报，确保每个成员都对一般性地、常见的业务需求达成一致很有好处  
还应当标识出那些存在误解的地方，可以在后续的访谈中搞清楚这些问题，可以保留一封尚未解决问题的动态清单    
如果团队成员具有源工作系统的工作经验，应当将自己对已知需求的可行性的看法提出来与大家分享  
每次访谈之后都要尽快审查访谈记录，并且依据自己的喜好进行注释  
访谈记录中不可避免会有一些摇摆不定的想法或者不完整的语句，如果在访谈结束后推迟一整天才进行上述活动，那么访谈记录中很多内容就可能会变得无法理解 5  
__围绕业务过程进行合成__
为了对了解到的内容进行合成，可以试着围绕机构的核心业务过程来对需求结果进行聚类  
业务过程是业务执行中最底层的事件或活动，在确定业务过程应当注意以下几个常见特征：
+ 业务过程常常用行为动词来表达，代表业务要执行的各项活动
+ 业务过程一般由工作系统或模块支持，如订单系统
+ 业务过程生成或者捕捉关键性能度量，有时度量是业务过程的直接结果，而有时度量则是这些直接结果的派生 5  
+ 注明业务过程"不是什么"也很有用，不是业务部门，不是机构，也不是功能，同样也不是报表或分析结果  

业务用户常常用战略业务提案来替代业务过程，一般是企业的运营方案，DW/BI天必须要同战略业务提案一致，以确保项目群/项目能够为业务社区提供有价值或有参考意义的信息  
将业务提案同业务过程联系起来是高层主管比较关注的问题，业务过程对DW/BI团队来说描述了一个项目大小的单元，需要将业务提案分解到底层的业务过程中去，这意味着要进行深入一些的研究  
单独的业务过程一般和一个生命周期项目的迭代相对应，这一过程的基本标准就是：业务机构和IT机构就关键业务过程达成共识（性能指标、业务提案和各自的分析方法）的速度越快越好  
审查需求会议记录是为了确定关键的业务过程，为综合已有结果提供一个有机地框架，可以用电子表格来对访谈结果分类或分组 5  
## 3.6 准备和发布项目群需求文档  
强烈您建议从需求分析过程就开始编制正式文档  
之所以必须将收集到的信息记录下来，主要基于以下几方面原因：
+ 这些文档会为新成员提供比较全面的参考资料
+ 如果有人退出团队却没有将信息留档，项目就会损失重要信息，此时由于团队人员缩减，您不可能再派替补人员重新对用户进行专访  
+ 文档会帮助项目群团队将访谈内容具体化，进而深入了解访谈的内容 5  

### 3.6.1 访谈书面说明
对每次访谈的书面说明要在访谈会结束后尽快进行整理，应当合理地组织书面说明的内容，要让没有参加访谈的人能够看懂  
一般应当按照下面几个部分来组织书面说明的内容： 
+ 受访者的工作职责以及业务和目标方面的背景知识  
+ 主要业务过程和相关性能度量，包括下面的细节内容：
    + 对使用这种由业务过程产生的度量所需的条件进行叙述性的描述
    + 当前的分析障碍
    + 访谈中使用的问题示例 5
    + 必需的专用数据或分析
    + 潜在的业务影响
+ 初步的成功标准  

该文档以供他人使用之前要经过受访者查验和签名 2

+ 主管人员概述

    + 对使用业务过程产生的度量所需的条件进行叙述性的描述，包括潜在的业务阻碍因素和当前的分析障碍,如图3-4所示

+ 初步的成功标准 5

业务需求调查结果文档非常重要，它确立了DW/BI项目的实用性和可信性，它将业务需求和实际数据的可用性联系起来  
大多数需求只会和单一业务过程有关,但遇到需要来自己多个业务过程的需求也不要感到惊讶，仅在以过程为中心的核心模块交付之后作为次要内容予以跟踪，这些依赖关系也应当标记为影响可行性的警示因素

## 3.7 区分业务优先次序和商定下步工作
此时您已经知道用户想要什么，以及根据这些需求应当提交什么样的系统，下一阶段理应重新审查需求分析结果，进而在范围和优先级别上达成共识  
### 3.7.1 以优先级的审查和确定结束会议
我们喜欢使用优先级区分表来划分整个项目群中单个项目的优先级  
审查和优先级确定会议首先介绍业务需求结果，除了综合好的调查结果文档之外，还应当借助幻灯片来介绍业务需求提案的基本情况，增加需求结果的可信度  
每个业务需求都用一两张幻灯片介绍，简述该业务过程涉及到的事项、核心度量、已经意识到的业务价值、对主要提案的支持情况以及相关的可行性评估  
这常常会使高级管理人员意识到机构内部的公共信息需求和现有各部门独立的解决方案所造成的机构臃肿和资源浪费 5  
随后描述一下企业各个部门的分析数据集成后的远景，而不是给出一个复杂的代表不同数据源的框图  
一旦所有人对机构的项目群需求有了共识，就可以拿出图3-7所示的2X2优先级区分表进行研究了  
![3-7](3-7.png)  
但我们应该暂时放下左上角这些风险较大的项目，集中解决最优区域里的项目，并让那些非DW/BI项目群的人员进一步研究和解决左上角区域中各个项目的可行性，另外不能一开始就启动位于底部区域的项目 5  
可以进一步改进这个2x2的优先级区分表格，使用不同大小的圆圈来分布标识每个项目所需工作量的相对大小  
审查与优先级区分会议是结束想需求定义的一种极好的方法，因为在这个会议中业务人员和DW/BI团队都会得到很多收获   
### 3.7.2 结束本轮访谈
如果业务主管和DW/BI团队已经为短期路线图确定了优先级，就需要将这些额外的信息补充到需求调查结果文档  
要安排一次会议，同其他的受访者一同审查调查结果，并就下一阶段的有关事项进行商谈  
接下来就需要回到生命周期第一个步骤——项目规划了，更新范围文档并继续进行第二轮访谈，针对指定起始项目，进一步深入了解分析性需求和数据的实际情况  
## 3.8 项目层需求的调整
这一节，将较为深入地研究项目需求，并且讨论项目群需求和项目需求的不同之处  
假设项目群需求涵盖了企业的各个方面，且机构已经同意开始进行第一个基于单个业务过程的项目  
### 3.8.1 走进项目层  
不论是对项目群层面还是对项目层面上的需求来说，本章一开始在需求收集方面的绝大多数建议都是适用的  
访谈会和促谈会都是非常好的需求收集方法，这两种方法有很多共同的核心要素，可以从下面的促谈会任务清单看出： 
+ 确定促谈小组和角色，促谈人应该很得力，确保会议不偏离正题 5  
+ 为会谈小组成员准备背景信息和会前情况介绍，在促谈会之前应当对业务和相关术语有比较扎实的了解  
+ 确定参加讨论会的人员
+ 确定讨论会的时间安排并且协调后勤保障
+ 开发讨论会材料
+ 讨论会的发起人要为每个参加讨论会的人员发送一份会前介绍邮件，简要说明项目目标和讨论会的目的，并且说明与会人员，以便管理参与人员的期望 5  
+ 主持讨论会，以精心组织的情况介绍开场，随后促谈人依据自己的风格、技巧和经验来推动与会人员的相互沟通，最后对会谈进行总结
+ 将讨论结果存档
+ 向每位与会人员分发一份讨论结果供其审查

### 3.8.2 为项目需求访谈做准备
项目层需求收集的准备工作与项目群层很类似，不过项目层访谈预先准备工作的范围要窄一些，仅关注某个特定的项目  
1. 选择受访者  
    在项目需求选择受访者时，需要在中层管理和单个职员身上花费更多的世界，主管人员可以少花些时间 5   
    由主要业务过程产生的性能度量会用于机构的多个部门，因此可能需要同来自市场营销、物流、客户服务以及财务部门的代表进行会谈  
    对于项目层面上的需求分析，建议纵向选择代表参加访谈，如果项目需求活动在企业项目群需求活动之后紧接着进行，就不需要再对主管进行回访，而如果项目群需求活动结束之后6个月以上才进行项目需求，就有必要做一次回访了  
    应当花大量的时间同中层管理人员沟通，因为他们既了解项目的总体规划又知道实际具备的条件，还必须花一定时间和最优秀、最聪明的分析员沟通，还应该横向选择一些不同部门的业务分析员代表   
    在IT数据审计方面，应该同负责核心工作系统的人员以及主要的数据库管理员和数据模型师进行沟通，审查已有的工作系统文档是访谈小组极为重要的一项准备工作  
2. 请受访者做好准备  
    对于项目层需求，召开一个用户启动会来向所有的受访者发布一致的讯息是十分可行的 5  
    用户启动会议日程安排示例如图3-8所示
    ![3-8](3-8.png)  
    这个会议是管理业务用户期望的重要工具，每位与会人员都应当获得一份有关初步项目范围和大体时间表的通告  
    可以告诉与会人员在本项目中访谈、需求验证、在每个检查点对数据模型或应用程序说明进行审查、培训等相关事项所需要的用户时间总量  
    还有个间接受益就是使项目成为业务用户和IT部门的联合项目，而不仅仅是一个技术驱动的IT项目  
    启动会不能代替图3-3所示的访前信函，这封信是重申项目总体目标的又一次机会 5  

### 3.8.3 进行访谈
收集项目需求和收集项目群需求所遵循的基本规则是相同的，需要为这些会谈做充分的准备  
对于比较详细的项目需求应当注意一点：不要进入得太深入和太迅速，建议从高层问题开始，逐步深入细节，而不是一开始就直接进入细节  
访谈会流程与早先讲述的很类似，在开场介绍之后，首先要询问业务部门的目标，随后要深入探讨项目中特别需要关注的问题  
_请描述一下您的产品，您是如何区分产品的，有没有进行产品分类的常用方法？这些主要的产品类目多久变更一次？如果有成千上万种产品，如何缩减这份清单以便找到想要的产品？_  
用户对这些问题的解答会告诉您它们是如何描述自己业务的主要方面的，数据仓库必须支持对这些描述性特征的剖析 5  
访谈小组应当尽量了解主要属性之间的关系，例如，“每个账户都仅对应一个经纪人吗？”如果问题的答案是不太常见，应当用红色记号标记出来  
在这些高层的业务目标和综述性问题之后，一般要通过下列问题转移到分析的需求上来：
*您目前都做哪些类型的分析? 使用哪些数据? 您目前是怎样获取那些数据的? 一旦得到了这些信息之后您会怎么做？目前使用哪种报告？ 报告中哪种数据比较重要？您如何使用信息？*  
当前标准报告上有很多数据从来都没有被人们留意过，应当尽量明确提出可能对当前的报告做哪些改进  
*您喜欢哪些分析方法？当前的方法/过程中有哪些可能的改进之处？* 5  
*您做过哪些即时分析?是谁请求这些特定的分析?他们用分析结果做什么？一般这样的分析要用多长时间？*
*您需要多少历史信息？这些数据需要多长时间更新一次才能反映出运营过程中捕捉到的新信息？这些数据要保留多长时间？在安全性、合规性或者审计方面有什么特别的需求？*
*改进信息访问之后会显著改进业务带来哪些机遇？有哪些财务方面的影响？如果您具有了刚才所说的功能，将会对业务有什么影响？*
不管当前项目是不是关注量化的业务影响，都应当向受访者提出这些关键问题  
1. 审阅已有的报告和分析结果  
    有时对业务分析员来说，描述已有的关键报告和分析结果比回答自由式的提问更为简单，仅收集他们当前所使用的标准报表和电子表格的复制品是不够的，还应该询问他们目前是如何使用这些文档的 5   
    不要忘记在每次访谈结束时请求受访者提供关键报表，记住要在这些报表上注明提供者的姓名以及收到报表的日期  
    当分析他们的当前报表和分析结果时，应该记住您要设计的是一个具有分析功能的DW/BI环境，而不仅时一个业务报表系统  
2. 完成访谈  
    在访谈中投入的精力越多，对有关项目成功的标准定义得也越细致，一般要对服务层的度量进行较多的讨论，例如：
    1. 基于数据库和BI服务器故障时间定义的可用性  
    2. 基于每百万次质量检查中的错误数定义的数据质量(例如，完备性和遵守转换业务规则) 5
    3. 基于从业务数据捕捉开始到数据在数据仓库中达到可用状态所需的时间总量定义的数据生效周期
    4. 基于某个查询和应用程序标准集合的平均响应时间定义的数据仓库响应能力
    5. 基于服务请求或者处理服务请求的平均响应时间定义的支持响应能力
    
   尽管访谈会理论上应当关注某个单独的业务过程，但是仍然不能保证在项目第一次发布时就提交会谈中所讨论的所有事项，会谈结束时一般要有一个否认声明（否认对某些事项富有责任）

### 3.8.4 深入调查数据
一旦把精力集中于单个业务过程，就应当对基本源数据进行深入研究 5  
记住您仍然时在做收集需求的工作，只不过对基本的源数据有比较可靠的了解会使您更为实际地管理用户期望  
现在是数据探查过程在生命周期中第一次出现，在维度建模和ETL设计活动中会进行更为详细的数据研究  
不需要了解有关数据的每一个细节，但必须要搞清楚数据是否可以满足需求，这就需要探查比开始设想的更为深入   
最好先假设数据源的质量不合格，即使这样会在很大程度上让您的业务发起人和用户感到失望，也比ETL开发过程中真的出现数据质量问题要好得多  
### 3.8.5 审查访谈结果
在访谈结束后马上对记录进行审查和注释依然很有益处，在页面空白处用标记注释数据、BI报告/分析性需求或者技术架构需求，以及它们会造成哪些影响 5  
需要审查从访谈会中收集到的关键报表和分析结果，因为从中可以得到维度模型建模所需的大部分要素，如图3-9所示  
![3-9](3-9.png)  
在离线状态下审阅那些收集到的报表常常对更深入地了解业务维和用户分析性需求很有帮助，有时可以在这些报告中找到访谈过程中没有提到的维度和属性  
### 3.8.6 准备和发布项目提交材料
再次建议总结个别访谈并且整理调查结果文档，这样可以使公布的内容更加详细  
项目层综合和整理文档与项目群层在文档格式上会有所不同，这是因为项目需求访谈关注的范围更窄，更为深入 5    
文档以标准的执行机构概述和项目/过程概述开始，转到调查结果上来，一般涉及三个方面：数据模型、BI报表/分析和技术架构  
### 3.8.7 协商下一步工作并结束本轮访谈
将项目需求定义工作整理成条理清晰的结论后，常常只要需要DW/BI项目经理讨论并予以签发，确保每个人对项目需求的认识都是一致的  
## 3.9 应对富有挑战性的受访者
访谈的过程中都可能会遇到一些常见的阻碍，我们将从下面几个部分阐述这些挑战以及应对方法  
### 3.9.1 受过打击的用户
这种类型的业务代表一般很不合作，曾接受过多次访谈，但是他们并有看到自己的努力能带来什么结果，因此可能不愿意再和访谈小组进行会谈  
对此应该提前了解哪些人曾经参与过早期的DW/BI方面的项目，并且应该审阅早期项目的需求文档  
最好先肯定他们对早期项目所做的贡献，让他知道您已经审阅了以前的文档，访谈中应当提出看法来请他确认，而不是像以前那样从头开始访谈  
这些用户仍然会坚持自己一起所提出的种种事项，但如果访谈人员把关注点放在如何确定当前这些需求的优先级上，那么他们就会很乐意沟通 5  
最后，这也可能是召开一次需求收集讨论会的好机会，最好使用那些早先整理的需求文档作为基础召开一个促谈会，重点收集那些在业务过程中发生改变的细节内容   
### 3.9.2 超负荷的用户/替换用户
这些散漫的业务用户经常会因为工作太忙而无法在预定时间接受访谈，他们会统一安排好的访谈时间，但随后又因为自己有事要求更改时间    
应当由项目群发起人给所有的受访者发邮件，告诉每个人他们对项目群的重要性，避免发生日程安排上的混乱  
如果这种情况像传染病一样扩散传播，致使访谈时间一改再改，建议先停下来，暂时也不要投入太多的精力  
如果一个业务用户没有时间接受访谈和告诉访谈小组他们的需求，你们也就没有时间参加培训并且将新的信息和分析结果整合到他们的日常工作中去 5  
建议在造成任何损失之前先暂时停止工作，可以在机构内部再找以为更愿意合作的业务伙伴  
### 3.9.3 昏昏欲睡的用户
这种类型的业务用户往往对于您提出的典型的、自由回答的问题仅用一两个字草草回答了事  
有时从否定的角度提一些问题会比较有效，有时让他们说出自己对整个业务过程的想法很难，但让他们说出已经收集的业务需求中有哪些错误却比较容易  
建议您先试着同他们沟通，如果仍然一无所获就停止对他们的访谈，重新找个可以代替他的业务代表进行访谈  
### 3.9.4 过分热心的用户
过分热心的用户又走了另一个极端，比如约了2个业务用户，最终却到场了7个，且都嚷嚷着说希望DW/BI团队能够倾听他们的需求 5  
用户热心和投入当然是极好的，但如果试图花一个小时同时访谈七个人就不是一件好事了  
建议您对这群人的成分做一个迅速的评估：他们是在做同样的工作并且每个人的想法都一样吗？他们的工作属于不同类型或具有不同的功能吗？  
通常都是后一种情况，应当把他们分成几个小组，为他们安排一定的世界分别进行访谈，确保时间得以充分利用，收集到您所需要的细节内容  
### 3.9.5 自以为无所不知的用户
这种类型的用户一般都是IT人员或者最终的实际用户，会自认为是把关的好手，常常认为不需要再去麻烦更多的业务用户来参与需求收集   
有时，那些自以为什么都知道的用户确实知道很多，但是有时他们的看法却是有失偏颇的 5  
建议认真倾听这些用户，甚至要重视他们的作用，但不要过于依赖他们  
要注意这些用户有时还想客串IT人员的角色，除了对项目群团队与其他业务部门的访谈造成一定的限制之外，还想履行IT设计人员的职责，为他们所提倡的系统解决方案指定数据布局  
### 3.9.6 一窍不通的用户
您是否碰到过并不能为您提供什么信息的用户？您是否觉得同他们进行会谈是一件没有意义的事情，因为他们没有需求？  
99.9%的情况下，一窍不通的用户都是IT职业人员自己主观臆断的，他们通常都能够清洗地描述自己在做什么、为什么这么做以及以后还想做什么  
向用户提正确的问题对获得相关有用的指导十分必要的 5  
### 3.9.7 用户的缺位
这种障碍对于DW/BI项目群来说是典型的致命因素，IT机构成员宣称他们已经非常了解用户的需求就会出现这种情况，IT机构自行其事，基于自己构建的源数据布局进行数据建模，随后却纳闷业务用户对使用数据仓库不那么热心  
## 3.10 小结
Kimball生命周期方法有一条基本信念：数据仓库和商业智能必须把着眼点放在业务及其需求上  
本章提到的工具进行业务需求收集十分重要，随着DW/BI设计与开发工作的进一步展开，它们的价值更是无法估量了  
## 3.11 管理工作和降低风险
需求定义过程的最大风险就是跳过了需求收集过程，如果没有坚实的业务需求基础，建立DW/BI系统就会面临极大的风险  
从管理的角度来看，需求过程进行有限次迭代是非常重要的，设计团队和开发团队也应该对业务群体和他们的需求做进一步了解，另外建议需求过程不要没有止境地扩展  
区分优先级的活动对于管理需求定义也十分重要，需要搞清楚哪些需求是关键需求，哪些是次要需求，这将对生命周期中各项工作的推进非常有益 
## 3.12 保证质量
与合适的人员进行访谈对确保需求定义过程产生高质量的结果十分重要  
需求定义的过程可以从横向和纵向两个方向描述，缺乏纵向的考察，团队就会容易变得近视；缺乏横向的视角，团队就容易孤立自己  
要同受访者一起确认单独访谈的内容，同业务用户一起审查综合整理后的调查结果，还要和他们一起制定下一阶段的决策  
## 3.13 关键角色
+ 业务分析员驱动需求收集和整理存档
+ 项目经理、业务项目领导者协助各项需求收集，在项目层的需求分析中让数据模型师以及BI应用程序开发人员参与进来  
+ 有代表性的业务用户和管理人员应当主动参与到访谈和优先级区分活动中来，对源系统有深入了解的IT人员也应当以数据审计受访者的身份参与进来    

## 3.14 关键提交内容
+ 访谈问卷
+ 业务用户启动会日程安排
+ 给受访者的访前情况介绍信函
+ 项目/项目群访谈摘要
+ 项目群调查结果文档，包括企业数据仓库总线矩阵和达成一致的优先级表
+ 项目需求调查结果文档

## 3.15 工作量估计
收集需求的工作量受项目规模和范围影响，一般来说应当尽量在3-4周内完成需求定义工作，取决于以下因素：
+ 需求会谈的数量，每天最多3-4次，每次会谈要持续60-90分组，安排30分钟进行汇报和休息
+ 避免和受访者日程表冲突，尽量避免在比较繁忙的时节，如制定年度计划或者其他活动期间，这些活动会影响到需求日程安排，也影响到会谈的质量
+ 需要用到的源系统数量，IT数据审计访谈会的数量受到源系统数量的影响
+ 生成文档的级别，大约要未每次会谈分配4个小时的时间来完成访谈总结，这其中包括审查和更新的过程，在项目计划中要为分析访谈结果和整理调查结果文档安排一定的时间
+ 安排促谈会进行优先级区分并达成共识，在全部的调查结果文档完成之前就应当把促谈会的日程确定下来，与会人员的日程安排将会对会议的时间又较大的影响，不要忘记还要花点时间为会议准备材料
