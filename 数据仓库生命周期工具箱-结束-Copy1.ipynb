## 13.2 文档和培训
开发人员可能希望业务用户不需要大量的文档和培训就能够使用新系统，但是，毕竟您要花费大量的时间和精力用某种合乎情理的方式来组织和命名这些事项  
根据我们的经验，用户确实需要一些关于新特性和新资源方面的指导。开发好的面向用户的系统文档并不困难，大部分文档其实都是披着华丽外衣的元数据  
正如我们在第11章和第12章中讲述的，BI门户是机构报表、分析和相关信息的唯一的来源。围绕着BI门户的主页，用户应该能够找到这里所讲述的文档和工具的链接    
### 13.2.1 核心文档
首先对数据编制文档。应该描述业务过程的主题范围，包括事实、维度、表、列、计算，还包括构成这些主题范围的其他规则  
1. 业务过程维度模型描述  
    大多数BI文档出发点是业务过程维度模型。应该为系统中每个维度模型写一个清晰而简洁的描述 5  
    例如，如果在总线矩阵的初始行选择是订单处理，那么要编制一个描述订单主题范围的文档，可以通过回答以下问题完成：  
    + 在该数据中捕捉到的业务过程的本质是什么？  
    + 主要的业务规则是什么？
    + 事实表的粒度是什么？
    + 包含在事实表中的数据的数据范围是什么？ 5
    + 哪些数据被忽略(为什么要忽略)？
    + 哪些维度参与了这次业务过程？很多维度都要有自己的描述性文档，而且本文档能够链接到这些文档
   
   这个文档应该有一些能够以图形形式显示屏幕捕捉的目标模式（正如第7章中讲的高层模型图那样）、一些示例值和少量证明该文档能够解决业务问题类型的演示报表  
2. 表描述和列描述  
    一旦人们对某个特定模式有了大体的了解，他们就要求能够逐表逐行地向下钻取获取细节    
    在建立初始目标模型时捕捉地描述性元数据就时用在这里地。表名、列名、描述、样本值，甚至所有注释，都会帮助用户理解该表格的内容 5  
    如果在一个可查询地数据库存储了大量有关表和列的元数据，应该建几个报表，以便用户浏览这些信息  
    大部分OLAP数据库都能够自行编制文档，尤其如果您已经特意填充了所有描述性字段  
3. 报表描述  
    每个报表都必须有一套基本的描述性信息作为第12章中讲述地标准模板的一部分。在报表地定义中可以包含一些这种描述性信息，如报表的标题和描述  

### 13.2.2 补充文档
数据文档和报表文档当然时最常用的，但是其他文件也很重要。最有价值的补充文档都是以在线教程和服务指南形式等形式出现的，或者时使用该系统的同事清单和其他可能有帮助的文档形式    
正如本章后面要讨论的，您应该为业务用户提供培训 5  
服务指南能告诉您的业务用户在他们有问题的时候应该找谁  
### 13.2.3 用户培训
BI应用程序的一个主要目的是向80%-90%不会学习如何直接查询数据的结构提供对数据仓库中数据的访问  
您应该为将要用到BI应用程序的所有人员开设一期短期培训班，即使他们从未接触过即席工具  
培训高级用户直接查询数据仓库分为两个阶段：开发阶段和提交阶段。开发阶段包括设计课程和编写培训教材的全部工作。、教材和练习的实际课堂演示，以及基于Web的教材发布    
您必须指定大部分培训教材，因为培训过程中最重要的是您的数据特征和环境特征 5  
当数据库稳定下来并且已经选定前端工具之后，您才需要开始着手业务用户培训的工作，但是要在实际部署工作全面铺开，能够创建并测试一套可靠的课程教材之前就开始这项工作     
1. 设计与方法  
    在您真正坐下来写教材之前，花几分钟时间仔细考虑一下您的听众时哪些人，他们具有什么样的技能水平，以及他们使用工具的熟练程度如何  
    这将有助于您确定自己的总体课程：您将如何授课以及它们之间的相互关系如何  
    系统首次亮相的时候通常需要开设两个班：一个持续一到两天的介绍性的即席查询班和一个持续一到两个小时的短期BI应用程序班  
    提供一个单独以数据为中心的班，这对每个加入到数据仓库的新业务过程主题范围也很有用处 5  
    设计过程的部分内容包括拟定各个班的教学提纲，参见如图13-1所示的即席查询班。在根据要讲授的内容以及讲授时间进行课程的开发、测试和提交时，提纲也是不断演化和充实的  
    ![13-1](../syn_pic/data_warehouse_life_cycle/13-1.png)  
2. 开发培训教材  
    编写高品质的用于时机操作的课程教材是极其困难的。许多经典的沟通原则在此处非常适用：每个培训模块应该不超过成年人平均注意力的时间跨度，也就是大约在45分钟-1小时之内  
    各个模块也应具有相同的结构，以课程概要和学员应该该模块中学到的要点知识开始。模块主体应使用一个相关的业务问题作为贯穿教材的动机  
    这些模块应逐步变得更加复杂，由早期提供的一步一步的说明到后来提供的更高层次的指导。每个模块快要结束时要进行额外练习来巩固初学者掌握的内容 5  
3. 建立培训数据库  
    直接针对生产数据仓库的数据库来设计课程是不错的主意。然而，通常也可以创建一个培训数据库。培训数据库能屏蔽机密数据并确保查询性能始终如一  
    应该及早在课程开发过程中创建培训数据库，其使用的典型数据集反映了当人们结束课程后回到各自工作岗位上的真实情况  
    确保将培训数据库中内容的讲授作为第一个模块的一部分。如果培训是非常具体的业务过程，就在开班之前增加一个完整的数据内容模块  
4. 工作量水平计划  
    开设好的培训课程要花很多功夫。您应该至少花8小时来创建1小时的上课教材 5  

## 13.3 维护与支持
让您的DW/BI系统保持在最佳状态并不难，但您需要一开始就计划并构建一个可维护的系统  
不论是前台还是后台都必须规划维护活动。在前台，应该制定与用户沟通、处理用户疑问和安全管理规程和机制  
后台团队应当制定合理的计划来运行ETL作业、监视资源、进行运行性能调整，以及备份和恢复重要数据  
遗憾的是，许多CIO还是在新系统投入运行后自作主张地将DW/BI团队地人员减半。这是一种不切实际地期望，原因有两点  
首先，成功的数据仓库是始终变化的，因此仍然经常需要进行一些开发方面的工作。其次，即使没有变化，您也应该看到人员从后台开发转移到前台支持 5  
### 13.3.1 管理前台
我们在本节详细讲述前台运营团队的主要任务  
1. 提供用户支持  
    精心设计和实现的DW/BI系统比设计不当的更易于使用，但使用起来仍然有难度。DW/BI团队需要向用户社区提供运行支持。我们推荐一个三层支持方法  
    第一层是网站和自服务支持；第二层是您的业务分组中的超级用户；第三层是DW/BI团队的前端人员  
    注意：对于真正的帮助，您需要让业务用户与那些了解业务问题、数据内容和BI工具及应用程序的人员沟通   
    + **第一层：网站**。具有丰富的内容和用于查找它们的工具（导航，搜索以及元数据浏览器）是通过网站提供支持的基础 5  
    + **第二层：超级用户**  
        如果某个用户需要获得帮助来创建一个即席查询，或者需要使用某个当前不存在的报表，那么他们需要寻找有专业技能的人以获得帮助  
        要将用户的期望设定为他能够与某个了解业务的人进行初步接洽，而且最好就是在那个人所在的部门。我们将这种能够提供帮助的联络人称作超级用户  
    + **第三层：DW/BI团队**  
        当网站和本地专家也不能解决问题时，DW/BI团队应提供最后一种支持资源。前端团队实际上有责任维护所有的支持层。例如，一个8人的DW/BI团队，至少要有4个人负责前台工作  

   注意：在一些大型机构里，DW/BI团队中的BI部分从数据仓库人力资源中分离出去成为独立的实体  
   让团队人员专注于前端工作很重要，但从长远来看把他们分离出自己的机构一般都会降低生产率 5  
   正如我们在第12章中所说，BI应用程序应该是通过下拉菜单，选择列表以及帮助屏幕实现自支持的  
   然后，BI团队也需要监视BI应用程序的使用情况，当数据和数据结构发生改变后要对它们进行维护，并且在有了可用的补充数据后及时扩展并加强它们  
   BI团队也需要向用户提供途径，便于他们对已有BI应用程序提供反馈信息和请求新的应用程序   
   许多IT机构在开始构建他们的DW/BI系统时的目的是让用户自己创建报表。而真正的目的是提供更多的自服务——IT人员想从报表生成的跑步机上下来  
   遗憾的是，即使数据访问更加容易了，多数知识工作者还是没有时间和兴趣从头开始学习如何满足自己的信息需求 5  
   BI团队要把创建自定义报表包括在其责任清单里，并确保有足够的人手来满足最重要的请求  
   
   
   
   
# 编辑点   
2. 维护BI门户  
    门户上应该出现的操作集和维护信息集如下：
    + 数据仓库状态：每个业务过程最近几天的数据是说明
    + 计划停机的时间安排
    + 任何非计划停机的诚实状况和完成状况  
    + 明确警告用户系统存在的问题，如数据质量问题，还包括诚实和完成状况的评估问题何时可以解决 5  
    + 系统当前的运行状况，其中包括：
        + 上一小时回复了多少查询
        + 已经生成了多少报表
        + 当前的活动用户的数量
        + 按照活动查询生成的报表，包括这些报表运行了多久以及由谁来运行它们
    + 关于系统容量的证明点，用于培养用户对系统的信心：
        + 上一次数据加载花了多长时间 5  
        + 最大数据装载的数据量由多大，以及装载要花多长时间
        + 允许同时存在的最大用户数  
    
   每12-18个月就应该审查一下整个DW/BI系统，评估需要进行哪些变更才能很好地为用户工作  
3. 管理安全性  
    安全措施应该在数据库层面上实现，部分安全管理员运行职责是确保报表和其他BI应用程序受到恰当地保护    
    大多数报表工具都由相当直观的用户界面来管理角色和特权 5  
    在数据库中实现的安全措施通常由DW/BI安全管理员处理，一个相对简单的自定义工具能在安全管理中极大地降低负担和潜在的错误  
4. 监视器的使用  
    DW/BI系统的使用状况对其性能有巨大的影响，应该开发一个监视系统，收集查询运行数和应用程序类型返回的行数  
    也要收集一些查询样本，样本有助于根据新的所有或聚集来调整系统性能  
    一些机构收集所有提交给系统的查询文本，这实际上会给系统带来巨大的性能负担，供选方案是偶尔在日志中记录查询文本  
    您应该记录有哪些用户登录到了系统，包括BI门户、BI应用程序以及数据库的用户验证登陆程序，查询文本是非常有价值的，单存储所有的查询文本的代价就太大了 5  
5. 关于使用情况的报表  
    BI门户网站应该划出少量屏幕区域来生成有关系统使用情况的报表，特别是分部门的使用情况  
    还应该知道哪些人正在对系统做什么操作、进行了多少操作，以及何时进行的操作，对系统性能调整至关重要  

### 13.3.2 管理后台
后台系统管理由数据调谐支持、ETL系统执行和监视、资源监视、数据增长和磁盘空间管理、性能调整、备份和恢复组成  
DW/BI系统的管理越自动化越好，越小的团队越应该做自动化  
1. 支持数据调谐  
    数据质量验证和调谐必须在所有用户访问之前就进行 5  
    尤其当发现以前的报表信息不正确时，可能需要进行一些挖掘来帮助用户充分理解数年来运营系统使用的信息是不准确的  
2. 执行和监视ETL系统  
    ETL系统应该自动防范攻击和自动修复可预测的错误  
    可以设计一个错误事件表，您必须制定和遵守一定的规程来惊恐研究和修正错误行  
3. 监视资源  
    您需要实现理解两类监视：
    1. 您的系统要发生什么？正在使用什么资源？何时使用以及由哪个进程使用?  
    2. 哪些人在访问系统？何时访问系统？他们在运行什么样的查询以及这些查询用了多长事件？ 5  
    
   DW/BI供应商提供了很多监视工具，监视DW/BI系统每个组件的资源使用情况  
   除了DW/BI部件的集成信息，还需要组合信息，甚至是同一个组件内的组合信息以便于使用和分析  
   运营计划的一部分应该建立一个系统基线性能配置文件  
   跟踪性能基线最重要的独立：内存、处理器以及磁盘使用，制定计划15分钟为间隔将监视信息存储在数据库  
4. 管理磁盘空间  
    磁盘空间用户是ETL作业最常见也最容易预防的 5  
    操作系统要设置一个报警器，两个警报：将一个月内用完磁盘空间发出警报，一周内用完发出警报  
    重点关注您所积累的大文件来检查剩余空间：
    + 用于事实表抽取的登台文件，按照每月数据量考虑，做以恶搞保守的估计，在运营计划上进行记录以便定期重新检查  
    + 关系数据库文件分组，包括临时空间，监视数据库文件存放的磁盘
    + OLAP数据库，包括临时空间  
    
   每个月您所使用的增量磁盘空间都会变成事实数据存储 5  
5. 性能调整  
    应该把性能调整作为系统进入生产环境时部署和测试过程的一部分  
    制定一项计划，按季度、半年或年度来安排运行性能调整，一份有关关键统计数据的简单报表有助于评估系统性能
6. 备份和恢复  
    不论系统的可用性需求如何，都需要一份备份和恢复计划  
    测试这些规程也同样重要，至少应当练习一次数据库恢复，做大量笔记  
    您要为每日或每月的装载周期制定故障计划 5  
    关系数据库在每次装载后要备份以下数据集：
    + 关系数据仓库数据库
    + 登台数据库
    + 文件系统中的登台数据
    + 元数据数据库  
    
   需要为数据仓库日志数据库建立常规的备份调度 5  
    关系数据库通常很大，在夜间装载周期末尾运行备份通常有挑战性，这里有几种替代方案  
    + 把数据仓库存储到一个存储区域网络（SAN)并使用SAM软件来执行备份 
    + 对大事实数据表进行分区，并将老分区设置为只读，主要依靠各个分区不同的备份策略
   
   日志数据库被不断写入，更像事务处理数据库而不是数据仓库数据库，需要制定严格的备份策略和恢复策略  
   登台数据库的备份策略和恢复策略有很多不同，多数仅包含最近加载的数据 5  
   元数据数据库也应像一个事务处理数据库那样使用完全恢复模式和日志备份  
   您的ETL工具、OLAP数据库、报表工具或者作业调度程序可能都会将信息存储到数据库，这些工具信息应该进行一致的备份  
   对于OLAP数据库的备份，您需要一个基于文件的备份机制或使用基于SAN的策略
7. 长期存档  
    许多企业都有一些数据由于法律和制度的原因必须存档很长事件  
    我们推荐一个进行长期存档的两部分策略：
    1. 不要把赌注押在任何一种物理媒介技术上，应该制定一个定期的迁移和刷新程序 5  
    2. 在虚拟机上封装当前应用运行环境，存档改虚拟机镜像，并提交迁移和刷新周期
    
## 13.4 小结
成功的DW/BI系统部署和运营需要在开发工作完成之前进行充分的规划和协调  
首先要关注的技术是：测试应用程序，并且将组件从开发阶段转到测试阶段然后投入生产阶段  
一个可靠的测试流程包括不同种类的测试，只有完成系统测试后才能准备部署  
您还需要关注桌面安装、集中式的用户培训和支持，一旦系统投入生产环境，就需要组建支持团队  
## 13.5 管理工作和降低风险  
最大风险是投入的工作量不足，因为多数CIO很难相信在运行操作过程中需要的团队应该与开始部署系统时具有一样的规模  
新DW/BI系统的最大部署风险是未能诚实地评估您地部署准备情况  
健康的系统不可避免会发生变化，要在实际事件之前部署变更，计划、记录和测试整个过程  
面向用户的部署难题是低风险的，可以通过使用团队以外的资源来提升教材质量  
前台运营需要一个规模较大的团队，CIO和业务发起人要确保安排好所需的各种人员，后台维护应尽量自动化  
## 13.6 质量保证
开发和质量保证团队必须进行比较平稳的部署，高质量的培训和文档教材展示了DW/BI系统的专业形象
## 13.7 关键角色
+ 主测试员负责制定和监督测试计划，制定测试方法，选择测试工具，并与所有其他测试员进行协调  
+ 质量保证分析员负责大多数预部署测试，包括功能测试，部署流程测试以及重要的数据质量测试  
+ 数据库管理员负责性能测试和调整，要在部署之前和持续运营过程中进行
+ 数据仓库施训人员的职责是开发和提交业务用户培训课程
+ 项目经理和业务项目指导者基于管理层和项目发起人的指示或者反馈意见来确定用户支持机构的结构和职员

## 13.8 关键提交内容
+ 测试计划：测试方法，测试工具，测试脚本和测试数据集
+ 测试系统部署方案：要详细说明如何把新系统从测试环境迁移到生产环境
+ 业务用户文档，利用整个生命周期中开发的文档或元数据
+ 培训教材和数据库
+ 运营计划：数据库维护活动和时间表、备份策略和恢复策略
+ 前台支持计划：服务台计划配备的职员，DW/BI团队中面向用户的BI分析员以及业务用户社区的超级用户

## 13.9 工作量估计
预部署测试阶段通常会持续四到八周  
有了好的元数据，就可以花几天来写总价、开发元数据报表并把他们集成到BI门户中  
如果您还没有编制单独的文档来描述表、列或者预定义报表，那么您的各种还需要花费好几周  
需要花8天事件来为为期一天的培训会议做准备  
前台运行支持需要相当大规模团队，这个团队应当与最初的DW/BI开发团队规模一样大
